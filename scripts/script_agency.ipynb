```sql
/****** Object:  UserDefinedFunction [dbo].[Split]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE FUNCTION [dbo].[Split](@String varchar(8000), @Delimiter char(1))       
returns @temptable TABLE (items varchar(8000))       
as       
begin       
    declare @idx int       
    declare @slice varchar(8000)       
      
    select @idx = 1       
        if len(@String)<1 or @String is null  return       
      
    while @idx!= 0       
    begin       
        set @idx = charindex(@Delimiter,@String)       
        if @idx!=0       
            set @slice = left(@String,@idx - 1)       
        else       
            set @slice = @String       
          
        if(len(@slice)>0)  
            insert into @temptable(Items) values(@slice)       
  
        set @String = right(@String,len(@String) - @idx)       
        if len(@String) = 0 break       
    end   
return       
end
GO
```

# [dbo].[CheckLogin]

```sql
/****** Object:  StoredProcedure [dbo].[CheckLogin]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





















































CREATE PROCEDURE [dbo].[CheckLogin]
	@UserID nvarchar(100),
	@Password nvarchar(100)
AS
BEGIN
	-- Update Last Login
	IF EXISTS (SELECT ID FROM UserMaster WHERE UserId = @UserID AND Password = @Password)
	UPDATE UserMaster Set LastLogin = GETDATE() WHERE UserId = @UserID AND Password = @Password
	
	--Return Total Record
	SELECT * FROM UserMaster WHERE UserId = @UserID AND Password = @Password
END



















































GO
```

# [dbo].[Convert_Estimate_To_Invoice]

```sql
/****** Object:  StoredProcedure [dbo].[Convert_Estimate_To_Invoice]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






CREATE PROCEDURE [dbo].[Convert_Estimate_To_Invoice] 	
	@EstimateId numeric(18,0),
	@VoucherDate nvarchar(50),
	@F_CompanyYears numeric(18,0),
	@F_UserMaster numeric(18,0)
AS
BEGIN
	--Checking if already converted
	DECLARE @F_SaleInvoiceH NUMERIC(18,0) = ISNULL((SELECT Id FROM SaleInvoiceH WHERE ISNULL(F_EstimateH,0)=@EstimateId),0)
	IF(@F_SaleInvoiceH>0)
		BEGIN
			SELECT -1 AS Id
		END
	ELSE
		BEGIN
			--Making entry in SaleInvoiceH
			DECLARE @VoucherNo NVARCHAR(50) = (SELECT (ISNULL(MAX(CAST(VH.VoucherNo AS NUMERIC)),0)+1) AS VoucherNo
												FROM SaleInvoiceH VH
													LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster = VH.F_CompanyMaster AND CY.Id=@F_CompanyYears
												WHERE ISNULL(VoucherNoPrefix,'')='' AND (VH.VoucherDate>=CY.FinYearFrom AND VH.VoucherDate<=CY.FinYearTo))

			INSERT INTO SaleInvoiceH([DateOfCreation],[VoucherNo],[VoucherDate],[F_LedgerMaster],[Remark],[F_UserMaster],[F_CompanyMaster],[IsVATIncluded],[Cancelled],
				[TotalProDisAmt],[TotalDisplayAmt],[TotalCashDisAmt],[TotalReplacementAmt],[TotalGrossAmt],[TotalTaxationAmt],[TotalNetAmt],[TotalLessSpecialAmt],
				[F_OtherLedgerMaster],[IsScheme],[TotalQty],[TotalFree],[TotalQtyxRate],[TotalAmount],[TotalQPSDisAmt],[TotalFinalAmt],[IsChecked],[F_ItemBrandMaster],
				[LedgerName],[VoucherNoPrefix],[F_LedgerGroupMaster],[IsDisputed],[IsOtherState],[RoundOff],[BillNo],[TotalAdditionalTaxAmt],
				[F_EstimateH])
			SELECT GETDATE(),@VoucherNo,@VoucherDate,[F_LedgerMaster],[Remark],@F_UserMaster,[F_CompanyMaster],[IsVATIncluded],[Cancelled],
				[TotalProDisAmt],[TotalDisplayAmt],[TotalCashDisAmt],[TotalReplacementAmt],[TotalGrossAmt],[TotalTaxationAmt],[TotalNetAmt],[TotalLessSpecialAmt],
				[F_OtherLedgerMaster],[IsScheme],[TotalQty],[TotalFree],[TotalQtyxRate],[TotalAmount],[TotalQPSDisAmt],[TotalFinalAmt],[IsChecked],[F_ItemBrandMaster],
				[LedgerName],[VoucherNoPrefix],[F_LedgerGroupMaster],[IsDisputed],[IsOtherState],[RoundOff],[BillNo],[TotalAdditionalTaxAmt],
				@EstimateId
			FROM EstimateH
			WHERE Id = @EstimateId
			SET @F_SaleInvoiceH = (SELECT SCOPE_IDENTITY() AS Id)

			--Making entry in SaleInvoiceH_AdditionalTaxation
			INSERT INTO SaleInvoiceH_AdditionalTaxation([F_LedgerMaster_Tax],[AddAmt],[TaxableAmt],[F_SaleInvoiceH])
			SELECT [F_LedgerMaster_Tax],[AddAmt],[TaxableAmt],@F_SaleInvoiceH
			FROM EstimateH_AdditionalTaxation
			WHERE F_EstimateH = @EstimateId

			--Making entry in SaleInvoiceH_CreditNote
			INSERT INTO SaleInvoiceH_CreditNote([F_SaleInvoiceH],[F_VoucherH],[F_SaleReturnH])
			SELECT @F_SaleInvoiceH,[F_VoucherH],[F_SaleReturnH]
			FROM EstimateH_CreditNote
			WHERE F_EstimateH = @EstimateId
			--Updating VoucherH For CreditNote
			UPDATE VH
			SET PurVoucherNo = SH.VoucherNo,
				PurVoucherDate = SH.VoucherDate
			FROM VoucherH VH
			LEFT JOIN SaleInvoiceH_CreditNote SHC ON VH.Id = SHC.F_VoucherH
			LEFT JOIN SaleInvoiceH SH ON SH.Id = SHC.F_SaleInvoiceH
			WHERE SH.Id = @F_SaleInvoiceH
	
			--Making entry in SaleInvoiceH_DebitNote
			INSERT INTO SaleInvoiceH_DebitNote([F_SaleInvoiceH],[F_VoucherH],[F_SaleReturnH])
			SELECT @F_SaleInvoiceH,[F_VoucherH],[F_SaleReturnH]
			FROM EstimateH_DebitNote
			WHERE F_EstimateH = @EstimateId
			--Updating VoucherH For DebitNote
			UPDATE VH
			SET PurVoucherNo = SH.VoucherNo,
				PurVoucherDate = SH.VoucherDate
			FROM VoucherH VH
			LEFT JOIN SaleInvoiceH_DebitNote SHC ON VH.Id = SHC.F_VoucherH
			LEFT JOIN SaleInvoiceH SH ON SH.Id = SHC.F_SaleInvoiceH
			WHERE SH.Id = @F_SaleInvoiceH
	
			--Making entry in SaleInvoiceL
			INSERT INTO SaleInvoiceL(
				[F_SaleInvoiceH],[F_ItemMaster],[Remark],[MRP],[F_UnitMasterQ],[Qty],[Rate],[F_UnitMasterF],[Free],[Display],[F_LedgerMaster_Tax],[QtyxRate],[Amount],[Replacement],
				[FinalAmt],[IsManualProDis],[IsManualCashDis],[TaxAmt],[TaxableAmt],[TaxRate],[ProductDiscRate],[ProductDiscAmt],[CashDiscRate],[CashDiscAmt],[MinQty],[MinQtyF],
				[QPSRate],[QPSAmt],[IsManualQPSDis],[F_ItemMasterF],[SNo],[TmpTypeId],[F_EstimateL],[SellingRate],[F_LedgerMaster_Tax2],[TaxAmt2],[TaxableAmt2],[TaxRate2],
				[F_LedgerMaster_Tax3],[TaxAmt3],[TaxableAmt3],[TaxRate3],[TaxableAmt4],[TaxRate4])
			SELECT
				@F_SaleInvoiceH,[F_ItemMaster],[Remark],[MRP],[F_UnitMasterQ],[Qty],[Rate],[F_UnitMasterF],[Free],[Display],[F_LedgerMaster_Tax],[QtyxRate],[Amount],[Replacement],
				[FinalAmt],[IsManualProDis],[IsManualCashDis],[TaxAmt],[TaxableAmt],[TaxRate],[ProductDiscRate],[ProductDiscAmt],[CashDiscRate],[CashDiscAmt],[MinQty],[MinQtyF],
				[QPSRate],[QPSAmt],[IsManualQPSDis],[F_ItemMasterF],[SNo],[TmpTypeId],Id AS F_EstimateL,[SellingRate],[F_LedgerMaster_Tax2],[TaxAmt2],[TaxableAmt2],[TaxRate2],
				[F_LedgerMaster_Tax3],[TaxAmt3],[TaxableAmt3],[TaxRate3],[TaxableAmt4],[TaxRate4]
			FROM EstimateL
			WHERE F_EstimateH = @EstimateId
	
			--Making entry in SaleInvoiceL_Replacement
			INSERT INTO SaleInvoiceL_Replacement([F_SaleInvoiceH],[Item],[MRP],[Qty],[Rate],[Amount])
			SELECT @F_SaleInvoiceH,[Item],[MRP],[Qty],[Rate],[Amount]
			FROM EstimateL_Replacement
			WHERE F_EstimateH = @EstimateId
	
			--Making entry in SaleInvoiceL_TypeDetailMaster
			INSERT INTO SaleInvoiceL_TypeDetailMaster([F_SaleInvoiceH],[F_TypeDetailMaster],[Qty],[F_ItemMaster],[TmpTypeId])
			SELECT @F_SaleInvoiceH,[F_TypeDetailMaster],[Qty],[F_ItemMaster],[TmpTypeId]
			FROM EstimateL_TypeDetailMaster
			WHERE F_EstimateH = @EstimateId
	
			--Making entry in SaleInvoiceL_TypeDetailMaster_TMP
			INSERT INTO SaleInvoiceL_TypeDetailMaster_TMP([F_SaleInvoiceH],[F_TypeDetailMaster],[Qty],[F_ItemMaster],[TmpTypeId])
			SELECT @F_SaleInvoiceH,[F_TypeDetailMaster],[Qty],[F_ItemMaster],[TmpTypeId]
			FROM EstimateL_TypeDetailMaster_TMP
			WHERE F_EstimateH = @EstimateId
	
			--Making entry in SaleInvoiceLTax
			INSERT INTO SaleInvoiceLTax([F_SaleInvoiceH],[F_LedgerMaster],[AddAmt],[LessAmt],[TaxableAmt])
			SELECT @F_SaleInvoiceH,[F_LedgerMaster],[AddAmt],[LessAmt],[TaxableAmt]
			FROM EstimateLTax
			WHERE F_EstimateH = @EstimateId
	
			SELECT @F_SaleInvoiceH AS Id
		END
END



















































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Insert_AdditionalTaxation]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO







CREATE PROCEDURE [dbo].[Insert_AdditionalTaxation]
	@TableNameH nvarchar(50),
	@F_InvoiceH_Field nvarchar(50),
	@F_InvoiceH numeric(18, 0),
	@F_LedgerMaster_Tax numeric(18, 0),
	@AddAmt numeric(18, 2),
	@TaxableAmt numeric(18, 2)
AS
BEGIN
	DECLARE @Id NUMERIC(18,0)
	IF(@TableNameH = 'SaleInvoiceH_AdditionalTaxation')
		BEGIN
			SET @Id = ISNULL((SELECT Id FROM SaleInvoiceH_AdditionalTaxation WHERE F_SaleInvoiceH = @F_InvoiceH AND F_LedgerMaster_Tax = @F_LedgerMaster_Tax),0)
		END
	ELSE IF(@TableNameH = 'SaleReturnH_AdditionalTaxation')
		BEGIN
			SET @Id = ISNULL((SELECT Id FROM SaleReturnH_AdditionalTaxation WHERE F_SaleReturnH = @F_InvoiceH AND F_LedgerMaster_Tax = @F_LedgerMaster_Tax),0)
		END
	ELSE IF(@TableNameH = 'PurchaseInvoiceH_AdditionalTaxation')
		BEGIN
			SET @Id = ISNULL((SELECT Id FROM PurchaseInvoiceH_AdditionalTaxation WHERE F_PurchaseInvoiceH = @F_InvoiceH AND F_LedgerMaster_Tax = @F_LedgerMaster_Tax),0)
		END
	ELSE IF(@TableNameH = 'PurchaseReturnH_AdditionalTaxation')
		BEGIN
			SET @Id = ISNULL((SELECT Id FROM PurchaseReturnH_AdditionalTaxation WHERE F_PurchaseReturnH = @F_InvoiceH AND F_LedgerMaster_Tax = @F_LedgerMaster_Tax),0)
		END

	IF(@Id=0)
		BEGIN
			EXEC('INSERT INTO [dbo].[' + @TableNameH + ']([' + @F_InvoiceH_Field + '],[F_LedgerMaster_Tax],[AddAmt],[TaxableAmt])
					VALUES (' + @F_InvoiceH + ',' + @F_LedgerMaster_Tax + ',' + @AddAmt + ',' + @TaxableAmt + ')
				  SELECT SCOPE_IDENTITY() AS Id')
		END
	ELSE
		BEGIN
			EXEC('UPDATE ' + @TableNameH + '
				  SET AddAmt = ' + @AddAmt + ',
					TaxableAmt = ' + @TaxableAmt + '
				  WHERE ' + @F_InvoiceH_Field + ' = ' + @F_InvoiceH + ' AND F_LedgerMaster_Tax = ' + @F_LedgerMaster_Tax + '')
		END
END
GO
```

# [dbo].[Insert_Update_EInvoiceCancelAknowledgementH]

```sql
/****** Object:  StoredProcedure [dbo].[Insert_Update_EInvoiceCancelAknowledgementH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





CREATE PROCEDURE [dbo].[Insert_Update_EInvoiceCancelAknowledgementH]
	@F_UserMaster numeric(18, 0),
	@F_CompanyMaster numeric(18, 0),
	@IRN nvarchar(150),
	@Cancelled nvarchar(50),
	@CancelUser nvarchar(50),
	@Status nvarchar(50)
AS
BEGIN
--Getting existing IRN details	
	DECLARE @Id NUMERIC(18,0) = ISNULL((SELECT Id FROM EInvoiceAknowledgementH WHERE IRN=@IRN),0)
	IF(@Id>0)
		BEGIN
		--Updating cancelled details in existing IRN details
			UPDATE EInvoiceAknowledgementH
			SET [LastUpdateOn] = GETDATE(),
				[CancelledDate] = @Cancelled,
				[CancelUser] = @CancelUser,
				[CancelledStatus] = @Status
			WHERE [IRN] = @IRN AND Id = @Id
			
		--Getting Doc Type
			DECLARE @DocType nvarchar(50) = ISNULL((SELECT DocTyp FROM EInvoiceAknowledgementH WHERE IRN = @IRN AND Id = @Id) ,0)
		--Getting Doc No.
			DECLARE @DocNo nvarchar(50) = ISNULL((SELECT DocNo FROM EInvoiceAknowledgementH WHERE IRN = @IRN AND Id = @Id) ,0)
		--Getting Doc Date
			DECLARE @DocDate nvarchar(50) = ISNULL((SELECT DocDate FROM EInvoiceAknowledgementH WHERE IRN = @IRN AND Id = @Id) ,0)

		--Updating cancellation information in
			DECLARE @VId AS NUMERIC(18,0)
			IF(@DocType = 'Tax Invoice')			--Sale Invoice			
				BEGIN
					UPDATE SaleInvoiceH SET Cancelled=1 WHERE VoucherNo = @DocNo AND VoucherDate = @DocDate
					--Deleting Account Posting
					SET @VId = (SELECT Id FROM SaleInvoiceH WHERE VoucherNo = @DocNo AND VoucherDate = @DocDate)
					EXEC Delete_VoucherL @VId,9
				END
			ELSE IF(@DocType = 'Credit Note')		--Credit Note (Sale Return)
				BEGIN
					UPDATE SaleReturnH SET Cancelled=1 WHERE VoucherNo = @DocNo AND VoucherDate = @DocDate
					--Deleting Account Posting
					SET @VId = (SELECT Id FROM SaleReturnH WHERE VoucherNo = @DocNo AND VoucherDate = @DocDate)
					EXEC Delete_VoucherL @VId,6
				END
			ELSE IF(@DocType = 'Debit Note')		--Debit Note (Purchase Return)
				BEGIN
					UPDATE PurchaseReturnH SET Cancelled=1 WHERE VoucherNo = @DocNo AND VoucherDate = @DocDate
					--Deleting Account Posting
					SET @VId = (SELECT Id FROM PurchaseReturnH WHERE VoucherNo = @DocNo AND VoucherDate = @DocDate)
					EXEC Delete_VoucherL @VId,8
				END

			SELECT @Id AS Id
		END
END



GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Insert_Update_MoneyReceiptBankH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE PROCEDURE [dbo].[Insert_Update_MoneyReceiptBankH]
	@VoucherNo nvarchar(50),
	@VoucherDate nvarchar(50),
	@ChequeNo nvarchar(50),
	@ChequeDate nvarchar(50),
	@F_LedgerMaster numeric(18, 0),
	@F_UserMaster numeric(18, 0),
	@F_CompanyMaster numeric(18, 0),
	
	@F_LedgerMaster_Sales numeric(18, 0),
	@F_BankMaster numeric(18, 0),
	@F_BranchMaster numeric(18, 0),
	@F_LedgerMaster_Discount numeric(18, 0),
	@Remark nvarchar(250),
	@IsFullPaid smallint,
	@TotalAmount numeric(18, 2),
	@Discount numeric(18, 2),
	@TotalInvoiceAmt numeric(18, 2),
	@TotalDiffAmt numeric(18, 2),
	@F_OtherLedgerMaster numeric(18,0),
	@NetAmount numeric(18,2),
	@IsChequeCancel smallint,
	@ChequeCancelCharges numeric(18,2),
	@F_LedgerMaster_ChequeCancel numeric(18,0),
	
	@Id numeric(18,0),
	@F_CompanyYears numeric(18,0),
	@F_VoucherH_Journal numeric(18,0),
	@F_LedgerGroupMaster numeric(18,0),
	@Id_DebitNote numeric(18,0),
	@ChequeReceivedDate nvarchar(50)='',
	@ChequeReceivedNo nvarchar(50)=''
AS
BEGIN

	IF(@Id=0)
		BEGIN
			SET @VoucherNo=(SELECT (ISNULL(MAX(CAST(VH.VoucherNo AS NUMERIC)),0)+1) AS VoucherNo
								FROM MoneyReceiptBankH VH LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster = VH.F_CompanyMaster AND CY.Id=@F_CompanyYears
								WHERE ISNULL(VoucherNoPrefix,'')='' AND (VH.VoucherDate>=CY.FinYearFrom AND VH.VoucherDate<=CY.FinYearTo)
							)

			INSERT INTO [dbo].[MoneyReceiptBankH]([DateOfCreation],[VoucherNo],[VoucherDate],[F_LedgerMaster],[F_UserMaster],[F_CompanyMaster],[F_LedgerMaster_Sales],[F_BankMaster],[F_BranchMaster],[F_LedgerMaster_Discount],[Remark],[IsFullPaid],[TotalAmount],[Discount],[TotalInvoiceAmt],[TotalDiffAmt],[F_OtherLedgerMaster],[NetAmount],[ChequeNo],[ChequeDate],[IsChequeCancel],[ChequeCancelCharges],[F_LedgerMaster_ChequeCancel],[F_VoucherH_Journal],[F_LedgerGroupMaster],[Id_DebitNote],[ChequeReceivedDate],[ChequeReceivedNo])
				VALUES (GETDATE(),@VoucherNo,@VoucherDate,@F_LedgerMaster,@F_UserMaster,@F_CompanyMaster,@F_LedgerMaster_Sales,@F_BankMaster,@F_BranchMaster,@F_LedgerMaster_Discount,@Remark,@IsFullPaid,@TotalAmount,@Discount,@TotalInvoiceAmt,@TotalDiffAmt,@F_OtherLedgerMaster,@NetAmount,@ChequeNo,@ChequeDate,@IsChequeCancel,@ChequeCancelCharges,@F_LedgerMaster_ChequeCancel,@F_VoucherH_Journal,@F_LedgerGroupMaster,@Id_DebitNote,@ChequeReceivedDate,@ChequeReceivedNo)
			SELECT SCOPE_IDENTITY() AS Id
		END
	ELSE
		BEGIN
			UPDATE [dbo].[MoneyReceiptBankH]
			SET [LastUpdateOn] = GETDATE(),
				[F_UserMaster] = @F_UserMaster,
				[F_CompanyMaster] = @F_CompanyMaster,		
				
				[VoucherNo] = @VoucherNo,
				[VoucherDate] = @VoucherDate,			
				[ChequeNo] = @ChequeNo,
				[ChequeDate] = @ChequeDate,
				[F_LedgerMaster] = @F_LedgerMaster,
				
				[F_LedgerMaster_Sales] = @F_LedgerMaster_Sales,
				[F_BankMaster] = @F_BankMaster,
				[F_BranchMaster] = @F_BranchMaster,
				[F_LedgerMaster_Discount] = @F_LedgerMaster_Discount,
				[Remark] = @Remark,
				[IsFullPaid] = @IsFullPaid,
				[TotalAmount] = @TotalAmount,
				[Discount] = @Discount,
				[TotalInvoiceAmt] = @TotalInvoiceAmt,
				[TotalDiffAmt] = @TotalDiffAmt,
				[F_OtherLedgerMaster] = @F_OtherLedgerMaster,
				[NetAmount] = @NetAmount,
				[IsChequeCancel] = @IsChequeCancel,
				[ChequeCancelCharges] = @ChequeCancelCharges,
				[F_LedgerMaster_ChequeCancel] = @F_LedgerMaster_ChequeCancel,
				[F_VoucherH_Journal] = @F_VoucherH_Journal,
				[F_LedgerGroupMaster] = @F_LedgerGroupMaster,
				[Id_DebitNote] = @Id_DebitNote,
				[ChequeReceivedDate] = @ChequeReceivedDate,
				[ChequeReceivedNo] = @ChequeReceivedNo
			WHERE [Id] = @Id

			/** UPDATING LOG DETAILS **/
			INSERT INTO LogDetails([DateOfCreation],[F_CompanyMaster],[F_UserMaster],[Type],[LogDetails],[SDate],[FormName],[F_LedgerMaster],[TableName])
				VALUES(GETDATE(),@F_CompanyMaster,@F_UserMaster,'Edit'
					,'Ref. No. :' + @VoucherNo + ', Date :' + (SUBSTRING(@VoucherDate,7,2)+'/'+SUBSTRING(@VoucherDate,5,2)+'/'+SUBSTRING(@VoucherDate,1,4))
					,CONVERT(VARCHAR(8),GETDATE(),112),'Money Receipt Bank',@F_LedgerMaster,'MoneyReceiptBankH')
			/****/

			SELECT @Id AS Id
		END
END
GO
```

# [dbo].[Insert_Update_SaleInvoiceItemMapMaster]

```sql
/****** Object:  StoredProcedure [dbo].[Insert_Update_SaleInvoiceItemMapMaster]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO












CREATE PROCEDURE [dbo].[Insert_Update_SaleInvoiceItemMapMaster]
	@SaleInvoiceItemId NVARCHAR(50),
	@MRP NUMERIC(18,2),
	@F_ItemMaster numeric(18,0),
	@F_UserMaster numeric(18,0),
	@F_CompanyMaster numeric(18,0),
	@F_ItemTypeMaster numeric(18,0),
	@F_TypeDetailMaster numeric(18,0)
AS
BEGIN
	DECLARE @Id AS NUMERIC(18,0)
	SET @Id = ISNULL((SELECT Id FROM SaleInvoiceItemMapMaster WHERE SaleInvoiceItemId=@SaleInvoiceItemId),0)

	IF(@F_ItemMaster=0)
		BEGIN
			IF(@Id>0)
				DELETE FROM SaleInvoiceItemMapMaster WHERE SaleInvoiceItemId = @SaleInvoiceItemId
		END
	ELSE
		BEGIN
			IF(@Id=0)
				BEGIN
					INSERT INTO SaleInvoiceItemMapMaster(F_UserMaster,F_CompanyMaster,DateOfCreation,MRP,F_ItemMaster,F_ItemTypeMaster,F_TypeDetailMaster,SaleInvoiceItemId)
					VALUES(@F_UserMaster,@F_CompanyMaster,GETDATE(),@MRP,@F_ItemMaster,@F_ItemTypeMaster,@F_TypeDetailMaster,@SaleInvoiceItemId)
				END
			ELSE
				BEGIN
					UPDATE SaleInvoiceItemMapMaster
					SET F_ItemMaster = @F_ItemMaster,
						F_ItemTypeMaster = @F_ItemTypeMaster,
						F_TypeDetailMaster = @F_TypeDetailMaster,
						MRP=@MRP,LastUpdateOn=GETDATE(),F_UserMaster=@F_UserMaster,F_CompanyMaster=@F_CompanyMaster
					WHERE Id=@Id
				END
		END
END



GO
```

# [dbo].[LockVouchers]

# [dbo].[Select_Acc_BalanceSheet_Cr]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_BalanceSheet_Cr]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











































































CREATE PROCEDURE [dbo].[Select_Acc_BalanceSheet_Cr] 
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0)
AS
BEGIN			
SELECT TAB3.LedgerGroup,TAB3.Ledger,ABS(TAB3.DebitAmount-TAB3.CreditAmount) AS CreditAmount,(SUBSTRING(@ToDate,7,2)+'/'+SUBSTRING(@ToDate,5,2)+'/'+SUBSTRING(@ToDate,1,4)) AS ToDate
FROM(	
	SELECT TAB2.LedgerGroup,TAB2.Ledger
		,SUM(TAB2.DebitAmount) AS DebitAmount,SUM(TAB2.CreditAmount) AS CreditAmount
	FROM(		
		--Opening
		SELECT LGM.Name AS LedgerGroup, LM.Name AS Ledger		
			,(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount
			,(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount					
		FROM LedgerMaster LM
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		WHERE LM.F_CompanyMaster=@F_CompanyMaster AND LGM.IsBalanceSheet=1
		UNION ALL
		--Transaction
		SELECT LGM.Name AS LedgerGroup,LM.Name AS Ledger
			,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS DebitAmount
			,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS CreditAmount			
		FROM (
				
				SELECT VH.VoucherNo,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate
				,LM.Name AS Ledger,VH.Id
				,LM.Id AS LedgerId
				,VL.DrCr,VL.Amount,VTM.Name AS VoucherType
				FROM [dbo].[VoucherH] VH	
					LEFT JOIN VoucherL VL ON VL.F_VoucherH=VH.Id
					LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id				
					LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VH.F_VoucherTypeMaster
				WHERE (VH.VoucherDate<=@ToDate ) AND VH.F_CompanyMaster=@F_CompanyMaster
					
		) AS TAB1
		LEFT JOIN VoucherL VL ON VL.F_VoucherH=TAB1.Id
		LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		WHERE VL.F_LedgerMaster!=TAB1.LedgerId AND LGM.IsBalanceSheet=1
		GROUP BY LGM.Name,LM.Name,LM.Id
	) AS TAB2
	GROUP BY TAB2.LedgerGroup,TAB2.Ledger
) AS TAB3
WHERE (TAB3.CreditAmount-TAB3.DebitAmount)>0
ORDER BY TAB3.LedgerGroup,TAB3.Ledger
END


















































GO
```

# [dbo].[Select_Acc_BankReconciliation]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_BankReconciliation]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




























CREATE PROCEDURE [dbo].[Select_Acc_BankReconciliation] 
	@F_LedgerMaster numeric(18,0),
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),	
	@F_CompanyMaster numeric(18,0),
	@F_OtherLedgerMaster NUMERIC(18,0)=0,
	@F_VoucherTypeMaster NUMERIC(18,0)=0,
	@OnlyPending smallint=0,
	@F_LedgerMaster_Party numeric(18,0)=0,
	@Amount numeric(18,2)=0,
	@IsFilterByReconcileDate smallint=0
AS
BEGIN
-- OPENING BALANCE
	DECLARE @WHERE_VoucherDate NVARCHAR(MAX)
	DECLARE @WHERE_VoucherDate2 NVARCHAR(MAX)
	DECLARE @WHERE_VoucherDate3 NVARCHAR(MAX)
	DECLARE @ROW_NUMBER NVARCHAR(MAX)
	DECLARE @FinYearFromDate NVARCHAR(MAX)
	DECLARE @FinYearFromDate_Min NVARCHAR(MAX)
	SET @FinYearFromDate = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC)	
	SET @FinYearFromDate_Min = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)

	IF(@IsFilterByReconcileDate=0)
		BEGIN
			SET @WHERE_VoucherDate = '(VL.VoucherDate<' + @FromDate + ' AND VL.VoucherDate>=ISNULL(' + @FinYearFromDate_Min + ',0) )'
			SET @WHERE_VoucherDate2 = '(VL.VoucherDate>=' + @FinYearFromDate + ' )'
			SET @WHERE_VoucherDate3 = '(VL.VoucherDate>=' + @FromDate + ' AND VL.VoucherDate<=' + @ToDate + ' )
										AND 1 = CASE WHEN(' + CAST(@OnlyPending AS nvarchar)+ '=1) THEN
											(CASE WHEN(ISNULL(VL.ReconciliationDate,'''')='''') THEN 1 ELSE 0 END)
										  ELSE
											(CASE WHEN(ISNULL(VL.ReconciliationDate,'''')<>'''') THEN 1 ELSE 0 END)
										  END'
			SET @ROW_NUMBER='ROW_NUMBER() OVER (ORDER BY CAST(VL.VoucherDate AS NUMERIC) ASC,CAST(LEFT([VoucherNo], Patindex(''%[^0-9]%'', [VoucherNo] + ''x'') - 1) AS INT),VL.Id)'
		END
	ELSE
		BEGIN
			SET @WHERE_VoucherDate = '(VL.ReconciliationDate<' + @FromDate + ' AND VL.ReconciliationDate>=ISNULL(' + @FinYearFromDate_Min + ',0) )'
			SET @WHERE_VoucherDate2 = '(VL.ReconciliationDate>=' + @FinYearFromDate + ' )'
			SET @WHERE_VoucherDate3 = '(VL.ReconciliationDate>=' + @FromDate + ' AND VL.ReconciliationDate<=' + @ToDate + ' )
										AND 1 = CASE WHEN(' + CAST(@OnlyPending AS nvarchar) + '=1) THEN
											(CASE WHEN(ISNULL(VL.ReconciliationDate,'''')='''') THEN 1 ELSE 0 END)
										  ELSE
											(CASE WHEN(ISNULL(VL.ReconciliationDate,'''')<>'''') THEN 1 ELSE 0 END)
										  END'
			SET @ROW_NUMBER='ROW_NUMBER() OVER (ORDER BY CAST(VL.ReconciliationDate AS NUMERIC) ASC,CAST(LEFT([VoucherNo], Patindex(''%[^0-9]%'', [VoucherNo] + ''x'') - 1) AS INT),VL.Id)'
		END
--**************EMPTY TMP LEDGER TABLE**************
	DELETE FROM TMPLedger
--**************OPENING START**************
	EXEC('
	INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
	SELECT 0 AS RowNumber,0 AS ID,
		ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
		,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
		,''Opening'' AS AgLedger,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
		(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
			-
		ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
		) AS Balance
	FROM (
		--Opening
		SELECT SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		FROM LedgerMaster LM
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		WHERE LM.Id =' + @F_LedgerMaster + ' AND LM.F_CompanyMaster=' + @F_CompanyMaster + ' AND LGM.IsBalanceSheet = 1 AND ISNULL(' + @FinYearFromDate_Min + ',0)>0
		GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			
		UNION ALL
		
		--Transaction
		SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		FROM VoucherL VL
			LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id		
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
			LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
		WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = ' + @F_LedgerMaster + '
			AND 
			(
				' + @WHERE_VoucherDate + '
			)
			AND VL.F_CompanyMaster=' + @F_CompanyMaster + '
			AND 1 = (CASE WHEN(' + @F_OtherLedgerMaster + '=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = ' + @F_OtherLedgerMaster + ') THEN 1 ELSE 0 END) END)
			AND 1 = (CASE WHEN(' + @F_VoucherTypeMaster + '=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = ' + @F_VoucherTypeMaster + ') THEN 1 ELSE 0 END) END)
			AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 
					AND 
					(
						' + @WHERE_VoucherDate2 + '
					)
			))
		GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
	) AS TAB4
	GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
	ORDER BY LedgerGroup,Ledger')
--**************OPENING END**************
--**************TRANSACTION START************
	EXEC('
	INSERT INTO TMPLedger(ReconciliationDate,ReconciliationRemark,IsOpening,RowNumber,Id,VoucherDate,VoucherNo,VoucherType,AgLedger,AgLedgerId,DrAmt,CrAmt,VoucherDate1,F_VoucherTypeMaster,Narration,Ledger,LedgerId,LedgerGroup,LedgerGroupId,Balance,AgLedgerDetail,Discount,ChequeCancelCharges)
	SELECT  VL.ReconciliationDate,VL.ReconciliationRemark,1,' + @ROW_NUMBER + ' AS RowNumber
		,VL.F_VoucherH AS Id
		,(SUBSTRING(VL.VoucherDate,7,2)+''/''+SUBSTRING(VL.VoucherDate,5,2)+''/''+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate
		,VL.VoucherNo
		,VTM.Name AS VoucherType
		,(SELECT TOP 1 (LM1.Name)
			FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
			WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
				AND VL1.F_LedgerMaster != ' + @F_LedgerMaster + '
			ORDER BY VL1.Amount DESC
			) AS AgLedger
		,(SELECT TOP 1 (LM1.Id)
			FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
			WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
				AND VL1.F_LedgerMaster != ' + @F_LedgerMaster + '
			ORDER BY VL1.Amount DESC
			) AS AgLedgerId		
		,(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) AS DrAmt
		,(CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END) AS CrAmt
		,VL.VoucherDate AS VoucherDate1
		,VL.F_VoucherTypeMaster,VL.Narration
		,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		,((CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END)
			-
		  (CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END)
		 )
		 ,AgLedgerDetail = STUFF(			
			(SELECT (LM.Name + ''   '' + CAST(VL1.Amount AS NVARCHAR) + '' '' + CASE WHEN(VL1.DrCr=0) THEN ''Dr'' ELSE ''Cr'' END + '''') + CHAR(10)
				FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM ON LM.Id = VL1.F_LedgerMaster
				WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherH = VL.F_VoucherH AND VL1.Id<>VL.Id AND VL1.F_VoucherTypeMaster = VL.F_VoucherTypeMaster
			FOR XML PATH('''')), 1, 0, '''')
		,(SELECT MRB.Discount FROM MoneyReceiptBankH MRB WHERE MRB.Id = VL.F_VoucherH) AS Discount
		,(SELECT MRB.ChequeCancelCharges FROM MoneyReceiptBankH MRB WHERE MRB.Id = VL.F_VoucherH) AS ChequeCancelCharges
	FROM VoucherL VL
		LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = ' + @F_LedgerMaster + '
		AND 
		(
			' + @WHERE_VoucherDate3 + '			
		)
		AND VL.F_CompanyMaster=' + @F_CompanyMaster + '
		AND 1 = (CASE WHEN(' + @F_OtherLedgerMaster + '=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = ' + @F_OtherLedgerMaster + ') THEN 1 ELSE 0 END) END)
		AND 1 = (CASE WHEN(' + @F_VoucherTypeMaster + '=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = ' + @F_VoucherTypeMaster + ') THEN 1 ELSE 0 END) END)
	ORDER BY LedgerGroup,Ledger,VoucherDate1,RowNumber,VoucherType')
--**************TRANSACTION END**************

--**************FINAL SELECTION START**************
	SELECT [Id]
			,(SUBSTRING(LT.ReconciliationDate,7,2)+'/'+SUBSTRING(LT.ReconciliationDate,5,2)+'/'+SUBSTRING(LT.ReconciliationDate,1,4)) AS ReconciliationDate
			,[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[Narration],[DrAmt],[CrAmt]		
			,ABS(LTView.Balance) AS [Balance]
			,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
			,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening],[AgLedgerDetail],[ReconciliationRemark]
			,[Discount],[ChequeCancelCharges]
	FROM [TMPLedger] LT
		CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    					FROM [TMPLedger]
    					WHERE ROWNUMBER <= LT.ROWNUMBER
    				) AS LTView
    WHERE 
	1 = CASE WHEN(@OnlyPending=1) THEN
				(CASE WHEN(ISNULL(LT.ReconciliationDate,'')='') THEN 1 ELSE 0 END)
			  ELSE
				(CASE WHEN(ISNULL(LT.ReconciliationDate,'')<>'' OR IsOpening=0) THEN 1 ELSE 0 END)
			  END AND 
	1 = (CASE WHEN(@F_LedgerMaster_Party=0) THEN 1 ELSE (CASE WHEN(@F_LedgerMaster_Party=LT.AgLedgerId) THEN 1 ELSE 0 END) END)
	AND 1 = (CASE WHEN(@Amount=0) THEN 1 ELSE (CASE WHEN(@Amount=LT.DrAmt OR @Amount=LT.CrAmt) THEN 1 ELSE 0 END) END)
	ORDER BY [LedgerGroup],[Ledger],[IsOpening],[VoucherDate1],RowNumber
--**************FINAL SELECTION END**************
END







































GO
```

# [dbo].[Select_Acc_BookPrinting]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_BookPrinting]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO












CREATE PROCEDURE [dbo].[Select_Acc_BookPrinting]
	@F_LedgerGroupMaster numeric(18,0),
	@F_LedgerMaster numeric(18,0),
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),	
	@F_CompanyMaster numeric(18,0),
	@F_OtherLedgerMaster NUMERIC(18,0)=0,
	@F_VoucherTypeMaster NUMERIC(18,0)=0,
	@IsOpBalAbove smallint=1,
	@LedgerIds nvarchar(max)='',
	@View smallint=0,
	@DetailedView smallint=0
AS
BEGIN
-- OPENING BALANCE
	DECLARE @FinYearFromDate NVARCHAR(MAX)
	DECLARE @FinYearFromDate_Min NVARCHAR(MAX)
	SET @FinYearFromDate = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC)	
	SET @FinYearFromDate_Min = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)

	DECLARE @IsLedgerBalance_Split SMALLINT
	DECLARE @FinYearFromDate_Split NVARCHAR(MAX)
	DECLARE @F_CompanyYears_Split NUMERIC(18,0)
	--DECLARE @F_CompanyYears NUMERIC(18,0)
	SET @IsLedgerBalance_Split = (SELECT TOP 1 ISNULL(IsLedgerBalanceSplit,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @F_CompanyYears_Split = (SELECT TOP 1 ISNULL(Id,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @FinYearFromDate_Split = (SELECT TOP 1 ISNULL(FinYearFrom,0) FROM CompanyYears WHERE Id=@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
	--SET @F_CompanyYears = (SELECT TOP 1 Id FROM CompanyYears WHERE Id>@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster)
--**************EMPTY TMP LEDGER TABLE**************
	DELETE FROM TMPLedger	

--**************OPENING START**************
	IF(@View=0)
	BEGIN
		IF(@IsLedgerBalance_Split=1)
			BEGIN
				INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,AgLedgerDetail,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
				SELECT 0 AS RowNumber,0 AS ID,
					ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
					,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
					,'Opening' AS AgLedger,'Opening' AS AgLedgerDetail,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
					(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
					-
					ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
					) AS Balance
				FROM(
					SELECT (CASE WHEN (LML.DrCr=0) THEN LML.OpeningBalance ELSE 0 END) AS DebitAmount,
						(CASE WHEN (LML.DrCr=1) THEN LML.OpeningBalance ELSE 0 END) AS CreditAmount,
						LM.Name AS Ledger, LM.Id AS LedgerId,
						LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
					FROM LedgerMaster LM
					LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears_Split 
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
					WHERE
					1=(CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(LM.Id =@F_LedgerMaster) THEN 1 ELSE 0 END) END)
					AND 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LGM.Id=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)				
					AND LM.F_CompanyMaster=@F_CompanyMaster AND LGM.IsBalanceSheet = 1
					AND 1=(CASE WHEN (@LedgerIds='') THEN 1 ELSE (CASE WHEN (LM.Id  IN(SELECT * FROM Split(@LedgerIds,',')) ) THEN 1 ELSE 0 END) END)		
		
					UNION ALL
		
					--Transaction
					SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
					FROM VoucherL VL
						LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id		
						LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
						LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
					WHERE ISNULL(VL.Cancelled,0)=0 
						AND 1=(CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_LedgerMaster = @F_LedgerMaster) THEN 1 ELSE 0 END) END)
						AND 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LGM.Id=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)				
						AND (VL.VoucherDate<@FromDate AND VL.VoucherDate>=ISNULL(@FinYearFromDate_Split,0)) AND VL.F_CompanyMaster=@F_CompanyMaster
						AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
						AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
						AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=@FinYearFromDate))
						AND
						1=(CASE WHEN (@LedgerIds='') THEN 1 ELSE (CASE WHEN (LM.Id  IN(SELECT * FROM Split(@LedgerIds,',')) ) THEN 1 ELSE 0 END) END)		
					GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
				) AS TAB4
				GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
				ORDER BY LedgerGroup,Ledger
			END
		ELSE
			BEGIN
				INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,AgLedgerDetail,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
				SELECT 0 AS RowNumber,0 AS ID,
					ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
					,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
					,'Opening' AS AgLedger,'Opening' AS AgLedgerDetail,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
					(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
						-
					ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
					) AS Balance
				FROM(
					--Opening
					SELECT SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
					FROM LedgerMaster LM
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
					WHERE 
					1=(CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(LM.Id =@F_LedgerMaster) THEN 1 ELSE 0 END) END)
					AND 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LGM.Id=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)				
					AND LM.F_CompanyMaster=@F_CompanyMaster AND LGM.IsBalanceSheet = 1 AND ISNULL(@FinYearFromDate_Min,0)>0
					GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			
					UNION ALL
		
					--Transaction
					SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
					FROM VoucherL VL
						LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id		
						LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
						LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
					WHERE ISNULL(VL.Cancelled,0)=0
						AND 1=(CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_LedgerMaster = @F_LedgerMaster) THEN 1 ELSE 0 END) END)
						AND 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LGM.Id=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)				
						AND (VL.VoucherDate<@FromDate AND VL.VoucherDate>=ISNULL(@FinYearFromDate_Min,0)) AND VL.F_CompanyMaster=@F_CompanyMaster
						AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
						AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
						AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=@FinYearFromDate))
						AND
						1=(CASE WHEN (@LedgerIds='') THEN 1 ELSE (CASE WHEN (LM.Id  IN(SELECT * FROM Split(@LedgerIds,',')) ) THEN 1 ELSE 0 END) END)		
					GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
				) AS TAB4
				GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
				ORDER BY LedgerGroup,Ledger
		END
	END
--**************OPENING END**************

--**************TRANSACTION START************ (voucherno order by please write on rownumber)
	INSERT INTO TMPLedger(IsOpening,RowNumber,Id,VoucherDate,VoucherNo,VoucherType,AgLedger,DrAmt,CrAmt,VoucherDate1,F_VoucherTypeMaster,Narration,Ledger,LedgerId,LedgerGroup,LedgerGroupId,Balance,AgLedgerDetail)
	SELECT  1,ROW_NUMBER() OVER (order by CAST(VL.VoucherDate AS NUMERIC) ASC,CAST(LEFT([VoucherNo], Patindex('%[^0-9]%', [VoucherNo] + 'x') - 1) AS BIGINT),VL.Id) as RowNumber
		,VL.F_VoucherH AS Id
		,(SUBSTRING(VL.VoucherDate,7,2)+'/'+SUBSTRING(VL.VoucherDate,5,2)+'/'+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate
		,VL.VoucherNo
		,VTM.Name AS VoucherType
		,(SELECT TOP 1 (LM1.Name)
			FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
			WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
				AND VL1.F_LedgerMaster != @F_LedgerMaster
			ORDER BY VL1.Amount DESC
			) AS AgLedger
		,(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) AS DrAmt
		,(CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END) AS CrAmt
		,VL.VoucherDate AS VoucherDate1
		,VL.F_VoucherTypeMaster,VL.Narration
		,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		,((CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END)
			-
		  (CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END)
		 )
		 ,
		 AgLedgerDetail = (CASE WHEN(@DetailedView=0) THEN '' ELSE 
		 STUFF(			
			(SELECT (LM.Name + '   ' + CAST(VL1.Amount AS NVARCHAR) + ' ' + CASE WHEN(VL1.DrCr=0) THEN 'Dr' ELSE 'Cr' END + '') + CHAR(10)
				FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM ON LM.Id = VL1.F_LedgerMaster
				WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherH = VL.F_VoucherH AND VL1.Id<>VL.Id AND VL1.F_VoucherTypeMaster = VL.F_VoucherTypeMaster
			FOR XML PATH('')), 1, 0, '')
		END)
	FROM VoucherL VL
		LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	WHERE ISNULL(VL.Cancelled,0)=0
		AND 1=(CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_LedgerMaster = @F_LedgerMaster) THEN 1 ELSE 0 END) END)
		AND 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LGM.Id=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)				
		AND (VL.VoucherDate>=@FromDate AND VL.VoucherDate<=@ToDate)
		AND VL.F_CompanyMaster=@F_CompanyMaster
		AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
		AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@LedgerIds='') THEN 1 ELSE (CASE WHEN (LM.Id  IN(SELECT * FROM Split(@LedgerIds,',')) ) THEN 1 ELSE 0 END) END)		
	ORDER BY LedgerGroup,Ledger,VoucherDate1,RowNumber,VoucherType
--**************TRANSACTION END**************

--**************FINAL SELECTION START************** (please donot change this query, this will may affect balance amt)
	DECLARE @SQL NVARCHAR(MAX)
	DECLARE @ORDER_BY NVARCHAR(MAX)
	--IF(@IsOpBalAbove=1)
	--	BEGIN
			IF(@View=0)
				BEGIN
					SET @ORDER_BY = '[LedgerGroup],[Ledger],[IsOpening],[VoucherDate1],ROWNUMBER'
				END
			ELSE
				BEGIN
					SET @ORDER_BY = '[VoucherDate1],ROWNUMBER'
				END
			SET @SQL='SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[AgLedgerDetail],[Narration],[DrAmt],[CrAmt]		
							,ABS(LTView.Balance) AS [Balance]
							,(CASE WHEN(LTView.Balance>=0) THEN ''Dr'' ELSE ''Cr'' END) AS DrCr
							,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening]
						FROM [TMPLedger] LT
							CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    								FROM [TMPLedger]
    								WHERE ROWNUMBER <= LT.ROWNUMBER AND [LedgerId]=LT.[LedgerId]
    							) AS LTView
						ORDER BY ' + @ORDER_BY
	--	END
	--ELSE
	--	BEGIN
	--		SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[AgLedgerDetail],[Narration],[DrAmt],[CrAmt]		
	--				,ABS(LTView.Balance) AS [Balance]
	--				,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
	--				,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening]
	--		FROM [TMPLedger] LT
	--		 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
 --   						FROM [TMPLedger]
 --   						WHERE ROWNUMBER <= LT.ROWNUMBER
 --   					) AS LTView
	--		ORDER BY [LedgerGroup],[Ledger],[IsOpening] DESC,[VoucherDate1],ROWNUMBER
	--	END
	EXEC(@SQL)
--**************FINAL SELECTION END**************
END

















































GO
```

# [dbo].[Select_Acc_DayBook]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_DayBook]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO









CREATE PROCEDURE [dbo].[Select_Acc_DayBook] 
	@F_VoucherTypeMaster numeric(18,0)=0,
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),	
	@F_CompanyMaster numeric(18,0),
	@F_LedgerMaster numeric(18,0),
	@Cancelled smallint,
	@SearchType smallint=0,
	@SearchParameters nvarchar(max)='',
	@IsMatchExact smallint=0
AS
BEGIN
	DELETE FROM TMPLedger
	INSERT INTO TMPLedger(RowNumber,Id,VoucherNo,VoucherDate,Ledger,Narration,VoucherType,RefNo,RefDate,DrAmt,CrAmt,F_VoucherTypeMaster
	--,AgLedger,VoucherDate1,LedgerId,LedgerGroup,LedgerGroupId,AgLedgerDetail
	)
	SELECT ROW_NUMBER() OVER (order by CAST(VH.VoucherDate AS NUMERIC) ASC,CAST(LEFT(VH.[VoucherNo], Patindex('%[^0-9]%', VH.[VoucherNo] + 'x') - 1) AS BIGINT)) AS RowNumber
		,(VH.F_VoucherH) AS Id
		,VH.VoucherNo
		,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate
		,LEDGER_DETAILS.Ledger
		,VH.Narration
		,VTM.Name AS VoucherType
		,VHM.PurVoucherNo AS RefNo
		,(SUBSTRING(VHM.PurVoucherDate,7,2)+'/'+SUBSTRING(VHM.PurVoucherDate,5,2)+'/'+SUBSTRING(VHM.PurVoucherDate,1,4)) AS RefDate
		,LEDGER_DETAILS.DebitAmount,LEDGER_DETAILS.CreditAmount,VH.F_VoucherTypeMaster
	FROM VoucherL VH
	LEFT JOIN VoucherH VHM ON VHM.Id=VH.F_VoucherH
	LEFT JOIN LedgerMaster LM ON LM.Id=VH.F_LedgerMaster
	LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VH.F_VoucherTypeMaster
	OUTER APPLY	(SELECT TOP 1 (LM1.Name) AS Ledger
					,ISNULL((CASE WHEN((VL1.DrCr)=0) THEN (VL1.Amount) END),0) AS DebitAmount
					,ISNULL((CASE WHEN((VL1.DrCr)=1) THEN (VL1.Amount) END),0) AS CreditAmount
				FROM VoucherL VL1
					LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
				WHERE 1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VL1.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END) AND VL1.F_VoucherTypeMaster=VH.F_VoucherTypeMaster AND VL1.F_VoucherH=VH.F_VoucherH
				ORDER BY VL1.Id
		) LEDGER_DETAILS
	WHERE 1=CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VH.F_VoucherTypeMaster=@F_VoucherTypeMaster) THEN 1 ELSE 0 END) END
		AND
		1=CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(VH.F_LedgerMaster=@F_LedgerMaster) THEN 1 ELSE 0 END) END
		AND VH.F_CompanyMaster = @F_CompanyMaster
		AND VH.VoucherDate>=@FromDate AND VH.VoucherDate <=@ToDate
		AND
		1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VHM.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN(@SearchParameters='') THEN 1 ELSE
				(CASE WHEN(@IsMatchExact=0) THEN
					(
						(CASE WHEN(@SearchType=0
								AND (CHARINDEX(@SearchParameters,VH.VoucherNo)>0
								OR CHARINDEX(@SearchParameters,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)))>0
								OR CHARINDEX(@SearchParameters,VHM.PurVoucherNo)>0
								OR CHARINDEX(@SearchParameters,(SUBSTRING(VHM.PurVoucherDate,7,2)+'/'+SUBSTRING(VHM.PurVoucherDate,5,2)+'/'+SUBSTRING(VHM.PurVoucherDate,1,4)))>0
								OR CHARINDEX(@SearchParameters,VH.Amount)>0
								OR CHARINDEX(@SearchParameters,VH.Narration)>0)
							)
						THEN 1 ELSE
							CASE WHEN(@SearchType=1 AND CHARINDEX(@SearchParameters,VH.VoucherNo)>0) THEN 1
								WHEN(@SearchType=2 AND CHARINDEX(@SearchParameters,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)))>0) THEN 1
								WHEN(@SearchType=3 AND CHARINDEX(@SearchParameters,VHM.PurVoucherNo)>0) THEN 1
								WHEN(@SearchType=4 AND CHARINDEX(@SearchParameters,(SUBSTRING(VHM.PurVoucherDate,7,2)+'/'+SUBSTRING(VHM.PurVoucherDate,5,2)+'/'+SUBSTRING(VHM.PurVoucherDate,1,4)))>0) THEN 1
								WHEN(@SearchType=5 AND CHARINDEX(@SearchParameters,VH.Amount)>0) THEN 1
								WHEN(@SearchType=6 AND CHARINDEX(@SearchParameters,VH.Narration)>0) THEN 1
								ELSE 0
							END
						END)
					)
				ELSE
					(
						(CASE WHEN(@SearchType=0
								AND (@SearchParameters=VH.VoucherNo
								OR @SearchParameters=(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4))
								OR @SearchParameters=VHM.PurVoucherNo
								OR @SearchParameters=(SUBSTRING(VHM.PurVoucherDate,7,2)+'/'+SUBSTRING(VHM.PurVoucherDate,5,2)+'/'+SUBSTRING(VHM.PurVoucherDate,1,4))
								OR @SearchParameters=VH.Amount
								OR @SearchParameters=VH.Narration)
							)
						THEN 1 ELSE
							CASE WHEN(@SearchType=1 AND (@SearchParameters=VH.VoucherNo)) THEN 1
								WHEN(@SearchType=2 AND (@SearchParameters=(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)))) THEN 1
								WHEN(@SearchType=3 AND (@SearchParameters=VHM.PurVoucherNo)) THEN 1
								WHEN(@SearchType=4 AND (@SearchParameters=(SUBSTRING(VHM.PurVoucherDate,7,2)+'/'+SUBSTRING(VHM.PurVoucherDate,5,2)+'/'+SUBSTRING(VHM.PurVoucherDate,1,4)))) THEN 1
								WHEN(@SearchType=5 AND (@SearchParameters=VH.Amount)) THEN 1
								WHEN(@SearchType=6 AND (@SearchParameters=VH.Narration)) THEN 1
								ELSE 0
							END
						END)
					)
				END)
			END)
	GROUP BY VH.F_VoucherH,VH.VoucherDate,VTM.Name,VH.VoucherNo,VHM.PurVoucherNo,VHM.PurVoucherDate,VH.F_VoucherTypeMaster,VH.Narration,VHM.PurVoucherDate,LEDGER_DETAILS.Ledger,LEDGER_DETAILS.DebitAmount,LEDGER_DETAILS.CreditAmount
	ORDER BY CASE WHEN(VH.F_VoucherTypeMaster=7 AND @F_VoucherTypeMaster>0) THEN VHM.PurVoucherDate ELSE VH.VoucherDate END
	
	
--**************FINAL SELECTION START************** (please donot change this query, this will may affect balance amt)
	--RowNumber,Id,VoucherDate,Ledger,Narration,VoucherType,VoucherNo,DrAmt,CrAmt,F_VoucherTypeMaster,Balance
	SELECT [Id],[VoucherNo],[VoucherDate],[Ledger],[Narration],[VoucherType],[RefNo],[RefDate],[DrAmt] AS DebitAmount,[CrAmt] AS CreditAmount,[F_VoucherTypeMaster],ABS(LTView.Balance) AS [Balance],(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
	--,[LedgerId],[LedgerGroup],[LedgerGroupId],[AgLedger],[VoucherDate1],[IsOpening],[AgLedgerDetail]
	FROM [TMPLedger] LT
		CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    				FROM [TMPLedger]
    				WHERE ROWNUMBER <= LT.ROWNUMBER
    			) AS LTView
	ORDER BY [VoucherDate1],ROWNUMBER
--**************FINAL SELECTION END**************
END


GO
```

# [dbo].[Select_Acc_Depreciation]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Depreciation]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO













CREATE PROCEDURE [dbo].[Select_Acc_Depreciation]
	@FromDate NVARCHAR(50),
	@ToDate NVARCHAR(50),
	@F_CompanyMaster NVARCHAR(50),
	@F_CompanyYears NUMERIC(18,0),
	@Type smallint=0
AS
BEGIN
	DECLARE @DefaultDepreciationPostingLedger NUMERIC(18,0)
	DECLARE @HalfDate NVARCHAR(50)
	SET @HalfDate = SUBSTRING(@FromDate,1,4) + '0930'
	SET @DefaultDepreciationPostingLedger = (SELECT DefaultDepreciationPostingLedger FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
	--Opening
	SELECT TAB5.Ledger, TAB5.LedgerId,MAX(TAB5.DepRate) AS DepRate
		,SUM(TAB5.Opening) AS Opening
		,SUM(TAB5.Addition_More_Than_180) AS Addition_More_Than_180
		,SUM(TAB5.Addition_Less_Than_180) AS Addition_Less_Than_180
		,(SUM(TAB5.SalesDuringYear)-ISNULL(MAX(DEP.Amount),0)) AS SalesDuringYear
		,(SUM(TAB5.Opening) + SUM(TAB5.Addition_More_Than_180) + SUM(TAB5.Addition_Less_Than_180) - (SUM(TAB5.SalesDuringYear)-ISNULL(MAX(DEP.Amount),0))) AS TotalBefore
		,CAST(ROUND(
				--More Than 180 Days
				(SUM(TAB5.Addition_Less_Than_180) * ((MAX(TAB5.DepRate)/2)/100)) +
				--Less Than 180 Days
				((SUM(TAB5.Addition_More_Than_180) + SUM(TAB5.Opening)) * (MAX(TAB5.DepRate)/100)),0) AS NUMERIC(18,2))
		--SalesDuringYear_VoucherDate
			AS DepreciationAmt
		,(SUM(TAB5.Opening) + SUM(TAB5.Addition_More_Than_180) + SUM(TAB5.Addition_Less_Than_180) - (SUM(TAB5.SalesDuringYear)-ISNULL(MAX(DEP.Amount),0)) -
			ROUND(
				--More Than 180 Days
				(SUM(TAB5.Addition_Less_Than_180) * ((MAX(TAB5.DepRate)/2)/100)) +
				--Less Than 180 Days
				((SUM(TAB5.Addition_More_Than_180) + SUM(TAB5.Opening)) * (MAX(TAB5.DepRate)/100)),0)
		) AS Closing
		,(CAST(ROUND(
				--More Than 180 Days
				(SUM(TAB5.Addition_Less_Than_180) * ((MAX(TAB5.DepRate)/2)/100)) +
				--Less Than 180 Days
				((SUM(TAB5.Addition_More_Than_180) + SUM(TAB5.Opening)) * (MAX(TAB5.DepRate)/100)),0) AS NUMERIC(18,2))- ISNULL(MAX(DEP.Amount),0)) AS Diff
		,ISNULL(MAX(DEP.F_VoucherH),0) AS F_VoucherH
		,ISNULL(MAX(DEP.VoucherNo),'') AS VoucherNo
		,ISNULL(MAX((SUBSTRING(DEP.VoucherDate,7,2)+'/'+SUBSTRING(DEP.VoucherDate,5,2)+'/'+SUBSTRING(DEP.VoucherDate,1,4))),'') AS VoucherDate
		,ISNULL(MAX(DEP.Amount),0) AS PostingAmt
	FROM (
		SELECT TAB4.Ledger, TAB4.LedgerId
			,ISNULL((SELECT LML.DepRate FROM LedgerMasterL LML WHERE LML.F_LedgerMaster = TAB4.LedgerId AND LML.F_CompanyYears = @F_CompanyYears),0) AS DepRate
			,(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
				-
				ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
				) AS Opening
			,0 AS Addition_More_Than_180
			,0 AS Addition_Less_Than_180
			,0 AS SalesDuringYear,'' AS SalesDuringYear_VoucherDate
			,0 AS TotalBefore
			,0 AS DepreciationAmt
			,0 AS Closing
		FROM(
			SELECT 0 AS DepRate,SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
			FROM LedgerMaster LM
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
			WHERE LM.F_CompanyMaster=@F_CompanyMaster AND LM.F_LedgerGroupMaster = 15		
			GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
				
			UNION ALL
			
			--Transaction
			SELECT 0 AS DepRate,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
			FROM VoucherL VL
				LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
				LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
			WHERE ISNULL(VL.Cancelled,0)=0 AND (VL.VoucherDate<@FromDate) AND VL.F_CompanyMaster=@F_CompanyMaster AND LM.F_LedgerGroupMaster = 15			
			GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
		) AS TAB4
		GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,TAB4.DepRate
		
		UNION ALL
		
	--Transaction
		SELECT LM.Name AS Ledger, LM.Id AS LedgerId
			,ISNULL((SELECT LML.DepRate FROM LedgerMasterL LML WHERE LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears),0) AS DepRate
			,0 AS Opening
			,ISNULL((SUM(CASE WHEN (VL.DrCr=0 AND VL.VoucherDate<=@HalfDate) THEN VL.Amount ELSE 0 END)),0) AS Addition_More_Than_180
			,ISNULL((SUM(CASE WHEN (VL.DrCr=0 AND VL.VoucherDate>@HalfDate) THEN VL.Amount ELSE 0 END)),0) AS Addition_Less_Than_180
			,ISNULL((SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)),0) AS SalesDuringYear
			,ISNULL((MAX(CASE WHEN (VL.DrCr=1) THEN VL.VoucherDate ELSE 0 END)),0) AS SalesDuringYear_VoucherDate
			,0 AS TotalBefore
			,0 AS DepreciationAmt
			,0 AS Closing
		FROM VoucherL VL
			LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
		WHERE ISNULL(VL.Cancelled,0)=0 AND LM.F_LedgerGroupMaster = 15
			AND (VL.VoucherDate >= @FromDate AND VL.VoucherDate <= @ToDate ) AND VL.F_CompanyMaster= @F_CompanyMaster		
		GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
	) AS TAB5
	OUTER APPLY(SELECT TOP 1 
				VL.VoucherNo,VL.F_VoucherH,VL.VoucherDate,VL.Amount
			FROM VoucherL VL
				LEFT JOIN VoucherH VH ON VH.Id = VL.F_VoucherH
			WHERE VL.VoucherDate>=@FromDate AND VL.VoucherDate<=@ToDate AND VL.F_LedgerMaster = TAB5.LedgerId
				 AND VL.F_CompanyMaster= @F_CompanyMaster AND VL.F_VoucherTypeMaster = 4 AND VH.F_LedgerMaster = @DefaultDepreciationPostingLedger
	) DEP
	GROUP BY TAB5.Ledger, TAB5.LedgerId
	ORDER BY TAB5.Ledger, TAB5.LedgerId
END






















GO
```

# [dbo].[Select_Acc_Expenditure]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Expenditure]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





























CREATE PROCEDURE [dbo].[Select_Acc_Expenditure]
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0),
	@F_LedgerGroupMaster numeric(18,0),
	@F_LedgerMaster numeric(18,0)=0,
	@Cancelled smallint=-1,
	@F_VoucherTypeMaster NUMERIC(18,0)=0,
	@EditPassword nvarchar(50)='',
	@F_LedgerMasterAg NUMERIC(18,0)=0
AS
BEGIN
	IF(@F_LedgerGroupMaster = 30)
		BEGIN
			-- OPENING BALANCE
			DECLARE @IsLedgerBalance_Split SMALLINT
			DECLARE @FinYearFromDate_Split NVARCHAR(MAX)
			DECLARE @F_CompanyYears_Split NUMERIC(18,0)
			DECLARE @F_CompanyYears2 NUMERIC(18,0)
			SET @IsLedgerBalance_Split = (SELECT TOP 1 ISNULL(IsLedgerBalanceSplit,0) FROM CompanyYears WHERE FinYearFrom<@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
			--SET @F_CompanyYears_Split = (SELECT TOP 1 ISNULL(Id,0) FROM CompanyYears WHERE FinYearFrom<@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
			--SET @FinYearFromDate_Split = (SELECT TOP 1 ISNULL(FinYearFrom,0) FROM CompanyYears WHERE Id>@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
			SET @F_CompanyYears2 = (SELECT TOP 1 Id FROM CompanyYears WHERE Id>@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster)
			--

			DECLARE @YearId NUMERIC(18,0)
			DECLARE @OpeningDate NVARCHAR(50)
			DECLARE @OpeningAmt NUMERIC(18,2)			
			SET @YearId = (SELECT FinYearFrom FROM CompanyYears WHERE FinYearFrom<@FromDate AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1)
			IF(@YearId>0)
				BEGIN
					SET @OpeningDate = ''
					SET @OpeningAmt = 0
				END
			ELSE
				BEGIN
					SET @OpeningDate = (SUBSTRING(@FromDate,7,2)+'/'+SUBSTRING(@FromDate,5,2)+'/'+SUBSTRING(@FromDate,1,4))
					SET @OpeningAmt = ( SELECT LM.OpeningBalance
										FROM LedgerMaster LM
											LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @YearId
										WHERE LM.Id = @F_LedgerMaster)
				END
			
			DECLARE @TransId NUMERIC(18,0)
			SET @TransId = (SELECT TOP 1 Id FROM VoucherH VH WHERE VH.F_CompanyMaster = @F_CompanyMaster AND VH.F_LedgerMaster = @F_LedgerMaster
								AND VH.VoucherDate>=@FromDate AND VH.VoucherDate <=@ToDate
								AND 1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END))
			IF(@TransId>0)
				BEGIN	
					SELECT VH.Id
						,VH.VoucherNo
						,(CASE WHEN(ISNULL(STOCKOP.OpeningDate,'')='') THEN @OpeningDate ELSE STOCKOP.OpeningDate END) AS OpeningDate				
						,(CASE WHEN(ISNULL(STOCKOP.OpeningAmt,0)=0) THEN @OpeningAmt ELSE STOCKOP.OpeningAmt END) AS OpeningAmt
						,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS ClosingDate
						,VH.TotalNetAmt AS ClosingAmt
						,(CASE WHEN (ISNULL(VH.Cancelled,0)=1) THEN 'Cancelled' ELSE '' END) AS Cancelled
					FROM VoucherH VH
					LEFT JOIN VoucherL VL ON VH.Id = VL.F_VoucherH
					LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
					LEFT JOIN ManagerMaster MM ON MM.Id=VH.F_ManagerMaster
					OUTER APPLY(
							SELECT TOP 1 VH1.TotalNetAmt AS OpeningAmt,
								('01/' + SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH1.VoucherDate,1,4)) AS OpeningDate
							FROM VoucherH VH1
							WHERE VH1.VoucherDate<@FromDate
							AND VH1.F_LedgerMaster=VH.F_LedgerMaster
							AND VH1.F_CompanyMaster = @F_CompanyMaster
							AND
							1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
							
							ORDER BY VH1.VoucherDate DESC
					) STOCKOP
					WHERE 1=CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LM.F_LedgerGroupMaster=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END
						AND VH.F_LedgerMaster = @F_LedgerMaster
						AND VH.F_CompanyMaster = @F_CompanyMaster
						AND VH.VoucherDate>=@FromDate AND VH.VoucherDate <=@ToDate
						AND 1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
					ORDER BY VH.VoucherDate,CAST(VH.VoucherNo AS NUMERIC)
				END
			ELSE
				BEGIN	
					SELECT 0 AS Id
						,'' AS VoucherNo
						,@OpeningDate AS OpeningDate
						,@OpeningAmt AS OpeningAmt
						,'' AS ClosingDate
						,0 AS ClosingAmt
						,'' AS Cancelled
					FROM LedgerMaster LM
					WHERE 1=CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LM.F_LedgerGroupMaster=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END
						AND LM.Id = @F_LedgerMaster
						AND LM.F_CompanyMaster = @F_CompanyMaster
				END
		END
	ELSE
		BEGIN
			SELECT VH.Id
				,VH.VoucherNo
				,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate
				,(SUBSTRING(VH.PostingDate,7,2)+'/'+SUBSTRING(VH.PostingDate,5,2)+'/'+SUBSTRING(VH.PostingDate,1,4)) AS PostingDate
				,VH.PurVoucherNo
				,(SUBSTRING(VH.PurVoucherDate,7,2)+'/'+SUBSTRING(VH.PurVoucherDate,5,2)+'/'+SUBSTRING(VH.PurVoucherDate,1,4)) AS PurVoucherDate
				,LM.Name AS Ledger
				,VL.F_LedgerMaster
				,VH.TotalNetAmt
				,AgLedger = REPLACE(STUFF((SELECT CHAR(10) + (LM1.Name) 
								FROM VoucherL VL1
									LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
								WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
									AND VH.F_LedgerMaster != VL1.F_LedgerMaster 
								ORDER BY VL1.Amount DESC
								FOR XML PATH('')), 1, 1, ''),'&amp;','&')
				,AgLedgerAmt = STUFF((SELECT CHAR(10) + CAST(VL1.Amount AS NVARCHAR) 
								FROM VoucherL VL1
									LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
								WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
									AND VL1.F_LedgerMaster != @F_LedgerMaster
								ORDER BY VL1.Amount DESC
								FOR XML PATH('')), 1, 1, '')
				,VH.ChequeInFavour
				,VH.Narration
				,(CASE WHEN (ISNULL(VH.Cancelled,0)=1) THEN 'Cancelled' ELSE '' END) AS Cancelled
				,MM.Name AS Manager
				,LGM.Name AS LedgerGroup
				,LM.F_LedgerGroupMaster
				,((SUBSTRING(VH.VoucherDate,1,4)) + ' - ' +  SUBSTRING(DateName(mm,DATEADD(mm,CAST((SUBSTRING(VH.VoucherDate,5,2)) AS NUMERIC)-1,0)),0,4)) AS [MonthName]
				,SUBSTRING(VH.VoucherDate,1,6) AS YearMonth
				,ManagerDetails = STUFF((SELECT CHAR(10) + MM1.Name
								FROM VoucherL3 VL1
									LEFT JOIN ManagerMaster MM1 ON VL1.F_ManagerMaster=MM1.Id
								WHERE VL1.F_VoucherH=VL.F_VoucherH
								ORDER BY VL1.Amount DESC
								FOR XML PATH('')), 1, 1, '')
				,ManagerDetailsAmt = STUFF((SELECT CHAR(10) + CAST(VL1.Amount AS NVARCHAR) 
								FROM VoucherL3 VL1
									LEFT JOIN ManagerMaster MM1 ON VL1.F_ManagerMaster=MM1.Id
								WHERE VL1.F_VoucherH=VL.F_VoucherH
								ORDER BY VL1.Amount DESC
								FOR XML PATH('')), 1, 1, '')
				,AgLedgerGroup = REPLACE(STUFF((SELECT CHAR(10) + (LGM1.Name)
								FROM VoucherL VL1
									LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
									LEFT JOIN LedgerGroupMaster LGM1 ON LGM1.ID = LM1.F_LedgerGroupMaster
								WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
									AND VL1.F_LedgerMaster != @F_LedgerMaster
								ORDER BY VL1.Amount DESC
								FOR XML PATH('')), 1, 1, ''),'&amp;','&')
			FROM VoucherL VL
			LEFT JOIN VoucherH VH ON VH.Id = VL.F_VoucherH
			LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
			LEFT JOIN ManagerMaster MM ON MM.Id=VH.F_ManagerMaster
			WHERE 
				1=CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LM.F_LedgerGroupMaster=@F_LedgerGroupMaster OR LGM.GroupType=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END
				AND 1=CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VH.F_VoucherTypeMaster=@F_VoucherTypeMaster AND VL.F_VoucherTypeMaster=@F_VoucherTypeMaster) THEN 1 ELSE 0 END) END				
				AND 1=CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(VH.F_LedgerMaster=@F_LedgerMaster ) THEN 1 ELSE 0 END) END
				AND VH.F_CompanyMaster = @F_CompanyMaster
				AND VH.VoucherDate>=@FromDate AND VH.VoucherDate <=@ToDate
				AND
				1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
				AND ISNULL(LGM.EditPassword,'') = @EditPassword
				AND VH.IsDeposit<>2
			ORDER BY VH.VoucherDate DESC,CAST(VH.VoucherNo AS NUMERIC) DESC
		END
END
























GO
```

# [dbo].[Select_Acc_Ledger]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Ledger]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO























































CREATE PROCEDURE [dbo].[Select_Acc_Ledger] 
	@F_LedgerMaster numeric(18,0),
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0)
AS
BEGIN

	SELECT 0 AS RowNumber,0 AS Id,'' AS VoucherDate,''  AS VoucherNo,'' AS VoucherType,'Opening' AS Ledger
		,(CASE WHEN ((SUM(TAB3.DebitAmount)-SUM(TAB3.CreditAmount))>=0) THEN 1 ELSE 0 END) AS DrCr	
		,ABS(SUM(TAB3.DebitAmount)-SUM(TAB3.CreditAmount)) AS Amount	
		,'0' AS VoucherDate1,0 AS F_VoucherTypeMaster,'' AS Narration
	FROM(	
		SELECT TAB2.LedgerGroup,TAB2.Ledger,SUM(TAB2.DebitAmount) AS DebitAmount,SUM(TAB2.CreditAmount) AS CreditAmount
		
		FROM(		
			--Opening
			SELECT LGM.Name AS LedgerGroup, LM.Name AS Ledger		
				,(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount
				,(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount					
			FROM LedgerMaster LM
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
			WHERE LM.Id =@F_LedgerMaster AND LM.F_CompanyMaster=@F_CompanyMaster 
			UNION ALL
			--Transaction
			SELECT LGM.Name AS LedgerGroup,LM.Name AS Ledger
				,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount
				,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount			
			FROM (
					
					SELECT VL.VoucherNo,(SUBSTRING(VL.VoucherDate,7,2)+'/'+SUBSTRING(VL.VoucherDate,5,2)+'/'+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate
					,LM.Name AS Ledger,VL.F_VoucherH AS Id
					,LM.Id AS LedgerId
					,VL.DrCr,VL.Amount,VTM.Name AS VoucherType
					FROM VoucherL VL
						LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id				
						LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
					WHERE (VL.VoucherDate<@FromDate ) AND  LM.Id =@F_LedgerMaster AND VL.F_CompanyMaster=@F_CompanyMaster
			) AS TAB1
			LEFT JOIN VoucherL VL ON VL.F_VoucherH=TAB1.Id
			LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
			WHERE VL.F_LedgerMaster!=TAB1.LedgerId 
			GROUP BY LGM.Name,LM.Name,LM.Id
		) AS TAB2
		GROUP BY TAB2.LedgerGroup,TAB2.Ledger
	) AS TAB3
		
	UNION ALL
	
	SELECT ROW_NUMBER() OVER (order by TAB1.VoucherDate ASC,VL.Id) as RowNumber,TAB1.Id,TAB1.VoucherDate, TAB1.VoucherNo,VTM.Name AS VoucherType
			,LM.Name AS Ledger,VL.DrCr,VL.Amount,TAB1.VoucherDate1,TAB1.F_VoucherTypeMaster,TAB1.Narration
	FROM (
			SELECT  VL.VoucherNo,(SUBSTRING(VL.VoucherDate,7,2)+'/'+SUBSTRING(VL.VoucherDate,5,2)+'/'+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate
				,LM.Name AS Ledger,VL.F_VoucherH AS Id
				,LM.Id AS LedgerId
				,VL.DrCr,VL.Amount,VTM.Name AS VoucherType,VL.VoucherDate AS VoucherDate1,VL.F_VoucherTypeMaster
				,VL.Narration
			FROM VoucherL VL
				LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id				
				LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
			WHERE VL.F_LedgerMaster = @F_LedgerMaster
				AND (VL.VoucherDate>=@FromDate AND VL.VoucherDate<=@ToDate) AND VL.F_CompanyMaster=@F_CompanyMaster
	) AS TAB1
	LEFT JOIN VoucherL VL ON (VL.F_VoucherH=TAB1.Id AND VL.F_VoucherTypeMaster=TAB1.F_VoucherTypeMaster)
	LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster				
	LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	WHERE VL.F_LedgerMaster!=TAB1.LedgerId
	ORDER BY VoucherDate1,RowNumber,VoucherType
END




































































GO
```

# [dbo].[Select_Acc_Ledger_Detail]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Ledger_Detail]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

















































CREATE PROCEDURE [dbo].[Select_Acc_Ledger_Detail] 
	@F_LedgerMaster numeric(18,0),
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),	
	@F_CompanyMaster numeric(18,0)
AS
BEGIN
--**************EMPTY TMP LEDGER TABLE**************
	DELETE FROM TMPLedger
	--DELETE FROM TMPLedgerView
--**************OPENING START**************
	INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
	SELECT 0 AS RowNumber,0 AS ID,
		ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
		,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
		,'Opening' AS AgLedger,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
		(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
			-
		ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
		) AS Balance
	FROM(
		--Opening
		SELECT SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		FROM LedgerMaster LM
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		WHERE LM.Id =@F_LedgerMaster AND LM.F_CompanyMaster=@F_CompanyMaster
		GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			
		UNION ALL
		
		--Transaction
		SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		FROM VoucherL VL
			LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id		
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
			LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
		WHERE VL.F_LedgerMaster = @F_LedgerMaster AND (VL.VoucherDate<@FromDate) AND VL.F_CompanyMaster=@F_CompanyMaster
		GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
	) AS TAB4
	GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
	ORDER BY LedgerGroup,Ledger
--**************OPENING END**************
--**************TRANSACTION START************
	INSERT INTO TMPLedger(IsOpening,RowNumber,Id,VoucherDate,VoucherNo,VoucherType,AgLedger,DrAmt,CrAmt,VoucherDate1,F_VoucherTypeMaster,Narration,Ledger,LedgerId,LedgerGroup,LedgerGroupId,Balance)
	SELECT 1 AS IsOpening,ROW_NUMBER() OVER (order by TAB1.VoucherDate1 ASC,VL.Id) AS RowNumber,TAB1.Id,TAB1.VoucherDate, TAB1.VoucherNo,VTM.Name AS VoucherType
			,LM.Name AS AgLedger				
			,(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) AS DrAmt
			,(CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END) AS CrAmt
			,TAB1.VoucherDate1,TAB1.F_VoucherTypeMaster,TAB1.Narration
			,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
			,((CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END)
				-
			  (CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END)
			 ) AS Balance
	FROM (
			SELECT  ROW_NUMBER() OVER (order by CAST(VL.VoucherDate AS NUMERIC) ASC,VL.Id) AS RowNumber
				,VL.F_VoucherH AS Id
				,(SUBSTRING(VL.VoucherDate,7,2)+'/'+SUBSTRING(VL.VoucherDate,5,2)+'/'+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate
				,VL.VoucherNo
				,VTM.Name AS VoucherType
				,(SELECT TOP 1 (LM1.Name)
					FROM VoucherL VL1
						LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
					WHERE VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
						AND VL1.F_LedgerMaster != @F_LedgerMaster
					ORDER BY VL1.Amount DESC
					) AS AgLedger		
				,(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) AS DrAmt
				,(CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END) AS CrAmt
				,VL.VoucherDate AS VoucherDate1
				,VL.F_VoucherTypeMaster,VL.Narration
				,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
				,((CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END)
					-
				  (CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END)
				 ) AS Balance
			FROM VoucherL VL
				LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
				LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster				
				LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
			WHERE VL.F_LedgerMaster = @F_LedgerMaster
				AND (VL.VoucherDate>=@FromDate AND VL.VoucherDate<=@ToDate)
				AND VL.F_CompanyMaster=@F_CompanyMaster 
	) AS TAB1
	LEFT JOIN VoucherL VL ON (VL.F_VoucherH=TAB1.Id AND VL.F_VoucherTypeMaster=TAB1.F_VoucherTypeMaster)
	LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster			
	LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
	LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	WHERE VL.F_LedgerMaster!=TAB1.LedgerId
	ORDER BY VoucherDate1,RowNumber,VoucherType
--**************TRANSACTION END**************

--**************FINAL SELECTION START**************
	SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[Narration],[DrAmt],[CrAmt]		
			,ABS(LTView.Balance) AS [Balance]
			,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
			,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening]
	FROM [TMPLedger] LT
	 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    				FROM [TMPLedger]
    				WHERE ROWNUMBER <= LT.ROWNUMBER
    			) AS LTView
	ORDER BY [IsOpening],[VoucherDate1]
	--[LedgerGroup],
--**************FINAL SELECTION END**************
END







































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Ledger_Summary_TDSReport]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




















CREATE PROCEDURE [dbo].[Select_Acc_Ledger_Summary_TDSReport] 
	@F_LedgerMaster numeric(18,0),
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),	
	@F_CompanyMaster numeric(18,0),
	@F_OtherLedgerMaster NUMERIC(18,0)=0,
	@F_VoucherTypeMaster NUMERIC(18,0)=0,
	@IsOpBalAbove smallint=1
AS
BEGIN
-- OPENING BALANCE
	DECLARE @FinYearFromDate NVARCHAR(MAX)
	DECLARE @FinYearFromDate_Min NVARCHAR(MAX)
	SET @FinYearFromDate = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC)	
	SET @FinYearFromDate_Min = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
--**************EMPTY TMP LEDGER TABLE**************
	DELETE FROM TMPLedger	
--**************OPENING START**************
	--INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
	--SELECT 0 AS RowNumber,0 AS ID,
	--	ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
	--	,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
	--	,'Opening' AS AgLedger,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
	--	(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
	--		-
	--	ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
	--	) AS Balance
	--FROM(
	--	--Opening
	--	SELECT SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
	--	FROM LedgerMaster LM
	--	LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
	--	WHERE LM.Id =@F_LedgerMaster AND LM.F_CompanyMaster=@F_CompanyMaster AND LGM.IsBalanceSheet = 1 AND ISNULL(@FinYearFromDate_Min,0)>0
	--	GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			
	--	UNION ALL
		
	--	--Transaction
	--	SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
	--	FROM VoucherL VL
	--		LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id		
	--		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
	--		LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	--	WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = @F_LedgerMaster AND (VL.VoucherDate<@FromDate AND VL.VoucherDate>=ISNULL(@FinYearFromDate_Min,0)) AND VL.F_CompanyMaster=@F_CompanyMaster
	--		AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
	--		AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
	--		AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=@FinYearFromDate))
	--	GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
	--) AS TAB4
	--GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
	--ORDER BY LedgerGroup,Ledger
--**************OPENING END**************
--**************TRANSACTION START************ (voucherno order by please write on rownumber)
	INSERT INTO TMPLedger(IsOpening,RowNumber,Id,VoucherDate,VoucherNo,VoucherType,AgLedger,DrAmt,CrAmt,VoucherDate1,F_VoucherTypeMaster,Narration,Ledger,LedgerId,LedgerGroup,LedgerGroupId,Balance,AgLedgerDetail,PANNo,TotalNetAmt,TaxRate,Interest)
	SELECT  1,ROW_NUMBER() OVER (order by CAST(VL.VoucherDate AS NUMERIC) ASC,CAST(LEFT(VL.[VoucherNo], Patindex('%[^0-9]%', VL.[VoucherNo] + 'x') - 1) AS BIGINT),VL.Id) as RowNumber
		,VL.F_VoucherH AS Id
		,(SUBSTRING(VL.VoucherDate,7,2)+'/'+SUBSTRING(VL.VoucherDate,5,2)+'/'+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate
		,VL.VoucherNo
		,VTM.Name AS VoucherType
		,(SELECT TOP 1 (LM1.Name)
			FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
			WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
				AND VL1.F_LedgerMaster != @F_LedgerMaster AND LM1.F_LedgerGroupMaster IN(31,32)
			ORDER BY VL1.Amount DESC
			) AS AgLedger
		,(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) AS DrAmt
		,(CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END) AS CrAmt
		,VL.VoucherDate AS VoucherDate1
		,VL.F_VoucherTypeMaster,VL.Narration
		,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		,((CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END)
			-
		  (CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END)
		 )
		 ,AgLedgerDetail = STUFF(			
			(SELECT (LM.Name + '   ' + CAST(VL1.Amount AS NVARCHAR) + ' ' + CASE WHEN(VL1.DrCr=0) THEN 'Dr' ELSE 'Cr' END + '') + CHAR(10)
				FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM ON LM.Id = VL1.F_LedgerMaster
				WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherH = VL.F_VoucherH AND VL1.Id<>VL.Id AND VL1.F_VoucherTypeMaster = VL.F_VoucherTypeMaster
			FOR XML PATH('')), 1, 0, '')
		,(SELECT TOP 1 (LM1.PANNo)
			FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
			WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
				AND VL1.F_LedgerMaster != @F_LedgerMaster AND LM1.F_LedgerGroupMaster IN(31,32)
			ORDER BY VL1.Amount DESC
			) AS PANNo
		,VH.TotalNetAmt
		,LM.TaxRate
		,(ROUND((VL.Amount*(1.5/100)),2)*3) AS Interest
	FROM VoucherL VL
		LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
		LEFT JOIN VoucherH VH ON VH.Id=VL.F_VoucherH
	WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = @F_LedgerMaster
		AND (VL.VoucherDate>=@FromDate AND VL.VoucherDate<=@ToDate)
		AND VL.F_CompanyMaster=@F_CompanyMaster
		AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
		AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
		AND VH.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL.F_VoucherTypeMaster=4
	ORDER BY LedgerGroup,Ledger,VoucherDate1,RowNumber,VoucherType
--**************TRANSACTION END**************

--**************FINAL SELECTION START************** (please donot change this query, this will may affect balance amt)
IF(@IsOpBalAbove=1)
	BEGIN
		SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[Narration],[DrAmt],[CrAmt]		
				,ABS(LTView.Balance) AS [Balance]
				,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
				,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening],[AgLedgerDetail],[PANNo],[TotalNetAmt],[TaxRate],[Interest]
		FROM [TMPLedger] LT
		 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    					FROM [TMPLedger]
    					WHERE ROWNUMBER <= LT.ROWNUMBER
    				) AS LTView
		ORDER BY [AgLedger],[IsOpening],[VoucherDate1],ROWNUMBER
	END
ELSE
	BEGIN
		SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[Narration],[DrAmt],[CrAmt]		
				,ABS(LTView.Balance) AS [Balance]
				,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
				,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening],[AgLedgerDetail],[PANNo],[TotalNetAmt],[TaxRate],[Interest]
		FROM [TMPLedger] LT
		 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    					FROM [TMPLedger]
    					WHERE ROWNUMBER <= LT.ROWNUMBER
    				) AS LTView
		ORDER BY [AgLedger],[IsOpening] DESC,[VoucherDate1],ROWNUMBER
	END
--**************FINAL SELECTION END**************
END

























































GO
```

# [dbo].[Select_Acc_Ledger_SummaryAll]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Ledger_SummaryAll]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





CREATE PROCEDURE [dbo].[Select_Acc_Ledger_SummaryAll] 
	@F_LedgerMaster numeric(18,0),
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),	
	@F_CompanyMaster numeric(18,0),
	@LedgerIds nvarchar(max)='',
	@OtherLedgerIds nvarchar(max)='',
	@F_OtherLedgerMaster numeric(18,0)=0
AS
BEGIN
-- OPENING BALANCE	
	DECLARE @FinYearFromDate NVARCHAR(MAX)
	SET @FinYearFromDate = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC)	
	DECLARE @FinYearFromDate_First NVARCHAR(MAX)
	SET @FinYearFromDate_First = ISNULL((SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom),0)
	IF(@FinYearFromDate_First=0)
		SET @FinYearFromDate_First = ISNULL((SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom),0)

--**************EMPTY TMP LEDGER TABLE**************
	DELETE FROM TMPLedger
	--DELETE FROM TMPLedgerView
--**************OPENING START**************
	INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
	SELECT 0 AS RowNumber,0 AS ID,
		ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
		,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
		,'Opening' AS AgLedger,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
		(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
			-
		ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
		) AS Balance
	FROM(
		----Opening
		--SELECT SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		--FROM LedgerMaster LM
		--LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		--WHERE 1 = (CASE WHEN(@F_LedgerMaster=0 AND @LedgerIds<>'') THEN (CASE WHEN(LM.Id IN(SELECT * FROM Split(@LedgerIds,',')) ) THEN 1 ELSE 0 END) ELSE (CASE WHEN(LM.Id =@F_LedgerMaster) THEN 1 ELSE 0 END) END)
		--AND LM.F_CompanyMaster=@F_CompanyMaster AND LGM.IsBalanceSheet = 1 AND @FromDate>=@FinYearFromDate_First
		--GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			
		--UNION ALL
		
		--Transaction
		SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,
			(LM.Name
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CM.Name,'''') <> '''') THEN ', ' + CM.Name END),''))
			AS Ledger,
			LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		FROM VoucherL VL
			LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
			LEFT JOIN CityMaster CM ON CM.Id = LM.F_CityMaster
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
			LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
		WHERE ISNULL(VL.Cancelled,0)=0 AND 
			1 = (CASE WHEN(@F_LedgerMaster=0 AND @LedgerIds<>'') THEN (CASE WHEN(VL.F_LedgerMaster IN(SELECT * FROM Split(@LedgerIds,',')) ) THEN 1 ELSE 0 END) ELSE (CASE WHEN(VL.F_LedgerMaster = @F_LedgerMaster) THEN 1 ELSE 0 END) END)
			AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0 AND @LedgerIds<>'' AND @OtherLedgerIds<>'') THEN 
						(CASE WHEN(VL.F_OtherLedgerMaster IN(SELECT * FROM Split(@OtherLedgerIds,',')) ) THEN 1 ELSE 0 END)
					ELSE
						(CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
					END)
			--AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0 AND @OtherLedgerIds<>'' AND @F_LedgerMaster>0) THEN (CASE WHEN(VL.F_OtherLedgerMaster IN(SELECT * FROM Split(@OtherLedgerIds,',')) ) THEN 1 ELSE 0 END) ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster AND @F_OtherLedgerMaster>0) THEN 1 ELSE 0 END) END)
			AND (VL.VoucherDate<@FromDate) AND VL.F_CompanyMaster=@F_CompanyMaster
			AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=@FinYearFromDate))
		GROUP BY LM.Name,LM.Address1,LM.Address2,CM.Name,LM.Id,LGM.Name,LGM.Id
	) AS TAB4
	GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
	ORDER BY LedgerGroup,Ledger
--**************OPENING END**************
--**************TRANSACTION START************
	INSERT INTO TMPLedger(IsOpening,RowNumber,Id,VoucherDate,VoucherNo,VoucherType,AgLedger,DrAmt,CrAmt,VoucherDate1,F_VoucherTypeMaster,Narration,Ledger,LedgerId,LedgerGroup,LedgerGroupId,Balance,AgLedgerDetail)
	SELECT  1,ROW_NUMBER() OVER (order by CAST(VL.VoucherDate AS NUMERIC) ASC,VL.Id) as RowNumber
		,VL.F_VoucherH AS Id
		,(SUBSTRING(VL.VoucherDate,7,2)+'/'+SUBSTRING(VL.VoucherDate,5,2)+'/'+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate
		,VL.VoucherNo
		,VTM.Name AS VoucherType
		,(SELECT TOP 1 (LM1.Name)
			FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
			WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
				AND VL1.F_LedgerMaster NOT IN(SELECT * FROM Split(@LedgerIds,','))
				AND VL1.F_OtherLedgerMaster NOT IN(SELECT * FROM Split(@OtherLedgerIds,','))
			ORDER BY VL1.Amount DESC
			) AS AgLedger		
		,(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) AS DrAmt
		,(CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END) AS CrAmt
		,VL.VoucherDate AS VoucherDate1
		,VL.F_VoucherTypeMaster,VL.Narration
		,(LM.Name
			+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
			+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
			+ ISNULL((CASE WHEN(ISNULL(CM.Name,'''') <> '''') THEN ', ' + CM.Name END),''))
		AS Ledger,
		LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		,((CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END)
			-
		  (CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END)
		 )
		 ,AgLedgerDetail = STUFF(			
			(SELECT (LM.Name + '   ' + CAST(VL1.Amount AS NVARCHAR) + ' ' + CASE WHEN(VL1.DrCr=0) THEN 'Dr' ELSE 'Cr' END + '') + CHAR(10)
				FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM ON LM.Id = VL1.F_LedgerMaster
				WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherH = VL.F_VoucherH AND VL1.Id<>VL.Id AND VL1.F_VoucherTypeMaster = VL.F_VoucherTypeMaster
			FOR XML PATH('')), 1, 0, '')		
	FROM VoucherL VL
		LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
		LEFT JOIN CityMaster CM ON CM.Id = LM.F_CityMaster
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster				
		LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	WHERE ISNULL(VL.Cancelled,0)=0 AND 
		1 = (CASE WHEN(@F_LedgerMaster=0 AND @LedgerIds<>'') THEN (CASE WHEN(VL.F_LedgerMaster IN(SELECT * FROM Split(@LedgerIds,',')) ) THEN 1 ELSE 0 END) ELSE (CASE WHEN(VL.F_LedgerMaster = @F_LedgerMaster) THEN 1 ELSE 0 END) END)
			AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0 AND @LedgerIds<>'' AND @OtherLedgerIds<>'') THEN 
					(CASE WHEN(VL.F_OtherLedgerMaster IN(SELECT * FROM Split(@OtherLedgerIds,',')) ) THEN 1 ELSE 0 END)
				ELSE
					(CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
				END)
		--AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0 AND @OtherLedgerIds<>'' AND @F_LedgerMaster>0) THEN (CASE WHEN(VL.F_OtherLedgerMaster IN(SELECT * FROM Split(@OtherLedgerIds,',')) ) THEN 1 ELSE 0 END) ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster AND @F_OtherLedgerMaster>0) THEN 1 ELSE 0 END) END)
		AND (VL.VoucherDate>=@FromDate AND VL.VoucherDate<=@ToDate)
		AND VL.F_CompanyMaster=@F_CompanyMaster 
	ORDER BY LedgerGroup,Ledger,VoucherDate1,RowNumber,VoucherType
--**************TRANSACTION END**************

--**************FINAL SELECTION START**************
	SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[Narration],[DrAmt],[CrAmt]		
			,ABS(LTView.Balance) AS [Balance]
			,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
			,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening],[AgLedgerDetail]
	FROM [TMPLedger] LT
	 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    				FROM [TMPLedger]
    				WHERE ROWNUMBER <= LT.ROWNUMBER
    			) AS LTView	
	ORDER BY [LedgerGroup],[Ledger],[IsOpening],[VoucherDate1]
--**************FINAL SELECTION END**************
END






















GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Ledger_SummaryNew]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO








CREATE PROCEDURE [dbo].[Select_Acc_Ledger_SummaryNew] 
	@F_LedgerMaster numeric(18,0),
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),	
	@F_CompanyMaster numeric(18,0),
	@F_OtherLedgerMaster NUMERIC(18,0)=0,
	@F_VoucherTypeMaster NUMERIC(18,0)=0,
	@IsOpBalAbove smallint=1
AS
BEGIN
--
	DECLARE @F_CompanyYears numeric(18,0)
	SET @F_CompanyYears = ISNULL((SELECT TOP 1 Id FROM CompanyYears WHERE FinYearFrom>=@FromDate AND FinYearTo<=@ToDate),0)
	IF (@F_CompanyYears =0)
		SET @F_CompanyYears = ISNULL((SELECT TOP 1 Id FROM CompanyYears WHERE FinYearFrom>=@FromDate),0)
-- OPENING BALANCE
	DECLARE @FinYearFromDate NVARCHAR(MAX)
	DECLARE @FinYearFromDate_Min NVARCHAR(MAX)
	SET @FinYearFromDate = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC)	
	SET @FinYearFromDate_Min = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)

	DECLARE @IsLedgerBalance_Split SMALLINT
	DECLARE @FinYearFromDate_Split NVARCHAR(MAX)
	DECLARE @F_CompanyYears_Split NUMERIC(18,0)
	SET @IsLedgerBalance_Split = (SELECT TOP 1 ISNULL(IsLedgerBalanceSplit,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @F_CompanyYears_Split = (SELECT TOP 1 ISNULL(Id,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @FinYearFromDate_Split = (SELECT TOP 1 ISNULL(FinYearFrom,0) FROM CompanyYears WHERE Id=@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
--**************EMPTY TMP LEDGER TABLE**************
	DELETE FROM TMPLedger	
--**************OPENING START**************
	IF(@IsLedgerBalance_Split=1)
		BEGIN
			INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
			SELECT 0 AS RowNumber,0 AS ID,
				ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
				,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
				,'Opening' AS AgLedger,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
				(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
				-
				ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
				) AS Balance
			FROM(
				SELECT (CASE WHEN (LML.DrCr=0) THEN LML.OpeningBalance ELSE 0 END) AS DebitAmount,
					(CASE WHEN (LML.DrCr=1) THEN LML.OpeningBalance ELSE 0 END) AS CreditAmount,
					LM.Name AS Ledger, LM.Id AS LedgerId,
					LGM2.Name AS LedgerGroup,LGM2.Id AS LedgerGroupId
				FROM LedgerMaster LM
					LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears_Split
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
					LEFT JOIN LedgerMasterL LML2 ON LML2.F_LedgerMaster = LM.Id AND LML2.F_CompanyYears = @F_CompanyYears
					LEFT JOIN LedgerGroupMaster LGM2 ON LGM2.Id=LML2.F_LedgerGroupMaster
				WHERE LM.Id =@F_LedgerMaster AND LM.F_CompanyMaster=@F_CompanyMaster AND LGM.IsBalanceSheet = 1
			
				UNION ALL
		
				--Transaction
				SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
				FROM VoucherL VL
					LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
					LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
					LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
				WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = @F_LedgerMaster AND (VL.VoucherDate<@FromDate AND VL.VoucherDate>=ISNULL(@FinYearFromDate_Split,0)) AND VL.F_CompanyMaster=@F_CompanyMaster
					AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
					AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
					AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=@FinYearFromDate))
				GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			) AS TAB4
			GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
			ORDER BY LedgerGroup,Ledger
		END
	ELSE
		BEGIN
			INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
			SELECT 0 AS RowNumber,0 AS ID,
				ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
				,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
				,'Opening' AS AgLedger,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
				(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
					-
				ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
				) AS Balance
			FROM(
				--Opening
				SELECT SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
				FROM LedgerMaster LM
				LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
				WHERE LM.Id =@F_LedgerMaster AND LM.F_CompanyMaster=@F_CompanyMaster AND LGM.IsBalanceSheet = 1 AND ISNULL(@FinYearFromDate_Min,0)>0
				GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			
				UNION ALL
		
				--Transaction
				SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
				FROM VoucherL VL
					LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
					LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster	
					LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
				WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = @F_LedgerMaster AND (VL.VoucherDate<@FromDate AND VL.VoucherDate>=ISNULL(@FinYearFromDate_Min,0)) AND VL.F_CompanyMaster=@F_CompanyMaster
					AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
					AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
					AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=@FinYearFromDate))
				GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			) AS TAB4
			GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
			ORDER BY LedgerGroup,Ledger
	END
--**************OPENING END**************
--**************TRANSACTION START************ (voucherno order by please write on rownumber)
	INSERT INTO TMPLedger(IsOpening,RowNumber,Id,VoucherDate,VoucherNo,VoucherType,AgLedger,DrAmt,CrAmt,VoucherDate1,F_VoucherTypeMaster,Narration,Ledger,LedgerId,LedgerGroup,LedgerGroupId,Balance,AgLedgerDetail)
	SELECT  1,ROW_NUMBER() OVER (order by CAST(VL.VoucherDate AS NUMERIC) ASC,CAST(LEFT([VoucherNo], Patindex('%[^0-9]%', [VoucherNo] + 'x') - 1) AS BIGINT),VL.Id) as RowNumber
		,VL.F_VoucherH AS Id
		,(SUBSTRING(VL.VoucherDate,7,2)+'/'+SUBSTRING(VL.VoucherDate,5,2)+'/'+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate
		,VL.VoucherNo
		,VTM.Name AS VoucherType
		,(SELECT TOP 1 (LM1.Name)
			FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
			WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
				AND VL1.F_LedgerMaster != @F_LedgerMaster
			ORDER BY VL1.Amount DESC
			) AS AgLedger		
		,(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) AS DrAmt
		,(CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END) AS CrAmt
		,VL.VoucherDate AS VoucherDate1
		,VL.F_VoucherTypeMaster,VL.Narration
		,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		,((CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END)
			-
		  (CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END)
		 )
		 ,AgLedgerDetail = STUFF(			
			(SELECT (LM.Name + '   ' + CAST(VL1.Amount AS NVARCHAR) + ' ' + CASE WHEN(VL1.DrCr=0) THEN 'Dr' ELSE 'Cr' END + '') + CHAR(10)
				FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM ON LM.Id = VL1.F_LedgerMaster
				WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherH = VL.F_VoucherH AND VL1.Id<>VL.Id AND VL1.F_VoucherTypeMaster = VL.F_VoucherTypeMaster
			FOR XML PATH('')), 1, 0, '')
	FROM VoucherL VL
		LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = @F_LedgerMaster
		AND (VL.VoucherDate>=@FromDate AND VL.VoucherDate<=@ToDate)
		AND VL.F_CompanyMaster=@F_CompanyMaster
		AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
		AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
	ORDER BY LedgerGroup,Ledger,VoucherDate1,RowNumber,VoucherType
--**************TRANSACTION END**************

--**************FINAL SELECTION START************** (please donot change this query, this will may affect balance amt)
IF(@IsOpBalAbove=1)
	BEGIN
		SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[Narration],[DrAmt],[CrAmt]		
				,ABS(LTView.Balance) AS [Balance]
				,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
				,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening],[AgLedgerDetail]
		FROM [TMPLedger] LT
		 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    					FROM [TMPLedger]
    					WHERE ROWNUMBER <= LT.ROWNUMBER
    				) AS LTView
		ORDER BY [Ledger],[IsOpening],[VoucherDate1],ROWNUMBER
	END
ELSE
	BEGIN
		SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[Narration],[DrAmt],[CrAmt]		
				,ABS(LTView.Balance) AS [Balance]
				,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
				,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening],[AgLedgerDetail]
		FROM [TMPLedger] LT
		 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    					FROM [TMPLedger]
    					WHERE ROWNUMBER <= LT.ROWNUMBER
    				) AS LTView
		ORDER BY [Ledger],[IsOpening] DESC,[VoucherDate1],ROWNUMBER
	END
--**************FINAL SELECTION END**************
END


GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Ledger_SummaryNew2]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO









CREATE PROCEDURE [dbo].[Select_Acc_Ledger_SummaryNew2] 
	@F_LedgerMaster numeric(18,0),
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),	
	@F_CompanyMaster numeric(18,0),
	@F_OtherLedgerMaster NUMERIC(18,0)=0,
	@F_VoucherTypeMaster NUMERIC(18,0)=0,
	@IsOpBalAbove smallint=1
AS
BEGIN
--
	DECLARE @F_CompanyYears numeric(18,0)
	SET @F_CompanyYears = ISNULL((SELECT TOP 1 Id FROM CompanyYears WHERE FinYearFrom>=@FromDate AND FinYearTo<=@ToDate),0)
	IF (@F_CompanyYears =0)
		SET @F_CompanyYears = ISNULL((SELECT TOP 1 Id FROM CompanyYears WHERE FinYearFrom>=@FromDate),0)
-- OPENING BALANCE
	DECLARE @FinYearFromDate NVARCHAR(MAX)
	DECLARE @FinYearFromDate_Min NVARCHAR(MAX)
	SET @FinYearFromDate = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC)	
	SET @FinYearFromDate_Min = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)

	DECLARE @IsLedgerBalance_Split SMALLINT
	DECLARE @FinYearFromDate_Split NVARCHAR(MAX)
	DECLARE @F_CompanyYears_Split NUMERIC(18,0)
	SET @IsLedgerBalance_Split = (SELECT TOP 1 ISNULL(IsLedgerBalanceSplit,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @F_CompanyYears_Split = (SELECT TOP 1 ISNULL(Id,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @FinYearFromDate_Split = (SELECT TOP 1 ISNULL(FinYearFrom,0) FROM CompanyYears WHERE Id=@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
--**************EMPTY TMP LEDGER TABLE**************
	DELETE FROM TMPLedger
--**************TRANSACTION START************ (voucherno order by please write on rownumber)
	INSERT INTO TMPLedger(IsOpening,RowNumber,Id,VoucherDate,VoucherNo,VoucherType,AgLedger,DrAmt,CrAmt,VoucherDate1,F_VoucherTypeMaster,Narration,Ledger,LedgerId,LedgerGroup,LedgerGroupId,Balance,AgLedgerDetail)
	SELECT  1,ROW_NUMBER() OVER (order by CAST(VL.VoucherDate AS NUMERIC) ASC,CAST(LEFT([VoucherNo], Patindex('%[^0-9]%', [VoucherNo] + 'x') - 1) AS BIGINT),VL.Id) as RowNumber
		,VL.F_VoucherH AS Id
		,(SUBSTRING(VL.VoucherDate,7,2)+'/'+SUBSTRING(VL.VoucherDate,5,2)+'/'+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate
		,VL.VoucherNo
		,VTM.Name AS VoucherType
		,(SELECT TOP 1 (LM1.Name)
			FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM1 ON VL1.F_LedgerMaster=LM1.Id
			WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherTypeMaster=VL.F_VoucherTypeMaster AND VL1.F_VoucherH=VL.F_VoucherH
				AND VL1.F_LedgerMaster != @F_LedgerMaster
			ORDER BY VL1.Amount DESC
			) AS AgLedger		
		,(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) AS DrAmt
		,(CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END) AS CrAmt
		,VL.VoucherDate AS VoucherDate1
		,VL.F_VoucherTypeMaster,VL.Narration
		,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
		,((CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END)
			-
		  (CASE WHEN(VL.DrCr<>0) THEN VL.Amount ELSE 0 END)
		 )
		 ,AgLedgerDetail = STUFF(			
			(SELECT (LM.Name + '   ' + CAST(VL1.Amount AS NVARCHAR) + ' ' + CASE WHEN(VL1.DrCr=0) THEN 'Dr' ELSE 'Cr' END + '') + CHAR(10)
				FROM VoucherL VL1
				LEFT JOIN LedgerMaster LM ON LM.Id = VL1.F_LedgerMaster
				WHERE ISNULL(VL1.Cancelled,0)=0 AND VL1.F_VoucherH = VL.F_VoucherH AND VL1.Id<>VL.Id AND VL1.F_VoucherTypeMaster = VL.F_VoucherTypeMaster
			FOR XML PATH('')), 1, 0, '')
	FROM VoucherL VL
		LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = @F_LedgerMaster
		AND (VL.VoucherDate>=@FromDate AND VL.VoucherDate<=@ToDate)
		AND VL.F_CompanyMaster=@F_CompanyMaster
		AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
		AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
	ORDER BY LedgerGroup,Ledger,VoucherDate1,RowNumber,VoucherType
--**************TRANSACTION END**************
--**************OPENING START**************
	IF(@IsLedgerBalance_Split=1)
		BEGIN
			INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
			SELECT 0 AS RowNumber,0 AS ID,
				ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
				,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
				,'Opening' AS AgLedger,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
				(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
				-
				ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
				) AS Balance
			FROM(
				SELECT (CASE WHEN (LML.DrCr=0) THEN LML.OpeningBalance ELSE 0 END) AS DebitAmount,
					(CASE WHEN (LML.DrCr=1) THEN LML.OpeningBalance ELSE 0 END) AS CreditAmount,
					LM.Name AS Ledger, LM.Id AS LedgerId,
					LGM2.Name AS LedgerGroup,LGM2.Id AS LedgerGroupId
				FROM LedgerMaster LM
					LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears_Split
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
					LEFT JOIN LedgerMasterL LML2 ON LML2.F_LedgerMaster = LM.Id AND LML2.F_CompanyYears = @F_CompanyYears
					LEFT JOIN LedgerGroupMaster LGM2 ON LGM2.Id=LML2.F_LedgerGroupMaster
				WHERE LM.Id =@F_LedgerMaster AND LM.F_CompanyMaster=@F_CompanyMaster AND LGM.IsBalanceSheet = 1
			
				UNION ALL
		
				--Transaction
				SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
				FROM VoucherL VL
					LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
					LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
					LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
				WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = @F_LedgerMaster AND (VL.VoucherDate<@FromDate AND VL.VoucherDate>=ISNULL(@FinYearFromDate_Split,0)) AND VL.F_CompanyMaster=@F_CompanyMaster
					AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
					AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
					AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=@FinYearFromDate))
				GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			) AS TAB4
			GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
			ORDER BY LedgerGroup,Ledger
		END
	ELSE
		BEGIN
			INSERT INTO TMPLedger(RowNumber,Id,DrAmt,CrAmt,AgLedger,Ledger,LedgerId,LedgerGroup,LedgerGroupId,IsOpening,Balance)
			SELECT 0 AS RowNumber,0 AS ID,
				ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS DrAmt
				,ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) END),0) AS CrAmt
				,'Opening' AS AgLedger,TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId,0 AS IsOpening,
				(ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
					-
				ISNULL((CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))<=0) THEN ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) ELSE 0 END),0)
				) AS Balance
			FROM(
				--Opening
				SELECT SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
				FROM LedgerMaster LM
				LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
				WHERE LM.Id =@F_LedgerMaster AND LM.F_CompanyMaster=@F_CompanyMaster AND LGM.IsBalanceSheet = 1 AND ISNULL(@FinYearFromDate_Min,0)>0
				GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			
				UNION ALL
		
				--Transaction
				SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LM.Name AS Ledger, LM.Id AS LedgerId,LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId
				FROM VoucherL VL
					LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
					LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster	
					LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
				WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = @F_LedgerMaster AND (VL.VoucherDate<@FromDate AND VL.VoucherDate>=ISNULL(@FinYearFromDate_Min,0)) AND VL.F_CompanyMaster=@F_CompanyMaster
					AND 1 = (CASE WHEN(@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_OtherLedgerMaster = @F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
					AND 1 = (CASE WHEN(@F_VoucherTypeMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_VoucherTypeMaster = @F_VoucherTypeMaster) THEN 1 ELSE 0 END) END)
					AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=@FinYearFromDate))
				GROUP BY LM.Name, LM.Id,LGM.Name,LGM.Id
			) AS TAB4
			GROUP BY TAB4.Ledger, TAB4.LedgerId,TAB4.LedgerGroup,TAB4.LedgerGroupId
			ORDER BY LedgerGroup,Ledger
	END
--**************OPENING END**************

--**************FINAL SELECTION START************** (please donot change this query, this will may affect balance amt)
IF(@IsOpBalAbove=1)
	BEGIN
		SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[Narration],[DrAmt],[CrAmt]		
				,ABS(LTView.Balance) AS [Balance]
				,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
				,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening],[AgLedgerDetail]
		FROM [TMPLedger] LT
		 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    					FROM [TMPLedger]
    					WHERE ROWNUMBER <= LT.ROWNUMBER
    				) AS LTView
		ORDER BY [Ledger],[IsOpening],[VoucherDate1],ROWNUMBER
	END
ELSE
	BEGIN
		SELECT [Id],[Ledger],[LedgerId],[LedgerGroup],[LedgerGroupId],[VoucherDate],[VoucherNo],[VoucherType],[AgLedger],[Narration],[DrAmt],[CrAmt]		
				,ABS(LTView.Balance) AS [Balance]
				,(CASE WHEN(LTView.Balance>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
				,[F_VoucherTypeMaster],[VoucherDate1],[IsOpening],[AgLedgerDetail]
		FROM [TMPLedger] LT
		 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    					FROM [TMPLedger]
    					WHERE ROWNUMBER <= LT.ROWNUMBER
    				) AS LTView
		ORDER BY [Ledger],[IsOpening] DESC,[VoucherDate1],ROWNUMBER
	END
--**************FINAL SELECTION END**************
END

GO
```

# [dbo].[Select_Acc_LedgerMonthSummary]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_LedgerMonthSummary]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO







CREATE PROCEDURE [dbo].[Select_Acc_LedgerMonthSummary] 
	@F_LedgerGroupMaster numeric(18,0),
	@F_LedgerMaster numeric(18,0)=0,
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0)	
AS
BEGIN
--**************EMPTY TMP LEDGER TABLE START**************
	DELETE FROM TMPLedger
--**************EMPTY TMP LEDGER TABLE END**************
--**************OPENING STOCK START**************
	DECLARE @FinYearFromDate NVARCHAR(MAX)
	DECLARE @IsLedgerBalance_Split SMALLINT
	DECLARE @FinYearFromDate_Split NVARCHAR(MAX)
	DECLARE @F_CompanyYears_Split NUMERIC(18,0)
	--DECLARE @F_CompanyYears NUMERIC(18,0)
	SET @FinYearFromDate = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC)	
	SET @IsLedgerBalance_Split = (SELECT TOP 1 ISNULL(IsLedgerBalanceSplit,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @F_CompanyYears_Split = (SELECT TOP 1 ISNULL(Id,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @FinYearFromDate_Split = (SELECT TOP 1 ISNULL(FinYearFrom,0) FROM CompanyYears WHERE Id=@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
	--SET @F_CompanyYears = (SELECT TOP 1 Id FROM CompanyYears WHERE Id>@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster)
	
	IF(@IsLedgerBalance_Split=1)
		BEGIN
			INSERT INTO TMPLedger(ROWNUMBER,IsOpening,AgLedger,VoucherDate,DrAmt,CrAmt)
			SELECT 0 AS RowNumber,0,' Opening' AS MonthN,'' AS MonthNo,
					(CASE WHEN (SUM(TAB_OP.DebitAmount)-SUM(TAB_OP.CreditAmount)>0) THEN ABS(SUM(TAB_OP.DebitAmount)-SUM(TAB_OP.CreditAmount))
						ELSE 0 END) AS DebitAmount,
					
					(CASE WHEN (SUM(TAB_OP.CreditAmount)-SUM(TAB_OP.DebitAmount)>0) THEN ABS(SUM(TAB_OP.CreditAmount)-SUM(TAB_OP.DebitAmount))
						ELSE 0 END) AS CreditAmount
			FROM (
				SELECT
					SUM(CASE WHEN (LML.DrCr=0) THEN LML.OpeningBalance ELSE 0 END) AS DebitAmount,
					SUM(CASE WHEN (LML.DrCr=1) THEN LML.OpeningBalance ELSE 0 END) AS CreditAmount
				FROM LedgerMaster LM
					LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears_Split
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
				WHERE
					1= (CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(LM.Id =@F_LedgerMaster) THEN 1 ELSE 0 END) END)
					AND 1= (CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LM.F_LedgerGroupMaster =@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)		
					AND LM.F_CompanyMaster=@F_CompanyMaster
			
				UNION ALL
		
				--Transaction
				SELECT (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount,
					(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount
				FROM VoucherL VL
					LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id		
					LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster		
					LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
				WHERE ISNULL(VL.Cancelled,0)=0 AND VL.F_LedgerMaster = @F_LedgerMaster AND VL.VoucherDate >= @FinYearFromDate_Split
					AND (VL.VoucherDate<@FromDate) AND VL.F_CompanyMaster=@F_CompanyMaster
					AND LM.F_LedgerGroupMaster<>30 AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=@FinYearFromDate))
			) AS TAB_OP
		END
	ELSE
		BEGIN
			INSERT INTO TMPLedger(ROWNUMBER,IsOpening,AgLedger,VoucherDate,DrAmt,CrAmt)
			SELECT 0 AS RowNumber,0,' Opening' AS MonthN,'' AS MonthNo,
				SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount,
				SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount
			FROM LedgerMaster LM
				LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
			WHERE
				1= (CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(LM.Id =@F_LedgerMaster) THEN 1 ELSE 0 END) END)
				AND 1= (CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LM.F_LedgerGroupMaster =@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)		
				AND LM.F_CompanyMaster=@F_CompanyMaster
		END
--**************OPENING STOCK END**************
--**************TRANSACTION START**************
	INSERT INTO TMPLedger(ROWNUMBER,IsOpening,AgLedger,VoucherDate,DrAmt,CrAmt)
	SELECT ROW_NUMBER() OVER (order by SUBSTRING(VL.VoucherDate,1,6)) AS RowNumber,1,
		((SUBSTRING(VL.VoucherDate,1,4)) + ' - ' +  SUBSTRING(DateName(mm,DATEADD(mm,CAST((SUBSTRING(VL.VoucherDate,5,2)) AS NUMERIC)-1,0)),0,4)) AS MonthN,
		SUBSTRING(VL.VoucherDate,1,6) AS MonthNo,
		(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount ,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount
	FROM VoucherL VL
		LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id				
		LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	WHERE ISNULL(VL.Cancelled,0)=0 AND
		1= (CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(VL.F_LedgerMaster = @F_LedgerMaster) THEN 1 ELSE 0 END) END)
		AND 1= (CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LM.F_LedgerGroupMaster =@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)
		AND (VL.VoucherDate>=@FromDate AND VL.VoucherDate<=@ToDate) AND VL.F_CompanyMaster=@F_CompanyMaster
	GROUP BY  SUBSTRING(VL.VoucherDate,1,6),SUBSTRING(VL.VoucherDate,1,4),SUBSTRING(VL.VoucherDate,5,2)
--**************TRANSACTION END**************

--**************MONTH ADDING START**************
	DECLARE @ROWNUMBER NUMERIC(18,0)
	DECLARE @MonthStart INT
	DECLARE @MonthEnd INT
	DECLARE @Year NVARCHAR(10)
	DECLARE @Month NVARCHAR(10)
	DECLARE @I INT
	SET @MonthStart = SUBSTRING(@FromDate,1,6)
	SET @MonthEnd = SUBSTRING(@ToDate,1,6)
	SET @I = @MonthStart
	WHILE (@I <= @MonthEnd)
		BEGIN
			SET @ROWNUMBER = (SELECT ROWNUMBER FROM [TMPLedger] WHERE VoucherDate = @I)
			SET @Year =SUBSTRING(CAST(@I AS NVARCHAR),1,4)
			SET @Month =SUBSTRING(CAST(@I AS NVARCHAR),5,2)
						
			IF(ISNULL(@ROWNUMBER,0)<=0)
				BEGIN
					INSERT INTO TMPLedger(ROWNUMBER,IsOpening,AgLedger,VoucherDate,DrAmt,CrAmt)
					SELECT CAST(@Month AS NUMERIC),1
							,((SUBSTRING(CAST(@I AS NVARCHAR),1,4)) + ' - ' +  SUBSTRING(DateName(mm,DATEADD(mm,CAST((SUBSTRING(CAST(@I AS NVARCHAR),5,2)) AS NUMERIC)-1,0)),0,4)) AS MonthN
							,CAST(@I AS NVARCHAR),0,0
				END
			
			IF(@Month=12)
				BEGIN
					SET @Year = @Year + 1
					SET @Month=1
				END
			ELSE
				BEGIN
					SET @Month = @Month + 1
				END
			
			
			IF(@Month<10)
				SET @I = CAST(CAST(@Year AS NVARCHAR)+ '0' + CAST(@Month AS NVARCHAR) AS NUMERIC)
			ELSE
				SET @I = CAST(CAST(@Year AS NVARCHAR) + CAST(@Month AS NVARCHAR) AS NUMERIC)
		END
--**************MONTH ADDING END****************

--**************FINAL SELECTION START**************
	SELECT AgLedger AS MonthN
			,SUM(DrAmt) AS DebitAmount
			,SUM(CrAmt) AS CreditAmount			
			,ABS(SUM(LTView.Balance)) AS [Balance]
			,(CASE WHEN (SUM(LTView.Balance)>=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
			,VoucherDate AS MonthNo
	FROM [TMPLedger] LT
	 CROSS APPLY (SELECT SUM([DrAmt]-[CrAmt]) AS Balance
    				FROM [TMPLedger]
    				WHERE VoucherDate <= LT.VoucherDate
    			) AS LTView
	GROUP BY AgLedger,VoucherDate
	ORDER BY VoucherDate
--**************FINAL SELECTION END**************

END











































































GO
```

# [dbo].[Select_Acc_PostingDetails]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_PostingDetails]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE PROCEDURE [dbo].[Select_Acc_PostingDetails] 	
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_VoucherTypeMaster numeric(18,0),
	@F_CompanyMaster numeric(18,0),
	@Type smallint=-1
AS
BEGIN
	DECLARE @SQL NVARCHAR(MAX)
	DECLARE @TableNameH NVARCHAR(50)
	DECLARE @FIELDS NVARCHAR(MAX)
	DECLARE @WHERE NVARCHAR(MAX)
	DECLARE @FIELDS_INNER NVARCHAR(MAX)
	DECLARE @WHERE_INNER NVARCHAR(MAX)
	DECLARE @WHERE_INNER2 NVARCHAR(MAX)
	DECLARE @GROUPBY_INNER NVARCHAR(MAX)
	DECLARE @AmountFieldName NVARCHAR(MAX)
--	
	SET @TableNameH = (SELECT TableNameH FROM VoucherTypeMaster WHERE Id = @F_VoucherTypeMaster)
	SET @WHERE = '(TAB1.TotalNetAmt-TAB1.CrAmt)<>0 OR (TAB1.TotalNetAmt-TAB1.DrAmt)<>0 OR TAB1.DrAmt=0 OR TAB1.CrAmt=0'
	SET @FIELDS = 'TAB1.Id,TAB1.VoucherNo,(SUBSTRING(TAB1.VoucherDate,7,2)+''/''+SUBSTRING(TAB1.VoucherDate,5,2)+''/''+SUBSTRING(TAB1.VoucherDate,1,4)) AS VoucherDate,TAB1.PurVoucherNo,TAB1.Ledger,TAB1.TotalNetAmt,TAB1.DrAmt,TAB1.CrAmt,TAB1.Diff'
	
	IF EXISTS(SELECT * FROM sys.columns 
				WHERE Name = N'TotalNetAmt' AND Object_ID = Object_ID(@TableNameH))    
		BEGIN
			SET @AmountFieldName = ',SH.TotalNetAmt '
		END
	ELSE IF EXISTS(SELECT * FROM sys.columns 
				WHERE Name = N'NetAmt' AND Object_ID = Object_ID(@TableNameH))    
		BEGIN
			SET @AmountFieldName = ',SH.NetAmt '
		END
	ELSE
		BEGIN
			SET @AmountFieldName = ',SH.NetAmount '
		END
		
	IF(@TableNameH='MoneyReceiptCashH' OR @TableNameH='MoneyReceiptBankH')
		BEGIN
			SET @FIELDS_INNER = 'SH.VoucherDate,'''' AS PurVoucherNo,SH.VoucherNo,SH.Id,0 AS F_OtherLedgerMaster,SH.F_LedgerMaster,LM.Name AS Ledger,SH.TotalAmount AS TotalNetAmt'
			SET @GROUPBY_INNER = 'SH.VoucherDate,SH.VoucherNo,SH.Id,SH.F_LedgerMaster,SH.TotalAmount,LM.Name'
		END
	ELSE IF(@TableNameH='PurchaseInvoiceH' OR @TableNameH='PurchaseReturnH')
		BEGIN
			SET @FIELDS_INNER = 'SH.VoucherDate,SH.PurVoucherNo,SH.VoucherNo,SH.Id,0 AS F_OtherLedgerMaster,SH.F_LedgerMaster,LM.Name AS Ledger' + @AmountFieldName + ' AS TotalNetAmt'
			SET @GROUPBY_INNER = 'SH.VoucherDate,SH.PurVoucherNo,SH.VoucherNo,SH.Id,SH.F_LedgerMaster' + @AmountFieldName + ',LM.Name'
		END
	ELSE
		BEGIN
			SET @FIELDS_INNER = 'SH.VoucherDate,'''' AS PurVoucherNo,SH.VoucherNo,SH.Id,0 AS F_OtherLedgerMaster,SH.F_LedgerMaster,LM.Name AS Ledger' + @AmountFieldName + ' AS TotalNetAmt'
			SET @GROUPBY_INNER = 'SH.VoucherDate,SH.VoucherNo,SH.Id,SH.F_LedgerMaster' + @AmountFieldName + ',LM.Name'
		END
		
	IF(@TableNameH = 'VoucherH')
		BEGIN
			SET @WHERE_INNER = 'SH.VoucherDate>=' + @FromDate + ' AND SH.VoucherDate<=' + @ToDate + ' AND ISNULL(SH.Cancelled,0)=0 AND SH.F_VoucherTypeMaster = ' + CAST(@F_VoucherTypeMaster AS NVARCHAR)
			SET @WHERE_INNER2 = 'SH.VoucherDate>=' + @FromDate + ' AND SH.VoucherDate<=' + @ToDate + ' AND SH.F_VoucherTypeMaster=' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + ' AND VL.F_VoucherTypeMaster=' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + ' AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + ' AND SH.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR)
		END
	ELSE
		BEGIN
			SET @WHERE_INNER = 'SH.VoucherDate>=' + @FromDate + ' AND SH.VoucherDate<=' + @ToDate + ' AND ISNULL(SH.Cancelled,0)=0'
			SET @WHERE_INNER2 = 'SH.VoucherDate>=' + @FromDate + ' AND SH.VoucherDate<=' + @ToDate + ' AND VL.F_VoucherTypeMaster=' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + ' AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + ' AND SH.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR)
		END
--		
	IF(@Type=0)
		BEGIN		-- POSTING DIFFERENCE
			SET @SQL = 'SELECT ' + @FIELDS + '
						FROM (
							SELECT ' + @FIELDS_INNER + '
								,ISNULL((SELECT SUM(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) FROM VoucherL VL WHERE VL.F_VoucherH = SH.Id AND VL.F_VoucherTypeMaster=' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '),0) AS DrAmt
								,ISNULL((SELECT SUM(CASE WHEN(VL.DrCr=1) THEN VL.Amount ELSE 0 END) FROM VoucherL VL WHERE VL.F_VoucherH = SH.Id AND VL.F_VoucherTypeMaster=' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '),0) AS CrAmt
								,(
									ISNULL((SELECT SUM(CASE WHEN(VL.DrCr=0) THEN VL.Amount ELSE 0 END) FROM VoucherL VL WHERE VL.F_VoucherH = SH.Id AND VL.F_VoucherTypeMaster=' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '),0)
									-
									ISNULL((SELECT SUM(CASE WHEN(VL.DrCr=1) THEN VL.Amount ELSE 0 END) FROM VoucherL VL WHERE VL.F_VoucherH = SH.Id AND VL.F_VoucherTypeMaster=' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '),0)
								  ) AS Diff									
							FROM ' + @TableNameH + ' SH
							LEFT JOIN LedgerMaster LM ON LM.Id=SH.F_LedgerMaster
							WHERE ' + @WHERE_INNER + '
						) AS TAB1
						WHERE ' + @WHERE + '
						ORDER BY VoucherDate,CAST(VoucherNo AS NUMERIC)'
		END
	ELSE IF(@Type=1)
		BEGIN		-- WRONG POSTING	 AND VL.F_LedgerMaster <> SH.F_LedgerMaster AND VL.DrCr=1
			SET @SQL = 'SELECT ' + @FIELDS + ' FROM (
							SELECT ' + @FIELDS_INNER + '
								,SUM(CASE WHEN(VL.DrCr=0) THEN (VL.Amount) ELSE 0 END) AS DrAmt
								,SUM(CASE WHEN(VL.DrCr=1) THEN (VL.Amount) ELSE 0 END) AS CrAmt
								,(SUM(CASE WHEN(VL.DrCr=0) THEN (VL.Amount) ELSE 0 END) - SUM(CASE WHEN(VL.DrCr=1) THEN (VL.Amount) ELSE 0 END)) AS Diff
							FROM ' + @TableNameH + ' SH
							LEFT JOIN VoucherL VL ON SH.Id = VL.F_VoucherH
							LEFT JOIN LedgerMaster LM ON LM.Id = SH.F_LedgerMaster
							WHERE ' + @WHERE_INNER2 + '
							GROUP BY ' + @GROUPBY_INNER + '
						) AS TAB1
						WHERE TAB1.Diff<>0'
		END
	ELSE IF(@Type=2)
		BEGIN		-- WRONG POSTING	 AND VL.F_LedgerMaster <> SH.F_LedgerMaster AND VL.DrCr=1
			SET @SQL = 'SELECT ' + @FIELDS + ' FROM (
							SELECT ' + @FIELDS_INNER + '
								,SUM(CASE WHEN(VL.DrCr=0) THEN (VL.Amount) ELSE 0 END) AS DrAmt
								,SUM(CASE WHEN(VL.DrCr=1) THEN (VL.Amount) ELSE 0 END) AS CrAmt
								,(SUM(CASE WHEN(VL.DrCr=0) THEN (VL.Amount) ELSE 0 END) - SUM(CASE WHEN(VL.DrCr=1) THEN (VL.Amount) ELSE 0 END)) AS Diff
							FROM ' + @TableNameH + ' SH
							LEFT JOIN VoucherL VL ON SH.Id = VL.F_VoucherH
							LEFT JOIN LedgerMaster LM ON LM.Id = SH.F_LedgerMaster
							WHERE ' + @WHERE_INNER + ' AND VL.F_LedgerMaster NOT IN(SELECT Id FROM LedgerMaster)
							GROUP BY ' + @GROUPBY_INNER + '
						) AS TAB1
						WHERE TAB1.Diff<>0'
		END
	EXEC(@SQL)
END
















































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_ProfitLoss]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO










CREATE PROCEDURE [dbo].[Select_Acc_ProfitLoss]	
	@DataSetName nvarchar(50)='',
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0),
	@F_LedgerGroupMaster numeric(18,0)=0,
	@F_LedgerMaster numeric(18,0)=0,
	@Type nvarchar(max)=' AND LGM.IsTrading=1',
	@Where nvarchar(max)='(TAB4.DebitAmount-TAB4.CreditAmount) >0',
	@AmountField nvarchar(max)='DebitAmount',
	@GroupBy nvarchar(max)='TAB4.LedgerGroup,TAB4.LedgerGroupId,TAB4.Ledger,TAB4.LedgerId,TAB4.Annuxure',
	@OrderBy nvarchar(max)='TAB5.LedgerGroup,TAB5.LedgerGroupId,TAB5.Ledger,TAB5.LedgerId',
	@IsBalanceSheet smallint=0,
	@OnlyOpening smallint=0
AS
BEGIN
--
	DECLARE @F_CompanyYears_Curr numeric(18,0)
	SET @F_CompanyYears_Curr = (SELECT TOP 1 Id FROM CompanyYears WHERE @FromDate >=FinYearFrom AND @ToDate<=FinYearTo)
--
	DECLARE @DefaultStockPostingLedger NUMERIC(18,0)
	SET @DefaultStockPostingLedger = (SELECT DefaultStockPostingLedger FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
	
	DECLARE @FinYearFromDate_First NVARCHAR(MAX)
	SET @FinYearFromDate_First = ISNULL((SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom),0)
	IF(@FinYearFromDate_First=0)
		SET @FinYearFromDate_First = ISNULL((SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster ORDER BY FinYearFrom),0)

-- QUERY FROM LEDGER
	DECLARE @FinYearFromDate NVARCHAR(MAX)
	DECLARE @FinYearFromDate_Min NVARCHAR(MAX)	
	SET @FinYearFromDate = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC)	
	SET @FinYearFromDate_Min = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)

--- Date Where Clause
DECLARE @DateWhere NVARCHAR(MAX)
DECLARE @DateWhere1 NVARCHAR(MAX)
DECLARE @Transaction NVARCHAR(MAX)
IF(@OnlyOpening=0)
	BEGIN
		SET @DateWhere = 'AND (VL.VoucherDate<=' + @ToDate + ' AND VL.VoucherDate>=' + @FinYearFromDate_First + ')'
		SET @DateWhere1 = 'AND (VL.VoucherDate<' + @FromDate + ' AND VL.VoucherDate>=ISNULL(' + @FinYearFromDate_Min + ',0))'
	END
ELSE
	BEGIN
		SET @DateWhere = 'AND (VL.VoucherDate<' + @ToDate + ' AND VL.VoucherDate>=' + @FinYearFromDate_First + ')'
		SET @DateWhere1 = 'AND (VL.VoucherDate<' + @FromDate + ' AND VL.VoucherDate>=' + @FinYearFromDate_First + ')'
	END
---

--Opening Stock (Starting Session)
	DECLARE @YearId NUMERIC(18,0)
	DECLARE @OpeningDate NVARCHAR(50)
	DECLARE @OpeningAmt NUMERIC(18,2)
	DECLARE @LedgerGroup NVARCHAR(50)
	DECLARE @LedgerGroupId NUMERIC(18,0)
	SET @YearId = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC)
	IF(@YearId>0)
		BEGIN
			SET @OpeningDate = ''
			SET @OpeningAmt = 0
			SET @LedgerGroup = ''
			SET @LedgerGroupId = 0
		END
	ELSE
		BEGIN
			SET @OpeningDate = (SUBSTRING(@FromDate,7,2)+'/'+SUBSTRING(@FromDate,5,2)+'/'+SUBSTRING(@FromDate,1,4))
			SET @OpeningAmt = (SELECT OpeningBalance FROM LedgerMaster WHERE Id = @DefaultStockPostingLedger)
			SET @LedgerGroup = (SELECT LGM.Name FROM LedgerMaster LM LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster = LGM.Id WHERE LM.Id = @DefaultStockPostingLedger)
			SET @LedgerGroupId = (SELECT LGM.Id FROM LedgerMaster LM LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster = LGM.Id WHERE LM.Id = @DefaultStockPostingLedger)
		END

--Opening Stock (Starting Session)		
	DECLARE @OpeningStock NVARCHAR(MAX)
	DECLARE @ClosingStock NVARCHAR(MAX)
	SET @OpeningStock = ''
	SET @ClosingStock = ''
	IF(@IsBalanceSheet=0 OR @Type=' AND LGM.IsTrading=1')
		BEGIN
			IF(@OpeningDate<>'')
				BEGIN
					SET @OpeningStock = '--OPENING STOCK
								SELECT 
										''' + @LedgerGroup + ''' AS LedgerGroup								 
										,' + CAST(@LedgerGroupId AS NVARCHAR) + ' AS LedgerGroupId
										,''OPENING STOCK'' AS Ledger
										,' + CAST(@DefaultStockPostingLedger AS NVARCHAR) + ' AS LedgerId
										,' + CAST(@OpeningAmt AS NVARCHAR) + ' AS DebitAmount
										,0 AS CreditAmount,LGM.Annuxure
								FROM LedgerGroupMaster LGM
								WHERE LGM.Id = ' + CAST(@LedgerGroupId AS NVARCHAR) + ' ' + @Type + '
					
								UNION ALL	'
				END
			ELSE
				BEGIN
					SET @OpeningStock = '--OPENING STOCK
						SELECT TOP 1 LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId,''OPENING STOCK'' AS Ledger,LM.Id AS LedgerId
								,VL.Amount AS DebitAmount
								,0 AS CreditAmount,LGM.Annuxure
						FROM VoucherL VL
							LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
							LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears_Curr AS NVARCHAR) + '
							LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster			
							LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
						WHERE ISNULL(VL.Cancelled,0)=0 ' + @DateWhere1 + ' AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR)+ ' ' + @Type + '
							AND 1=(CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR)+ '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR)+ '=LM.Id) THEN 1 ELSE 0 END) END)
							AND 1=(CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR)+ '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR)+ '=LGM.Id) THEN 1 ELSE 0 END) END)
							AND LM.F_LedgerGroupMaster=30
						ORDER BY VL.VoucherDate DESC
			
						UNION ALL	'
				END
		END
	IF(@IsBalanceSheet=1 AND @Type<>' AND LGM.IsTrading=1')
		BEGIN
			SET @ClosingStock = '--CLOSING STOCK
						SELECT TOP 1 LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId,''CLOSING STOCK'' AS Ledger,LM.Id AS LedgerId
								,VL.Amount AS DebitAmount
								,0 AS CreditAmount,LGM.Annuxure
						FROM VoucherL VL
							LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
							LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears_Curr AS NVARCHAR) + '
							LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster			
							LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
						WHERE ISNULL(VL.Cancelled,0)=0 ' + @DateWhere + ' AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + ' ' + @Type + '
							AND 1=(CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerMaster  AS NVARCHAR)+ '=LM.Id) THEN 1 ELSE 0 END) END)
							AND 1=(CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=LGM.Id) THEN 1 ELSE 0 END) END)
							AND LM.F_LedgerGroupMaster=30
						ORDER BY VL.VoucherDate DESC

						UNION ALL	'
		END
	ELSE
		BEGIN
			SET @ClosingStock = '--CLOSING STOCK
						SELECT TOP 1 LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId,''CLOSING STOCK'' AS Ledger,LM.Id AS LedgerId
								,0 AS DebitAmount
								,VL.Amount AS CreditAmount,LGM.Annuxure
						FROM VoucherL VL
							LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
							LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears_Curr AS NVARCHAR) + '	
							LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster			
							LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
						WHERE ISNULL(VL.Cancelled,0)=0 ' + @DateWhere + ' AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + ' ' + @Type + '
							AND 1=(CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=LM.Id) THEN 1 ELSE 0 END) END)
							AND 1=(CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=LGM.Id) THEN 1 ELSE 0 END) END)
							AND LM.F_LedgerGroupMaster=30
						ORDER BY VL.VoucherDate DESC
			
						UNION ALL	'
		END

-- OPENING BALANCE
	DECLARE @OpeningBalance NVARCHAR(MAX)	

	DECLARE @IsLedgerBalance_Split SMALLINT
	DECLARE @FinYearFromDate_Split NVARCHAR(MAX)
	DECLARE @F_CompanyYears_Split NUMERIC(18,0)
	--DECLARE @F_CompanyYears NUMERIC(18,0)
	SET @IsLedgerBalance_Split = (SELECT TOP 1 ISNULL(IsLedgerBalanceSplit,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @F_CompanyYears_Split = (SELECT TOP 1 ISNULL(Id,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @FinYearFromDate_Split = (SELECT TOP 1 ISNULL(FinYearFrom,0) FROM CompanyYears WHERE Id=@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
	--SET @F_CompanyYears = (SELECT TOP 1 Id FROM CompanyYears WHERE Id>@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster)
	
	IF(@IsLedgerBalance_Split=1)
		BEGIN
			SET @OpeningBalance ='SELECT TAB3.LedgerGroup,TAB3.LedgerGroupId,TAB3.Ledger,TAB3.LedgerId
											,SUM(TAB3.DebitAmount) AS DebitAmount
											,SUM(TAB3.CreditAmount) AS CreditAmount,TAB3.Annuxure
									FROM (
										SELECT LGM2.Name AS LedgerGroup,LGM2.Id AS LedgerGroupId,LM.Name AS Ledger,LM.Id AS LedgerId
											,SUM(CASE WHEN (LML.DrCr=0) THEN LML.OpeningBalance ELSE 0 END) AS DebitAmount
											,SUM(CASE WHEN (LML.DrCr=1) THEN LML.OpeningBalance ELSE 0 END) AS CreditAmount,LGM2.Annuxure
										FROM LedgerMaster LM
										LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears_Split AS NVARCHAR) + '
										LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
										LEFT JOIN LedgerMasterL LML2 ON LML2.F_LedgerMaster = LM.Id AND LML2.F_CompanyYears = ' + CAST(@F_CompanyYears_Curr AS NVARCHAR) + '
										LEFT JOIN LedgerGroupMaster LGM2 ON LGM2.Id=LML2.F_LedgerGroupMaster
										WHERE LM.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + ' ' + @Type + '
											AND 1=(CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=LM.Id) THEN 1 ELSE 0 END) END)
											AND 1=(CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=LGM.Id) THEN 1 ELSE 0 END) END)
											AND LM.F_LedgerGroupMaster<>30 AND ISNULL(' + CAST(@FinYearFromDate_Min AS NVARCHAR) + ',0)>0
										GROUP BY LGM.Name,LGM.Id,LM.Name,LM.Id,LGM.Annuxure,LGM2.Name,LGM2.Id,LGM2.Annuxure
										UNION ALL
										--Transaction
										SELECT LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId,LM.Name AS Ledger,LM.Id AS LedgerId
												,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount
												,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LGM.Annuxure
										FROM VoucherL VL
											LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
											LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears_Curr AS NVARCHAR) + '
											LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
											LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
										WHERE ISNULL(VL.Cancelled,0)=0 ' + @Type + '
											AND 1=(CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=LM.Id) THEN 1 ELSE 0 END) END)
											AND 1=(CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=LGM.Id) THEN 1 ELSE 0 END) END)
											AND (VL.VoucherDate<' + @FromDate + ' AND VL.VoucherDate>=' + @FinYearFromDate_Split + ') AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + '
											AND LM.F_LedgerGroupMaster<>30 AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=' + @FinYearFromDate + ' ))
										GROUP BY LGM.Name,LGM.Id,LM.Name,LM.Id,LGM.Annuxure
									) AS TAB3
									WHERE (TAB3.DebitAmount-TAB3.CreditAmount) <>0
									GROUP BY TAB3.LedgerGroup,TAB3.LedgerGroupId,TAB3.Ledger,TAB3.LedgerId,TAB3.Annuxure '
		END
	ELSE
		BEGIN
			SET @OpeningBalance ='SELECT TAB3.LedgerGroup,TAB3.LedgerGroupId,TAB3.Ledger,TAB3.LedgerId
											,SUM(TAB3.DebitAmount) AS DebitAmount
											,SUM(TAB3.CreditAmount) AS CreditAmount,TAB3.Annuxure
									FROM (
										SELECT LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId,LM.Name AS Ledger,LM.Id AS LedgerId
												,SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount
												,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount,LGM.Annuxure
										FROM LedgerMaster LM
										LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears_Curr AS NVARCHAR) + '
										LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
										WHERE LM.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + ' ' + @Type + '
											AND 1=(CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=LM.Id) THEN 1 ELSE 0 END) END)
											AND 1=(CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=LGM.Id) THEN 1 ELSE 0 END) END)
											AND LM.F_LedgerGroupMaster<>30 AND ISNULL(' + @FinYearFromDate_Min + ',0)>0
										GROUP BY LGM.Name,LGM.Id,LM.Name,LM.Id,LGM.Annuxure
										UNION ALL
										--Transaction
										SELECT LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId,LM.Name AS Ledger,LM.Id AS LedgerId
												,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount
												,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LGM.Annuxure
										FROM VoucherL VL
											LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
											LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears_Curr AS NVARCHAR) + '
											LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
											LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
										WHERE ISNULL(VL.Cancelled,0)=0 ' + @Type + '
											AND 1=(CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=LM.Id) THEN 1 ELSE 0 END) END)
											AND 1=(CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=LGM.Id) THEN 1 ELSE 0 END) END)
											' + @DateWhere1 + ' AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + '
											AND LM.F_LedgerGroupMaster<>30 AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=' + @FinYearFromDate + ' AND VL.VoucherDate>=' + @FinYearFromDate_First + '))
										GROUP BY LGM.Name,LGM.Id,LM.Name,LM.Id,LGM.Annuxure
									) AS TAB3
									WHERE (TAB3.DebitAmount-TAB3.CreditAmount) <>0
									GROUP BY TAB3.LedgerGroup,TAB3.LedgerGroupId,TAB3.Ledger,TAB3.LedgerId,TAB3.Annuxure '
		END

-- Transaction
IF(@OnlyOpening = 0)
	SET @Transaction = '
		UNION ALL
		
		--Transaction
		SELECT LGM.Name AS LedgerGroup,LGM.Id AS LedgerGroupId,LM.Name AS Ledger,LM.Id AS LedgerId
				,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount
				,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LGM.Annuxure
		FROM VoucherL VL
			LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
			LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears_Curr AS NVARCHAR) + '	
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster			
			LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
		WHERE ISNULL(VL.Cancelled,0)=0 AND (VL.VoucherDate>=' + @FromDate + ' AND VL.VoucherDate<=' + @ToDate + ') AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR)+ ' ' + @Type + '
			AND 1=(CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR)+ '=LM.Id) THEN 1 ELSE 0 END) END)
			AND 1=(CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=LGM.Id) THEN 1 ELSE 0 END) END)
			AND LM.F_LedgerGroupMaster<>30
		GROUP BY LGM.Name,LGM.Id,LM.Name,LM.Id,LGM.Annuxure'
ELSE
	SET @Transaction=''
--
	EXEC('
		SELECT ' + @DataSetName + 'TAB5.LedgerGroup,TAB5.LedgerGroupId,TAB5.Ledger,TAB5.LedgerId,TAB5.DrCr,TAB5.' + @AmountField + ',TAB5.Annuxure
		FROM (	
			SELECT TAB4.LedgerGroup,TAB4.LedgerGroupId
					,MAX(TAB4.Ledger) AS Ledger,MAX(TAB4.LedgerId) AS LedgerId
					,(CASE WHEN ((SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount))>=0) THEN 0 ELSE 1 END) AS DrCr
					,ABS(SUM(TAB4.DebitAmount)-SUM(TAB4.CreditAmount)) AS ' + @AmountField + '
					,SUM(TAB4.DebitAmount) AS DrAmt
					,SUM(TAB4.CreditAmount) AS CrAmt,TAB4.Annuxure
			FROM(
			' + @OpeningStock + @ClosingStock + '		
				--Opening
				' + @OpeningBalance + '
				' + @Transaction + '
			) AS TAB4
			GROUP BY ' + @GroupBy + '
		) AS TAB5
		WHERE ' + @Where + ' AND (TAB5.DrAmt>0 OR TAB5.CrAmt>0)
		ORDER BY ' + @OrderBy
	)
END
















GO
```

# [dbo].[Select_Acc_ProfitLoss_Cr]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_ProfitLoss_Cr]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO












































































CREATE PROCEDURE [dbo].[Select_Acc_ProfitLoss_Cr] 
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0)
AS
BEGIN	
SELECT TAB3.LedgerGroup,TAB3.Ledger,ABS(TAB3.DebitAmount-TAB3.CreditAmount) AS DebitAmount,(SUBSTRING(@ToDate,7,2)+'/'+SUBSTRING(@ToDate,5,2)+'/'+SUBSTRING(@ToDate,1,4)) AS ToDate
FROM(		
	SELECT TAB2.LedgerGroup,TAB2.Ledger
		,SUM(TAB2.DebitAmount) AS DebitAmount,SUM(TAB2.CreditAmount) AS CreditAmount
	FROM(		
		--Opening
		SELECT LGM.Name AS LedgerGroup, LM.Name AS Ledger		
			,(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount
			,(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount					
		FROM LedgerMaster LM
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		WHERE LM.F_CompanyMaster=@F_CompanyMaster
		UNION ALL
		--Transaction
		SELECT LGM.Name AS LedgerGroup,LM.Name AS Ledger
			,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount
			,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount			
		FROM (
				
				SELECT VH.VoucherNo,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate
				,LM.Name AS Ledger,VH.Id
				,LM.Id AS LedgerId
				,VL.DrCr,VL.Amount,VTM.Name AS VoucherType
				FROM [dbo].[VoucherH] VH	
					LEFT JOIN VoucherL VL ON VL.F_VoucherH=VH.Id
					LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id				
					LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VH.F_VoucherTypeMaster
				WHERE (VH.VoucherDate<=@ToDate ) AND VH.F_CompanyMaster=@F_CompanyMaster
					
		) AS TAB1
		LEFT JOIN VoucherL VL ON VL.F_VoucherH=TAB1.Id
		LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		WHERE VL.F_LedgerMaster!=TAB1.LedgerId AND LGM.IsProfitLoss =1
		GROUP BY LGM.Name,LM.Name,LM.Id
	) AS TAB2
	GROUP BY TAB2.LedgerGroup,TAB2.Ledger
) AS TAB3
WHERE (TAB3.CreditAmount-TAB3.DebitAmount)>0
ORDER BY TAB3.LedgerGroup,TAB3.Ledger
END












































































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_ProfitLossDetails]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO













CREATE PROCEDURE [dbo].[Select_Acc_ProfitLossDetails] 
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0),
	@F_LedgerGroupMaster numeric(18,0)=0,
	@F_LedgerMaster numeric(18,0)=0
AS
BEGIN
	--DataSet1
	--@FromDate nvarchar(50)='',
	--@ToDate nvarchar(50),
	--@F_CompanyMaster numeric(18,0),
	--@F_LedgerGroupMaster numeric(18,0)=0,
	--@F_LedgerMaster numeric(18,0)=0,
	--@Type nvarchar(max)=' AND LGM.IsTrading=1',
	--@Where nvarchar(max)='(TAB4.DebitAmount-TAB4.CreditAmount) >0',
	--@AmountField nvarchar(max)='DebitAmount',
	--@GroupBy nvarchar(max)=,
	--@IsBalanceSheet smallint=0,
	--@DataSetName nvarchar(50)=''
	--TAB5.LedgerGroup,TAB5.LedgerGroupId,TAB5.Ledger,TAB5.LedgerId,TAB5.DrCr,TAB5.' + @AmountField + '
	DELETE FROM TMPProfitLoss
	
	INSERT INTO TMPProfitLoss(DataSetName,LedgerGroup,LedgerGroupId,Ledger,LedgerId,DrCr,DebitAmount,Annuxure)
	EXEC dbo.Select_Acc_ProfitLoss '''DataSet1'' AS DataSetName,',@FromDate,@ToDate,@F_CompanyMaster,@F_LedgerGroupMaster,@F_LedgerMaster,' AND LGM.IsTrading=1','TAB5.DrCr=0','DebitAmount'

	--DataSet2
	INSERT INTO TMPProfitLoss(DataSetName,LedgerGroup,LedgerGroupId,Ledger,LedgerId,DrCr,CreditAmount,Annuxure)
	EXEC dbo.Select_Acc_ProfitLoss '''DataSet2'' AS DataSetName,',@FromDate,@ToDate,@F_CompanyMaster,@F_LedgerGroupMaster,@F_LedgerMaster,' AND LGM.IsTrading=1','TAB5.DrCr=1','CreditAmount'

	--DataSet3
	INSERT INTO TMPProfitLoss(DataSetName,LedgerGroup,LedgerGroupId,Ledger,LedgerId,DrCr,DebitAmount,Annuxure)
	EXEC dbo.Select_Acc_ProfitLoss '''DataSet3'' AS DataSetName,',@FromDate,@ToDate,@F_CompanyMaster,@F_LedgerGroupMaster,@F_LedgerMaster,' AND LGM.IsProfitLoss=1','TAB5.DrCr=0','DebitAmount'

	--DataSet4
	INSERT INTO TMPProfitLoss(DataSetName,LedgerGroup,LedgerGroupId,Ledger,LedgerId,DrCr,CreditAmount,Annuxure)
	EXEC dbo.Select_Acc_ProfitLoss '''DataSet4'' AS DataSetName,',@FromDate,@ToDate,@F_CompanyMaster,@F_LedgerGroupMaster,@F_LedgerMaster,' AND LGM.IsProfitLoss=1','TAB5.DrCr=1','CreditAmount'


	--Gross Profit (B/F)
	DECLARE @GrossProfit NUMERIC(18,2)
	SET @GrossProfit = (SELECT (CASE WHEN(SUM(CreditAmount)-SUM(DebitAmount)>=0) THEN SUM(CreditAmount)-SUM(DebitAmount) ELSE 0 END) AS GrossProfit FROM TMPProfitLoss WHERE DataSetName IN('DataSet1','DataSet2'))

	--Net Profit (B/F
	DECLARE @NetProfit NUMERIC(18,2)
	SET @NetProfit = (SELECT (CASE WHEN(@GrossProfit>0) THEN
								(CASE WHEN(@GrossProfit + SUM(CreditAmount)-SUM(DebitAmount)>0) THEN (@GrossProfit + SUM(CreditAmount)-SUM(DebitAmount)) ELSE 0 END)
							 ELSE 0 END) AS NetProfit
					  FROM TMPProfitLoss WHERE DataSetName IN('DataSet3','DataSet4'))

	--Gross Loss (B/F)
	DECLARE @GrossLoss NUMERIC(18,2)
	SET @GrossLoss = (SELECT (CASE WHEN(SUM(CreditAmount)-SUM(DebitAmount)<0) THEN SUM(DebitAmount)-SUM(CreditAmount) ELSE 0 END) AS GrossLoss FROM TMPProfitLoss WHERE DataSetName IN('DataSet1','DataSet2'))

	--Net Loss (B/F)
	DECLARE @NetLoss NUMERIC(18,2)
	SET @NetLoss = (SELECT (CASE WHEN(@GrossLoss>0) THEN
								(CASE WHEN(@GrossLoss + SUM(DebitAmount) - SUM(CreditAmount)	>0) THEN (@GrossLoss + SUM(DebitAmount) - SUM(CreditAmount)) ELSE 0 END)
						   ELSE 0 END) AS NetLoss
					FROM TMPProfitLoss WHERE DataSetName IN('DataSet3','DataSet4'))

	IF(@NetProfit>0)
		BEGIN
			SELECT 'GROSS PROFIT' AS HEAD,@GrossProfit AS GrossProfit
			UNION ALL
			SELECT 'NET PROFIT' AS HEAD,@NetProfit AS NetProfit
		END
	ELSE
		BEGIN
			SELECT 'GROSS LOSS' AS HEAD,@GrossLoss AS GrossLoss
			UNION ALL
			SELECT 'NET LOSS' AS HEAD,@NetLoss AS NetLoss
		END
END

























GO
```

# [dbo].[Select_Acc_ProfitLossPost]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_ProfitLossPost]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO













CREATE PROCEDURE [dbo].[Select_Acc_ProfitLossPost]
	@F_CompanyMaster NVARCHAR(50),
	@F_CompanyYears NUMERIC(18,0),
	@Total smallint=0
AS
BEGIN
	IF(@Total=0)
		BEGIN
			SELECT LM.Name AS Ledger,LM.Id AS LedgerId,NL.IntRate,NL.Amount,NL.Id
				,POST.F_VoucherH,POST.VoucherNo,(SUBSTRING(POST.VoucherDate,7,2)+'/'+SUBSTRING(POST.VoucherDate,5,2)+'/'+SUBSTRING(POST.VoucherDate,1,4)) AS VoucherDate
			FROM NetProfitL NL
			LEFT JOIN LedgerMaster LM ON LM.Id = NL.F_LedgerMaster
				OUTER APPLY(
					SELECT TOP 1
						VL.VoucherNo,VL.F_VoucherH,VL.VoucherDate,VL.Amount
					FROM VoucherL VL
						LEFT JOIN VoucherH VH ON VH.Id = VL.F_VoucherH
						LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster = VH.F_CompanyMaster AND CY.Id=@F_CompanyYears
					WHERE VL.F_LedgerMaster = LM.Id AND VL.F_CompanyMaster= 1 AND VL.F_VoucherTypeMaster = 4 AND VH.F_LedgerMaster IN(-1,-2)
					AND (VH.VoucherDate>=CY.FinYearFrom AND VH.VoucherDate<=CY.FinYearTo)
					AND ISNULL(VL.Cancelled,0)=0
				) POST
			WHERE NL.F_CompanyMaster=1 AND NL.F_CompanyYears=@F_CompanyYears
		END
	ELSE
		BEGIN
			SELECT SUM(NL.Amount) AS NetProfit_Posted
			FROM NetProfitL NL
				LEFT JOIN LedgerMaster LM ON LM.Id = NL.F_LedgerMaster
			WHERE NL.F_CompanyMaster=@F_CompanyMaster AND NL.F_CompanyYears=@F_CompanyYears
		END
END










GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_StockInHand]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO









CREATE PROCEDURE [dbo].[Select_Acc_StockInHand]
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0),
	@F_LedgerGroupMaster numeric(18,0),
	@F_LedgerMaster numeric(18,0)=0,
	@Cancelled smallint=-1,
	@F_VoucherTypeMaster NUMERIC(18,0)=0,
	@EditPassword nvarchar(50)='',
	@F_LedgerMasterAg NUMERIC(18,0)=0
AS
BEGIN
	-- OPENING BALANCE
	DECLARE @IsLedgerBalance_Split SMALLINT
	DECLARE @FinYearFromDate_Split NVARCHAR(MAX)
	DECLARE @F_CompanyYears_Split NUMERIC(18,0)
	DECLARE @F_CompanyYears2 NUMERIC(18,0)
	SET @IsLedgerBalance_Split = (SELECT TOP 1 ISNULL(IsLedgerBalanceSplit,0) FROM CompanyYears WHERE FinYearFrom<@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @F_CompanyYears_Split = (SELECT TOP 1 ISNULL(Id,0) FROM CompanyYears WHERE FinYearFrom<@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @FinYearFromDate_Split = (SELECT TOP 1 ISNULL(FinYearFrom,0) FROM CompanyYears WHERE Id>@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
	SET @F_CompanyYears2 = (SELECT TOP 1 Id FROM CompanyYears WHERE Id>@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster)
	--

	--DECLARE @YearId NUMERIC(18,0)
	DECLARE @OpeningDate NVARCHAR(50)
	DECLARE @OpeningAmt NUMERIC(18,2)			
	--SET @YearId = (SELECT FinYearFrom FROM CompanyYears WHERE FinYearFrom<@FromDate AND ISNULL(IsAdditional,0)=0)
	--IF(@F_CompanyYears_Split>0)
	--	BEGIN
	--		SET @OpeningDate = ''
	--		SET @OpeningAmt = 0
	--	END
	--ELSE
	--	BEGIN
			SET @OpeningDate = (SUBSTRING(@FromDate,7,2)+'/'+SUBSTRING(@FromDate,5,2)+'/'+SUBSTRING(@FromDate,1,4))
			SET @OpeningAmt = ( SELECT LML.OpeningBalance
								FROM LedgerMaster LM
									LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears2
								WHERE LM.Id = @F_LedgerMaster)
		--END
			
	DECLARE @TransId NUMERIC(18,0)
	SET @TransId = (SELECT TOP 1 Id FROM VoucherH VH WHERE VH.F_CompanyMaster = @F_CompanyMaster AND VH.F_LedgerMaster = @F_LedgerMaster
						AND VH.VoucherDate>=@FromDate AND VH.VoucherDate <=@ToDate
						AND 1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END))
	IF(@TransId>0)
		BEGIN	
			SELECT VH.Id
				,VH.VoucherNo
				,(CASE WHEN(ISNULL(STOCKOP.OpeningDate,'')='') THEN @OpeningDate ELSE STOCKOP.OpeningDate END) AS OpeningDate				
				,(CASE WHEN(ISNULL(STOCKOP.OpeningAmt,0)=0) THEN @OpeningAmt ELSE STOCKOP.OpeningAmt END) AS OpeningAmt
				,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS ClosingDate
				,VH.TotalNetAmt AS ClosingAmt
				,(CASE WHEN (ISNULL(VH.Cancelled,0)=1) THEN 'Cancelled' ELSE '' END) AS Cancelled
			FROM VoucherH VH
			LEFT JOIN VoucherL VL ON VH.Id = VL.F_VoucherH
			LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
			LEFT JOIN ManagerMaster MM ON MM.Id=VH.F_ManagerMaster
			OUTER APPLY(
					SELECT TOP 1 VH1.TotalNetAmt AS OpeningAmt,
						('01/' + SUBSTRING(@FromDate,5,2)+'/'+SUBSTRING(VH1.VoucherDate,1,4)) AS OpeningDate
					FROM VoucherH VH1
					WHERE VH1.VoucherDate<@FromDate
					AND VH1.F_LedgerMaster=VH.F_LedgerMaster
					AND VH1.F_CompanyMaster = @F_CompanyMaster
					AND
					1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
							
					ORDER BY VH1.VoucherDate DESC
			) STOCKOP
			WHERE 1=CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LM.F_LedgerGroupMaster=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END
				AND VH.F_LedgerMaster = @F_LedgerMaster
				AND VH.F_CompanyMaster = @F_CompanyMaster
				AND VH.VoucherDate>=@FromDate AND VH.VoucherDate <=@ToDate
				AND 1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(VH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
			ORDER BY VH.VoucherDate,CAST(VH.VoucherNo AS NUMERIC)
		END
	ELSE
		BEGIN	
			SELECT 0 AS Id
				,'' AS VoucherNo
				,@OpeningDate AS OpeningDate
				,@OpeningAmt AS OpeningAmt
				,'' AS ClosingDate
				,0 AS ClosingAmt
				,'' AS Cancelled
			FROM LedgerMaster LM
			WHERE 1=CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(LM.F_LedgerGroupMaster=@F_LedgerGroupMaster) THEN 1 ELSE 0 END) END
				AND LM.Id = @F_LedgerMaster
				AND LM.F_CompanyMaster = @F_CompanyMaster
		END
END

GO
```

# [dbo].[Select_Acc_Trading_Cr]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Trading_Cr]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO












































































CREATE PROCEDURE [dbo].[Select_Acc_Trading_Cr] 
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0)
AS
BEGIN			
	SELECT TAB2.LedgerGroup,TAB2.Ledger,ABS(TAB2.DebitAmount-TAB2.CreditAmount) AS CreditAmount,TAB2.ToDate FROM(		
		SELECT LGM.Name AS LedgerGroup,LM.Name AS Ledger,
			(
				SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)
				+
				(CASE WHEN (MAX(LM.DrCr)=0) THEN MAX(LM.OpeningBalance) ELSE 0 END)
			) AS DebitAmount,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)
			+
			(CASE WHEN (SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)=0) THEN
			MAX(LM.OpeningBalance) ELSE 0 END)
			) AS CreditAmount,
						
			(SUBSTRING(@ToDate,7,2)+'/'+SUBSTRING(@ToDate,5,2)+'/'+SUBSTRING(@ToDate,1,4)) AS ToDate
		FROM (
				
				SELECT VH.VoucherNo,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate
				,LM.Name AS Ledger,VH.Id
				,LM.Id AS LedgerId
				,VL.DrCr,VL.Amount,VTM.Name AS VoucherType
				FROM [dbo].[VoucherH] VH	
					LEFT JOIN VoucherL VL ON VL.F_VoucherH=VH.Id
					LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id				
					LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VH.F_VoucherTypeMaster
				WHERE (VH.VoucherDate<=@ToDate ) AND VH.F_CompanyMaster=@F_CompanyMaster
					
		) AS TAB1
		LEFT JOIN VoucherL VL ON VL.F_VoucherH=TAB1.Id
		LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		WHERE VL.F_LedgerMaster!=TAB1.LedgerId AND LGM.IsTrading =1
		GROUP BY LGM.Name,LM.Name,LM.Id
	) AS TAB2
	WHERE (TAB2.CreditAmount-TAB2.DebitAmount)>0
	ORDER BY TAB2.LedgerGroup,TAB2.Ledger
END












































































GO
```

# [dbo].[Select_Acc_Trading_Dr]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_Trading_Dr]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO












































































CREATE PROCEDURE [dbo].[Select_Acc_Trading_Dr] 	
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0)
AS
BEGIN			
	SELECT TAB2.LedgerGroup,TAB2.Ledger,ABS(TAB2.DebitAmount-TAB2.CreditAmount) AS DebitAmount,TAB2.ToDate FROM(		
		SELECT LGM.Name AS LedgerGroup,LM.Name AS Ledger
			,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)
			+
			(CASE WHEN (SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)=0) THEN
			MAX(LM.OpeningBalance) ELSE 0 END)
			) AS DebitAmount,
			(
				SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)
				+
				(CASE WHEN (MAX(LM.DrCr)=1) THEN MAX(LM.OpeningBalance) ELSE 0 END)
			) AS CreditAmount,
						
			(SUBSTRING(@ToDate,7,2)+'/'+SUBSTRING(@ToDate,5,2)+'/'+SUBSTRING(@ToDate,1,4)) AS ToDate
		FROM (
				
				SELECT VH.VoucherNo,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate
				,LM.Name AS Ledger,VH.Id
				,LM.Id AS LedgerId
				,VL.DrCr,VL.Amount,VTM.Name AS VoucherType
				FROM [dbo].[VoucherH] VH	
					LEFT JOIN VoucherL VL ON VL.F_VoucherH=VH.Id
					LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id				
					LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VH.F_VoucherTypeMaster
				WHERE (VH.VoucherDate<=@ToDate )
					AND VH.F_CompanyMaster=@F_CompanyMaster
					
		) AS TAB1
		LEFT JOIN VoucherL VL ON VL.F_VoucherH=TAB1.Id
		LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
		WHERE VL.F_LedgerMaster!=TAB1.LedgerId AND LGM.IsTrading=1
		GROUP BY LGM.Name,LM.Name,LM.Id
	) AS TAB2
	WHERE (TAB2.DebitAmount-TAB2.CreditAmount) >0
	ORDER BY TAB2.LedgerGroup,TAB2.Ledger
END












































































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_TrialBalance]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO














CREATE PROCEDURE [dbo].[Select_Acc_TrialBalance]
	@F_LedgerGroupMaster numeric(18,0)=0,
	@F_LedgerMaster numeric(18,0)=0,
	@F_CityMaster numeric(18,0)=0,
	@F_AreaMaster numeric(18,0)=0,
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0),
	@Options smallint=1,
	@BalanceStatus smallint=0,
	@TaxType smallint=-1,
	@IsPostingView smallint=0,
	@IsExport smallint=-1,
	@F_VoucherTypeMaster numeric(18,0)=0,
	@VoucherType smallint=-1,
	@IsReverseChargesApplicable smallint=-1,
	@IsTaxation smallint=-1
AS
BEGIN
--
	DECLARE @F_CompanyYears numeric(18,0)
	SET @F_CompanyYears = (SELECT TOP 1 Id FROM CompanyYears WHERE @FromDate >=FinYearFrom AND @ToDate<=FinYearTo)
--Posting View
	DECLARE @POSTING_FIELDS NVARCHAR(MAX)
	DECLARE @POSTING_JOIN NVARCHAR(MAX)
	IF(@IsPostingView=1)
		BEGIN
			SET @POSTING_FIELDS = ',POST.Amount AS PostingAmt,0 AS Diff,POST.F_VoucherH,POST.VoucherNo,(SUBSTRING(POST.VoucherDate,7,2)+''/''+SUBSTRING(POST.VoucherDate,5,2)+''/''+SUBSTRING(POST.VoucherDate,1,4)) AS VoucherDate'
			SET @POSTING_JOIN = ' OUTER APPLY(SELECT TOP 1
											VL.VoucherNo,VL.F_VoucherH,VL.VoucherDate,VL.Amount
										FROM VoucherL VL
											LEFT JOIN VoucherH VH ON VH.Id = VL.F_VoucherH
										WHERE VL.VoucherDate>=' + @FromDate + ' AND VL.VoucherDate<=' + @ToDate + ' AND VL.F_LedgerMaster = TAB5.LedgerId
												AND VL.F_CompanyMaster= ' + CAST(@F_CompanyMaster AS nvarchar) + ' AND VL.F_VoucherTypeMaster = 4 AND VH.F_LedgerMaster IN(-1,-2)
								  ) POST'
		END
	ELSE
		BEGIN
			SET @POSTING_FIELDS = ',0 AS PostingAmt,0 AS Diff,0 AS F_VoucherH,'''' AS VoucherNo,'''' AS VoucherDate'
			SET @POSTING_JOIN = ''
		END
	DECLARE @DefaultStockPostingLedger NUMERIC(18,0)
	SET @DefaultStockPostingLedger = (SELECT DefaultStockPostingLedger FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
--Opening Stock (Starting Session)
	DECLARE @YearId NUMERIC(18,0)
	DECLARE @OpeningDate NVARCHAR(50)
	DECLARE @OpeningAmt NUMERIC(18,2)
	DECLARE @LedgerGroup NVARCHAR(50)
	DECLARE @Annuxure NVARCHAR(50)
	DECLARE @LedgerGroupId NUMERIC(18,0)
	SET @YearId = ISNULL((SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC),0)
	IF(@YearId>0)
		BEGIN
			SET @OpeningDate = ''
			SET @OpeningAmt = 0
			SET @LedgerGroup = ''
			SET @LedgerGroupId = 0
			SET @Annuxure = ''
		END
	ELSE
		BEGIN
			SET @OpeningDate = (SUBSTRING(@FromDate,7,2)+'/'+SUBSTRING(@FromDate,5,2)+'/'+SUBSTRING(@FromDate,1,4))
			SET @OpeningAmt = (SELECT OpeningBalance FROM LedgerMaster WHERE Id = @DefaultStockPostingLedger)
			SET @LedgerGroup =
					(SELECT LGM.Name
						FROM LedgerMaster LM
							LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears
							LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
						WHERE LM.Id = @DefaultStockPostingLedger)
			SET @LedgerGroupId = 
					(SELECT LGM.Id
						FROM LedgerMaster LM
							LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears
							LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
						WHERE LM.Id = @DefaultStockPostingLedger)
			SET @Annuxure = 
					(SELECT ISNULL(LGM.Annuxure,'') AS Annuxure
						FROM LedgerMaster LM
							LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = @F_CompanyYears
							LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
						WHERE LM.Id = @DefaultStockPostingLedger)
		END
--Opening Stock (Starting Session)
	DECLARE @OpeningStock NVARCHAR(MAX)
	DECLARE @ClosingStock NVARCHAR(MAX)
	SET @OpeningStock = ''
	SET @ClosingStock = ''
	IF(@OpeningDate<>'')
		BEGIN
			SET @OpeningStock = '--OPENING STOCK
						SELECT 
								''' + @LedgerGroup + ''' AS LedgerGroup			 
								,' + CAST(@LedgerGroupId AS NVARCHAR) + ' AS GroupId
								,''OPENING STOCK'' AS Ledger
								,' + CAST(@DefaultStockPostingLedger AS NVARCHAR) + ' AS LedgerId
								,' + CAST(@OpeningAmt AS NVARCHAR) + ' AS OpBal
								,0 AS DebitAmount,0 AS CreditAmount
								,''' + @Annuxure + ''' AS Annuxure,(CASE WHEN(LGM.DrCr=0) THEN ''Dr'' ELSE ''Cr'' END) AS DefaultDrCr
								,'''' AS TransferTo,0 AS TaxRate
						FROM LedgerGroupMaster LGM
						WHERE LGM.Id = ' + CAST(@LedgerGroupId AS NVARCHAR)+ '
						UNION ALL	'
		END
	ELSE
		BEGIN
			SET @OpeningStock = '--OPENING STOCK
						SELECT TOP 1 LGM.Name AS LedgerGroup,LGM.Id AS GroupId,''OPENING STOCK'' AS Ledger,LM.Id AS LedgerId
								,VL.Amount AS OpBal
								,0 AS DebitAmount,0 AS CreditAmount,LGM.Annuxure,(CASE WHEN(LGM.DrCr=0) THEN ''Dr'' ELSE ''Cr'' END) AS DefaultDrCr
								,'''' AS TransferTo,LM.TaxRate
						FROM VoucherL VL
							LEFT JOIN VoucherH VH ON VH.Id = VL.F_VoucherH
							LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
							LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears AS NVARCHAR) + '
							LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
							LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
						WHERE ISNULL(VL.Cancelled,0)=0 AND (VL.VoucherDate<' + @FromDate + ') AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + '
							AND 1=(CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=LM.F_CityMaster) THEN 1 ELSE 0 END) END)
							AND 1=(CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=LM.F_AreaMaster) THEN 1 ELSE 0 END) END)
							AND 1=(CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=LM.Id) THEN 1 ELSE 0 END) END)
							AND 1=(CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=LGM.Id) THEN 1 ELSE 0 END) END)
							AND 1=(CASE WHEN(' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '=VTM.Id) THEN 1 ELSE 0 END) END)							
							AND 1=(CASE WHEN(' + CAST(@VoucherType AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(' + CAST(@VoucherType AS NVARCHAR) + '=ISNULL(VTM.VoucherType,0)) THEN 1 ELSE 0 END) END)							
							AND 1=(CASE WHEN(' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsReverseChargesApplicable,0)=' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)														
							AND 1=(CASE WHEN(' + CAST(@IsTaxation AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(VH.IsTaxation,0)=' + CAST(@IsTaxation AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)				
							AND LM.F_LedgerGroupMaster=30
						ORDER BY VL.VoucherDate DESC
			
						UNION ALL	'
		END		
--	
-- GROUP BY
	DECLARE @GroupBy1 NVARCHAR(MAX)
	DECLARE @GroupBy2 NVARCHAR(MAX)
	DECLARE @GroupBy3 NVARCHAR(MAX)
	DECLARE @OrderBy NVARCHAR(MAX)
	IF(@Options=0)
		BEGIN
			SET @GroupBy1 = 'GROUP BY LGM.Name,LGM.Id,LGM.Annuxure,LGM.DrCr,LGM.IsTrading,LGM.IsProfitLoss,LGM.IsBalanceSheet'
			SET @GroupBy2 = 'GROUP BY TAB3.LedgerGroup,TAB3.GroupId,TAB3.Annuxure,TAB3.DefaultDrCr,TAB3.TransferTo'
			SET @GroupBy3 = 'GROUP BY TAB4.LedgerGroup,TAB4.GroupId,TAB4.Annuxure,TAB4.DefaultDrCr,TAB4.TransferTo'
			SET @OrderBy = 'ORDER BY TAB5.LedgerGroup'
		END
	ELSE IF(@Options=1)
		BEGIN
			SET @GroupBy1 = 'GROUP BY LM.Name,LM.Id,LGM.Name,LGM.Id,LGM.Annuxure,LGM.DrCr,LGM.IsTrading,LGM.IsProfitLoss,LGM.IsBalanceSheet,LM.TaxRate'
			SET @GroupBy2 = 'GROUP BY TAB3.LedgerGroup,TAB3.GroupId,TAB3.Ledger,TAB3.LedgerId,TAB3.Annuxure,TAB3.DefaultDrCr,TAB3.TransferTo,TAB3.TaxRate'
			SET @GroupBy3 = 'GROUP BY TAB4.Ledger,TAB4.LedgerId,TAB4.LedgerGroup,TAB4.GroupId,TAB4.Annuxure,TAB4.DefaultDrCr,TAB4.TransferTo,TAB4.TaxRate'
			SET @OrderBy = 'ORDER BY TAB5.LedgerGroup,TAB5.Ledger'
		END		
	ELSE IF(@Options=2)
		BEGIN
			SET @GroupBy1 = 'GROUP BY LM.Name,LM.Id,LGM.Name,LGM.Id,LGM.Annuxure,LGM.DrCr,LGM.IsTrading,LGM.IsProfitLoss,LGM.IsBalanceSheet,LM.TaxRate'
			SET @GroupBy2 = 'GROUP BY TAB3.LedgerGroup,TAB3.GroupId,TAB3.Ledger,TAB3.LedgerId,TAB3.Annuxure,TAB3.DefaultDrCr,TAB3.TransferTo,TAB3.TaxRate'
			SET @GroupBy3 = 'GROUP BY TAB4.Ledger,TAB4.LedgerId,TAB4.LedgerGroup,TAB4.GroupId,TAB4.Annuxure,TAB4.DefaultDrCr,TAB4.TransferTo,TAB4.TaxRate'
			SET @OrderBy = 'ORDER BY TAB5.LedgerGroup,TAB5.Ledger'
		END
--
-- OPENING BALANCE
	DECLARE @OpeningBalance NVARCHAR(MAX)	
	DECLARE @FinYearFromDate NVARCHAR(MAX)
	DECLARE @FinYearFromDate_Min NVARCHAR(MAX)
	SET @FinYearFromDate = ISNULL((SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC),0)
	SET @FinYearFromDate_Min = (SELECT TOP 1 FinYearFrom FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
	
	DECLARE @IsLedgerBalance_Split SMALLINT
	DECLARE @FinYearFromDate_Split NVARCHAR(MAX)
	DECLARE @F_CompanyYears_Split NUMERIC(18,0)
	--DECLARE @F_CompanyYears2 NUMERIC(18,0)
	SET @IsLedgerBalance_Split = (SELECT TOP 1 ISNULL(IsLedgerBalanceSplit,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @F_CompanyYears_Split = (SELECT TOP 1 ISNULL(Id,0) FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 AND IsLedgerBalanceSplit=1 ORDER BY FinYearFrom DESC)
	SET @FinYearFromDate_Split = (SELECT TOP 1 ISNULL(FinYearFrom,0) FROM CompanyYears WHERE Id=@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom ASC)
	--SET @F_CompanyYears2 = (SELECT TOP 1 Id FROM CompanyYears WHERE Id>@F_CompanyYears_Split AND F_CompanyMaster = @F_CompanyMaster)

	IF(@IsLedgerBalance_Split=1)
		BEGIN
			SET @OpeningBalance ='--Opening
									SELECT TAB3.LedgerGroup,TAB3.GroupId,MAX(TAB3.Ledger) AS Ledger,MAX(TAB3.LedgerId) AS LedgerId
										,(SUM(TAB3.DebitAmount)-SUM(TAB3.CreditAmount)) AS OpBal		
										,0 AS DebitAmount,0 AS CreditAmount,TAB3.Annuxure,TAB3.DefaultDrCr,TAB3.TransferTo,MAX(TAB3.TaxRate) AS TaxRate
									FROM(
									--
										SELECT LGM2.Name AS LedgerGroup,LGM2.Id AS GroupId,MAX(LM.Name) AS Ledger,MAX(LM.Id) AS LedgerId
											,SUM(CASE WHEN (LML.DrCr=0) THEN LML.OpeningBalance ELSE 0 END) AS DebitAmount
											,SUM(CASE WHEN (LML.DrCr=1) THEN LML.OpeningBalance ELSE 0 END) AS CreditAmount
											,LGM2.Annuxure,(CASE WHEN(LGM2.DrCr=0) THEN ''Dr'' ELSE ''Cr'' END) AS DefaultDrCr
											,'''' AS TransferTo,MAX(LM.TaxRate) AS TaxRate
										FROM LedgerMaster LM
											LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears_Split AS NVARCHAR) + '
											LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
											LEFT JOIN LedgerMasterL LML2 ON LML2.F_LedgerMaster = LM.Id AND LML2.F_CompanyYears = ' + CAST(@F_CompanyYears AS NVARCHAR) + '
											LEFT JOIN LedgerGroupMaster LGM2 ON LGM2.Id=LML2.F_LedgerGroupMaster
										WHERE
										1=CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(LM.Id = ' + CAST(@F_LedgerMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END
										AND 1=(CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=LM.F_CityMaster) THEN 1 ELSE 0 END) END)
										AND 1=(CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=LM.F_AreaMaster) THEN 1 ELSE 0 END) END)
										AND 1=CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(LGM.Id = ' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END
										AND LM.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + '
										AND LM.F_LedgerGroupMaster<>30 AND LGM.IsBalanceSheet = 1
										AND 1=(CASE WHEN(' + CAST(@TaxType AS NVARCHAR) + ' = -1) THEN 1 ELSE (CASE WHEN (LM.TaxType = ' + CAST(@TaxType AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
										AND 1=(CASE WHEN(' + CAST(@IsExport AS NVARCHAR) + ' = -1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsExport,0)= ' + CAST(@IsExport AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
										AND 1=(CASE WHEN(' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsReverseChargesApplicable,0)=' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)'
										+ @GroupBy1 + ',LGM2.Name,LGM2.Id,LGM2.Annuxure,LGM2.DrCr

										UNION ALL

										--Transaction
										SELECT LGM.Name AS LedgerGroup,LGM.Id AS GroupId,MAX(LM.Name) AS Ledger,MAX(LM.Id) AS LedgerId
												,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount
												,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount
												,LGM.Annuxure,(CASE WHEN(LGM.DrCr=0) THEN ''Dr'' ELSE ''Cr'' END) AS DefaultDrCr
												,'''' AS TransferTo,MAX(LM.TaxRate) AS TaxRate
										FROM VoucherL VL
											LEFT JOIN VoucherH VH ON VH.Id = VL.F_VoucherH
											LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
											LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears AS NVARCHAR) + '
											LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster	
											LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
										WHERE ISNULL(VL.Cancelled,0)=0 AND
											1=CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(VL.F_LedgerMaster = ' + CAST(@F_LedgerMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END
											AND 1=(CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=LM.F_CityMaster) THEN 1 ELSE 0 END) END)
											AND 1=(CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=LM.F_AreaMaster) THEN 1 ELSE 0 END) END)
											AND 1=CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(LGM.Id = ' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END
											AND (VL.VoucherDate<' + @FromDate + ' AND VL.VoucherDate>=' + @FinYearFromDate_Split + ') AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + '
											AND LM.F_LedgerGroupMaster<>30 AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=' + @FinYearFromDate + '))
											AND 1=(CASE WHEN(' + CAST(@TaxType AS NVARCHAR) + ' = -1) THEN 1 ELSE (CASE WHEN (LM.TaxType = ' + CAST(@TaxType AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
											AND 1=(CASE WHEN(' + CAST(@IsExport AS NVARCHAR) + ' = -1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsExport,0) = ' + CAST(@IsExport AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
											AND 1=(CASE WHEN(' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '=VTM.Id) THEN 1 ELSE 0 END) END)
											AND 1=(CASE WHEN(' + CAST(@VoucherType AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(' + CAST(@VoucherType AS NVARCHAR) + ',0)=VTM.VoucherType) THEN 1 ELSE 0 END) END)							
											AND 1=(CASE WHEN(' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsReverseChargesApplicable,0)=' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
											AND 1=(CASE WHEN(' + CAST(@IsTaxation AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(VH.IsTaxation,0)=' + CAST(@IsTaxation AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
										' + @GroupBy1 + '
									) AS TAB3
									' + @GroupBy2 + ''
		END
	ELSE
		BEGIN
			SET @OpeningBalance = '--Opening
							SELECT TAB3.LedgerGroup,TAB3.GroupId,MAX(TAB3.Ledger) AS Ledger,MAX(TAB3.LedgerId) AS LedgerId
								,(SUM(TAB3.DebitAmount)-SUM(TAB3.CreditAmount)) AS OpBal
								,0 AS DebitAmount,0 AS CreditAmount,TAB3.Annuxure,TAB3.DefaultDrCr,TAB3.TransferTo,MAX(TAB3.TaxRate) AS TaxRate
							FROM(
							--
								SELECT LGM.Name AS LedgerGroup,LGM.Id AS GroupId,MAX(LM.Name) AS Ledger,MAX(LM.Id) AS LedgerId
										,SUM(CASE WHEN (LM.DrCr=0) THEN LM.OpeningBalance ELSE 0 END) AS DebitAmount
										,SUM(CASE WHEN (LM.DrCr=1) THEN LM.OpeningBalance ELSE 0 END) AS CreditAmount
										,LGM.Annuxure,(CASE WHEN(LGM.DrCr=0) THEN ''Dr'' ELSE ''Cr'' END) AS DefaultDrCr
										,'''' AS TransferTo,MAX(LM.TaxRate) AS TaxRate
								FROM LedgerMaster LM
									LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears AS NVARCHAR) + '
									LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
								WHERE
									1=CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(LM.Id = ' + CAST(@F_LedgerMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END
									AND 1=(CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=LM.F_CityMaster) THEN 1 ELSE 0 END) END)
									AND 1=(CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=LM.F_AreaMaster) THEN 1 ELSE 0 END) END)
									AND 1=CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(LGM.Id = ' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END
									AND LM.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + '
									AND LM.F_LedgerGroupMaster<>30 AND LGM.IsBalanceSheet = 1
									AND 1=(CASE WHEN(' + CAST(@TaxType AS NVARCHAR) + ' = -1) THEN 1 ELSE (CASE WHEN (LM.TaxType = ' + CAST(@TaxType AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
									AND 1=(CASE WHEN(' + CAST(@IsExport AS NVARCHAR) + ' = -1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsExport,0)= ' + CAST(@IsExport AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
									AND 1=(CASE WHEN(' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsReverseChargesApplicable,0)=' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
								' + @GroupBy1 + '
								UNION ALL
								--Transaction
								SELECT LGM.Name AS LedgerGroup,LGM.Id AS GroupId,MAX(LM.Name) AS Ledger,MAX(LM.Id) AS LedgerId
										,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount
										,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LGM.Annuxure,(CASE WHEN(LGM.DrCr=0) THEN ''Dr'' ELSE ''Cr'' END) AS DefaultDrCr
										,'''' AS TransferTo,MAX(LM.TaxRate) AS TaxRate
								FROM VoucherL VL
									LEFT JOIN VoucherH VH ON VH.Id = VL.F_VoucherH
									LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
									LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + CAST(@F_CompanyYears AS NVARCHAR) + '
									LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster		
									LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
								WHERE ISNULL(VL.Cancelled,0)=0 AND
									1=CASE WHEN(' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(VL.F_LedgerMaster = ' + CAST(@F_LedgerMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END
									AND 1=(CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_CityMaster AS NVARCHAR) + '=LM.F_CityMaster) THEN 1 ELSE 0 END) END)
									AND 1=(CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_AreaMaster AS NVARCHAR) + '=LM.F_AreaMaster) THEN 1 ELSE 0 END) END)
									AND 1=CASE WHEN(' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(LGM.Id = ' + CAST(@F_LedgerGroupMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END
									AND (VL.VoucherDate<' + @FromDate + ' AND VL.VoucherDate>=' + @FinYearFromDate_Min + ') AND VL.F_CompanyMaster=' + CAST(@F_CompanyMaster AS NVARCHAR) + '
									AND LM.F_LedgerGroupMaster<>30 AND (LGM.IsBalanceSheet = 1 OR (LGM.IsBalanceSheet = 0 AND VL.VoucherDate>=' + @FinYearFromDate + '))
									AND 1=(CASE WHEN(' + CAST(@TaxType AS NVARCHAR) + ' = -1) THEN 1 ELSE (CASE WHEN (LM.TaxType = ' + CAST(@TaxType AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
									AND 1=(CASE WHEN(' + CAST(@IsExport AS NVARCHAR) + ' = -1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsExport,0) = ' + CAST(@IsExport AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
									AND 1=(CASE WHEN(' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN(' + CAST(@F_VoucherTypeMaster AS NVARCHAR) + '=VTM.Id) THEN 1 ELSE 0 END) END)
									AND 1=(CASE WHEN(' + CAST(@VoucherType AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(' + CAST(@VoucherType AS NVARCHAR) + ',0)=VTM.VoucherType) THEN 1 ELSE 0 END) END)							
									AND 1=(CASE WHEN(' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsReverseChargesApplicable,0)=' + CAST(@IsReverseChargesApplicable AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)														
									AND 1=(CASE WHEN(' + CAST(@IsTaxation AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(VH.IsTaxation,0)=' + CAST(@IsTaxation AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
								' + @GroupBy1 + '
							) AS TAB3
							' + @GroupBy2 + ''
		END
--
-- BALANCE STATUS
DECLARE @WHERE NVARCHAR(MAX)
SET @WHERE = ''
IF(@BalanceStatus=0)		-- All
	SET @WHERE = ''
ELSE IF(@BalanceStatus=1)	-- Only Zero Balance
	SET @WHERE = ' WHERE TAB5.OpBal=0 AND TAB5.DebitAmount=0 AND TAB5.CreditAmount=0 AND TAB5.ClBalance=0'
ELSE IF(@BalanceStatus=2)	-- Omit Zero Balance
	SET @WHERE = ' WHERE TAB5.OpBal>0 OR TAB5.DebitAmount>0 OR TAB5.CreditAmount>0 OR TAB5.ClBalance>0'
ELSE IF(@BalanceStatus=3)	-- Annexure Grid
	SET @WHERE = ' WHERE TAB5.ClBalance>0'
ELSE IF(@BalanceStatus=4)	-- Annexure Print
	SET @WHERE = ' WHERE TAB5.ClBalance>0 AND TAB5.DrCr=CAST(TAB5.DefaultDrCr AS NVARCHAR)'
--
	EXEC ('
	SELECT TAB5.LedgerGroup,TAB5.GroupId,TAB5.Annuxure,TAB5.Ledger,TAB5.LedgerId,TAB5.OpBal,TAB5.OpDrCr,TAB5.DebitAmount,TAB5.CreditAmount,TAB5.ClBalance,TAB5.DrCr,TAB5.DefaultDrCr,TAB5.TransferTo
			' + @POSTING_FIELDS + ',TAB5.TaxRate
	FROM(
		SELECT TAB4.LedgerGroup,TAB4.GroupId,MAX(TAB4.Ledger) AS Ledger,MAX(TAB4.LedgerId) AS LedgerId
			,ABS(ISNULL(SUM(TAB4.OpBal),0)) AS OpBal
			,
			CASE WHEN(ISNULL(SUM(TAB4.OpBal),0) =0 AND SUM(TAB4.DebitAmount)=0 AND SUM(TAB4.CreditAmount)=0)
				THEN
					(CASE WHEN((SELECT LM.DrCr FROM LedgerMaster LM WHERE LM.Id=MAX(TAB4.LedgerId))=0 ) THEN ''Dr'' ELSE ''Cr'' END)
				ELSE
					(CASE WHEN(ISNULL(SUM(TAB4.OpBal),0)>=0) THEN ''Dr'' ELSE ''Cr'' END)
				END
			
			AS OpDrCr
			,SUM(TAB4.DebitAmount) AS DebitAmount
			,SUM(TAB4.CreditAmount) AS CreditAmount
			,ABS(
				(SUM(TAB4.OpBal)+SUM(TAB4.DebitAmount))-SUM(TAB4.CreditAmount)) AS ClBalance
			,
			CASE WHEN(ISNULL(SUM(TAB4.OpBal),0) =0 AND SUM(TAB4.DebitAmount)=0 AND SUM(TAB4.CreditAmount)=0)
				THEN
					(CASE WHEN((SELECT LM.DrCr FROM LedgerMaster LM WHERE LM.Id=MAX(TAB4.LedgerId))=0 ) THEN ''Dr'' ELSE ''Cr'' END)
				ELSE
					(CASE WHEN (
						(SUM(TAB4.OpBal)
						+SUM(TAB4.DebitAmount))-SUM(TAB4.CreditAmount)>=0) THEN ''Dr'' ELSE ''Cr'' END)
				END
			AS DrCr,TAB4.Annuxure,TAB4.DefaultDrCr,TAB4.TransferTo,MAX(TAB4.TaxRate) AS TaxRate
		FROM(
			' + @OpeningStock + @OpeningBalance + '
				
			UNION
			
			--Transaction
			SELECT LGM.Name AS LedgerGroup,LGM.Id AS GroupId,MAX(LM.Name) AS Ledger,MAX(LM.Id) AS LedgerId
					,0 AS OpBal
					,(SUM(CASE WHEN (VL.DrCr=0) THEN VL.Amount ELSE 0 END)) AS DebitAmount
					,(SUM(CASE WHEN (VL.DrCr=1) THEN VL.Amount ELSE 0 END)) AS CreditAmount,LGM.Annuxure,(CASE WHEN(LGM.DrCr=0) THEN ''Dr'' ELSE ''Cr'' END) AS DefaultDrCr
					,'''' AS TransferTo,MAX(LM.TaxRate) AS TaxRate
			FROM VoucherL VL
				LEFT JOIN VoucherH VH ON VH.Id = VL.F_VoucherH
				LEFT JOIN LedgerMaster LM ON VL.F_LedgerMaster=LM.Id
				LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id AND LML.F_CompanyYears = ' + @F_CompanyYears + '
				LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
				LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
			WHERE ISNULL(VL.Cancelled,0)=0 AND
				1=CASE WHEN(' + @F_LedgerMaster + '=0) THEN 1 ELSE (CASE WHEN(VL.F_LedgerMaster = ' + @F_LedgerMaster + ') THEN 1 ELSE 0 END) END
				AND 1=(CASE WHEN(' + @F_CityMaster + '=0) THEN 1 ELSE (CASE WHEN(' + @F_CityMaster + '=LM.F_CityMaster) THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(' + @F_AreaMaster + '=0) THEN 1 ELSE (CASE WHEN(' + @F_AreaMaster + '=LM.F_AreaMaster) THEN 1 ELSE 0 END) END)
				AND 1=CASE WHEN(' + @F_LedgerGroupMaster + '=0) THEN 1 ELSE (CASE WHEN(LGM.Id = ' + @F_LedgerGroupMaster + ') THEN 1 ELSE 0 END) END
				AND (VL.VoucherDate>=' + @FromDate + ' AND VL.VoucherDate<=' + @ToDate + ') AND VL.F_CompanyMaster=' + @F_CompanyMaster + '
				AND LM.F_LedgerGroupMaster<>30
				AND 1=(CASE WHEN(' + @TaxType + ' = -1) THEN 1 ELSE (CASE WHEN (LM.TaxType = ' + @TaxType + ') THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(' + @IsExport + ' = -1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsExport,0) = ' + @IsExport + ') THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(' + @F_VoucherTypeMaster + '=0) THEN 1 ELSE (CASE WHEN(' + @F_VoucherTypeMaster + '=VTM.Id) THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(' + @VoucherType + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(' + @VoucherType + ',0)=VTM.VoucherType) THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(' + @IsReverseChargesApplicable + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(LM.IsReverseChargesApplicable,0)=' + @IsReverseChargesApplicable + ') THEN 1 ELSE 0 END) END)						
				AND 1=(CASE WHEN(' + @IsTaxation + '=-1) THEN 1 ELSE (CASE WHEN(ISNULL(VH.IsTaxation,0)=' + @IsTaxation + ') THEN 1 ELSE 0 END) END)
			' + @GroupBy1 + '
		) AS TAB4
		' + @GroupBy3 + '
	) AS TAB5 ' + @POSTING_JOIN + '
	' + @WHERE + @OrderBy )
END





GO
```

# [dbo].[Select_Acc_VoucherEntry]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Acc_VoucherEntry]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



CREATE PROCEDURE [dbo].[Select_Acc_VoucherEntry] 
	@Id numeric(18,0)
AS
BEGIN
	SELECT ((CASE WHEN(VTM.Prefix<>'') THEN VTM.Prefix ELSE '' END) + VH.VoucherNo) AS VoucherNo,
		(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate,
		VL.DrCr,LM.Name AS Ledger,VL.Amount,VH.Narration, LM1.Name AS LedgerMain, VH.TotalNetAmt,
		VL.F_VoucherTypeMaster,VTM.Name AS VoucherType,		
		(CASE WHEN(LM.F_LedgerGroupMaster IN(31,32)) THEN LM.Name ELSE LM1.Name END) AS ChequePrint,
		VH.PurVoucherNo,
		(SUBSTRING(VH.PurVoucherDate,7,2)+'/'+SUBSTRING(VH.PurVoucherDate,5,2)+'/'+SUBSTRING(VH.PurVoucherDate,1,4)) AS PurVoucherDate,
		LM1.GSTNo,
		(LM1.Address1 +
		(CASE WHEN(LM1.Address2<>'') THEN ', ' + LM1.Address2 ELSE '' END) +
		(CASE WHEN(LM1.Address3<>'') THEN ', ' + LM1.Address3 ELSE '' END) +
		(CASE WHEN(CM.Name<>'') THEN ', ' + CM.Name ELSE '' END) ) AS LedgerAddress,
		SM.Name AS [State]
	FROM VoucherH VH
	LEFT JOIN VoucherL VL ON VH.Id=VL.F_VoucherH
	LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
	LEFT JOIN LedgerMaster LM1 ON LM1.Id=VH.F_LedgerMaster	
	LEFT JOIN VoucherTypeMaster VTM ON VTM.Id=VL.F_VoucherTypeMaster
	LEFT JOIN CityMaster CM ON CM.Id = LM1.F_CityMaster
	LEFT JOIN StateMaster SM ON SM.Id = CM.F_StateMaster

	WHERE VH.Id=@Id AND VL.F_LedgerMaster <> VH.F_LedgerMaster AND VH.F_VoucherTypeMaster=VL.F_VoucherTypeMaster
	ORDER BY VL.Id
END























GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_AreaWiseItemSummary]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






















































CREATE PROCEDURE [dbo].[Select_AS_AreaWiseItemSummary] 		
	@FromDate nvarchar(50)=NULL,
	@ToDate nvarchar(50)=NULL,	
	@LedgerIds nvarchar(max)='',
	@ItemIds nvarchar(max)='',
	@OtherLedgerIds nvarchar(max)='',
	@AreaIds nvarchar(max)='',
	@BrandIds nvarchar(max)=''
AS
BEGIN
	SELECT 
		LDM.Id,AM.Name AS Area,(LDM.Name + '( ' +
		+ ISNULL((CASE WHEN(ISNULL(CAST(LDM.Address1 AS NVARCHAR),'') <> '') THEN CAST(LDM.Address1 AS NVARCHAR) END),'')
		+ ISNULL((CASE WHEN(ISNULL(CAST(LDM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LDM.Address2 AS NVARCHAR) END),'')
		+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')
		+ ISNULL((CASE WHEN(ISNULL(OLM.Name,'') <> '') THEN ' ( ' 
			+ CAST(OLM.Name AS NVARCHAR) 
				+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address1 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address2 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(OCM.Name,'') <> '') THEN ', ' + OCM.Name END),'')
			+ ' )' END),'')+ ' )'
		) AS Ledger
		,IBM.Name AS Brand
		,IM.Name AS Item
		,SUM(GIL.Qty) AS Qty
		,SUM(GIL.TaxableAmt) AS TaxableAmt,SUM(GIL.TaxAmt) AS TaxAmt,SUM(GIL.TaxAmt+GIL.TaxableAmt) AS Amount,AM.Id AS AreaId
		,( DENSE_RANK() OVER (ORDER BY AM.Name,LDM.Name )) As SNo
	FROM SaleInvoiceH GIH 
		LEFT JOIN SaleInvoiceL GIL ON GIH.Id=GIL.F_SaleInvoiceH	
		LEFT JOIN LedgerMaster LDM ON LDM.Id = GIH.F_LedgerMaster
		LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=LDM.F_CityMaster
		LEFT JOIN AreaMaster AM ON AM.Id=LDM.F_AreaMaster
		
		LEFT JOIN OtherLedgerMaster OLM ON OLM .Id = GIH.F_OtherLedgerMaster
		LEFT JOIN [dbo].[CityMaster] OCM ON OCM.Id=OLM.F_CityMaster
		
		LEFT JOIN ItemMaster IM ON IM.Id = GIL.F_ItemMaster
		LEFT JOIN ItemBrandMaster IBM ON IBM.Id = IM.F_ItemBrandMaster	
	WHERE
		1=(CASE WHEN(@FromDate='') THEN 1 ELSE
			(CASE WHEN(GIH.VoucherDate>=@FromDate AND GIH.VoucherDate<=@ToDate) THEN 1 ELSE 0 END)
			END)
	AND 1=(CASE WHEN (@AreaIds ='') THEN 1 ELSE (CASE WHEN (AM.Id  IN(SELECT * FROM Split(@AreaIds,',')) ) THEN 1 ELSE 0 END) END)
	AND 1=(CASE WHEN (@BrandIds ='') THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster IN(SELECT * FROM Split(@BrandIds,',')) ) THEN 1 ELSE 0 END) END)
	AND 1=(CASE WHEN (@LedgerIds ='') THEN 1 ELSE (CASE WHEN (LDM.Id  IN(SELECT * FROM Split(@LedgerIds,',')) ) THEN 1 ELSE 0 END) END)
	AND 1=(CASE WHEN (@OtherLedgerIds='') THEN 1 ELSE (CASE WHEN (OLM.Id  IN(SELECT * FROM Split(@OtherLedgerIds,',')) ) THEN 1 ELSE 0 END) END)		
	AND 1=(CASE WHEN (@ItemIds='') THEN 1 ELSE (CASE WHEN (IM.Id  IN(SELECT * FROM Split(@ItemIds,',')) ) THEN 1 ELSE 0 END) END)
	
	GROUP BY LDM.Id,LDM.Name,IBM.Id,IBM.Name,IM.Name,LDM.Address1,LDM.Address2,CM.Name,OLM.Name,OLM.Address1,OLM.Address2,OCM.Name,AM.Id,AM.Name
	ORDER BY AM.Name,LDM.Name,IBM.Name,IM.Name

END














































GO
```

# [dbo].[Select_AS_BillCollectionH]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_BillCollectionH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






















CREATE PROCEDURE [dbo].[Select_AS_BillCollectionH] 	
	@F_SalesmanMaster numeric(18,0)=0,
	@F_BillCollectionStatusMaster numeric(18,0)=0,
	@F_SaleInvoiceH numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Cancelled smallint=-1,
	@Other nvarchar(50)='',
	@Id numeric(18,0)=0,
	@ReceivedStatus smallint=-1
AS
BEGIN

SELECT TAB1.Id,TAB1.VoucherNo,TAB1.VoucherDate,TAB1.BillFromDate,TAB1.BillToDate,TAB1.Salesman,TAB1.Other,TAB1.NoOfBills,TAB1.Received,TAB1.Cancelled
	FROM (
	SELECT DSH.Id,DSH.VoucherNo,
		(SUBSTRING(DSH.VoucherDate,7,2)+'/'+SUBSTRING(DSH.VoucherDate,5,2)+'/'+SUBSTRING(DSH.VoucherDate,1,4)) AS VoucherDate,
		(SUBSTRING(DSH.BillFromDate,7,2)+'/'+SUBSTRING(DSH.BillFromDate,5,2)+'/'+SUBSTRING(DSH.BillFromDate,1,4)) AS BillFromDate,
		(SUBSTRING(DSH.BillToDate,7,2)+'/'+SUBSTRING(DSH.BillToDate,5,2)+'/'+SUBSTRING(DSH.BillToDate,1,4)) AS BillToDate,
		DM.Name AS Salesman,
		DSH.Other,
		COUNT(DSL.Id) AS NoOfBills,
		SUM(CASE WHEN(ISNULL(DSL.F_BillCollectionStatusMaster,0)>0) THEN 1 ELSE 0 END) AS Received,
		(CASE WHEN (ISNULL(DSH.Cancelled,0)=1) THEN 'Cancelled' ELSE '' END) AS Cancelled
	FROM BillCollectionH DSH
		LEFT JOIN BillCollectionL DSL ON DSH.Id=DSL.F_BillCollectionH		
		LEFT JOIN SalesmanMaster DM ON DM.Id = DSH.F_SalesmanMaster
	WHERE 
		1=(CASE WHEN (@Id=0) THEN 1 ELSE (CASE WHEN (DSH.Id =@Id) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@Other='') THEN 1 ELSE (CASE WHEN (DSH.Other =@Other) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_SalesmanMaster=0) THEN 1 ELSE (CASE WHEN (DSH.F_SalesmanMaster =@F_SalesmanMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_BillCollectionStatusMaster=0) THEN 1 ELSE (CASE WHEN (DSL.F_BillCollectionStatusMaster =@F_BillCollectionStatusMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_SaleInvoiceH=0) THEN 1 ELSE (CASE WHEN (DSL.F_SaleInvoiceH =@F_SaleInvoiceH) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN(@FromDate='') THEN 1 ELSE
			(CASE WHEN(DSH.VoucherDate>=@FromDate AND DSH.VoucherDate<=@ToDate) THEN 1 ELSE 0 END)
			END)
		AND
		1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(DSH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
	GROUP BY DSH.Id,DSH.VoucherNo,DSH.VoucherDate,DSH.VoucherDate,DSH.BillFromDate,DSH.BillToDate,DM.Name,DSH.Cancelled,DSH.Other
) AS TAB1	
	WHERE 1=
		(CASE WHEN(@ReceivedStatus=-1) THEN 1 ELSE
			(CASE WHEN((TAB1.NoOfBills-TAB1.Received)>0) THEN 1 ELSE 0 END)
		END)
	ORDER BY TAB1.VoucherDate,CAST(TAB1.VoucherNo AS NUMERIC),TAB1.Id,TAB1.Salesman
END
















































GO
```

# [dbo].[Select_AS_DeliveryProcessH]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_DeliveryProcessH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





















































CREATE PROCEDURE [dbo].[Select_AS_DeliveryProcessH] 	
	@F_DeliveryByMaster numeric(18,0)=0,
	@F_DeliveryStatusMaster numeric(18,0)=0,
	@F_CheckedByMaster numeric(18,0)=0,
	@F_SaleInvoiceH numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Cancelled smallint=-1,
	@Other nvarchar(50)='',
	@Id numeric(18,0)=0,
	@ReceivedStatus smallint=-1
AS
BEGIN

SELECT TAB1.Id,TAB1.VoucherNo,TAB1.VoucherDate,TAB1.BillFromDate,TAB1.BillToDate,TAB1.DeliveryBy,TAB1.Other,TAB1.TripNo,TAB1.NoOfBills,TAB1.Received,TAB1.Cancelled
	FROM (
	SELECT DSH.Id,DSH.VoucherNo,
		(SUBSTRING(DSH.VoucherDate,7,2)+'/'+SUBSTRING(DSH.VoucherDate,5,2)+'/'+SUBSTRING(DSH.VoucherDate,1,4)) AS VoucherDate,
		(SUBSTRING(DSH.BillFromDate,7,2)+'/'+SUBSTRING(DSH.BillFromDate,5,2)+'/'+SUBSTRING(DSH.BillFromDate,1,4)) AS BillFromDate,
		(SUBSTRING(DSH.BillToDate,7,2)+'/'+SUBSTRING(DSH.BillToDate,5,2)+'/'+SUBSTRING(DSH.BillToDate,1,4)) AS BillToDate,
		DM.Name AS DeliveryBy,
		DSH.Other,
		DSH.TripNo,
		COUNT(DSL.Id) AS NoOfBills,
		SUM(CASE WHEN(ISNULL(DSL.F_DeliveryStatusMaster,0)>0) THEN 1 ELSE 0 END) AS Received,
		(CASE WHEN (ISNULL(DSH.Cancelled,0)=1) THEN 'Cancelled' ELSE '' END) AS Cancelled
	FROM DeliveryProcessH DSH
		LEFT JOIN DeliveryProcessL DSL ON DSH.Id=DSL.F_DeliveryProcessH		
		LEFT JOIN DeliveryByMaster DM ON DM.Id = DSH.F_DeliveryByMaster
	WHERE 
		1=(CASE WHEN (@Id=0) THEN 1 ELSE (CASE WHEN (DSH.Id =@Id) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@Other='') THEN 1 ELSE (CASE WHEN (DSH.Other =@Other) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_DeliveryByMaster=0) THEN 1 ELSE (CASE WHEN (DSH.F_DeliveryByMaster =@F_DeliveryByMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_DeliveryStatusMaster=0) THEN 1 ELSE (CASE WHEN (DSL.F_DeliveryStatusMaster =@F_DeliveryStatusMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_CheckedByMaster=0) THEN 1 ELSE (CASE WHEN (DSL.F_CheckedByMaster =@F_CheckedByMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_SaleInvoiceH=0) THEN 1 ELSE (CASE WHEN (DSL.F_SaleInvoiceH =@F_SaleInvoiceH) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN(@FromDate='') THEN 1 ELSE
			(CASE WHEN(DSH.VoucherDate>=@FromDate AND DSH.VoucherDate<=@ToDate) THEN 1 ELSE 0 END)
			END)
		AND
		1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(DSH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
	GROUP BY DSH.Id,DSH.VoucherNo,DSH.VoucherDate,DSH.VoucherDate,DSH.BillFromDate,DSH.BillToDate,DM.Name,DSH.Cancelled,DSH.Other,DSH.TripNo
) AS TAB1	
	WHERE 1=
		(CASE WHEN(@ReceivedStatus=-1) THEN 1 ELSE
			(CASE WHEN((TAB1.NoOfBills-TAB1.Received)>0) THEN 1 ELSE 0 END)
		END)
	ORDER BY TAB1.VoucherDate,CAST(TAB1.VoucherNo AS NUMERIC),TAB1.Id,TAB1.DeliveryBy
END
















































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_E_Way_Bill]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO







CREATE PROCEDURE [dbo].[Select_AS_E_Way_Bill]
	@VoucherNo NVARCHAR(50)='',
	@FromDate NVARCHAR(50),
	@ToDate NVARCHAR(50)
AS
BEGIN
	SELECT SH.Id,SH.VoucherNo,
	(SUBSTRING(SH.VoucherDate,7,2)+'/'+SUBSTRING(SH.VoucherDate,5,2)+'/'+SUBSTRING(SH.VoucherDate,1,4)) AS VoucherDate,
	LM.Name AS Ledger,LM.GSTNo,MAX(IM.Name) AS ItemGroup,HM.Name AS HSNCode
	,SUM(SL.MinQty) AS Qty,UM.Name AS Unit,SUM(SL.FinalAmt) AS TaxableAmt,SL.TaxRate,SL.TaxRate2,SL.TaxRate3
	FROM SaleInvoiceH SH
	LEFT JOIN SaleInvoicel SL ON SH.Id=SL.F_SaleInvoiceH
	LEFT JOIN ItemMaster IM ON IM.Id=SL.F_ItemMaster
	LEFT JOIN LedgerMaster LM ON LM.Id=SH.F_LedgerMaster
	LEFT JOIN UnitMaster UM ON UM.Id=SL.F_UnitMasterQ
	LEFT JOIN HSNCodeMaster HM ON HM.Id=IM.F_HSNCodeMaster
	WHERE SH.VoucherDate >= @FromDate AND SH.VoucherDate <= @ToDate
		AND 1=(CASE WHEN(@VoucherNo='') THEN 1 ELSE (CASE WHEN(SH.VoucherNo = @VoucherNo) THEN 1 ELSE 0 END) END)
		AND ISNULL(SH.Cancelled,0)=0
		AND ISNULL(LM.GSTNo,'')<>''
	GROUP BY SH.Id,SH.VoucherNo,SH.VoucherDate,LM.Name,LM.GSTNo,HM.Name,UM.Name,SL.TaxRate,SL.TaxRate2,SL.TaxRate3
	ORDER BY CAST(SH.VoucherNo AS NUMERIC),HSNCode
END



GO
```

# [dbo].[Select_AS_G1BillDetails]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_G1BillDetails]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





































CREATE PROCEDURE [dbo].[Select_AS_G1BillDetails] 	
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@OnlyPending smallint=0
AS
BEGIN	
	SELECT SH.Id,SH.VoucherNo
		,(SUBSTRING(SH.VoucherDate,7,2)+'/'+SUBSTRING(SH.VoucherDate,5,2)+'/'+SUBSTRING(SH.VoucherDate,1,4)) AS VoucherDate
		,((CASE WHEN(LM.Name<>'') THEN LM.Name ELSE SH.LedgerName END)
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')
				+ ISNULL((CASE WHEN(ISNULL(OLM.Name,'') <> '') THEN ' ('
					+ CAST(OLM.Name AS NVARCHAR) 
						+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address1 AS NVARCHAR) END),'')
						+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address2 AS NVARCHAR) END),'')
						+ ISNULL((CASE WHEN(ISNULL(OCM.Name,'') <> '') THEN ', ' + OCM.Name END),'')
					+ ')' END),'')
				) AS Ledger
		,SH.TotalProDisAmt
		,SH.TotalNetAmt
		,SH.G1_BillNo
		,(CASE WHEN(SH.G1_IsSchemeMatched=1) THEN 
				'Yes'
			ELSE
				(CASE WHEN(ISNULL(SH.G1_BillNo,'')='') THEN '' ELSE 'No' END)
			END) AS G1_IsSchemeMatched
		,(CASE WHEN(SH.G1_IsAmountMatched=1) THEN 
				'Yes'
			ELSE
				(CASE WHEN(ISNULL(SH.G1_BillNo,'')='') THEN '' ELSE 'No' END)
			END) AS G1_IsAmountMatched
		,SH.G1_Remark
	FROM SaleInvoiceH SH
	LEFT JOIN SaleInvoiceL SL ON SL.F_SaleInvoiceH = SH.Id
	LEFT JOIN ItemMaster IM ON IM.Id= SL.F_ItemMaster
	LEFT JOIN LedgerMaster LM ON LM.Id=SH.F_LedgerMaster
	LEFT JOIN CityMaster CM ON CM.Id=LM.F_CityMaster
	LEFT JOIN StateMaster SM ON SM.Id=CM.F_StateMaster
	
	LEFT JOIN OtherLedgerMaster OLM ON OLM.Id = SH.F_OtherLedgerMaster
	LEFT JOIN CityMaster OCM ON OCM.Id=OLM.F_CityMaster
	
	WHERE SH.VoucherDate>=@FromDate AND SH.VoucherDate<=@ToDate
	AND ISNULL(Cancelled,0)=0
	AND 1 = CASE WHEN(@OnlyPending=0) THEN 1 ELSE (CASE WHEN(ISNULL(SH.G1_BillNo,'')='') THEN 1 ELSE 0 END) END
	AND IM.F_ItemBrandMaster = 2
	GROUP BY SH.Id,SH.VoucherNo,SH.VoucherDate,SH.TotalNetAmt,SH.G1_BillNo,SH.G1_IsSchemeMatched,SH.G1_IsAmountMatched,SH.G1_Remark
		,LM.Name,LM.Address1,LM.Address2,OLM.Name,OLM.Address1,OLM.Address2,SH.LedgerName,CM.Name,OCM.Name,SH.TotalProDisAmt
	ORDER BY CAST(VoucherNo AS NUMERIC)
END




























GO
```

# [dbo].[Select_AS_GST_HSNWiseSummary]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_GST_HSNWiseSummary]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE PROCEDURE [dbo].[Select_AS_GST_HSNWiseSummary] 
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@ProjectType SMALLINT,
	@Type smallint=0
AS
BEGIN
	DECLARE @TableNameH NVARCHAR(50)
	DECLARE @TableNameL NVARCHAR(50)
	DECLARE @TableNameH_Return NVARCHAR(50)
	DECLARE @TableNameL_Return NVARCHAR(50)
	DECLARE @GroupFields NVARCHAR(MAX)
	DECLARE @MinusReturn NVARCHAR(10)

	IF(@Type=0 OR @Type=1 OR @Type=2)			-- 0 -> Sale		-- 1 -> Sale Return			-- 2 -> Net Sale
		BEGIN
			SET @TableNameH = 'SaleInvoiceH'
			SET @TableNameL = 'SaleInvoiceL'
			SET @TableNameH_Return = 'SaleReturnH'
			SET @TableNameL_Return = 'SaleReturnL'
		END
	IF(@Type=3 OR @Type=4 OR @Type=5)			-- 3 -> Purchase	-- 4 -> Purchase Return		-- 5 -> Net Purchase
		BEGIN
			SET @TableNameH = 'PurchaseInvoiceH'
			SET @TableNameL = 'PurchaseInvoiceL'
			SET @TableNameH_Return = 'PurchaseReturnH'
			SET @TableNameL_Return = 'PurchaseReturnL'
		END
	IF(@Type=2 OR @Type=5)
		SET @MinusReturn = '*-1'
	ELSE
		SET @MinusReturn = ''

	DECLARE @SQL_Sale NVARCHAR(MAX)
	DECLARE @SQL_SaleReturn NVARCHAR(MAX)
	DECLARE @SQL_SaleVoucher NVARCHAR(MAX)
	DECLARE @SQL_NetSale NVARCHAR(MAX)
	SET @SQL_Sale = 'SELECT (CASE WHEN(MAX(ISNULL(HSN.Name,''''))='''') THEN MAX(IM.Name) ELSE SUBSTRING(MAX(HSN.Name),1,6) END) AS [HSN],
						LMTax_Sales.Name AS [Description],''PCS'' AS UQC,
							SUM(SL.MinQty+SL.MinQtyF) AS [Total Quantity],SUM(SL.TaxableAmt + SL.TaxableAmt3 + SL.TaxAmt + SL.TaxAmt2 + SL.TaxAmt3) AS [Total Value],SUM(SL.TaxableAmt+SL.TaxableAmt3) AS [Taxable Value],SUM(SL.TaxAmt3) AS [Integrated Tax Amount], SUM(SL.TaxAmt2) AS [Central Tax Amount],SUM(SL.TaxAmt) AS [State/UT Tax Amount],0 AS [Cess Amount]
						FROM ' + @TableNameH + ' SH
							LEFT JOIN ' + @TableNameL + ' SL ON SH.Id = SL.F_' + @TableNameH + '
							LEFT JOIN LedgerMaster LMTax ON LMTax.Id = SL.F_LedgerMaster_Tax
							LEFT JOIN LedgerMaster LMTax_Sales ON LMTax_Sales.Id = LMTax.TaxF_SalesLedger
							LEFT JOIN ItemMaster IM ON IM.Id = SL.F_ItemMaster
							LEFT JOIN HSNCodeMaster HSN ON HSN.Id = IM.F_HSNCodeMaster
							LEFT JOIN UnitMaster UM ON UM.Id = SL.F_UnitMasterQ
						WHERE SH.VoucherDate>=' + @FromDate + ' AND SH.VoucherDate<=' + @ToDate + ' AND ISNULL(SH.Cancelled,0)=0
						GROUP BY SUBSTRING(HSN.Name,1,6),LMTax_Sales.Name'
	IF(@ProjectType=0)						-- AGENCIES
		BEGIN
			SET @SQL_SaleVoucher = 'UNION ALL SELECT ''996111'' AS [HSN],'''' AS [Description],''PCS'' AS UQC
							,SUM(CASE WHEN(LMTax.F_LedgerGroupMaster<>12 AND VLT.F_LedgerMaster=VH.F_LedgerMaster) THEN 1 ELSE 0 END) AS [Total Quantity]
							,SUM(CASE WHEN(LMTax.F_LedgerGroupMaster<>12 AND VLT.F_LedgerMaster=VH.F_LedgerMaster) THEN VLT.Amount ELSE 0 END) AS [Total Value]
							,SUM(CASE WHEN(LMTax.F_LedgerGroupMaster<>12 AND VLT.F_LedgerMaster<>VH.F_LedgerMaster) THEN VLT.Amount ELSE 0 END) AS [Taxable Value]
							,SUM(CASE WHEN(LMTax.TaxType=8 AND LMTax.F_LedgerGroupMaster=12) THEN VLT.Amount ELSE 0 END) AS [Integrated Tax Amount]
							,SUM(CASE WHEN(LMTax.TaxType=7 AND LMTax.F_LedgerGroupMaster=12) THEN VLT.Amount ELSE 0 END) AS [Central Tax Amount]
							,SUM(CASE WHEN(LMTax.TaxType=6 AND LMTax.F_LedgerGroupMaster=12) THEN VLT.Amount ELSE 0 END) AS [State/UT Tax Amount],0 AS [Cess Amount]
						FROM VoucherH VH
						LEFT JOIN VoucherL VLT ON VH.Id = VLT.F_VoucherH
						LEFT JOIN LedgerMaster LMTax ON LMTax.Id = VLT.F_LedgerMaster
						WHERE VH.VoucherDate>=' + @FromDate + ' AND VH.VoucherDate<=' + @ToDate + ' AND ISNULL(VH.Cancelled,0)=0 AND VH.F_VoucherTypeMaster=33 AND VLT.F_VoucherTypeMaster=33'
		END
	ELSE
		BEGIN
			SET @SQL_SaleVoucher = ''
		END
	SET @SQL_SaleReturn = 'SELECT (CASE WHEN(MAX(ISNULL(HSN.Name,''''))='''') THEN MAX(IM.Name) ELSE SUBSTRING(MAX(HSN.Name),1,6) END) AS [HSN],
							LMTax_Sales.Name AS [Description],''PCS'' AS UQC,
							SUM(SL.MinQty+SL.MinQtyF)' + @MinusReturn + ' AS [Total Quantity],
							SUM(SL.TaxableAmt + SL.TaxableAmt3 + SL.TaxAmt + SL.TaxAmt2 + SL.TaxAmt3)' + @MinusReturn + ' AS [Total Value],
							SUM(SL.TaxableAmt+SL.TaxableAmt3)' + @MinusReturn + ' AS [Taxable Value],
							SUM(SL.TaxAmt3)' + @MinusReturn + ' AS [Integrated Tax Amount],
							SUM(SL.TaxAmt2)' + @MinusReturn + ' AS [Central Tax Amount],
							SUM(SL.TaxAmt)' + @MinusReturn + ' AS [State/UT Tax Amount],0 AS [Cess Amount]
						FROM ' + @TableNameH_Return + ' SH
							LEFT JOIN ' + @TableNameL_Return + ' SL ON SH.Id = SL.F_' + @TableNameH_Return + '
							LEFT JOIN LedgerMaster LMTax ON LMTax.Id = SL.F_LedgerMaster_Tax
							LEFT JOIN LedgerMaster LMTax_Sales ON LMTax_Sales.Id = LMTax.TaxF_SalesLedger
							LEFT JOIN ItemMaster IM ON IM.Id = SL.F_ItemMaster
							LEFT JOIN HSNCodeMaster HSN ON HSN.Id = IM.F_HSNCodeMaster
							LEFT JOIN UnitMaster UM ON UM.Id = SL.F_UnitMasterQ
						WHERE SH.VoucherDate>=' + @FromDate + ' AND SH.VoucherDate<=' + @ToDate + ' AND ISNULL(SH.Cancelled,0)=0
						GROUP BY SUBSTRING(HSN.Name,1,6),LMTax_Sales.Name'
	SET @GroupFields = 'TAB2.[HSN],
						MAX(TAB2.[Description]) AS [Description],MAX(TAB2.UQC) AS Unit,
						SUM(TAB2.[Total Quantity]) AS TotalQuantity,
						SUM(TAB2.[Total Value]) AS TotalValue,
						SUM(TAB2.[Taxable Value]) AS TaxableValue,
						SUM(TAB2.[Integrated Tax Amount]) AS IGST,
						SUM(TAB2.[Central Tax Amount]) AS CGST,
						SUM(TAB2.[State/UT Tax Amount]) AS SGST,
						0 AS Cess'
-- Final Execution
	IF(@Type=0 OR @Type=3)			-- 0 -> Sale			-- 3 -> Purchase
		BEGIN
			EXEC(@SQL_Sale)
		END
	ELSE IF(@Type=1 OR @Type=4)		-- 1 -> Sale Return		-- 4 -> Purchase Return
		BEGIN
			EXEC(@SQL_SaleReturn)
		END
	ELSE IF(@Type=2)				-- 5 -> Net Purchase
		BEGIN
			SET @SQL_NetSale = 'SELECT ' + @GroupFields + '
								FROM (
									' + @SQL_Sale + '
									' + @SQL_SaleVoucher + '
									UNION ALL
									' + @SQL_SaleReturn + '
								) AS TAB2
								GROUP BY TAB2.[HSN],TAB2.[Description]
								ORDER BY TAB2.[HSN]'
			EXEC(@SQL_NetSale)
		END
	ELSE IF(@Type=5)				-- 5 -> Net Purchase
		BEGIN
			SET @SQL_NetSale = 'SELECT ' + @GroupFields + '
								FROM (
									' + @SQL_Sale + '
									UNION ALL
									' + @SQL_SaleReturn + '
								) AS TAB2
								GROUP BY TAB2.[HSN],TAB2.[Description]
								ORDER BY TAB2.[HSN]'
			EXEC(@SQL_NetSale)
		END
END



GO
```

# [dbo].[Select_AS_InterestCalculation]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_InterestCalculation]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











CREATE PROCEDURE [dbo].[Select_AS_InterestCalculation]
	@F_CompanyMaster numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@F_LedgerMaster numeric(18,0)=0,
	@CurrentDate nvarchar(50)=''
AS
BEGIN
	SELECT RowNumber,Id,VoucherNo1,VoucherDate,LedgerName,DueAmt,DueDays,IsDisputed FROM (
		SELECT DENSE_RANK() OVER (order by TAB2.RowNumber) AS RowNumber,TAB2.VoucherDate,TAB2.LedgerName,TAB2.Id,TAB2.VoucherNo1,TAB2.DueAmt,DATEDIFF(day,CONVERT(datetime,TAB2.VoucherDate1),CONVERT(datetime,@CurrentDate)) AS DueDays,TAB2.IsDisputed FROM
		(
			SELECT (VH.TotalNetAmt -
				(	ISNULL(
						(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptCashL MRL LEFT JOIN MoneyReceiptCashH MRH ON MRH.Id=MRL.F_MoneyReceiptCashH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)
					,0)
					+ ISNULL(
						(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptBankL MRL LEFT JOIN MoneyReceiptBankH MRH ON MRH.Id=MRL.F_MoneyReceiptBankH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0 AND ISNULL(MRH.IsChequeCancel,0)=0)

					,0)
					+ ISNULL(
						(SELECT SUM(VH1.TotalNetAmt)
							FROM SaleInvoiceH_CreditNote SIC LEFT JOIN VoucherH VH1 ON VH1.Id=SIC.F_VoucherH
							WHERE SIC.F_SaleInvoiceH = VH.Id AND ISNULL(VH1.Cancelled,0)=0)
					,0)
					+ ISNULL(
						(SELECT SUM(MRL.Amount)
							FROM SaleReturnL_InvoiceDetails MRL LEFT JOIN SaleReturnH MRH ON MRH.Id=MRL.F_SaleReturnH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)

					,0)
					)) AS DueAmt
			,DENSE_RANK() OVER (order by VH.VoucherDate,CAST(VH.VoucherNo AS NUMERIC)) AS RowNumber
			,((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+
				VH.VoucherNo + '            ' + (SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) + '            ' + 
				CAST(VH.TotalNetAmt 
					-(	ISNULL(
							(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptCashL MRL LEFT JOIN MoneyReceiptCashH MRH ON MRH.Id=MRL.F_MoneyReceiptCashH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)
						,0) 
					+	ISNULL(
							(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptBankL MRL LEFT JOIN MoneyReceiptBankH MRH ON MRH.Id=MRL.F_MoneyReceiptBankH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0 AND ISNULL(MRH.IsChequeCancel,0)=0)
						,0) 
					+	ISNULL(
							(SELECT SUM(VH1.TotalNetAmt)
							FROM SaleInvoiceH_CreditNote SIC LEFT JOIN VoucherH VH1 ON VH1.Id=SIC.F_VoucherH
							WHERE SIC.F_SaleInvoiceH = VH.Id AND ISNULL(VH1.Cancelled,0)=0)
						,0)
					+	ISNULL(
							(SELECT SUM(MRL.Amount)
							FROM SaleReturnL_InvoiceDetails MRL LEFT JOIN SaleReturnH MRH ON MRH.Id=MRL.F_SaleReturnH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)
						,0)
						) AS NVARCHAR)) AS Name
			,VH.Id,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate,
			((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+VH.VoucherNo 
			+ '    (' + (SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) + ')') AS VoucherNo1,VH.VoucherDate AS VoucherDate1,VH.IsDisputed
			,LM.Name AS LedgerName
			FROM [dbo].[SaleInvoiceH] VH
				LEFT JOIN LedgerMaster LM ON VH.F_LedgerMaster=LM.Id
				
			WHERE VH.F_CompanyMaster =@F_CompanyMaster
				AND 1= CASE WHEN(@FromDate='' AND @F_LedgerMaster>0) THEN 1 ELSE (CASE WHEN(VH.VoucherDate <=@ToDate AND @F_LedgerMaster>0) THEN 1 ELSE 				
					(CASE WHEN(VH.VoucherDate >=@FromDate AND VH.VoucherDate <=@ToDate) THEN 1 ELSE 0 END)		
				END) END
				AND ISNULL(VH.Cancelled,0)=0
				AND 1=(CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(VH.F_LedgerMaster = @F_LedgerMaster) THEN 1 ELSE 0 END) END)
			GROUP BY VH.Id,VH.VoucherNo,VH.VoucherDate,LM.Name,VH.TotalNetAmt,VH.VoucherNoPrefix,VH.IsDisputed
		) AS TAB2
		WHERE TAB2.DueAmt>0
		) AS Tab1	
	ORDER BY RowNumber,LedgerName,Id
END





















GO
```

# [dbo].[Select_AS_InvoiceDetail]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_InvoiceDetail]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






















































CREATE PROCEDURE [dbo].[Select_AS_InvoiceDetail] 	
	@F_LedgerMaster numeric(18,0)=0,
	@FromDate nvarchar(50)=NULL,
	@ToDate nvarchar(50)=NULL
AS
BEGIN
	SELECT GIH.Id,CAST(GIH.VoucherNo AS NUMERIC) AS VoucherNo,
		(SUBSTRING(GIH.VoucherDate,7,2)+'/'+SUBSTRING(GIH.VoucherDate,5,2)+'/'+SUBSTRING(GIH.VoucherDate,1,4)) AS VoucherDate,
		(LM.Name
		+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
		+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
		+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')
		+ ISNULL((CASE WHEN(ISNULL(OLM.Name,'') <> '') THEN ' ( ' 
			+ CAST(OLM.Name AS NVARCHAR) 
				+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address1 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address2 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(OCM.Name,'') <> '') THEN ', ' + OCM.Name END),'')
			+ ' )' END),'')
		) AS Ledger
		,IM.Name AS Item,GIL.Qty,UMQ.Name AS UnitQ,GIL.Rate,GIL.Free,IMF.Name AS ItemF,UMF.Name AS UnitF
		,LMTax.Name AS TaxName,GIL.TaxableAmt,GIL.TaxAmt,(GIL.TaxAmt+GIL.TaxableAmt) AS Amount
		,TypeDetails = STUFF(			
			(SELECT   ', ' + (TDM.TypeDetail + '-' + RTRIM(LTRIM(STR(SL1.Qty)))) 
			FROM SaleInvoiceL_TypeDetailMaster SL1
				LEFT JOIN TypeDetailMaster TDM ON TDM.Id=SL1.F_TypeDetailMaster
			WHERE SL1.F_SaleInvoiceH = GIH.Id AND SL1.Qty>0 AND SL1.F_ItemMaster=GIL.F_ItemMaster
				AND SL1.TmpTypeId=GIL.TmpTypeId
			ORDER BY TDM.Id 
			FOR XML PATH('')), 1, 2, '')
	FROM [dbo].[SaleInvoiceH] GIH
		LEFT JOIN SaleInvoiceL GIL ON GIH.Id = GIL.F_SaleInvoiceH
		LEFT JOIN LedgerMaster LM ON LM.Id = GIH.F_LedgerMaster
		LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=LM.F_CityMaster
		LEFT JOIN [dbo].[StateMaster] SM ON SM.Id=CM.F_StateMaster	
		
		LEFT JOIN OtherLedgerMaster OLM ON OLM .Id = GIH.F_OtherLedgerMaster
		LEFT JOIN [dbo].[CityMaster] OCM ON OCM.Id=OLM.F_CityMaster
		LEFT JOIN [dbo].[StateMaster] OSM ON OSM.Id=OCM.F_StateMaster	
				
		LEFT JOIN ItemMaster IM ON IM.Id = GIL.F_ItemMaster
		LEFT JOIN UnitMaster UMQ ON UMQ.Id = GIL.F_UnitMasterQ
		LEFT JOIN UnitMaster UMF ON UMF.Id = GIL.F_UnitMasterF
		LEFT JOIN ItemMaster IMF ON IMF.Id = GIL.F_ItemMasterF
		
		LEFT JOIN LedgerMaster LMTax ON LMTax.Id = GIL.F_LedgerMaster_Tax
	WHERE 
		--1=(CASE WHEN (@F_LedgerMaster =0) THEN 1 ELSE (CASE WHEN (GIH.F_LedgerMaster =@F_LedgerMaster ) THEN 1 ELSE 0 END) END)
		(GIH.F_LedgerMaster =@F_LedgerMaster )
		AND
		1=(CASE WHEN(ISNULL(@FromDate ,'')='') THEN 1 ELSE
			(CASE WHEN(GIH.VoucherDate>=ISNULL( @FromDate ,'') AND GIH.VoucherDate<=ISNULL(@ToDate,'')) THEN 1 ELSE 0 END)
			END)			
		AND ISNULL(GIH.Cancelled,0)=0
	
	ORDER BY CAST(GIH.VoucherNo AS NUMERIC),GIH.Id,LM.Name	
END




















































GO
```

# [dbo].[Select_AS_InvoiceRollover]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_InvoiceRollover]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






































CREATE PROCEDURE [dbo].[Select_AS_InvoiceRollover] 	
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_LedgerMaster numeric(18,0)=0
AS
BEGIN	
	SELECT VH.Id,VH.VoucherNo
		,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate
		,((CASE WHEN(LM.Name<>'') THEN LM.Name ELSE VH.LedgerName END)
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')
				+ ISNULL((CASE WHEN(ISNULL(OLM.Name,'') <> '') THEN ' ('
					+ CAST(OLM.Name AS NVARCHAR) 
						+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address1 AS NVARCHAR) END),'')
						+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address2 AS NVARCHAR) END),'')
						+ ISNULL((CASE WHEN(ISNULL(OCM.Name,'') <> '') THEN ', ' + OCM.Name END),'')
					+ ')' END),'')
				) AS Ledger
		,VH.TotalNetAmt
		,MCH.VoucherNo AS ReceiptNoCash
		,(SUBSTRING(MCH.VoucherDate,7,2)+'/'+SUBSTRING(MCH.VoucherDate,5,2)+'/'+SUBSTRING(MCH.VoucherDate,1,4)) AS ReceiptDateCash		
		,MCL.Amount AS ReceiptAmountCash
		,MBH.VoucherNo AS ReceiptNoBank
		,(SUBSTRING(MBH.VoucherDate,7,2)+'/'+SUBSTRING(MBH.VoucherDate,5,2)+'/'+SUBSTRING(MBH.VoucherDate,1,4)) AS ReceiptDateBank		
		,MBL.Amount AS ReceiptAmountBank
		,(VH.TotalNetAmt-
			(
			ISNULL((SELECT SUM(MRL.Amount)
				FROM MoneyReceiptCashL MRL LEFT JOIN MoneyReceiptCashH MRH ON MRH.Id=MRL.F_MoneyReceiptCashH
				WHERE MRL.F_SaleInvoiceH = VH.Id),0)
			+
			ISNULL((SELECT SUM(MRL.Amount)
				FROM MoneyReceiptBankL MRL LEFT JOIN MoneyReceiptBankH MRH ON MRH.Id=MRL.F_MoneyReceiptBankH
				WHERE MRL.F_SaleInvoiceH = VH.Id),0)
			)
		) AS Diff
	FROM SaleInvoiceH VH
	LEFT JOIN LedgerMaster LM ON LM.Id=VH.F_LedgerMaster
	LEFT JOIN CityMaster CM ON CM.Id=LM.F_CityMaster
	LEFT JOIN StateMaster SM ON SM.Id=CM.F_StateMaster
	
	LEFT JOIN OtherLedgerMaster OLM ON OLM.Id = VH.F_OtherLedgerMaster
	LEFT JOIN CityMaster OCM ON OCM.Id=OLM.F_CityMaster
	
	LEFT JOIN MoneyReceiptCashL MCL ON MCL.F_SaleInvoiceH = VH.Id
	LEFT JOIN MoneyReceiptCashH MCH ON MCH.Id=MCL.F_MoneyReceiptCashH
	LEFT JOIN MoneyReceiptBankL MBL ON MBL.F_SaleInvoiceH = VH.Id
	LEFT JOIN MoneyReceiptBankH MBH ON MBH.Id=MBL.F_MoneyReceiptBankH

	WHERE VH.VoucherDate>=@FromDate AND VH.VoucherDate<=@ToDate
		AND 1 = CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(@F_LedgerMaster=LM.Id) THEN 1 ELSE 0 END) END
		AND ISNULL(VH.Cancelled,0)=0
	ORDER BY CAST(VH.VoucherNo AS NUMERIC)
END





























GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_ItemMaster]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






















































CREATE PROCEDURE [dbo].[Select_AS_ItemMaster]
	
AS
BEGIN
	SELECT 
		IGM.Name AS ItemGroup
		,IBM.Name AS ItemBrand
		,IM.Name AS Item
		,IM.Id AS ItemId
		,(CASE WHEN(IM.IsActive=1) THEN '' ELSE 'No' END) AS IsActive
		,ITM.Name AS TypeName
		,TypeDetails = STUFF(			
			(SELECT ', ' + (TDM.TypeDetail ) 
			FROM ItemTypeDetailMaster SL1
				LEFT JOIN TypeDetailMaster TDM ON TDM.Id=SL1.F_TypeDetailMaster
			WHERE SL1.F_ItemMaster = IM.Id AND SL1.IsSelected=1
			ORDER BY SL1.OrderNo,TDM.Id,TDM.TypeDetail
			FOR XML PATH('')), 1, 2, '')
		,UnitDate = STUFF(			
			(SELECT CHAR(10) + 
				((SUBSTRING(SL1.SDate,7,2)+'/'+SUBSTRING(SL1.SDate,5,2)+'/'+SUBSTRING(SL1.SDate,1,4))
					) 
			FROM ItemUnitMaster SL1
				LEFT JOIN UnitMaster TDM ON TDM.Id=SL1.F_UnitMaster
				LEFT JOIN UnitMaster TDMF ON TDMF.Id=SL1.F_UnitMasterEqual
			WHERE SL1.F_ItemMaster = IM.Id
			ORDER BY SL1.Id
			FOR XML PATH('')), 1,1, '')
		,Unit = STUFF(			
			(SELECT CHAR(10) + (TDM.Name) 
			FROM ItemUnitMaster SL1
				LEFT JOIN UnitMaster TDM ON TDM.Id=SL1.F_UnitMaster
				LEFT JOIN UnitMaster TDMF ON TDMF.Id=SL1.F_UnitMasterEqual
			WHERE SL1.F_ItemMaster = IM.Id
			ORDER BY SL1.Id
			FOR XML PATH('')), 1,1, '')
		,UnitQty = STUFF(			
			(SELECT CHAR(10) + CAST(SL1.Qty AS NVARCHAR) 
			FROM ItemUnitMaster SL1
				LEFT JOIN UnitMaster TDM ON TDM.Id=SL1.F_UnitMaster
				LEFT JOIN UnitMaster TDMF ON TDMF.Id=SL1.F_UnitMasterEqual
			WHERE SL1.F_ItemMaster = IM.Id
			ORDER BY SL1.Id
			FOR XML PATH('')), 1,1, '')
		,UnitEqual = STUFF(			
			(SELECT CHAR(10) + (TDMF.Name) 
			FROM ItemUnitMaster SL1
				LEFT JOIN UnitMaster TDM ON TDM.Id=SL1.F_UnitMaster
				LEFT JOIN UnitMaster TDMF ON TDMF.Id=SL1.F_UnitMasterEqual
			WHERE SL1.F_ItemMaster = IM.Id
			ORDER BY SL1.Id
			FOR XML PATH('')), 1,1, '')
		,TaxDate = STUFF(			
			(SELECT CHAR(10) + 
				((SUBSTRING(SL1.SDate,7,2)+'/'+SUBSTRING(SL1.SDate,5,2)+'/'+SUBSTRING(SL1.SDate,1,4))
					) 
			FROM ItemSalesTaxMaster SL1
				LEFT JOIN LedgerMaster TDM ON TDM.Id=SL1.F_LedgerMaster_Tax
			WHERE SL1.F_ItemMaster = IM.Id
			ORDER BY SL1.Id
			FOR XML PATH('')), 1,1, '')
		,TaxName = STUFF(			
			(SELECT CHAR(10) + 
				(TDM.Name ) 
			FROM ItemSalesTaxMaster SL1
				LEFT JOIN LedgerMaster TDM ON TDM.Id=SL1.F_LedgerMaster_Tax
			WHERE SL1.F_ItemMaster = IM.Id
			ORDER BY SL1.Id
			FOR XML PATH('')), 1,1, '')
		,TaxMRP = STUFF(			
			(SELECT CHAR(10) + 
				CAST(SL1.MRP AS NVARCHAR)
			FROM ItemSalesTaxMaster SL1
				LEFT JOIN LedgerMaster TDM ON TDM.Id=SL1.F_LedgerMaster_Tax
			WHERE SL1.F_ItemMaster = IM.Id
			ORDER BY SL1.Id
			FOR XML PATH('')), 1,1, '')
		,TaxSellingRate = STUFF(			
			(SELECT CHAR(10) + 
				CAST(SL1.SellingRate AS NVARCHAR) 
			FROM ItemSalesTaxMaster SL1
				LEFT JOIN LedgerMaster TDM ON TDM.Id=SL1.F_LedgerMaster_Tax
			WHERE SL1.F_ItemMaster = IM.Id
			ORDER BY SL1.Id
			FOR XML PATH('')), 1,1, '')
		,TaxBeforeVATRate = STUFF(			
			(SELECT CHAR(10) + 
				CAST(SL1.BeforeVATRate AS NVARCHAR) 
			FROM ItemSalesTaxMaster SL1
				LEFT JOIN LedgerMaster TDM ON TDM.Id=SL1.F_LedgerMaster_Tax
			WHERE SL1.F_ItemMaster = IM.Id
			ORDER BY SL1.Id
			FOR XML PATH('')), 1,1, '')
	FROM ItemMaster IM
		LEFT JOIN ItemGroupMaster IGM ON IGM.Id = IM.F_ItemGroupMaster
		LEFT JOIN ItemBrandMaster IBM ON IBM.Id = IM.F_ItemBrandMaster
		LEFT JOIN ItemTypeMaster ITM ON ITM.Id = IM.F_ItemTypeMaster	
	ORDER BY IM.Name
END




















































GO
```

# [dbo].[Select_AS_LedgerMaster]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_LedgerMaster]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO









CREATE PROCEDURE [dbo].[Select_AS_LedgerMaster]
	@F_LedgerGroupMaster numeric(18, 0)=0,
	@F_LedgerMaster numeric(18, 0)=0,
	@IsActive smallint=-1,
	@F_CompanyYears numeric(18,0),
	@F_CompanyMaster numeric(18,0)
AS
BEGIN
	--DECLARE @YearId NUMERIC(18,0)
	--SET @YearId = ISNULL((SELECT TOP 1 Id FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC),0)
	--DECLARE @YearId_Current NUMERIC(18,0)
	--SET @YearId_Current = ISNULL((SELECT TOP 1 Id FROM CompanyYears WHERE FinYearFrom<=@FromDate AND F_CompanyMaster = @F_CompanyMaster AND ISNULL(IsAdditional,0)=0 ORDER BY FinYearFrom DESC),0)
	
	DECLARE @IsLedgerBalanceSplit SMALLINT
	SET @IsLedgerBalanceSplit = (SELECT ISNULL(IsLedgerBalanceSplit,0) FROM CompanyYears WHERE Id=@F_CompanyYears)

	IF(@IsLedgerBalanceSplit=0)
		BEGIN
			SELECT LGM.Name AS LedgerGroup,LM.Id,LM.Name
				,0 AS LastOpeningBalance
				,'' AS LastDrCr
				,LM.OpeningBalance
				,(CASE WHEN(LM.DrCr=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
				,(CASE WHEN(LM.IsActive=1) THEN 'Active' ELSE 'In-active' END) AS Active,0 AS Updated
				,CreditLimit_NoOfDays,CreditLimit_NoOfBills,IntRate
			FROM LedgerMaster LM
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
			WHERE 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(@F_LedgerGroupMaster=LM.F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(@F_LedgerMaster=LM.Id) THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(@IsActive=-1) THEN 1 ELSE (CASE WHEN(@IsActive=LM.IsActive) THEN 1 ELSE 0 END) END)
			ORDER BY LGM.Name,LM.Name,LM.Id
		END
	ELSE
		BEGIN
			SELECT LGM.Name AS LedgerGroup,LM.Id,LM.Name
				,LML.LastOpeningBalance
				,(CASE WHEN(LML.LastDrCr=0) THEN 'Dr' ELSE 'Cr' END) AS LastDrCr
				,LML.OpeningBalance
				,(CASE WHEN(LML.DrCr=0) THEN 'Dr' ELSE 'Cr' END) AS DrCr
				,(CASE WHEN(LM.IsActive=1) THEN 'Active' ELSE 'In-active' END) AS Active,0 AS Updated
				,CreditLimit_NoOfDays,CreditLimit_NoOfBills,IntRate
			FROM LedgerMaster LM
			LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id
			LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LML.F_LedgerGroupMaster
			WHERE 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE (CASE WHEN(@F_LedgerGroupMaster=LML.F_LedgerGroupMaster) THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(@F_LedgerMaster=LM.Id) THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(@IsActive=-1) THEN 1 ELSE (CASE WHEN(@IsActive=LM.IsActive) THEN 1 ELSE 0 END) END)
				AND LML.F_CompanyYears = @F_CompanyYears
			ORDER BY LGM.Name,LM.Name,LM.Id
		END
END









GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_MoneyReceiptCashH_Detailed]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



























CREATE PROCEDURE [dbo].[Select_AS_MoneyReceiptCashH_Detailed] 	
	@F_LedgerMaster numeric(18,0)=0,
	@F_OtherLedgerMaster numeric(18,0)=0,
	@F_LedgerMaster_Sales numeric(18,0)=0,
	@F_ManagerMaster numeric(18,0)=0,
	@F_SalesmanMaster numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Cancelled smallint=-1,
	@Id numeric(18,0)=0,	
	@F_SaleInvoiceH numeric(18,0)=0,
	@Amount numeric(18,2)=0,
	@F_MoneyReceiptIssueBookH numeric(18,0)=0,
	@CheckingStatus smallint=-1,
	@ReceiptNo nvarchar(max)=''
AS
BEGIN
	SELECT DSH.Id
		,LMS.Name AS SalesLedger
		,MM.Name AS Manager
		,SM.Name AS Salesman
		,DSH.VoucherNo
		,(SUBSTRING(DSH.VoucherDate,7,2)+'/'+SUBSTRING(DSH.VoucherDate,5,2)+'/'+SUBSTRING(DSH.VoucherDate,1,4)) AS VoucherDate
		,(LM.Name
		+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
		+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
		+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),''))
		AS Ledger
		
		,(OLM.Name
			+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address1 AS NVARCHAR) END),'')
			+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address2 AS NVARCHAR) END),'')
			+ ISNULL((CASE WHEN(ISNULL(OCM.Name,'') <> '') THEN ', ' + OCM.Name END),''))		
		AS OtherLedger
		
		,(CASE WHEN (DSH.IsFullPaid=0) THEN '' ELSE 'Yes' END) AS FullPaid
		,(DSH.TotalAmount + ISNULL(DSH.ChequeCancelCharges,0)) AS TotalAmount
		,DSH.Discount
		,(DSH.NetAmount + ISNULL(DSH.ChequeCancelCharges,0)) AS NetAmount
		,DSH.TotalInvoiceAmt
		,DSH.TotalDiffAmt
		,DSH.ChequeCancelCharges
		,(CASE WHEN (ISNULL(DSH.Cancelled,0)=1) THEN 'Cancelled' ELSE '' END) AS Cancelled
		,(CASE WHEN(ISNULL(DSH.IsChecked,0)=0) THEN '' ELSE 'Checked' END) AS Checked
		,LGM.Name AS LedgerGroup
		,SH.VoucherNo AS InvNo
		,(SUBSTRING(SH.VoucherDate,7,2)+'/'+SUBSTRING(SH.VoucherDate,5,2)+'/'+SUBSTRING(SH.VoucherDate,1,4)) AS InvDate
		,MAX(DSL.Amount) AS InvAmt,DSL.Id AS InvId
	FROM MoneyReceiptCashH DSH
		LEFT JOIN MoneyReceiptCashL DSL ON DSH.Id=DSL.F_MoneyReceiptCashH
		LEFT JOIN SaleInvoiceH SH ON SH.Id=DSL.F_SaleInvoiceH
		LEFT JOIN LedgerMaster LM ON LM.Id = DSH.F_LedgerMaster
		LEFT JOIN LedgerGroupMaster LGM ON LGM.Id = LM.F_LedgerGroupMaster
		LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=LM.F_CityMaster
		
		LEFT JOIN OtherLedgerMaster OLM ON OLM .Id = DSH.F_OtherLedgerMaster
		LEFT JOIN [dbo].[CityMaster] OCM ON OCM.Id=OLM.F_CityMaster
		
		LEFT JOIN LedgerMaster LMS ON LMS.Id = DSH.F_LedgerMaster_Sales
		LEFT JOIN ManagerMaster MM ON MM.Id = DSH.F_ManagerMaster
		LEFT JOIN SalesmanMaster SM ON SM.Id = DSH.F_SalesmanMaster
	WHERE
		1=(CASE WHEN (@Id=0) THEN 1 ELSE (CASE WHEN (DSH.Id =@Id) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN (DSH.F_LedgerMaster =@F_LedgerMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_OtherLedgerMaster=0) THEN 1 ELSE (CASE WHEN (DSH.F_OtherLedgerMaster =@F_OtherLedgerMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_LedgerMaster_Sales=0) THEN 1 ELSE (CASE WHEN (DSH.F_LedgerMaster_Sales =@F_LedgerMaster_Sales) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_ManagerMaster=0) THEN 1 ELSE (CASE WHEN (DSH.F_ManagerMaster =@F_ManagerMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_SalesmanMaster=0) THEN 1 ELSE (CASE WHEN (DSH.F_SalesmanMaster =@F_SalesmanMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN(@FromDate='') THEN 1 ELSE
			(CASE WHEN(DSH.VoucherDate>=@FromDate AND DSH.VoucherDate<=@ToDate) THEN 1 ELSE 0 END)
			END)
		AND
		1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(DSH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_SaleInvoiceH=0) THEN 1 ELSE (CASE WHEN (DSL.F_SaleInvoiceH =@F_SaleInvoiceH) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@Amount=0) THEN 1 ELSE
			(CASE WHEN (DSH.TotalAmount =@Amount
						OR DSH.Discount =@Amount
						OR DSH.NetAmount =@Amount
						) THEN 1 ELSE 0 END)
		   END)
		AND
		1 =(CASE WHEN(@F_MoneyReceiptIssueBookH=0) THEN 1 ELSE
				(CASE WHEN (
							DSH.VoucherNo>=(SELECT TOP 1 ReceiptStartNo FROM MoneyReceiptIssueBookH WHERE Id=@F_MoneyReceiptIssueBookH)
							AND
							DSH.VoucherNo<=(SELECT TOP 1 ReceiptEndNo FROM MoneyReceiptIssueBookH WHERE Id=@F_MoneyReceiptIssueBookH)
							)
					THEN 1 ELSE 0
				END)
			END)
		AND
		1=(CASE WHEN (@CheckingStatus=-1) THEN 1 ELSE (CASE WHEN (ISNULL(DSH.IsChecked,0) =@CheckingStatus) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@ReceiptNo='') THEN 1 ELSE (CASE WHEN (DSH.Id  IN(SELECT * FROM Split(@ReceiptNo,',')) ) THEN 1 ELSE 0 END) END)		
			
	GROUP BY DSH.Id,LMS.Name,MM.Name,SM.Name,DSH.VoucherNo,DSH.VoucherDate,LM.Name,OLM.Name,DSH.IsFullPaid,DSH.TotalAmount,DSH.Discount,DSH.TotalInvoiceAmt,DSH.TotalDiffAmt,DSH.Cancelled,DSH.NetAmount,SH.VoucherNo,SH.VoucherDate,DSL.Id,
		LM.Address1,LM.Address2,CM.Name,
		OLM.Address1,OLM.Address2,OCM.Name,DSH.ChequeCancelCharges,
		DSH.IsChecked,LGM.Name
	ORDER BY DSH.VoucherDate,CAST(DSH.VoucherNo AS NUMERIC),DSH.Id,LM.Name
END





















































GO
```

# [dbo].[Select_AS_MoneyReceiptIssueBookH]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_MoneyReceiptIssueBookH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

























































CREATE PROCEDURE [dbo].[Select_AS_MoneyReceiptIssueBookH] 	
	@F_SalesmanMaster numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Cancelled smallint=-1,
	@Id numeric(18,0)=0
AS
BEGIN
	SELECT DSH.BookNo,('B-' + DSH.BookNo + ' ' + CAST(DSH.ReceiptStartNo AS NVARCHAR) + ' To ' + CAST(DSH.ReceiptEndNo AS NVARCHAR)
			+ ' ' + (SUBSTRING(DSH.VoucherDate,7,2)+'/'+SUBSTRING(DSH.VoucherDate,5,2)+'/'+SUBSTRING(DSH.VoucherDate,1,4))
			+ ' ' + SM.Name
			) AS BookHead,
			DSH.VoucherDate,
			DSH.ReceiptStartNo,DSH.ReceiptEndNo,
		ReceiptNoUsed = STUFF((SELECT ', ' + MRH.VoucherNo
							FROM MoneyReceiptCashH MRH
							WHERE MRH.VoucherNo >= DSH.ReceiptStartNo AND MRH.VoucherNo <=DSH.ReceiptEndNo AND ISNULL(MRH.Cancelled,0)=0
							ORDER BY CAST(MRH.VoucherNo AS NUMERIC)
						FOR XML PATH('')), 1, 2, ''),
		ReceiptNoCancelled = STUFF((SELECT ', ' + MRH.VoucherNo
							FROM MoneyReceiptCashH MRH
							WHERE MRH.VoucherNo >= DSH.ReceiptStartNo AND MRH.VoucherNo <=DSH.ReceiptEndNo AND ISNULL(MRH.Cancelled,0)=1
							ORDER BY CAST(MRH.VoucherNo AS NUMERIC)
						FOR XML PATH('')), 1, 2, ''),
		ReceiptNoUsed_Date = STUFF((SELECT ', ' + MRH.VoucherDate
							FROM MoneyReceiptCashH MRH
							WHERE MRH.VoucherNo >= DSH.ReceiptStartNo AND MRH.VoucherNo <=DSH.ReceiptEndNo AND ISNULL(MRH.Cancelled,0)=0
							ORDER BY CAST(MRH.VoucherNo AS NUMERIC)
						FOR XML PATH('')), 1, 2, ''),
		ReceiptNoCancelled_Date = STUFF((SELECT ', ' + MRH.VoucherDate
							FROM MoneyReceiptCashH MRH
							WHERE MRH.VoucherNo >= DSH.ReceiptStartNo AND MRH.VoucherNo <=DSH.ReceiptEndNo AND ISNULL(MRH.Cancelled,0)=1
							ORDER BY CAST(MRH.VoucherNo AS NUMERIC)
						FOR XML PATH('')), 1, 2, '')
	FROM MoneyReceiptIssueBookH DSH
		LEFT JOIN SalesmanMaster SM ON SM.Id=DSH.F_SalesmanMaster
	WHERE 
		1=(CASE WHEN (@Id=0) THEN 1 ELSE (CASE WHEN (DSH.Id =@Id) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_SalesmanMaster=0) THEN 1 ELSE (CASE WHEN (DSH.F_SalesmanMaster =@F_SalesmanMaster) THEN 1 ELSE 0 END) END)		
		AND
		1=(CASE WHEN(@FromDate='') THEN 1 ELSE
			(CASE WHEN(DSH.VoucherDate>=@FromDate AND DSH.VoucherDate<=@ToDate) THEN 1 ELSE 0 END)
			END)
		AND
		1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(DSH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)

	ORDER BY CAST(LEFT(DSH.BookNo, Patindex('%[^0-9]%', DSH.BookNo + 'x') - 1) AS INT),DSH.VoucherDate,DSH.Id
END























































GO
```

# [dbo].[Select_AS_PartyLoyaltyProgramH]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_PartyLoyaltyProgramH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




















CREATE PROCEDURE [dbo].[Select_AS_PartyLoyaltyProgramH]
	@F_PartyLoyaltyProgramH numeric(18,0),
	@F_ItemBrandMaster numeric(18,0)=0,
	@IsWithoutTax smallint=0
AS
BEGIN
	SELECT TAB1.Ledger,TAB1.F_LedgerMaster,TAB1.Target,MAX(TAB1.Achievement) AS Achievement,MAX(TAB1.BTG) AS BTG,MAX(TAB1.AchievementPercentage) AS AchievementPercentage
	FROM (
	SELECT (LM.Name + ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
					 + ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
					 + ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')) AS Ledger
		,GIL.F_LedgerMaster
		,GIL.Target
		,(CASE WHEN(@IsWithoutTax=0) THEN SUM(SL.TaxableAmt + SL.TaxAmt + ISNULL(SL.TaxAmt2,0)) ELSE SUM(SL.TaxableAmt) END) AS Achievement
		,((CASE WHEN(@IsWithoutTax=0) THEN SUM(SL.TaxableAmt + SL.TaxAmt + ISNULL(SL.TaxAmt2,0)) ELSE SUM(SL.TaxableAmt) END)-GIL.Target) AS BTG
		,CAST(ROUND(((CASE WHEN(@IsWithoutTax=0) THEN SUM(SL.TaxableAmt + SL.TaxAmt + ISNULL(SL.TaxAmt2,0)) ELSE SUM(SL.TaxableAmt) END)/GIL.Target)*100,2) AS NUMERIC(18,2)) AS AchievementPercentage
	FROM PartyLoyaltyProgramL GIL
		LEFT JOIN PartyLoyaltyProgramH GIH ON GIH.Id=GIL.F_PartyLoyaltyProgramH
		LEFT JOIN LedgerMaster LM ON LM.Id = GIL.F_LedgerMaster
		LEFT JOIN CityMaster CM ON CM.Id=LM.F_CityMaster
		LEFT JOIN SaleInvoiceH SH ON SH.F_LedgerMaster = LM.Id
		LEFT JOIN SaleInvoiceL SL ON SL.F_SaleInvoiceH = SH.Id
		LEFT JOIN ItemMaster IM ON IM.Id = SL.F_ItemMaster
		LEFT JOIN ItemBrandMaster IBM ON IBM.Id = IM.F_ItemBrandMaster
	WHERE GIL.F_PartyLoyaltyProgramH = @F_PartyLoyaltyProgramH AND IM.F_ItemBrandMaster = @F_ItemBrandMaster
		AND (SH.VoucherDate >=GIH.VoucherDate AND SH.VoucherDate<=GIH.ToDate)
	GROUP BY LM.Name,LM.Address1,LM.Address2,LM.Address3,CM.Name,GIL.F_LedgerMaster,GIL.Target

	UNION ALL

	SELECT (LM.Name + ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
					 + ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
					 + ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')) AS Ledger
		,GIL.F_LedgerMaster
		,GIL.Target
		,0 AS Achievement
		,(0-GIL.Target) AS BTG
		,0 AS AchievementPercentage
	FROM PartyLoyaltyProgramL GIL
		LEFT JOIN PartyLoyaltyProgramH GIH ON GIH.Id=GIL.F_PartyLoyaltyProgramH
		LEFT JOIN LedgerMaster LM ON LM.Id = GIL.F_LedgerMaster
		LEFT JOIN CityMaster CM ON CM.Id=LM.F_CityMaster
	WHERE GIL.F_PartyLoyaltyProgramH = @F_PartyLoyaltyProgramH
	GROUP BY LM.Name,LM.Address1,LM.Address2,LM.Address3,CM.Name,GIL.F_LedgerMaster,GIL.Target
	) AS TAB1
	GROUP BY TAB1.Ledger,TAB1.F_LedgerMaster,TAB1.Target
	ORDER BY TAB1.Ledger
END

















































GO
```

# [dbo].[Select_AS_PartyRateH]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_PartyRateH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

























































CREATE PROCEDURE [dbo].[Select_AS_PartyRateH] 	
	@F_LedgerMaster numeric(18,0)=0,
	@F_ItemMaster numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Cancelled smallint=-1,
	@Id numeric(18,0)=0
AS
BEGIN
	SELECT DSH.Id,DSH.VoucherNo,
		(SUBSTRING(DSH.VoucherDate,7,2)+'/'+SUBSTRING(DSH.VoucherDate,5,2)+'/'+SUBSTRING(DSH.VoucherDate,1,4)) AS VoucherDate,
		IBM.Name AS Ledger,
		(CASE WHEN (ISNULL(DSH.Cancelled,0)=1) THEN 'Cancelled' ELSE '' END) AS Cancelled
	FROM PartyRateH DSH
		LEFT JOIN PartyRateL DSL ON DSH.Id=DSL.F_PartyRateH
		LEFT JOIN LedgerMaster IBM ON IBM.Id = DSH.F_LedgerMaster
		LEFT JOIN ItemMaster IM ON IM.Id = DSL.F_ItemMaster
	WHERE 
		1=(CASE WHEN (@Id=0) THEN 1 ELSE (CASE WHEN (DSH.Id =@Id) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN (DSH.F_LedgerMaster =@F_LedgerMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_ItemMaster=0) THEN 1 ELSE (CASE WHEN (DSL.F_ItemMaster =@F_ItemMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN(@FromDate='') THEN 1 ELSE
			(CASE WHEN(DSH.VoucherDate>=@FromDate AND DSH.VoucherDate<=@ToDate) THEN 1 ELSE 0 END)
			END)
		AND
		1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(DSH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
	GROUP BY DSH.Id,DSH.VoucherNo,DSH.VoucherDate,IBM.Name,DSH.Cancelled
	ORDER BY DSH.VoucherDate,CAST(DSH.VoucherNo AS NUMERIC),DSH.Id,IBM.Name
END























































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_PartyWiseItemSaleDetail]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






















































CREATE PROCEDURE [dbo].[Select_AS_PartyWiseItemSaleDetail] 	
	@FromDate nvarchar(50)=NULL,
	@ToDate nvarchar(50)=NULL,
	@LedgerIds nvarchar(max)='',
	@ItemIds nvarchar(max)='',
	@OrderBy nvarchar(max)='CAST(GIH.VoucherNo AS NUMERIC)',
	@OtherLedgerIds nvarchar(max)='',
	@AreaIds nvarchar(max)='',
	@BrandIds nvarchar(max)=''
AS
BEGIN
	DECLARE @SQL nvarchar(max)
		
	SET @SQL = '
	SELECT GIH.Id,CAST(GIH.VoucherNo AS NUMERIC) AS VoucherNo,
		(SUBSTRING(GIH.VoucherDate,7,2)+''/''+SUBSTRING(GIH.VoucherDate,5,2)+''/''+SUBSTRING(GIH.VoucherDate,1,4)) AS VoucherDate,
		(LM.Name
		+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'''') <> '''') THEN '', '' + CAST(LM.Address1 AS NVARCHAR) END),'''')
		+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'''') <> '''') THEN '', '' + CAST(LM.Address2 AS NVARCHAR) END),'''')
		+ ISNULL((CASE WHEN(ISNULL(CM.Name,'''') <> '''') THEN '', '' + CM.Name END),'''')
		+ ISNULL((CASE WHEN(ISNULL(OLM.Name,'''') <> '''') THEN '' ('' 
			+ CAST(OLM.Name AS NVARCHAR) 
				+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address1 AS NVARCHAR),'''') <> '''') THEN '', '' + CAST(OLM.Address1 AS NVARCHAR) END),'''')
				+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address2 AS NVARCHAR),'''') <> '''') THEN '', '' + CAST(OLM.Address2 AS NVARCHAR) END),'''')
				+ ISNULL((CASE WHEN(ISNULL(OCM.Name,'''') <> '''') THEN '', '' + OCM.Name END),'''')
			+ '')'' END),'''')
		) AS Ledger
		,IM.Name AS Item,GIL.Qty,	
		( DENSE_RANK() OVER (ORDER BY CAST(GIH.VoucherNo AS NUMERIC) )) As SNo
		--,LMTax.Name AS TaxName,GIL.TaxableAmt,GIL.TaxAmt,(GIL.TaxAmt+GIL.TaxableAmt) AS Amount
	FROM [dbo].[SaleInvoiceH] GIH
		LEFT JOIN SaleInvoiceL GIL ON GIH.Id = GIL.F_SaleInvoiceH
		LEFT JOIN LedgerMaster LM ON LM.Id = GIH.F_LedgerMaster
		LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=LM.F_CityMaster
		LEFT JOIN [dbo].[StateMaster] SM ON SM.Id=CM.F_StateMaster	
		
		LEFT JOIN OtherLedgerMaster OLM ON OLM .Id = GIH.F_OtherLedgerMaster
		LEFT JOIN [dbo].[CityMaster] OCM ON OCM.Id=OLM.F_CityMaster
		LEFT JOIN [dbo].[StateMaster] OSM ON OSM.Id=OCM.F_StateMaster	
				
		LEFT JOIN ItemMaster IM ON IM.Id = GIL.F_ItemMaster
		LEFT JOIN UnitMaster UMQ ON UMQ.Id = GIL.F_UnitMasterQ
		LEFT JOIN UnitMaster UMF ON UMF.Id = GIL.F_UnitMasterF
		LEFT JOIN ItemMaster IMF ON IMF.Id = GIL.F_ItemMasterF
		
		LEFT JOIN LedgerMaster LMTax ON LMTax.Id = GIL.F_LedgerMaster_Tax
	WHERE 
		1=(CASE WHEN (''' + @AreaIds + '''='''') THEN 1 ELSE (CASE WHEN (LM.F_AreaMaster  IN(SELECT * FROM Split(''' + @AreaIds + ''','','')) ) THEN 1 ELSE 0 END) END)		
		AND
		1=(CASE WHEN (''' + @BrandIds + '''='''') THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster  IN(SELECT * FROM Split(''' + @BrandIds + ''','','')) ) THEN 1 ELSE 0 END) END)		
		AND
		1=(CASE WHEN (''' + @LedgerIds + '''='''') THEN 1 ELSE (CASE WHEN (LM.Id  IN(SELECT * FROM Split(''' + @LedgerIds + ''','','')) ) THEN 1 ELSE 0 END) END)		
		AND
		1=(CASE WHEN (''' + @OtherLedgerIds + '''='''') THEN 1 ELSE (CASE WHEN (OLM.Id  IN(SELECT * FROM Split(''' + @OtherLedgerIds + ''','','')) ) THEN 1 ELSE 0 END) END)		
		AND
		1=(CASE WHEN (''' + @ItemIds + '''='''') THEN 1 ELSE (CASE WHEN (IM.Id  IN(SELECT * FROM Split('''+ @ItemIds + ''','','')) ) THEN 1 ELSE 0 END) END)		
		AND
		1=(CASE WHEN(ISNULL(' + @FromDate + ','''')='''') THEN 1 ELSE
			(CASE WHEN(GIH.VoucherDate>=ISNULL( ' + @FromDate + ','''') AND GIH.VoucherDate<=ISNULL(' + @ToDate + ','''')) THEN 1 ELSE 0 END)
			END)			
		AND ISNULL(GIH.Cancelled,0)=0
	
	ORDER BY ' + @OrderBy + ',IM.Name'
	EXEC (@SQL)
END




















































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_PurchaseInvoiceH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
























CREATE PROCEDURE [dbo].[Select_AS_PurchaseInvoiceH] 	
	@F_LedgerMaster numeric(18,0)=0,
	@F_ItemMaster numeric(18,0)=0,
	@FromDate nvarchar(50)=NULL,
	@ToDate nvarchar(50)=NULL,
	@Cancelled smallint=-1,
	@VoucherNo nvarchar(50)='',
	@Id numeric(18,0)=0,
	@OrderBy nvarchar(50)='TAB1.VoucherDate2 DESC',
	@ShowOnlyDiffInvoices smallint=0,
	@F_ItemGroupMaster numeric(18,0)=0,
	@F_ItemBrandMaster numeric(18,0)=0,
	@VATReport smallint=-1,
	@HavingTINNo smallint=-1,
	@OwnState smallint=-1,
	@RowNumber nvarchar(max)=',DENSE_RANK() OVER (ORDER BY TAB1.Ledger ASC) AS RowNumber',
	@PrintBook smallint=0,
	@CheckingStatus smallint=-1,
	@VAT8 smallint,
	@MonthWiseFields nvarchar(max)='TAB1.MonthName,TAB1.MonthNo',
	@TableNameH nvarchar(50),
	@TableNameL nvarchar(50)
AS
BEGIN
--ITEM DETAILS
	DECLARE @GIL_FIELDS NVARCHAR(MAX)
	DECLARE @GIL_JOIN NVARCHAR(MAX)
	DECLARE @GIL_WHERE NVARCHAR(MAX)
	DECLARE @ITEM_FIELDS NVARCHAR(MAX)
	DECLARE @ITEM_JOIN NVARCHAR(MAX)
	SET @ITEM_FIELDS=''
	SET @ITEM_JOIN=''
--VOUCHER FIELDS
	IF(@TableNameH='PurchaseVoucherH' OR @TableNameH='PurchaseReturnVoucherH')	
		BEGIN
			SET @GIL_FIELDS = ',0 AS GridTaxAmt
							,0 AS DiffTaxableAmt
							,0 AS GridFinalAmt
							,0 AS DiffFinalAmt'
			SET @GIL_JOIN = ' LEFT JOIN ItemBrandMaster IBM ON IBM.Id=GIH.F_ItemBrandMaster'
			SET @GIL_WHERE = ' AND 1=(CASE WHEN (' + CAST(@F_ItemBrandMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (GIH.F_ItemBrandMaster =' + CAST(@F_ItemBrandMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)'
		END
	ELSE
		BEGIN
		
			SET @GIL_FIELDS = ',SUM(GIL.TaxAmt) AS GridTaxAmt
							,((GIH.TotalTaxationAmt)-SUM(GIL.TaxAmt)) AS DiffTaxableAmt
							,SUM(GIL.ActualAmount) AS GridFinalAmt
							,((GIH.TotalGrossAmt)-SUM(GIL.ActualAmount)) AS DiffFinalAmt'
			
			SET @GIL_JOIN = ' LEFT JOIN ' + @TableNameL + ' GIL ON GIH.Id = GIL.F_' + @TableNameH + '
							LEFT JOIN ItemMaster IM ON IM.Id=GIL.F_ItemMaster
							LEFT JOIN LedgerMaster LM_Tax ON LM_Tax.Id = GIL.F_LedgerMaster_Tax
							LEFT JOIN ItemGroupMaster IGM ON IGM.Id=IM.F_ItemGroupMaster
							LEFT JOIN ItemBrandMaster IBM ON IBM.Id=IM.F_ItemBrandMaster'
			SET @GIL_WHERE = ' AND
							1=(CASE WHEN (' + CAST(@F_ItemMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (GIL.F_ItemMaster =' + CAST(@F_ItemMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
							AND
							1=(CASE WHEN (' + CAST(@F_ItemGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemGroupMaster =' + CAST(@F_ItemGroupMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
							AND
							1=(CASE WHEN (' + CAST(@F_ItemBrandMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster =' + CAST(@F_ItemBrandMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)'
		END
--
	IF(@PrintBook>0)
		BEGIN
			SET @ITEM_FIELDS = ',IM.Name AS Item
								,IM.Id AS ItemId
								,IM.Remarks
								,GIL.MRP,UMQ.Name AS UnitQ,GIL.Qty,GIL.Rate
								,GIL.Free,IMF.Name AS ItemF,UMF.Name AS UnitF
								,GIL.ProductDiscRate,GIL.ProductDiscAmt,GIL.CashDiscRate,GIL.CashDiscAmt
								,LMTax.Name AS TaxName,GIL.Amount,GIL.ActualAmount AS FinalAmt
								,UMQ.Id AS UnitQId,UMF.Id AS UnitFId,LMTax.Id AS TaxId,GIL.Id,GIL.IsManualProDis,GIL.IsManualCashDis,GIL.TaxAmt,GIL.TaxableAmt,GIL.TaxRate,GIL.MinQty,GIL.MinQtyF
								,IMF.Id AS ItemFId								
								,GIL.TmpTypeId
								,TypeDetails = STUFF(			
									(SELECT   '', '' + (TDM.TypeDetail + ''-'' + RTRIM(LTRIM(STR(SL1.Qty)))) 
									FROM SaleInvoiceL_TypeDetailMaster SL1
										LEFT JOIN TypeDetailMaster TDM ON TDM.Id=SL1.F_TypeDetailMaster
									WHERE SL1.F_SaleInvoiceH = TAB1.Id AND SL1.Qty>0 AND SL1.F_ItemMaster=GIL.F_ItemMaster
										AND SL1.TmpTypeId=GIL.TmpTypeId
									ORDER BY TDM.Id 
									FOR XML PATH('''')), 1, 2, '''')'
			SET @ITEM_JOIN = ' LEFT JOIN ' + @TableNameL + ' GIL ON TAB1.Id = GIL.F_' + @TableNameH + '
								LEFT JOIN ItemMaster IM ON IM.Id = GIL.F_ItemMaster
								LEFT JOIN UnitMaster UMQ ON UMQ.Id = GIL.F_UnitMasterQ
								LEFT JOIN UnitMaster UMF ON UMF.Id = GIL.F_UnitMasterF
								LEFT JOIN ItemMaster IMF ON IMF.Id = GIL.F_ItemMasterF
								
								LEFT JOIN LedgerMaster LMTax ON LMTax.Id = GIL.F_LedgerMaster_Tax '
		END

	DECLARE @SQL nvarchar(max)
	DECLARE @WHERE nvarchar(max)
	IF(@ShowOnlyDiffInvoices=1)
		BEGIN
			SET @WHERE=' WHERE TAB1.DiffTaxableAmt<>0 OR TAB1.DiffFinalAmt<>0'
		END
	ELSE
		BEGIN
			SET @WHERE = ''
		END
--LM.Name AS Ledger,
	SET @SQL = '
	SELECT TAB1.Id,TAB1.VoucherNo,TAB1.VoucherDate,TAB1.PurVoucherNo,TAB1.PurVoucherDate,TAB1.Ledger,TAB1.IsOtherState
		,TAB1.TotalProDisAmt
		,TAB1.TotalCashDisAmt
		,TAB1.TotalGrossAmt
		,TAB1.TotalTaxationAmt,TAB1.AddLess
		,TAB1.TotalNetAmt	
		,TAB1.LedgerId	
		,TAB1.GridTaxAmt
		,TAB1.DiffTaxableAmt
		,TAB1.GridFinalAmt
		,TAB1.DiffFinalAmt
		,TAB1.Cancelled,TAB1.Checked
		' +  @ITEM_FIELDS + '
	FROM(		
		
		SELECT GIH.Id,GIH.VoucherNo,GIH.VoucherDate AS VoucherDate2,
		(SUBSTRING(GIH.VoucherDate,7,2)+''/''+SUBSTRING(GIH.VoucherDate,5,2)+''/''+SUBSTRING(GIH.VoucherDate,1,4)) AS VoucherDate
		,GIH.PurVoucherDate AS VoucherDate1
		,GIH.PurVoucherNo,
		(SUBSTRING(GIH.PurVoucherDate,7,2)+''/''+SUBSTRING(GIH.PurVoucherDate,5,2)+''/''+SUBSTRING(GIH.PurVoucherDate,1,4)) AS PurVoucherDate,
		(LM.Name
		+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'''') <> '''') THEN '', '' + CAST(LM.Address1 AS NVARCHAR) END),'''')
		+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'''') <> '''') THEN '', '' + CAST(LM.Address2 AS NVARCHAR) END),'''')
		+ ISNULL((CASE WHEN(ISNULL(CM.Name,'''') <> '''') THEN '', '' + CM.Name END),'''')
		
		) AS Ledger,
		(CASE WHEN (GIH.IsOtherState=0) THEN '''' ELSE ''Yes'' END) AS IsOtherState
		,GIH.TotalProDisAmt
		,GIH.TotalCashDisAmt
		,GIH.TotalGrossAmt		
		,(SELECT ISNULL(SUM(SLT.AddAmt),0)-ISNULL(SUM(SLT.LessAmt),0)
			FROM ' + @TableNameL + 'Tax SLT
			LEFT JOIN ' + @TableNameH + ' SH ON SH.Id = SLT.F_' + @TableNameH + '
			LEFT JOIN LedgerMaster LMT ON LMT.Id=SLT.F_LedgerMaster
			WHERE SLT.F_' + @TableNameH + ' = GIH.Id AND LMT.F_LedgerGroupMaster = 12) AS TotalTaxationAmt
		,(SELECT ISNULL(SUM(SLT.AddAmt),0)-ISNULL(SUM(SLT.LessAmt),0)
			FROM ' + @TableNameL + 'Tax SLT
			LEFT JOIN ' + @TableNameH + ' SH ON SH.Id = SLT.F_' + @TableNameH + '
			LEFT JOIN LedgerMaster LMT ON LMT.Id=SLT.F_LedgerMaster
			WHERE SLT.F_' + @TableNameH + ' = GIH.Id AND LMT.F_LedgerGroupMaster <> 12) AS AddLess
		,GIH.TotalNetAmt	
		,LM.Id AS LedgerId ' + @GIL_FIELDS + '		
		,(CASE WHEN (ISNULL(GIH.Cancelled,0)=1) THEN ''Cancelled'' ELSE '''' END) AS Cancelled
		,(CASE WHEN(ISNULL(GIH.IsChecked,0)=0) THEN '''' ELSE ''Checked'' END) AS Checked
		' + '
	FROM [dbo].[' + @TableNameH + '] GIH ' + @GIL_JOIN + '
		LEFT JOIN LedgerMaster LM ON LM.Id = GIH.F_LedgerMaster
		LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=LM.F_CityMaster
		LEFT JOIN [dbo].[StateMaster] SM ON SM.Id=CM.F_StateMaster	
	WHERE 
		1=(CASE WHEN (' + CAST(@Id AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (GIH.Id =' + CAST(@Id AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (''' + CAST(@VoucherNo AS NVARCHAR) + '''='''') THEN 1 ELSE (CASE WHEN (GIH.VoucherNo =''' + CAST(@VoucherNo AS NVARCHAR) + ''') THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (GIH.F_LedgerMaster =' + CAST(@F_LedgerMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
		' + @GIL_WHERE + '
		AND
		1=(CASE WHEN(ISNULL(' + @FromDate + ','''')='''') THEN 1 ELSE
			(CASE WHEN(GIH.VoucherDate>=ISNULL(' + @FromDate + ','''') AND GIH.VoucherDate<=ISNULL(' + @ToDate + ','''')) THEN 1 ELSE 0 END)
			END)			
		AND
		1=(CASE WHEN (' + CAST(@Cancelled AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN (ISNULL(GIH.Cancelled,0) =' + CAST(@Cancelled AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (' + CAST(@CheckingStatus AS NVARCHAR) + '=-1) THEN 1 ELSE (CASE WHEN (ISNULL(GIH.IsChecked,0) =' + CAST(@CheckingStatus AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
	GROUP BY GIH.Id,GIH.VoucherNo,GIH.VoucherDate,GIH.PurVoucherNo,GIH.PurVoucherDate,LM.Name,GIH.IsOtherState,GIH.TotalProDisAmt,GIH.TotalCashDisAmt,GIH.TotalGrossAmt,GIH.TotalTaxationAmt,GIH.TotalNetAmt,LM.Id,GIH.Cancelled
			,LM.Address1,LM.Address2,CM.Name,GIH.IsChecked
	
	) AS TAB1 ' + @ITEM_JOIN  +
	@WHERE +
	' ORDER BY ' + @OrderBy + ',TAB1.Id'
	EXEC (@SQL)

END




































GO
```

# [dbo].[Select_AS_StockLedger]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_StockLedger]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






















CREATE PROCEDURE [dbo].[Select_AS_StockLedger] 
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_LedgerMaster numeric(18,0)=0,
	@F_ItemMaster numeric(18,0)=0,
	@F_ItemBrandMaster numeric(18,0)=0,
	@F_ItemGroupMaster numeric(18,0)=0,
	@SMonth nvarchar(50)='',
	@MRP numeric(18,2)=0
AS
BEGIN
--QPSAmt -> TotalWeight
	DECLARE @FIELDS_IN NVARCHAR(MAX)
	DECLARE @FIELDS_OUT NVARCHAR(MAX)
	DECLARE @WHERE NVARCHAR(MAX)
	DECLARE @FIELDS_IN_OP NVARCHAR(MAX)
	DECLARE @FIELDS_OUT_OP NVARCHAR(MAX)
	DECLARE @WHERE_OP NVARCHAR(MAX)
	SET @FIELDS_IN='SH.Id,SH.VoucherNo
					,(SUBSTRING(SH.VoucherDate,7,2)+''/''+SUBSTRING(SH.VoucherDate,5,2)+''/''+SUBSTRING(SH.VoucherDate,1,4)) AS VoucherDate
					,LM.Name AS Ledger
					,(SL.MinQty+SL.MinQtyF) AS InQty,0 AS OutQty
					,(0) AS InWeight,0 AS OutWeight
					,(SL.ActualAmount) AS InAmt,0 AS OutAmt
					,(SL.MinQty+SL.MinQtyF) AS BalPcs,(0) AS BalWeight,(SL.ActualAmount) AS BalAmt
					,SH.VoucherDate AS VoucherDate1'
	SET @FIELDS_IN_OP='SUM(SL.MinQty+SL.MinQtyF) AS InQty,0 AS OutQty
					,0 AS InWeight,0 AS OutWeight
					,SUM(SL.ActualAmount) AS InAmt,0 AS OutAmt
					,SUM(SL.MinQty+SL.MinQtyF) AS BalPcs,0 AS BalWeight,SUM(SL.ActualAmount) AS BalAmt,'''' AS VoucherDate1
					,'''' AS BillNo'
	SET @FIELDS_OUT ='SH.Id,SH.VoucherNo
					,(SUBSTRING(SH.VoucherDate,7,2)+''/''+SUBSTRING(SH.VoucherDate,5,2)+''/''+SUBSTRING(SH.VoucherDate,1,4)) AS VoucherDate
					,LM.Name AS Ledger
					,0 AS InQty,(SL.MinQty+SL.MinQtyF) AS OutQty
					,0 AS InWeight,(0) AS OutWeight
					,0 AS InAmt,(SL.FinalAmt) AS OutAmt
					,((SL.MinQty+SL.MinQtyF)*-1) AS BalPcs,((0)*-1) AS BalWeight,(SL.FinalAmt*-1) AS BalAmt
					,SH.VoucherDate AS VoucherDate1'
	SET @FIELDS_OUT_OP ='0 AS InQty,SUM(SL.MinQty+SL.MinQtyF) AS OutQty
					,0 AS InWeight,0 AS OutWeight
					,0 AS InAmt,SUM(SL.FinalAmt) AS OutAmt
					,SUM((SL.MinQty+SL.MinQtyF)*-1) AS BalPcs,0 AS BalWeight,SUM(SL.FinalAmt*-1) AS BalAmt,'''' AS VoucherDate1,'''' AS BillNo'
	SET @WHERE=' LEFT JOIN LedgerMaster LM ON LM.Id=SH.F_LedgerMaster
				LEFT JOIN ItemMaster IM ON IM.Id=SL.F_ItemMaster
				
				WHERE (SH.VoucherDate>=' + @FromDate + 'AND SH.VoucherDate<=' + @ToDate + ')
				AND
					1=(CASE WHEN (' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SH.F_LedgerMaster =' + CAST(@F_LedgerMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SL.F_ItemMaster =' + CAST(@F_ItemMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemBrandMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster =' + CAST(@F_ItemBrandMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemGroupMaster =' + CAST(@F_ItemGroupMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)		
				AND 1=
					(CASE WHEN(''' + @SMonth + '''='''') THEN 1 ELSE
						(CASE WHEN(''' + @SMonth + '''=Substring(SH.VoucherDate,5,2)) THEN 1 ELSE 0 END)
					END)
				AND ISNULL(SH.Cancelled,0)=0
				AND
					1=(CASE WHEN (' + CAST(@MRP AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SL.MRP =' + CAST(@MRP AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				'
	SET @WHERE_OP=' LEFT JOIN LedgerMaster LM ON LM.Id=SH.F_LedgerMaster
				LEFT JOIN ItemMaster IM ON IM.Id=SL.F_ItemMaster
				
				WHERE (SH.VoucherDate<' + @FromDate + ')
				AND
					1=(CASE WHEN (' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SH.F_LedgerMaster =' + CAST(@F_LedgerMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SL.F_ItemMaster =' + CAST(@F_ItemMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemBrandMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster =' + CAST(@F_ItemBrandMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemGroupMaster =' + CAST(@F_ItemGroupMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)		
				AND 1=
					(CASE WHEN(''' + @SMonth + '''='''') THEN 1 ELSE
						(CASE WHEN(''' + @SMonth + '''=Substring(SH.VoucherDate,5,2)) THEN 1 ELSE 0 END)
					END)
				AND ISNULL(SH.Cancelled,0)=0
				AND
					1=(CASE WHEN (' + CAST(@MRP AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SL.MRP =' + CAST(@MRP AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				'
	EXEC('
		SELECT	0 AS Id,'''' AS VoucherNo,'''' AS VoucherDate,''Op. Bal.'' AS Ledger,'''' AS VoucherType
			,0 AS InQty,0 AS OutQty
			,0 AS InWeight,0 AS OutWeight
			,0 AS InAmt,0 AS OutAmt
			,SUM(TAB3.BalPcs) AS BalPcs,SUM(TAB3.BalWeight) AS BalWeight,SUM(TAB3.BalAmt) AS BalAmt
			,0 AS VoucherDate1,0 AS F_VoucherTypeMaster
		 FROM (
		 	SELECT (IM.OpeningQty) AS InQty,0 AS OutQty
					,(0) AS InWeight,0 AS OutWeight
					,(IM.LandingCost) AS InAmt,0 AS OutAmt
					,(IM.OpeningQty) AS BalPcs,(0) AS BalWeight,(IM.LandingCost) AS BalAmt,'''' AS VoucherDate1,'''' AS BillNo
			FROM ItemMaster IM
			WHERE 1=(CASE WHEN (' + @F_ItemMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.Id =' + @F_ItemMaster + ') THEN 1 ELSE 0 END) END)
			
			UNION ALL
			
			SELECT ' + @FIELDS_IN_OP + '
			FROM PurchaseInvoiceL SL LEFT JOIN PurchaseInvoiceH SH ON SH.Id=SL.F_PurchaseInvoiceH
			' + @WHERE_OP + '			
			
			UNION ALL
			
			SELECT  IM.Id AS ItemId,IM.Name AS Item,0 AS SNo
				,SL.Qty AS OpeningQty,0 AS OpWeight
				,0 AS InPcs,0 AS OutPcs
				,0 AS InWeight,0 AS OutWeight
				,(SL.Qty) AS BalPcs, (0) AS BalWeight
			FROM StockInL SL
			LEFT JOIN StockInH SH ON SH.Id=SL.F_StockInH
			LEFT JOIN ItemMaster IM ON IM.Id = SL.F_ItemMaster
			WHERE (SH.VoucherDate>=' + @FromDate + ' AND SH.VoucherDate<=' + @ToDate + ')
			AND 1=(CASE WHEN (' + @F_ItemMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.Id =' + @F_ItemMaster + ') THEN 1 ELSE 0 END) END)
			AND
			1=(CASE WHEN (' + @F_ItemBrandMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster =' + @F_ItemBrandMaster + ') THEN 1 ELSE 0 END) END)
			AND
			1=(CASE WHEN (' + @F_ItemGroupMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemGroupMaster =' + @F_ItemGroupMaster + ') THEN 1 ELSE 0 END) END)
			AND
				1=(CASE WHEN (' + @MRP + '=0) THEN 1 ELSE (CASE WHEN (SL.MRP =' + @MRP + ') THEN 1 ELSE 0 END) END)
			
			UNION ALL
			
			SELECT ' + @FIELDS_OUT_OP + '
			FROM SaleInvoiceL SL LEFT JOIN SaleInvoiceH SH ON SH.Id=SL.F_SaleInvoiceH
			' + @WHERE_OP + '	
		) AS TAB3
	UNION ALL
	
	 SELECT	TAB1.Id,TAB1.VoucherNo,TAB1.VoucherDate,TAB1.Ledger,TAB1.VoucherType
		,SUM(TAB1.InQty) AS InQty,SUM(TAB1.OutQty) AS OutQty
		,SUM(TAB1.InWeight) AS InWeight,SUM(TAB1.OutWeight) AS OutWeight
		,SUM(TAB1.InAmt) AS InAmt,SUM(TAB1.OutAmt) AS OutAmt
		,SUM(TAB1.BalPcs) AS BalPcs,SUM(TAB1.BalWeight) AS BalWeight,SUM(TAB1.BalAmt) AS BalAmt
		,TAB1.VoucherDate1,TAB1.F_VoucherTypeMaster
	 FROM (
	 	 
		SELECT ' + @FIELDS_IN + ',''Purchase Invoice'' AS VoucherType,11 AS F_VoucherTypeMaster
		FROM PurchaseInvoiceL SL LEFT JOIN PurchaseInvoiceH SH ON SH.Id=SL.F_PurchaseInvoiceH
		' + @WHERE + '
		
		
		UNION ALL
		 
		SELECT ' + @FIELDS_OUT + ' ,''SaleInvoice'' AS VoucherType,9 AS F_VoucherTypeMaster
		FROM SaleInvoiceL SL LEFT JOIN SaleInvoiceH SH ON SH.Id=SL.F_SaleInvoiceH
		' + @WHERE + '
	) AS TAB1
		
	GROUP BY TAB1.Id,TAB1.VoucherNo,TAB1.VoucherDate,TAB1.Ledger,TAB1.VoucherType,TAB1.VoucherDate1,TAB1.F_VoucherTypeMaster
	ORDER BY VoucherDate1')
	--
END






















GO
```

# [dbo].[Select_AS_StockPosition]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_StockPosition]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE PROCEDURE [dbo].[Select_AS_StockPosition] 
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_LedgerMaster numeric(18,0)=0,
	@F_ItemMaster numeric(18,0)=0,
	@F_ItemBrandMaster numeric(18,0)=0,
	@F_ItemGroupMaster numeric(18,0)=0,
	@SMonth nvarchar(50)='',
	@Type smallint=0,
	@AvoidFromDate smallint=0,
	@MRP numeric(18,2)=0,
	@Auto smallint=0
AS
BEGIN
--QPSAmt -> TotalWeight
--Display -> Repl. Qty
	DECLARE @FIELDS_IN NVARCHAR(MAX)
	DECLARE @FIELDS_OUT NVARCHAR(MAX)
	DECLARE @WHERE NVARCHAR(MAX)
	
	DECLARE @FIELDS_IN_OP NVARCHAR(MAX)
	DECLARE @FIELDS_OUT_OP NVARCHAR(MAX)
	DECLARE @WHERE_OP NVARCHAR(MAX)
	DECLARE @WHERE_Type NVARCHAR(MAX)
	
	IF(@Type=0)			-- Having Cl. Stock
		BEGIN
			IF(@Auto=1)
				SET @WHERE_Type='WHERE TAB2.BalPcs<>0'
			ELSE
				SET @WHERE_Type='WHERE TAB2.ActBalance<>0'
		END
	ELSE IF(@Type=1)	-- Having Transaction
		SET @WHERE_Type='WHERE TAB2.OpeningQty <>0 OR TAB2.InPcs <>0 OR TAB2.OutPcs <>0 '
	ELSE IF(@Type=-1)	-- All
		SET @WHERE_Type=' '
	ELSE IF(@Type=2)	-- No Transaction
		SET @WHERE_Type='WHERE TAB2.OpeningQty = 0 AND TAB2.InPcs = 0 AND TAB2.OutPcs = 0 '
					
	SET @FIELDS_IN=' IM.Id AS ItemId,IM.Name AS Item,0 AS SNo,0 AS OpeningQty,0 AS OpWeight
					,(SL.MinQty+SL.MinQtyF) AS InPcs,0 AS OutPcs
					,(0) AS InWeight,0 AS OutWeight
					,(SL.MinQty+SL.MinQtyF) AS BalPcs, (0) AS BalWeight,SL.MRP'
	SET @FIELDS_IN_OP=' IM.Id AS ItemId,IM.Name AS Item,0 AS SNo,0 AS OpeningQty,0 AS OpWeight
					,SUM(SL.MinQty+SL.MinQtyF) AS InPcs,0 AS OutPcs
					,0 AS InWeight,0 AS OutWeight
					,SUM(SL.MinQty+SL.MinQtyF) AS BalPcs, 0 AS BalWeight,SL.MRP'
	SET @FIELDS_OUT =' IM.Id AS ItemId,IM.Name AS Item,0 AS SNo,0 AS OpeningQty,0 AS OpWeight
					,0 AS InPcs,(SL.MinQty+SL.MinQtyF) AS OutPcs
					,0 AS InWeight,(0) AS OutWeight
					,((SL.MinQty+SL.MinQtyF)*-1) AS BalPcs,((0)*-1) AS BalWeight,SL.MRP'
	SET @FIELDS_OUT_OP =' IM.Id AS ItemId,IM.Name AS Item,0 AS SNo,0 AS OpeningQty,0 AS OpWeight
					,0 AS InPcs,SUM(SL.MinQty+SL.MinQtyF) AS OutPcs
					,0 AS InWeight,0 AS OutWeight
					,SUM((SL.MinQty+SL.MinQtyF)*-1) AS BalPcs,0 AS BalWeight,SL.MRP'
	SET @WHERE=' LEFT JOIN ItemMaster IM ON IM.Id=SL.F_ItemMaster
				 LEFT JOIN LedgerMaster LMT ON LMT.Id = SL.F_LedgerMaster_Tax
				WHERE ' +
				CASE WHEN(@AvoidFromDate=0) THEN
					'(SH.VoucherDate>=' + @FromDate + ' AND SH.VoucherDate<=' + @ToDate + ')'
				ELSE
					'(SH.VoucherDate<=' + @ToDate + ')'
				END 
				+ '
				AND
					1=(CASE WHEN (' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SH.F_LedgerMaster =' + CAST(@F_LedgerMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SL.F_ItemMaster =' + CAST(@F_ItemMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemBrandMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster =' + CAST(@F_ItemBrandMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemGroupMaster =' + CAST(@F_ItemGroupMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)		
				AND 1=
					(CASE WHEN(''' + @SMonth + '''='''') THEN 1 ELSE
						(CASE WHEN(''' + @SMonth + '''=Substring(SH.VoucherDate,5,2)) THEN 1 ELSE 0 END)
					END)
				AND ISNULL(SH.Cancelled,0)=0
				AND
					1=(CASE WHEN (' + CAST(@MRP AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SL.MRP =' + CAST(@MRP AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				'
	SET @WHERE_OP=' LEFT JOIN ItemMaster IM ON IM.Id=SL.F_ItemMaster
				WHERE (SH.VoucherDate<' + @FromDate + ')
				AND
					1=(CASE WHEN (' + CAST(@F_LedgerMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SH.F_LedgerMaster =' + CAST(@F_LedgerMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SL.F_ItemMaster =' + CAST(@F_ItemMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemBrandMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster =' + CAST(@F_ItemBrandMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				AND
					1=(CASE WHEN (' + CAST(@F_ItemGroupMaster AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemGroupMaster =' + CAST(@F_ItemGroupMaster AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)		
				AND 1=
					(CASE WHEN(''' + @SMonth + '''='''') THEN 1 ELSE
						(CASE WHEN(''' + @SMonth + '''=Substring(SH.VoucherDate,5,2)) THEN 1 ELSE 0 END)
					END)
				AND ISNULL(SH.Cancelled,0)=0
				AND
					1=(CASE WHEN (' + CAST(@MRP AS NVARCHAR) + '=0) THEN 1 ELSE (CASE WHEN (SL.MRP =' + CAST(@MRP AS NVARCHAR) + ') THEN 1 ELSE 0 END) END)
				'

	DECLARE @FromDateLast nvarchar(50)
	SET @FromDateLast = SUBSTRING(@FromDate,1,4)+ '0331'
--(IM.TotalWeight) OPENING WEIGHT
	EXEC('
	SELECT	TAB2.ItemId, TAB2.Item, TAB2.MRP
		,TAB2.OpeningQty,TAB2.InPcs,TAB2.OutPcs
		,TAB2.OpWeight,TAB2.InWeight,TAB2.OutWeight
		,TAB2.BalPcs,TAB2.BalWeight,TAB2.ActBalance AS ActBalance
		,0 AS Rate,0 AS Amount,TAB2.Tax
	FROM(
		 SELECT	TAB1.ItemId, TAB1.Item,TAB1.SNo	
			,SUM(TAB1.OpeningQty) AS OpeningQty,SUM(TAB1.InPcs) AS InPcs,SUM(TAB1.OutPcs) AS OutPcs
			,SUM(TAB1.OpWeight) AS OpWeight,SUM(TAB1.InWeight) AS InWeight,SUM(TAB1.OutWeight) AS OutWeight
			,(SUM(TAB1.OpeningQty)+SUM(TAB1.BalPcs)) AS BalPcs,(SUM(TAB1.OpWeight)+SUM(TAB1.BalWeight)) AS BalWeight
			,0 AS Amount,TAB1.MRP,SUM(TAB1.ActBalance) AS ActBalance,TAB1.Tax
		 FROM (
			SELECT IM.Id AS ItemId,IM.Name AS Item,0 AS SNo
				,SUM(SL.Qty) AS OpeningQty,0 AS OpWeight
				,0 AS InPcs,0 AS OutPcs
				,0 AS InWeight,0 AS OutWeight
				,0 AS BalPcs,0 AS BalWeight,SL.MRP,0 AS ActBalance,LMT.Name AS Tax
			FROM StockInL SL
			LEFT JOIN StockInH SH ON SH.Id=SL.F_StockInH
			LEFT JOIN ItemMaster IM ON IM.Id = SL.F_ItemMaster
			LEFT JOIN LedgerMaster LMT ON LMT.Id = SL.F_LedgerMaster_Tax
			WHERE (SH.VoucherDate=' + @FromDateLast + ')
				AND 1=(CASE WHEN (' + @F_ItemMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.Id =' + @F_ItemMaster + ') THEN 1 ELSE 0 END) END)
				AND
				1=(CASE WHEN (' + @F_ItemBrandMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster =' + @F_ItemBrandMaster + ') THEN 1 ELSE 0 END) END)
				AND
				1=(CASE WHEN (' + @F_ItemGroupMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemGroupMaster =' + @F_ItemGroupMaster + ') THEN 1 ELSE 0 END) END)
				AND
				1=(CASE WHEN (' + @MRP + '=0) THEN 1 ELSE (CASE WHEN (SL.MRP =' + @MRP + ') THEN 1 ELSE 0 END) END)
			
			GROUP BY IM.Id,IM.Name,SL.MRP,SL.F_LedgerMaster_Tax,LMT.Name
			
			UNION ALL
			
			SELECT IM.Id AS ItemId,IM.Name AS Item,0 AS SNo
				,0 AS OpeningQty,0 AS OpWeight
				,0 AS InPcs,0 AS OutPcs
				,0 AS InWeight,0 AS OutWeight
				,0 AS BalPcs,0 AS BalWeight,SL.MRP,SUM(SL.Qty) AS ActBalance,LMT.Name AS Tax
			FROM StockInL SL
			LEFT JOIN StockInH SH ON SH.Id=SL.F_StockInH
			LEFT JOIN ItemMaster IM ON IM.Id = SL.F_ItemMaster
			LEFT JOIN LedgerMaster LMT ON LMT.Id = SL.F_LedgerMaster_Tax
			WHERE (SH.VoucherDate=' + @ToDate + ')
				AND 1=(CASE WHEN (' + @F_ItemMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.Id =' + @F_ItemMaster + ') THEN 1 ELSE 0 END) END)
				AND
				1=(CASE WHEN (' + @F_ItemBrandMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemBrandMaster =' + @F_ItemBrandMaster + ') THEN 1 ELSE 0 END) END)
				AND
				1=(CASE WHEN (' + @F_ItemGroupMaster + '=0) THEN 1 ELSE (CASE WHEN (IM.F_ItemGroupMaster =' + @F_ItemGroupMaster + ') THEN 1 ELSE 0 END) END)
			GROUP BY IM.Id,IM.Name,SL.MRP,SL.F_LedgerMaster_Tax,LMT.Name
			
			UNION ALL
			
			SELECT ' + @FIELDS_IN + ',0 AS ActBalance,LMT.Name AS Tax
			FROM PurchaseInvoiceL SL LEFT JOIN PurchaseInvoiceH SH ON SH.Id=SL.F_PurchaseInvoiceH
			' + @WHERE + '			
			
			UNION ALL

			SELECT ' + @FIELDS_OUT + ',0 AS ActBalance,LMT.Name AS Tax
			FROM SaleInvoiceL SL LEFT JOIN SaleInvoiceH SH ON SH.Id=SL.F_SaleInvoiceH
			' + @WHERE + '
		) AS TAB1
		GROUP BY TAB1.SNo,TAB1.Item,TAB1.ItemId,TAB1.MRP,TAB1.Tax
	) AS TAB2
	' + @WHERE_Type + '
		ORDER BY TAB2.SNo,TAB2.Item,TAB2.ItemId,TAB2.MRP,TAB2.Tax')
END





















GO
```

# [dbo].[Select_AS_TransportStockAnalysis]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_TransportStockAnalysis]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





CREATE PROCEDURE [dbo].[Select_AS_TransportStockAnalysis] 	
	@F_TransportMaster numeric(18,0)=0,
	@F_ItemMaster numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Cancelled smallint=-1,
	@Id numeric(18,0)=0,
	@Type SMALLINT=0,
	@PurVoucherNo NVARCHAR(50)='',
	@PurVoucherDate NVARCHAR(50)=''
AS
BEGIN
	DECLARE @WHERE_IMPORT NVARCHAR(MAX)
	IF(@PurVoucherNo<>'')
		SET @WHERE_IMPORT=' AND TH.InvoiceNo=''' + @PurVoucherNo + ''' AND TH.InvoiceDate=''' + @PurVoucherDate + ''''
	ELSE
		SET @WHERE_IMPORT=''
	DECLARE @SQL_IMPORT NVARCHAR(MAX)
	--Import Transport
	SET @SQL_IMPORT='SELECT MAX(TH.Id) AS Id,TH.InvoiceNo AS PurVoucherNo
						,(SUBSTRING(TH.InvoiceDate,7,2)+''/''+SUBSTRING(TH.InvoiceDate,5,2)+''/''+SUBSTRING(TH.InvoiceDate,1,4)) AS PurVoucherDate
						,IM.Name AS Item,'''' AS UnitQ
						,SUM(TH.Qty_CS) AS Pur_Qty,0 AS TQty,0 AS LotSize,IM.Id AS F_ItemMaster,0 AS F_UnitMaster,TH.InvoiceDate AS InvoiceDate1
					FROM Transport_ImportH TH
						LEFT JOIN TransportItemMapMaster ISTM ON ISTM.TransportItemId=TH.Item_Id
						LEFT JOIN ItemMaster IM ON IM.Id = ISTM.F_ItemMaster
					WHERE (TH.InvoiceDate>='+@FromDate + ' AND TH.InvoiceDate<='+@ToDate + ')' + @WHERE_IMPORT + '
					GROUP BY TH.InvoiceNo,TH.InvoiceDate,IM.Id,IM.Name'
	--Transport Stock In
	DECLARE @WHERE_STOCKIN NVARCHAR(MAX)
	IF(@PurVoucherNo<>'')
		SET @WHERE_STOCKIN=' AND DSL.PurVoucherNo=''' + @PurVoucherNo + ''' AND DSL.PurVoucherDate=''' + @PurVoucherDate + ''''
	ELSE
		SET @WHERE_STOCKIN=''
	DECLARE @SQL_STOCKIN NVARCHAR(MAX)
	SET @SQL_STOCKIN='SELECT MAX(DSH.Id) AS Id
						,(CASE WHEN(DSL.PurVoucherDate='''') THEN ''N/A'' ELSE DSL.PurVoucherNo END) AS PurVoucherNo
						,(SUBSTRING(DSL.PurVoucherDate,7,2)+''/''+SUBSTRING(DSL.PurVoucherDate,5,2)+''/''+SUBSTRING(DSL.PurVoucherDate,1,4)) AS PurVoucherDate
						,IM.Name AS Item,MAX(UMQ.Name) AS UnitQ
						,0 AS Pur_Qty,SUM(DSL.Qty) AS TQty,MAX(DSL.LotSize) AS LotSize,IM.Id AS F_ItemMaster,MAX(UMQ.Id) AS F_UnitMaster
						,(CASE WHEN(DSL.PurVoucherDate='''') THEN ''N/A'' ELSE DSL.PurVoucherDate END) AS InvoiceDate1
					FROM TransportStockInH DSH
						LEFT JOIN TransportStockInL DSL ON DSH.Id=DSL.F_TransportStockInH
						LEFT JOIN TransportMaster TM ON TM.Id = DSH.F_TransportMaster
						LEFT JOIN ItemMaster IM ON IM.Id = DSL.F_ItemMaster
						LEFT JOIN UnitMaster UMQ ON UMQ.Id = DSL.F_UnitMaster
					WHERE ((DSL.PurVoucherDate>='+@FromDate + ' AND DSL.PurVoucherDate<='+@ToDate + ') OR DSL.PurVoucherDate='''')' + @WHERE_STOCKIN + '
					GROUP BY DSL.PurVoucherNo,DSL.PurVoucherDate,IM.Id,IM.Name'

	IF(@Type=0)
		BEGIN
			EXEC('SELECT  MAX(TAB1.Id) AS Id,TAB1.PurVoucherNo
				,TAB1.PurVoucherDate
				,TAB1.Item,MAX(TAB1.UnitQ) AS UnitQ
				,SUM(TAB1.Pur_Qty) AS Pur_Qty,SUM(TAB1.TQty) AS TQty,MAX(TAB1.LotSize) AS LotSize,TAB1.F_ItemMaster,MAX(TAB1.F_UnitMaster) AS F_UnitMaster
				,SUM(TAB1.Pur_Qty)-SUM(TAB1.TQty) AS Balance,TAB1.PurVoucherNo,TAB1.InvoiceDate1
			FROM (' + @SQL_IMPORT + '
				UNION ALL
				' + @SQL_STOCKIN + '
			) AS TAB1
			GROUP BY TAB1.PurVoucherNo,TAB1.PurVoucherDate,TAB1.InvoiceDate1,TAB1.Item,TAB1.F_ItemMaster
			ORDER BY TAB1.InvoiceDate1 DESC
				,(CASE WHEN IsNumeric(TAB1.PurVoucherNo) = 1 then Right(Replicate(''0'',21) + TAB1.PurVoucherNo, 20)
					WHEN IsNumeric(TAB1.PurVoucherNo) = 0 then Left(TAB1.PurVoucherNo + Replicate('''',21), 20)
				 ELSE TAB1.PurVoucherNo
				 END),TAB1.Item')
		 END
	ELSE IF(@Type=1)
		BEGIN
			EXEC(@SQL_IMPORT)
		END
	ELSE IF(@Type=2)
		BEGIN
			EXEC(@SQL_STOCKIN)
		END
END



GO
```

# [dbo].[Select_AS_TransportStockInH]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_TransportStockInH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO







CREATE PROCEDURE [dbo].[Select_AS_TransportStockInH] 	
	@F_TransportMaster numeric(18,0)=0,
	@F_ItemMaster numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Cancelled smallint=-1,
	@Id numeric(18,0)=0
AS
BEGIN
	SELECT DSH.Id,DSH.VoucherNo,
		(SUBSTRING(DSH.VoucherDate,7,2)+'/'+SUBSTRING(DSH.VoucherDate,5,2)+'/'+SUBSTRING(DSH.VoucherDate,1,4)) AS VoucherDate,
		TM.Name AS Transport,DSH.DriverName,
		DSH.Remark,
		DSH.TotalQty,
		(CASE WHEN (ISNULL(DSH.Cancelled,0)=1) THEN 'Cancelled' ELSE '' END) AS Cancelled
	FROM TransportStockInH DSH
		LEFT JOIN TransportStockInL DSL ON DSH.Id=DSL.F_TransportStockInH
		LEFT JOIN TransportMaster TM ON TM.Id = DSH.F_TransportMaster
		LEFT JOIN ItemMaster IM ON IM.Id = DSL.F_ItemMaster
		LEFT JOIN UnitMaster UMQ ON UMQ.Id = DSL.F_UnitMaster
		LEFT JOIN GodownMaster GM ON GM.Id = DSL.F_GodownMaster
	WHERE 
		1=(CASE WHEN (@Id=0) THEN 1 ELSE (CASE WHEN (DSH.Id =@Id) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_TransportMaster=0) THEN 1 ELSE (CASE WHEN (DSH.F_TransportMaster =@F_TransportMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN (@F_ItemMaster=0) THEN 1 ELSE (CASE WHEN (DSL.F_ItemMaster =@F_ItemMaster) THEN 1 ELSE 0 END) END)
		AND
		1=(CASE WHEN(@FromDate='') THEN 1 ELSE
			(CASE WHEN(DSH.VoucherDate>=@FromDate AND DSH.VoucherDate<=@ToDate) THEN 1 ELSE 0 END)
			END)
		AND
		1=(CASE WHEN (@Cancelled=-1) THEN 1 ELSE (CASE WHEN (ISNULL(DSH.Cancelled,0) =@Cancelled) THEN 1 ELSE 0 END) END)
	GROUP BY DSH.Id,DSH.VoucherNo,DSH.VoucherDate,DSH.Remark,TM.Name,DSH.DriverName,DSH.TotalQty,DSH.Cancelled
	ORDER BY DSH.VoucherDate DESC,CAST(DSH.VoucherNo AS NUMERIC),DSH.Id,TM.Name
END



GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_VAT8]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






















CREATE PROCEDURE [dbo].[Select_AS_VAT8] 	
	@F_VoucherTypeMaster numeric(18,0),
	@FromDate nvarchar(50)=NULL,
	@ToDate nvarchar(50)=NULL,
	@F_CompanyMaster numeric(18,0)=1,
	@WithTINNo smallint=-1,
	@F_LedgerMaster numeric(18,0)=0,
	@RowNumberA NVARCHAR(MAX)='DENSE_RANK() OVER (ORDER BY VL.VoucherDate,CAST(LEFT(VL.VoucherNo, Patindex(''%[^0-9]%'', VL.VoucherNo + ''x'') - 1) AS INT) ASC)',
--out fields	
	@RowNumber AS NUMERIC(18,0)=0 OUT,
	@Id NUMERIC(18,0)=0 OUT,
	@VoucherNo NVARCHAR(50)='' OUT,
	@VoucherDate NVARCHAR(50)='' OUT,
	@Ledger NVARCHAR(MAX)='' OUT,
	@TINNo NVARCHAR(50)='' OUT,
	@FinYear NVARCHAR(50)='' OUT,
	@TotalNetAmt NUMERIC(18,2)=0 OUT,
	@Tax1 NVARCHAR(50)='' OUT,
	@TaxAmt1 NUMERIC(18,2)=0 OUT,
	@TaxableAmt1 NUMERIC(18,2)=0 OUT,
	@Tax2 NVARCHAR(50)='' OUT,
	@TaxAmt2 NUMERIC(18,2)=0 OUT,
	@TaxableAmt2 NUMERIC(18,2)=0 OUT,
	@Tax3 NVARCHAR(50)='' OUT,
	@TaxAmt3 NUMERIC(18,2)=0 OUT,
	@TaxableAmt3 NUMERIC(18,2)=0 OUT,
	@Tax4 NVARCHAR(50)='' OUT,
	@TaxAmt4 NUMERIC(18,2)=0 OUT,
	@TaxableAmt4 NUMERIC(18,2)=0 OUT,
	@Tax5 NVARCHAR(50)='' OUT,
	@TaxAmt5 NUMERIC(18,2)=0 OUT,
	@TaxableAmt5 NUMERIC(18,2)=0 OUT,
	@GrossAmt_Ex_VAT NUMERIC(18,2)=0 OUT,
	@TotalTax  NUMERIC(18,2)=0 OUT,
	@GrossAmt_Inc_VAT NUMERIC(18,2)=0 OUT,
	@Difference NUMERIC(18,2)=0 OUT,
	@MonthName NVARCHAR(50)='' OUT,
	@MonthNo NVARCHAR(50)='' OUT,
	@VoucherDate2 NVARCHAR(50)='' OUT,
	@ReportNo NVARCHAR(50)='' OUT,
	@Report NVARCHAR(50)='' OUT,
	@Year NVARCHAR(50)='' OUT
AS
BEGIN
	
	DECLARE @TINNoNew NVARCHAR(MAX)
	SET @TINNoNew = ',(CASE WHEN(ISNULL(LM.TINNo3,'''') <>'''') THEN (CASE WHEN(VL.VoucherDate>=ISNULL(LM.TINNoFrom3,'''')) THEN ISNULL(LM.TINNo3,'''') ELSE
												(CASE WHEN(VL.VoucherDate>=ISNULL(LM.TINNoFrom2,'''')) THEN ISNULL(LM.TINNo2,'''') ELSE ISNULL(LM.TINNo,'''') END) END)
					ELSE (CASE WHEN(ISNULL(LM.TINNo2,'''') <>'''') THEN (CASE WHEN(VL.VoucherDate>=ISNULL(LM.TINNoFrom2,'''')) THEN ISNULL(LM.TINNo2,'''') ELSE ISNULL(LM.TINNo,'''') END)
										ELSE ISNULL(LM.TINNo,'''') END)
									 END) AS TINNo'

	-- SET NOCOUNT ON added to prevent extra result sets from
	-- interfering with SELECT statements.
	SET NOCOUNT ON;
	
	DECLARE @WHERE_WithTINNo NVARCHAR(MAX)
	IF(@WithTINNo=-1)
		SET @WHERE_WithTINNo=''
	ELSE IF(@WithTINNo=0)
		SET @WHERE_WithTINNo=' AND ISNULL(LM.TINNo,'''')<>'''' '
	ELSE
		SET @WHERE_WithTINNo=' AND ISNULL(LM.TINNo,'''')='''' '
		
	DECLARE @EXTRA_FIELD NVARCHAR(MAX)
	DECLARE @VoucherDate_Field nvarchar(MAX)
	SET @VoucherDate_Field='(SUBSTRING(VL.VoucherDate,7,2)+''/''+SUBSTRING(VL.VoucherDate,5,2)+''/''+SUBSTRING(VL.VoucherDate,1,4)) AS VoucherDate'
	IF(@F_VoucherTypeMaster=9 OR @F_VoucherTypeMaster=5)
		SET @EXTRA_FIELD =',1 AS ReportNo, ''Total Sale'' AS Report,TAB1.Year'
	ELSE IF(@F_VoucherTypeMaster=8 OR @F_VoucherTypeMaster=10)
		SET @EXTRA_FIELD =',2 AS ReportNo, ''Less Sale Return'' AS Report,TAB1.Year'
	ELSE IF(@F_VoucherTypeMaster=7 OR @F_VoucherTypeMaster=11)
		BEGIN
			SET @EXTRA_FIELD =',4 AS ReportNo, ''Less Purchase'' AS Report,TAB1.Year'
			SET @VoucherDate_Field='(SELECT 
			(SUBSTRING(PH.PurVoucherDate,7,2)+''/''+SUBSTRING(PH.PurVoucherDate,5,2)+''/''+SUBSTRING(PH.PurVoucherDate,1,4)) 
			FROM PurchaseInvoiceH PH
			WHERE PH.Id=VL.F_VoucherH AND VL.F_VoucherTypeMaster=11)
			AS VoucherDate'
		END
	ELSE IF(@F_VoucherTypeMaster=6 OR @F_VoucherTypeMaster=12)
		SET @EXTRA_FIELD =',5 AS ReportNo, ''Less Purchase Return'' AS Report,TAB1.Year'
		
	DECLARE @DontShowExemptedTax numeric(18,0)
	DECLARE @F_LedgerMaster_Tax1 numeric(18,0)
	DECLARE @F_LedgerMaster_Tax2 numeric(18,0)
	DECLARE @F_LedgerMaster_Tax3 numeric(18,0)
	DECLARE @F_LedgerMaster_Tax4 numeric(18,0)
	DECLARE @F_LedgerMaster_Tax5 numeric(18,0)
	DECLARE @F_LedgerMaster_Taxable1 numeric(18,0)
	DECLARE @F_LedgerMaster_Taxable2 numeric(18,0)
	DECLARE @F_LedgerMaster_Taxable3 numeric(18,0)
	DECLARE @F_LedgerMaster_Taxable4 numeric(18,0)
	DECLARE @F_LedgerMaster_Taxable5 numeric(18,0)

	DECLARE @Tax1_Name nvarchar(50)
	DECLARE @Tax2_Name nvarchar(50)
	DECLARE @Tax3_Name nvarchar(50)
	DECLARE @Tax4_Name nvarchar(50)
	DECLARE @Tax5_Name nvarchar(50)
				
	DECLARE @Taxable1_Name nvarchar(50)
	DECLARE @Taxable2_Name nvarchar(50)
	DECLARE @Taxable3_Name nvarchar(50)
	DECLARE @Taxable4_Name nvarchar(50)
	DECLARE @Taxable5_Name nvarchar(50)
		
	--AGAINST TAX ID
	IF(@F_VoucherTypeMaster=7 OR @F_VoucherTypeMaster=11 OR @F_VoucherTypeMaster=12)
		BEGIN
			--TAX ID
			SET @DontShowExemptedTax = (SELECT ISNULL(DontShowExemptedTax,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			SET @F_LedgerMaster_Tax1 = (SELECT ISNULL(F_LedgerMaster_Tax1,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			SET @F_LedgerMaster_Tax2 = (SELECT ISNULL(F_LedgerMaster_Tax2,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			SET @F_LedgerMaster_Tax3 = (SELECT ISNULL(F_LedgerMaster_Tax3,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			SET @F_LedgerMaster_Tax4 = (SELECT ISNULL(F_LedgerMaster_Tax4,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			SET @F_LedgerMaster_Tax5 = (SELECT ISNULL(F_LedgerMaster_Tax5,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
				
			--TAX NAME
			SET @Tax1_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax1)
			SET @Tax2_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax2)
			SET @Tax3_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax3)
			SET @Tax4_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax4)
			SET @Tax5_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax5)

			SET @F_LedgerMaster_Taxable1 = (SELECT TaxF_PurchaseLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax1)
			SET @F_LedgerMaster_Taxable2 = (SELECT TaxF_PurchaseLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax2)
			SET @F_LedgerMaster_Taxable3 = (SELECT TaxF_PurchaseLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax3)
			SET @F_LedgerMaster_Taxable4 = (SELECT TaxF_PurchaseLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax4)
			SET @F_LedgerMaster_Taxable5 = (SELECT TaxF_PurchaseLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax5)			
			
			IF(@DontShowExemptedTax=@F_LedgerMaster_Tax5)
				BEGIN
					SET @F_LedgerMaster_Tax5=0
					SET @Tax5_Name=''
					SET @F_LedgerMaster_Taxable5=0
				END
		END
	ELSE
		BEGIN
			--TAX ID
			SET @F_LedgerMaster_Tax1 = (SELECT ISNULL(F_LedgerMaster_Tax1,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			SET @F_LedgerMaster_Tax2 = (SELECT ISNULL(F_LedgerMaster_Tax2,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			SET @F_LedgerMaster_Tax3 = (SELECT ISNULL(F_LedgerMaster_Tax3,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			SET @F_LedgerMaster_Tax4 = (SELECT ISNULL(F_LedgerMaster_Tax4,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			SET @F_LedgerMaster_Tax5 = (SELECT ISNULL(F_LedgerMaster_Tax5,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
			--TAX NAME
			SET @Tax1_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax1)
			SET @Tax2_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax2)
			SET @Tax3_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax3)
			SET @Tax4_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax4)
			SET @Tax5_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax5)

			SET @F_LedgerMaster_Taxable1 = (SELECT TaxF_SalesLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax1)
			SET @F_LedgerMaster_Taxable2 = (SELECT TaxF_SalesLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax2)
			SET @F_LedgerMaster_Taxable3 = (SELECT TaxF_SalesLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax3)
			SET @F_LedgerMaster_Taxable4 = (SELECT TaxF_SalesLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax4)
			SET @F_LedgerMaster_Taxable5 = (SELECT TaxF_SalesLedger FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax5)
		END

	--TAX QUERY
	EXEC('
	SELECT TAB1.RowNumber,TAB1.Id,TAB1.VoucherNo,TAB1.VoucherDate,TAB1.Ledger,TAB1.TINNo,TAB1.FinYear,TAB1.TotalNetAmt
		,TAB1.Tax1,TAB1.TaxAmt1,TAB1.TaxableAmt1
		,TAB1.Tax2,TAB1.TaxAmt2,TAB1.TaxableAmt2
		,TAB1.Tax3,TAB1.TaxAmt3,TAB1.TaxableAmt3
		,TAB1.Tax4,TAB1.TaxAmt4,TAB1.TaxableAmt4
		,(TAB1.TaxableAmt1 + TAB1.TaxableAmt2 + TAB1.TaxableAmt3 + TAB1.TaxableAmt4 + TAB1.TaxableAmt5) AS GrossAmt_Ex_VAT
		,(TAB1.TaxAmt1 + TAB1.TaxAmt2 + TAB1.TaxAmt3 + TAB1.TaxAmt4 + TAB1.TaxAmt5) AS TotalTax
		,((TAB1.TaxAmt1 + TAB1.TaxAmt2 + TAB1.TaxAmt3 + TAB1.TaxAmt4 + TAB1.TaxAmt5) +
			(TAB1.TaxableAmt1 + TAB1.TaxableAmt2 + TAB1.TaxableAmt3 + TAB1.TaxableAmt4 + TAB1.TaxableAmt5)) AS GrossAmt_Inc_VAT
		,(TAB1.TotalNetAmt- ((TAB1.TaxAmt1 + TAB1.TaxAmt2 + TAB1.TaxAmt3 + TAB1.TaxAmt4 + TAB1.TaxAmt5) +
			(TAB1.TaxableAmt1 + TAB1.TaxableAmt2 + TAB1.TaxableAmt3 + TAB1.TaxableAmt4 + TAB1.TaxableAmt5))
		) AS Difference
		,TAB1.MonthName
		,TAB1.MonthNo
		,TAB1.VoucherDate2' + @EXTRA_FIELD + '
		,TAB1.Tax5,TAB1.TaxAmt5,TAB1.TaxableAmt5,TAB1.LedgerId
	FROM (	
	SELECT ' + @RowNumberA + ' AS RowNumber
		,VL.F_VoucherH AS Id, VL.VoucherNo
		,' + @VoucherDate_Field + '
		,LM.Name AS Ledger
		' + @TINNoNew + '
		,(SUBSTRING(VL.VoucherDate,1,4) + ''-'' + SUBSTRING(VL.VoucherDate,1,4)) AS FinYear
		,ISNULL(VL.Amount,0) AS TotalNetAmt
		,''' + @Tax1_Name + ''' AS Tax1
		,ISNULL(TAXDETAILS1.Amount,0) AS TaxAmt1
		,ISNULL(TAXABLE_DETAILS1.Amount,0) AS TaxableAmt1
		,''' + @Tax2_Name + ''' AS Tax2
		,ISNULL(TAXDETAILS2.Amount,0) AS TaxAmt2
		,ISNULL(TAXABLE_DETAILS2.Amount,0) AS TaxableAmt2
		,''' + @Tax3_Name + ''' AS Tax3
		,ISNULL(TAXDETAILS3.Amount,0) AS TaxAmt3
		,ISNULL(TAXABLE_DETAILS3.Amount,0) AS TaxableAmt3
		,''' + @Tax4_Name + ''' AS Tax4
		,ISNULL(TAXDETAILS4.Amount,0) AS TaxAmt4
		,ISNULL(TAXABLE_DETAILS4.Amount,0) AS TaxableAmt4
		,''' + @Tax5_Name + ''' AS Tax5
		,ISNULL(TAXDETAILS5.Amount,0) AS TaxAmt5
		,ISNULL(TAXABLE_DETAILS5.Amount,0) AS TaxableAmt5
		,((SUBSTRING(VL.VoucherDate,1,4)) + '' - '' +  SUBSTRING(DateName(mm,DATEADD(mm,CAST((SUBSTRING(VL.VoucherDate,5,2)) AS NUMERIC)-1,0)),0,4)) AS [MonthName]
		, CAST(SUBSTRING(VL.VoucherDate,1,4) + SUBSTRING(VL.VoucherDate,5,2) AS NUMERIC) AS MonthNo
		,VL.VoucherDate AS VoucherDate2,SUBSTRING(VL.VoucherDate,1,4) AS Year,LM.Id AS LedgerId
	FROM VoucherL VL
	LEFT JOIN LedgerMaster LM ON LM.Id=VL.F_LedgerMaster
	LEFT JOIN LedgerGroupMaster LGM ON LGM.Id = LM.F_LedgerGroupMaster
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Tax1 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXDETAILS1
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Tax2 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXDETAILS2
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Tax3 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXDETAILS3
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Tax4 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXDETAILS4
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Tax5 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXDETAILS5
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Taxable1 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXABLE_DETAILS1
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Taxable2 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXABLE_DETAILS2
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Taxable3 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXABLE_DETAILS3
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Taxable4 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXABLE_DETAILS4
	OUTER APPLY	(SELECT TOP 1 VL1.Amount
					FROM VoucherL VL1 LEFT JOIN LedgerMaster LM1 ON LM.Id=VL1.F_LedgerMaster
					WHERE VL1.F_LedgerMaster=' + @F_LedgerMaster_Taxable5 + ' AND VL1.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND VL1.F_VoucherH=VL.F_VoucherH AND VL1.F_CompanyMaster =' + @F_CompanyMaster + '
		) TAXABLE_DETAILS5
	WHERE VL.F_VoucherTypeMaster =' + @F_VoucherTypeMaster + ' AND LGM.GroupType IN (6,31,32,40) AND VL.F_CompanyMaster =' + @F_CompanyMaster + '
		AND (VL.VoucherDate>=' + @FromDate + ' AND VL.VoucherDate<=' + @ToDate + ')
		AND 1 = (CASE WHEN(' + @F_LedgerMaster + ' = 0) THEN 1 ELSE (CASE WHEN(LM.Id = ' + @F_LedgerMaster + ') THEN 1 ELSE 0 END) END)
		' + @WHERE_WithTINNo + '
	) AS TAB1
	ORDER BY TAB1.RowNumber,CAST(LEFT(TAB1.VoucherNo, Patindex(''%[^0-9]%'', TAB1.VoucherNo + ''x'') - 1) AS INT)')
END














GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_VAT8_MultiSale]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



























CREATE PROCEDURE [dbo].[Select_AS_VAT8_MultiSale]
	@F_VoucherTypeMaster nvarchar(max),
	@FromDate nvarchar(50)=NULL,
	@ToDate nvarchar(50)=NULL,
	@F_CompanyMaster numeric(18,0)=1,
	@WithTINNo smallint=-1,
	@RowNumber NVARCHAR(MAX)='DENSE_RANK() OVER (ORDER BY VL.VoucherDate,CAST(LEFT(VL.VoucherNo, Patindex(''%[^0-9]%'', VL.VoucherNo + ''x'') - 1) AS INT) ASC)',
	@F_LedgerMaster numeric(18,0)=0
AS
BEGIN
	-- SET NOCOUNT ON added to prevent extra result sets from
	-- interfering with SELECT statements.
	SET NOCOUNT ON;
	
	----Deleting tmp vat
	DELETE FROM TMPVAT
	
	IF(@F_VoucherTypeMaster='9,5')
		BEGIN
			--Total Sale
			INSERT INTO TMPVAT
			EXEC dbo.Select_AS_VAT8 9,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
			INSERT INTO TMPVAT
			EXEC dbo.Select_AS_VAT8 5,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
		END
	ELSE IF(@F_VoucherTypeMaster='11,7')
		BEGIN
			--Total Sale
			INSERT INTO TMPVAT
			EXEC dbo.Select_AS_VAT8 11,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
			INSERT INTO TMPVAT
			EXEC dbo.Select_AS_VAT8 7,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
		END
	ELSE
		BEGIN
			IF(@F_VoucherTypeMaster=9)
				BEGIN
					INSERT INTO TMPVAT
					EXEC dbo.Select_AS_VAT8 9,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
				END
			ELSE IF(@F_VoucherTypeMaster=5)
				BEGIN
					INSERT INTO TMPVAT
					EXEC dbo.Select_AS_VAT8 5,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
				END
			ELSE IF(@F_VoucherTypeMaster=8)
				BEGIN
					INSERT INTO TMPVAT
					EXEC dbo.Select_AS_VAT8 8,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
				END
			ELSE IF(@F_VoucherTypeMaster=10)
				BEGIN
					INSERT INTO TMPVAT
					EXEC dbo.Select_AS_VAT8 10,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
				END
			ELSE IF(@F_VoucherTypeMaster=7)
				BEGIN
					INSERT INTO TMPVAT
					EXEC dbo.Select_AS_VAT8 7,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
				END
			ELSE IF(@F_VoucherTypeMaster=11)
				BEGIN
					INSERT INTO TMPVAT
					EXEC dbo.Select_AS_VAT8 11,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
				END
			ELSE IF(@F_VoucherTypeMaster=6)
				BEGIN
					INSERT INTO TMPVAT
					EXEC dbo.Select_AS_VAT8 6,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
				END
			ELSE IF(@F_VoucherTypeMaster=12)
				BEGIN
					INSERT INTO TMPVAT
					EXEC dbo.Select_AS_VAT8 12,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster,@RowNumber
				END
		END
		
	SELECT TAB1.RowNumber,TAB1.Id,TAB1.VoucherNo,TAB1.VoucherDate,TAB1.Ledger,TAB1.TINNo,TAB1.FinYear,TAB1.TotalNetAmt
		,TAB1.Tax1,TAB1.TaxAmt1,TAB1.TaxableAmt1
		,TAB1.Tax2,TAB1.TaxAmt2,TAB1.TaxableAmt2
		,TAB1.Tax3,TAB1.TaxAmt3,TAB1.TaxableAmt3
		,TAB1.Tax4,TAB1.TaxAmt4,TAB1.TaxableAmt4
		,TAB1.Tax5,TAB1.TaxAmt5,TAB1.TaxableAmt5
		,TAB1.GrossAmt_Ex_VAT
		,TAB1.TotalTax
		,TAB1.GrossAmt_Inc_VAT
		,TAB1.Difference
		,TAB1.MonthName
		,TAB1.MonthNo
		,TAB1.VoucherDate2,TAB1.ReportNo,TAB1.Report,TAB1.Year
	FROM TMPVAT TAB1
	ORDER BY TAB1.VoucherDate2,CAST(LEFT(TAB1.VoucherNo, Patindex('%[^0-9]%', TAB1.VoucherNo + 'x') - 1) AS INT)
END















GO
```

# [dbo].[Select_AS_VATDifference]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_VATDifference]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





























CREATE PROCEDURE [dbo].[Select_AS_VATDifference] 	
	@FromDate nvarchar(50)=NULL,
	@ToDate nvarchar(50)=NULL,
	@F_CompanyMaster numeric(18,0)=1,
	@MonthWiseFields NVARCHAR(MAX)='',
	@WithTINNo smallint=-1,
	@F_LedgerMaster numeric(18,0)=0
AS
BEGIN
	DECLARE @WHERE_WithTINNo NVARCHAR(MAX)
	IF(@WithTINNo=-1)
		SET @WHERE_WithTINNo=''
	ELSE IF(@WithTINNo=0)
		SET @WHERE_WithTINNo=' AND ISNULL(LM.TINNo,'''')<>'''' '
	ELSE
		SET @WHERE_WithTINNo=' AND ISNULL(LM.TINNo,'''')='''' '
		
	DECLARE @MonthWiseFields_Order nvarchar(max)	
	IF(@MonthWiseFields<>'')
		BEGIN
			SET @MonthWiseFields_Order = @MonthWiseFields + ','
			SET @MonthWiseFields=',' + @MonthWiseFields
		END
	ELSE
	
	DECLARE @F_LedgerMaster_Tax1 numeric(18,0)
	DECLARE @F_LedgerMaster_Tax2 numeric(18,0)
	DECLARE @F_LedgerMaster_Tax3 numeric(18,0)
	DECLARE @F_LedgerMaster_Tax4 numeric(18,0)	
	DECLARE @F_LedgerMaster_Tax5 numeric(18,0)	
	DECLARE @Tax1_Name nvarchar(50)
	DECLARE @Tax2_Name nvarchar(50)
	DECLARE @Tax3_Name nvarchar(50)
	DECLARE @Tax4_Name nvarchar(50)
	DECLARE @Tax5_Name nvarchar(50)

	--TAX ID
	SET @F_LedgerMaster_Tax1 = (SELECT ISNULL(F_LedgerMaster_Tax1,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
	SET @F_LedgerMaster_Tax2 = (SELECT ISNULL(F_LedgerMaster_Tax2,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
	SET @F_LedgerMaster_Tax3 = (SELECT ISNULL(F_LedgerMaster_Tax3,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
	SET @F_LedgerMaster_Tax4 = (SELECT ISNULL(F_LedgerMaster_Tax4,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
	SET @F_LedgerMaster_Tax5 = (SELECT ISNULL(F_LedgerMaster_Tax5,0) FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
	--TAX NAME
	SET @Tax1_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax1)
	SET @Tax2_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax2)
	SET @Tax3_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax3)
	SET @Tax4_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax4)
	SET @Tax5_Name = (SELECT Name FROM LedgerMaster WHERE Id=@F_LedgerMaster_Tax5)

	------***************------
	------INSERTING RECORDS IN TMP VAT TABLE-----
	----Deleting tmp vat
	DELETE FROM TMPVAT
	--Total Sale
	INSERT INTO TMPVAT
	Exec dbo.Select_AS_VAT8 5,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster		--Sales Voucher
	INSERT INTO TMPVAT
	Exec dbo.Select_AS_VAT8 9,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster		--Sales Invoice
	--Less Sale Return
	--INSERT INTO TMPVAT
	--Exec dbo.Select_AS_VAT8 6,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster		--Credit Note
	INSERT INTO TMPVAT
	Exec dbo.Select_AS_VAT8 10,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster	--Credit Note Inventory (Sales Return)
	--Less Purchase
	INSERT INTO TMPVAT
	Exec dbo.Select_AS_VAT8 7,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster		--Purchase
	INSERT INTO TMPVAT
	Exec dbo.Select_AS_VAT8 11,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster	--Purchase Invoice
	--Less Purchase Return
	--INSERT INTO TMPVAT
	--Exec dbo.Select_AS_VAT8 8,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster		--Debit Note
	INSERT INTO TMPVAT
	Exec dbo.Select_AS_VAT8 12,@FromDate,@ToDate,@F_CompanyMaster,@WithTINNo,@F_LedgerMaster	--Debit Note Inventory (Purchase Return)
	------***************------
EXEC('
	SELECT (MIN(TAB1.Year) + ''-'' + MAX(TAB1.Year)) AS FinYear, TAB1.ReportNo,TAB1.Report' + @MonthWiseFields + '
		,''' + @Tax1_Name + ''' AS Tax1
		,''' + @Tax2_Name + ''' AS Tax2
		,''' + @Tax3_Name + ''' AS Tax3
		,''' + @Tax4_Name + ''' AS Tax4
		,''' + @Tax5_Name + ''' AS Tax5
		,SUM(TAB1.TotalNetAmt) AS TotalNetAmt
		,SUM(TAB1.Difference) AS Difference
		,SUM(TAB1.GrossAmt_Ex_VAT) AS GrossAmt_Ex_VAT
		,SUM(TAB1.TotalTax) AS TotalTax
		,SUM(TAB1.GrossAmt_Inc_VAT) AS GrossAmt_Inc_VAT
		,SUM(CASE WHEN(TAB1.ReportNo=6) THEN (TAB1.TotalTax)*-1 ELSE (CASE WHEN(TAB1.ReportNo=3) THEN (TAB1.TotalTax) ELSE 0 END) END) AS TotalTaxAct
		,SUM(CASE WHEN(TAB1.ReportNo=6) THEN (TAB1.TaxAmt1)*-1 ELSE (CASE WHEN(TAB1.ReportNo=3) THEN (TAB1.TaxAmt1) ELSE 0 END) END) AS TotalTaxAct1
		,SUM(CASE WHEN(TAB1.ReportNo=6) THEN (TAB1.TaxAmt2)*-1 ELSE (CASE WHEN(TAB1.ReportNo=3) THEN (TAB1.TaxAmt2) ELSE 0 END) END) AS TotalTaxAct2
		,SUM(CASE WHEN(TAB1.ReportNo=6) THEN (TAB1.TaxAmt3)*-1 ELSE (CASE WHEN(TAB1.ReportNo=3) THEN (TAB1.TaxAmt3) ELSE 0 END) END) AS TotalTaxAct3
		,SUM(CASE WHEN(TAB1.ReportNo=6) THEN (TAB1.TaxAmt4)*-1 ELSE (CASE WHEN(TAB1.ReportNo=3) THEN (TAB1.TaxAmt4) ELSE 0 END) END) AS TotalTaxAct4
		,SUM(CASE WHEN(TAB1.ReportNo=6) THEN (TAB1.TaxAmt5)*-1 ELSE (CASE WHEN(TAB1.ReportNo=3) THEN (TAB1.TaxAmt5) ELSE 0 END) END) AS TotalTaxAct5
		,SUM(TAB1.TaxableAmt1) AS TaxableAmt1,SUM(TAB1.TaxAmt1) AS TaxAmt1
		,SUM(TAB1.TaxableAmt2) AS TaxableAmt2,SUM(TAB1.TaxAmt2) AS TaxAmt2
		,SUM(TAB1.TaxableAmt3) AS TaxableAmt3,SUM(TAB1.TaxAmt3) AS TaxAmt3
		,SUM(TAB1.TaxableAmt4) AS TaxableAmt4,SUM(TAB1.TaxAmt4) AS TaxAmt4
		,SUM(TAB1.TaxableAmt5) AS TaxableAmt5,SUM(TAB1.TaxAmt5) AS TaxAmt5
		FROM (
		--Sales
			SELECT ReportNo,Report,[Year],[MonthName],MonthNo,TaxableAmt1,TaxAmt1,TaxableAmt2,TaxAmt2,TaxableAmt3,TaxAmt3,TaxableAmt4,TaxAmt4,TaxableAmt5,TaxAmt5,TotalNetAmt,Difference,GrossAmt_Ex_VAT,TotalTax,GrossAmt_Inc_VAT
			FROM TMPVAT WHERE ReportNo=1
			
			UNION ALL
		--Sales Return
			SELECT ReportNo,Report,[Year],[MonthName],MonthNo,TaxableAmt1,TaxAmt1,TaxableAmt2,TaxAmt2,TaxableAmt3,TaxAmt3,TaxableAmt4,TaxAmt4,TaxableAmt5,TaxAmt5,TotalNetAmt,Difference,GrossAmt_Ex_VAT,TotalTax,GrossAmt_Inc_VAT
			FROM TMPVAT WHERE ReportNo=2
				
			UNION ALL	
		--Net Sales			
			SELECT 3 AS ReportNo, ''Net Sales'' AS Report,MAX(TAB2.Year) AS [Year],TAB2.MonthName,TAB2.MonthNo
				,SUM(TAB2.TaxableAmt1) AS TaxableAmt1,SUM(TAB2.TaxAmt1) AS TaxAmt1
				,SUM(TAB2.TaxableAmt2) AS TaxableAmt2,SUM(TAB2.TaxAmt2) AS TaxAmt2
				,SUM(TAB2.TaxableAmt3) AS TaxableAmt3,SUM(TAB2.TaxAmt3) AS TaxAmt3
				,SUM(TAB2.TaxableAmt4) AS TaxableAmt4,SUM(TAB2.TaxAmt4) AS TaxAmt4
				,SUM(TAB2.TaxableAmt5) AS TaxableAmt5,SUM(TAB2.TaxAmt5) AS TaxAmt5
				,SUM(TAB2.TotalNetAmt) AS TotalNetAmt,SUM(TAB2.Difference) AS Difference
				,SUM(TAB2.GrossAmt_Ex_VAT) AS GrossAmt_Ex_VAT,SUM(TotalTax) AS TotalTax,SUM(GrossAmt_Inc_VAT) AS GrossAmt_Inc_VAT
			FROM (
					SELECT ReportNo,Report,[Year],[MonthName],MonthNo,TaxableAmt1,TaxAmt1,TaxableAmt2,TaxAmt2,TaxableAmt3,TaxAmt3,TaxableAmt4,TaxAmt4,TaxableAmt5,TaxAmt5,TotalNetAmt,Difference,GrossAmt_Ex_VAT,TotalTax,GrossAmt_Inc_VAT
					FROM TMPVAT WHERE ReportNo=1
					UNION ALL
					SELECT ReportNo,Report,[Year],[MonthName],MonthNo,(TaxableAmt1*-1) AS TaxableAmt1,(TaxAmt1*-1) AS TaxAmt1,(TaxableAmt2*-1) AS TaxableAmt2,(TaxAmt2*-1) AS TaxAmt2,(TaxableAmt3*-1) AS TaxableAmt3,(TaxAmt3*-1) AS TaxAmt3,(TaxableAmt4*-1) AS TaxableAmt4,(TaxAmt4*-1) AS TaxAmt4,(TaxableAmt5*-1) AS TaxableAmt5,(TaxAmt5*-1) AS TaxAmt5,(TotalNetAmt*-1) AS TotalNetAmt,(Difference * -1) AS Difference,(GrossAmt_Ex_VAT*-1) AS GrossAmt_Ex_VAT,(TotalTax*-1) AS TotalTax,(GrossAmt_Inc_VAT*-1) AS GrossAmt_Inc_VAT
					FROM TMPVAT WHERE ReportNo=2
			) AS TAB2			
			GROUP BY TAB2.MonthName,TAB2.MonthNo		
			
			UNION ALL
		--Purchase
			SELECT ReportNo,Report,[Year],[MonthName],MonthNo,TaxableAmt1,TaxAmt1,TaxableAmt2,TaxAmt2,TaxableAmt3,TaxAmt3,TaxableAmt4,TaxAmt4,TaxableAmt5,TaxAmt5,TotalNetAmt,Difference,GrossAmt_Ex_VAT,TotalTax,GrossAmt_Inc_VAT
			FROM TMPVAT WHERE ReportNo=4
			UNION ALL
		--Purchase Return
			SELECT ReportNo,Report,[Year],[MonthName],MonthNo,TaxableAmt1,TaxAmt1,TaxableAmt2,TaxAmt2,TaxableAmt3,TaxAmt3,TaxableAmt4,TaxAmt4,TaxableAmt5,TaxAmt5,TotalNetAmt,Difference,GrossAmt_Ex_VAT,TotalTax,GrossAmt_Inc_VAT
			FROM TMPVAT WHERE ReportNo=5
			UNION ALL
		--Net Purchase
			SELECT 6 AS ReportNo, ''Net Purchase'' AS Report,MAX(TAB2.Year) AS Year,TAB2.MonthName,TAB2.MonthNo
				,SUM(TAB2.TaxableAmt1) AS TaxableAmt1,SUM(TAB2.TaxAmt1) AS TaxAmt1
				,SUM(TAB2.TaxableAmt2) AS TaxableAmt2,SUM(TAB2.TaxAmt2) AS TaxAmt2
				,SUM(TAB2.TaxableAmt3) AS TaxableAmt3,SUM(TAB2.TaxAmt3) AS TaxAmt3
				,SUM(TAB2.TaxableAmt4) AS TaxableAmt4,SUM(TAB2.TaxAmt4) AS TaxAmt4
				,SUM(TAB2.TaxableAmt5) AS TaxableAmt5,SUM(TAB2.TaxAmt5) AS TaxAmt5
				,SUM(TAB2.TotalNetAmt) TotalNetAmt,SUM(TAB2.Difference) Difference
				,SUM(TAB2.GrossAmt_Ex_VAT) AS GrossAmt_Ex_VAT,SUM(TotalTax) AS TotalTax,SUM(GrossAmt_Inc_VAT) AS GrossAmt_Inc_VAT
			FROM (
					SELECT ReportNo,Report,[Year],[MonthName],MonthNo,TaxableAmt1,TaxAmt1,TaxableAmt2,TaxAmt2,TaxableAmt3,TaxAmt3,TaxableAmt4,TaxAmt4,TaxableAmt5,TaxAmt5,TotalNetAmt,Difference,GrossAmt_Ex_VAT,TotalTax,GrossAmt_Inc_VAT
					FROM TMPVAT WHERE ReportNo=4
					UNION ALL
					SELECT ReportNo,Report,[Year],[MonthName],MonthNo,(TaxableAmt1*-1) AS TaxableAmt1,(TaxAmt1*-1) AS TaxAmt1,(TaxableAmt2*-1) AS TaxableAmt2,(TaxAmt2*-1) AS TaxAmt2,(TaxableAmt3*-1) AS TaxableAmt3,(TaxAmt3*-1) AS TaxAmt3,(TaxableAmt4*-1) AS TaxableAmt4,(TaxAmt4*-1) AS TaxAmt4,(TaxableAmt5*-1) AS TaxableAmt5,(TaxAmt5*-1) AS TaxAmt5,(TotalNetAmt*-1) AS TotalNetAmt,(Difference*-1) AS Difference,(GrossAmt_Ex_VAT*-1) AS GrossAmt_Ex_VAT,(TotalTax*-1) AS TotalTax,(GrossAmt_Inc_VAT*-1) AS GrossAmt_Inc_VAT
					FROM TMPVAT WHERE ReportNo=5
			) AS TAB2			
			GROUP BY TAB2.MonthName,TAB2.MonthNo
			
		) AS TAB1
		
	GROUP BY TAB1.Report,TAB1.ReportNo' + @MonthWiseFields + '
	ORDER BY ' + @MonthWiseFields_Order + 'TAB1.ReportNo')
END


























GO
```

# [dbo].[Select_AS_VATPaid]

```sql
/****** Object:  StoredProcedure [dbo].[Select_AS_VATPaid]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO


























CREATE PROCEDURE [dbo].[Select_AS_VATPaid]
	@FromDate nvarchar(50),
	@ToDate nvarchar(50),
	@F_CompanyMaster numeric(18,0)
AS
BEGIN
	DECLARE @F_LedgerMaster_VATPayable NUMERIC(18,0)
	SET @F_LedgerMaster_VATPayable = (SELECT F_LedgerMaster_VATPayable FROM Options WHERE F_CompanyMaster = @F_CompanyMaster)
	
	SELECT LMP.Name TaxPayable,VL.Amount,VL.DrCr,VH.VoucherNo		
		,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate
		,LM.Name AS Tax
	FROM VoucherL VL
		LEFT JOIN VoucherH VH ON VH.Id=VL.F_VoucherH
		LEFT JOIN LedgerMaster LM ON LM.Id = VH.F_LedgerMaster
		LEFT JOIN LedgerMaster LMP ON LMP.Id = VL.F_LedgerMaster
	WHERE VL.F_LedgerMaster=@F_LedgerMaster_VATPayable AND VH.F_VoucherTypeMaster=4
		AND VH.VoucherDate>=@FromDate AND VH.VoucherDate<=@ToDate
		AND LM.F_LedgerGroupMaster=12 AND VL.F_CompanyMaster = 1 AND VH.F_CompanyMaster=@F_CompanyMaster
	ORDER BY VoucherDate
END



















GO
```

# [dbo].[Select_BillCollectionBillNo_DateWise]

```sql
/****** Object:  StoredProcedure [dbo].[Select_BillCollectionBillNo_DateWise]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






















CREATE PROCEDURE [dbo].[Select_BillCollectionBillNo_DateWise] 
	@F_CompanyMaster numeric(18,0),
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Id numeric(18,0)=0,
	@F_CompanyYears numeric(18,0)
AS
BEGIN
	SELECT Name,Id,VoucherNo1 FROM (
		SELECT 0 AS RowNumber,'' AS Name,0 AS Id,' [None]' AS VoucherNo1
		
		UNION ALL
		
		SELECT ROW_NUMBER() OVER (ORDER BY CAST(VH.VoucherNo AS NUMERIC) ASC) AS RowNumber,
		(VH.VoucherNo + '    ' + (SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4))) AS Name
		,VH.Id,VH.VoucherNo AS VoucherNo1
			FROM BillCollectionL DL
				LEFT JOIN SaleInvoiceH VH ON VH.Id=DL.F_SaleInvoiceH
				LEFT JOIN LedgerMaster LM ON VH.F_LedgerMaster=LM.Id
				LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=VH.F_CompanyMaster AND CY.Id=@F_CompanyYears
			WHERE VH.F_CompanyMaster =@F_CompanyMaster
				AND (VH.VoucherDate>=CY.FinYearFrom AND VH.VoucherDate<=CY.FinYearTo)
		) AS Tab1
	ORDER BY RowNumber,Name,Id
END




















































GO
```

# [dbo].[Select_BillNo_DateWise]

```sql
/****** Object:  StoredProcedure [dbo].[Select_BillNo_DateWise]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO























































CREATE PROCEDURE [dbo].[Select_BillNo_DateWise] 
	@F_CompanyMaster numeric(18,0),
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Id numeric(18,0)=0
AS
BEGIN
	SELECT Name,Id,VoucherNo1 FROM (
		SELECT 0 AS RowNumber,' [None]' AS Name,0 AS Id,' [None]' AS VoucherNo1
		UNION ALL
		SELECT ROW_NUMBER() OVER (order by CAST(VH.VoucherNo AS NUMERIC) ASC) as RowNumber,
	((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+
		VH.VoucherNo + ' ' + (SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) + ' ' + LM.Name) AS Name
	,VH.Id,
	((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+VH.VoucherNo) AS VoucherNo1
		FROM [dbo].[SaleInvoiceH] VH LEFT JOIN LedgerMaster LM ON VH.F_LedgerMaster=LM.Id
		WHERE VH.F_CompanyMaster =@F_CompanyMaster
			AND 1=
				CASE WHEN(@FromDate='') THEN 1 ELSE
					(CASE WHEN(VH.VoucherDate >= @FromDate AND VH.VoucherDate <=@ToDate) THEN 1 ELSE 0 END)
				END
			AND ISNULL(VH.Cancelled,0)=0
			AND 
				(--BILL HAVING NOT COMPLETED STATUS
					(VH.Id IN(SELECT DL.F_SaleInvoiceH
							FROM DeliveryProcessL DL LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DL.F_DeliveryStatusMaster
							WHERE ISNULL(DSM.IsCompleted,0)=0 AND isnull(DL.F_DeliveryStatusMaster,0)>0
							)
				AND
				--BILL HAVING COMPLETED STATUS
					VH.Id NOT IN(SELECT DL.F_SaleInvoiceH
							FROM DeliveryProcessL DL LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DL.F_DeliveryStatusMaster
							WHERE ISNULL(DSM.IsCompleted,0)=1 AND isnull(DL.F_DeliveryStatusMaster,0)>0
							))
				OR
				--BILL ARE ENTER IN DELIVERY PROESS ENTRY BOOK
				VH.Id NOT IN(SELECT DL.F_SaleInvoiceH
							FROM DeliveryProcessL DL
							)
				OR VH.Id IN(SELECT DL.F_SaleInvoiceH
							FROM DeliveryProcessL DL
							WHERE DL.F_DeliveryProcessH = @Id)
				)
		) AS Tab1
	ORDER BY RowNumber,Name,Id
END













































GO
```

# [dbo].[Select_DebitNoteBillNo_MoneyReceipt]

```sql
/****** Object:  StoredProcedure [dbo].[Select_DebitNoteBillNo_MoneyReceipt]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




























CREATE PROCEDURE [dbo].[Select_DebitNoteBillNo_MoneyReceipt] 
	@F_CompanyMaster numeric(18,0),
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Id numeric(18,0)=0,
	@F_LedgerMaster numeric(18,0)=0,
	@F_OtherLedgerMaster numeric(18,0),
	@CurrentDate nvarchar(50)='20160401',
	@AreaIds NVARCHAR(MAX)=''
AS
BEGIN
	SELECT RowNumber,Name,Id,VoucherNo1,VoucherDate,DueAmt,DueDays,IsDisputed FROM (
		SELECT 0 AS RowNumber,'' AS VoucherDate,'' AS Name,0 AS Id,'' AS VoucherNo1,0 AS DueAmt,0 AS DueDays,0 AS IsDisputed
		UNION ALL
		SELECT DENSE_RANK() OVER (order by TAB2.RowNumber) AS RowNumber,TAB2.VoucherDate,TAB2.Name,TAB2.Id,TAB2.VoucherNo1,TAB2.DueAmt,DATEDIFF(day,CONVERT(datetime,TAB2.VoucherDate1),CONVERT(datetime,@CurrentDate)) AS DueDays,TAB2.IsDisputed FROM
		(
			SELECT (VH.TotalNetAmt -
				(	ISNULL(
						(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptCashL MRL LEFT JOIN MoneyReceiptCashH MRH ON MRH.Id=MRL.F_MoneyReceiptCashH
							LEFT JOIN VoucherH VH1 ON VH1.Id = MRL.F_SaleInvoiceH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0 AND VH1.F_VoucherTypeMaster = 8 AND VH1.Id=VH.Id AND MRH.F_LedgerMaster = VH.F_LedgerMaster)
					,0)
					+ ISNULL(
						(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptBankL MRL LEFT JOIN MoneyReceiptBankH MRH ON MRH.Id=MRL.F_MoneyReceiptBankH
							LEFT JOIN VoucherH VH1 ON VH1.Id = MRL.F_SaleInvoiceH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0 AND ISNULL(MRH.IsChequeCancel,0)=0 AND VH1.F_VoucherTypeMaster = 8 AND VH1.Id=VH.Id AND MRH.F_LedgerMaster = VH.F_LedgerMaster)

					,0)
					--+ ISNULL(
					--	(SELECT SUM(VH1.TotalNetAmt)
					--		FROM SaleInvoiceH_CreditNote SIC LEFT JOIN VoucherH VH1 ON VH1.Id=SIC.F_VoucherH
					--		WHERE SIC.F_SaleInvoiceH = VH.Id AND ISNULL(VH1.Cancelled,0)=0)
					--,0)
					--+ ISNULL(
					--	(SELECT SUM(MRL.Amount)
					--		FROM SaleReturnL_InvoiceDetails MRL LEFT JOIN SaleReturnH MRH ON MRH.Id=MRL.F_SaleReturnH
					--		WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)

					--,0)
					)) AS DueAmt
			,DENSE_RANK() OVER (order by VH.VoucherDate,CAST(VH.VoucherNo AS NUMERIC)) AS RowNumber
			,((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+
				VH.VoucherNo + '            ' + (SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) + '            ' + 
				CAST(VH.TotalNetAmt 
					-(	ISNULL(
							(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptCashL MRL LEFT JOIN MoneyReceiptCashH MRH ON MRH.Id=MRL.F_MoneyReceiptCashH
							LEFT JOIN VoucherH VH1 ON VH1.Id = MRL.F_SaleInvoiceH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0 AND VH1.F_VoucherTypeMaster = 8 AND VH1.Id=VH.Id AND MRH.F_LedgerMaster = VH.F_LedgerMaster)
						,0) 
					+	ISNULL(
							(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptBankL MRL LEFT JOIN MoneyReceiptBankH MRH ON MRH.Id=MRL.F_MoneyReceiptBankH
							LEFT JOIN VoucherH VH1 ON VH1.Id = MRL.F_SaleInvoiceH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0 AND ISNULL(MRH.IsChequeCancel,0)=0 AND VH1.F_VoucherTypeMaster = 8 AND VH1.Id=VH.Id AND MRH.F_LedgerMaster = VH.F_LedgerMaster)
						,0) 
					--+	ISNULL(
					--		(SELECT SUM(VH1.TotalNetAmt)
					--		FROM SaleInvoiceH_CreditNote SIC LEFT JOIN VoucherH VH1 ON VH1.Id=SIC.F_VoucherH
					--		WHERE SIC.F_SaleInvoiceH = VH.Id AND ISNULL(VH1.Cancelled,0)=0)
					--	,0)
					--+	ISNULL(
					--		(SELECT SUM(MRL.Amount)
					--		FROM SaleReturnL_InvoiceDetails MRL LEFT JOIN SaleReturnH MRH ON MRH.Id=MRL.F_SaleReturnH
					--		WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)
					--	,0)
						) AS NVARCHAR)) AS Name
			,VH.Id,(SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate,
			((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+VH.VoucherNo + ' ') AS VoucherNo1,VH.VoucherDate AS VoucherDate1,0 AS IsDisputed
			FROM [dbo].[VoucherH] VH
				LEFT JOIN LedgerMaster LM ON VH.F_LedgerMaster=LM.Id				
			WHERE VH.F_CompanyMaster =@F_CompanyMaster AND VH.F_VoucherTypeMaster = 8
				AND 1= CASE WHEN(@FromDate='' AND @F_LedgerMaster>0) THEN 1 ELSE (CASE WHEN(VH.VoucherDate <=@ToDate AND @F_LedgerMaster>0) THEN 1 ELSE 				
					(CASE WHEN(VH.VoucherDate >=@FromDate AND VH.VoucherDate <=@ToDate) THEN 1 ELSE 0 END)		
				END) END
				AND ISNULL(VH.Cancelled,0)=0
				AND 1=(CASE WHEN(@F_LedgerMaster=0) THEN 1 ELSE (CASE WHEN(VH.F_LedgerMaster = @F_LedgerMaster) THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN (@AreaIds ='') THEN 1 ELSE (CASE WHEN (LM.F_AreaMaster IN (SELECT * FROM Split(@AreaIds ,',')) ) THEN 1 ELSE 0 END) END)
				--AND VH.F_OtherLedgerMaster = @F_OtherLedgerMaster
			GROUP BY VH.Id,VH.VoucherNo,VH.VoucherDate,VH.TotalNetAmt,VH.VoucherNoPrefix,VH.F_LedgerMaster
		) AS TAB2
		WHERE TAB2.DueAmt>0
		) AS TAB1
	ORDER BY RowNumber,Name,Id
END




















GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_EWayBillDetails]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO














CREATE PROCEDURE [dbo].[Select_EWayBillDetails]
	@VoucherNo nvarchar(50),
	@VoucherDate nvarchar(50)
AS
BEGIN
	SELECT SH.VoucherNo,SH.VoucherDate,LM.Name AS Ledger,LM.GSTNo,MAX(IM.Name) AS ItemGroup,HM.Name AS HSNCode
	,SUM(SL.MinQty) AS Qty,UM.Name AS Unit,SUM(SL.FinalAmt) AS TaxableAmt,SL.TaxRate,SL.TaxRate2,SL.TaxRate3
	FROM SaleInvoiceH SH
	LEFT JOIN SaleInvoiceL SL ON SH.Id=SL.F_SaleInvoiceH
	LEFT JOIN ItemMaster IM ON IM.Id=SL.F_ItemMaster
	LEFT JOIN LedgerMaster LM ON LM.Id=SH.F_LedgerMaster
	LEFT JOIN UnitMaster UM ON UM.Id=SL.F_UnitMasterQ
	LEFT JOIN HSNCodeMaster HM ON HM.Id=IM.F_HSNCodeMaster
	WHERE SH.VoucherNo=@VoucherNo AND VoucherDate>=@VoucherDate
	GROUP BY SH.VoucherNo,SH.VoucherDate,LM.Name,LM.GSTNo,HM.Name,UM.Name,SL.TaxRate,SL.TaxRate2,SL.TaxRate3
	ORDER BY HSNCode
END



















































GO
```

# [dbo].[Select_Import_SaleInvoice_Item]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Import_SaleInvoice_Item]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





CREATE PROCEDURE [dbo].[Select_Import_SaleInvoice_Item] 
	@InvoiceNo nvarchar(50)
AS
BEGIN
	SELECT TAB1.BillNo,
			TAB1.InvoiceDate,
			TAB1.Item,
			--MAX(TAB1.TypeDetail) AS TypeDetail,
			TypeDetail = STUFF(
			(SELECT ', ' + (TDM1.TypeDetail + '-' + RTRIM(LTRIM(STR(VL1.Qty_Pcs)))) 
				FROM SaleInvoice_ImportH VL1
					LEFT JOIN SaleInvoiceItemMapMaster ISTM1 ON ISTM1.SaleInvoiceItemId = VL1.SkuCode
					LEFT JOIN TypeDetailMaster TDM1 ON TDM1.Id=ISTM1.F_TypeDetailMaster
				WHERE VL1.BillNo = TAB1.BillNo AND VL1.SaleDate= TAB1.SaleDate AND ISTM1.F_ItemMaster=TAB1.F_ItemMaster AND ISTM1.MRP=TAB1.MRP
					AND VL1.Rate=TAB1.Rate
					AND CAST(ROUND(((VL1.Scheme_Value+VL1.QPS_Scheme_Value)/(VL1.Qty_Pcs*VL1.Rate))*100,2) AS numeric(18,2))=TAB1.Scheme_Rate
				ORDER BY TDM1.Id 
				FOR XML PATH('')), 1, 2, ''),
			SUM(TAB1.Qty_Pcs) AS Qty_Pcs,TAB1.MRP,TAB1.Rate,
			TAB1.Scheme_Rate,SUM(TAB1.Scheme_Value) AS Scheme_Value,SUM(TAB1.Taxable_Amount) AS Taxable_Amount,
			TAB1.F_ItemMaster,
			TypeDetailId = STUFF(
			(SELECT ', ' + (RTRIM(LTRIM(STR(TDM1.Id))) + '-' + RTRIM(LTRIM(STR(VL1.Qty_Pcs)))) 
				FROM SaleInvoice_ImportH VL1
					LEFT JOIN SaleInvoiceItemMapMaster ISTM1 ON ISTM1.SaleInvoiceItemId = VL1.SkuCode
					LEFT JOIN TypeDetailMaster TDM1 ON TDM1.Id=ISTM1.F_TypeDetailMaster
				WHERE VL1.BillNo = TAB1.BillNo AND VL1.SaleDate= TAB1.SaleDate AND ISTM1.F_ItemMaster=TAB1.F_ItemMaster AND ISTM1.MRP=TAB1.MRP
					AND VL1.Rate=TAB1.Rate
					AND CAST(ROUND(((VL1.Scheme_Value+VL1.QPS_Scheme_Value)/(VL1.Qty_Pcs*VL1.Rate))*100,2) AS numeric(18,2))=TAB1.Scheme_Rate
				ORDER BY TDM1.Id 
				FOR XML PATH('')), 1, 2, '')
	FROM (
		SELECT VL.Id,VL.BillNo,
			(SUBSTRING(VL.SaleDate,7,2)+'/'+SUBSTRING(VL.SaleDate,5,2)+'/'+SUBSTRING(VL.SaleDate,1,4)) AS InvoiceDate,
			IM.Name AS Item,TDM.TypeDetail,
			VL.Qty_Pcs,ISTM.MRP,VL.Rate,
			CAST(ROUND(((VL.Scheme_Value+VL.QPS_Scheme_Value)/(VL.Qty_Pcs*VL.Rate))*100,2) AS numeric(18,2)) AS Scheme_Rate,VL.Scheme_Value,VL.Taxable_Amount,
			ISTM.F_ItemMaster,VL.SaleDate
		FROM SaleInvoice_ImportH VL
		LEFT JOIN SaleInvoiceItemMapMaster ISTM ON ISTM.SaleInvoiceItemId = VL.SkuCode
		LEFT JOIN ItemMaster IM ON IM.Id = ISTM.F_ItemMaster
		LEFT JOIN TypeDetailMaster TDM ON TDM.Id = ISTM.F_TypeDetailMaster
		LEFT JOIN EstimateH EH ON ISNULL(EH.BizomInvoiceNo,0) = VL.BillNo
		WHERE VL.BillNo=@InvoiceNo AND ISNULL(EH.Id,0)=0
	) AS TAB1
	GROUP BY TAB1.BillNo,TAB1.InvoiceDate,TAB1.Item,TAB1.F_ItemMaster,TAB1.MRP,TAB1.Rate,TAB1.Scheme_Rate,TAB1.SaleDate
END
GO
```

# [dbo].[Select_ItemMaster]

```sql
/****** Object:  StoredProcedure [dbo].[Select_ItemMaster]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE PROCEDURE [dbo].[Select_ItemMaster]
	@SearchName nvarchar(150),
	@F_ItemBrandMaster numeric(18,0)=0,
	@F_CompanyMaster numeric(18,0)=1,
	@IsActive smallint=-1,
	@MRP numeric(18,2)=0,
	@SearchType bit=0
AS
BEGIN
	IF(@SearchType=0)
	BEGIN
		SELECT IM.Name,IM.Id
		FROM ItemMaster IM
		WHERE (charindex(@SearchName ,Name)>0 OR @SearchName='')
			AND IM.F_CompanyMaster = @F_CompanyMaster
			AND 1=(CASE WHEN(@F_ItemBrandMaster =0) THEN 1 ELSE (CASE WHEN(IM.F_ItemBrandMaster = @F_ItemBrandMaster) THEN 1 ELSE 0 END) END)
			AND 1=(CASE WHEN(@IsActive =-1) THEN 1 ELSE (CASE WHEN(IM.IsActive = @IsActive) THEN 1 ELSE 0 END) END)
		GROUP BY IM.Name,IM.Id
		ORDER BY IM.Name,IM.Id
	END
	ELSE
	BEGIN
		SELECT IM.Id,IGM.Name AS ItemGroup,IM.Name,ISTM.MRP,
		TypeDetails = STUFF(
					(SELECT ', ' + (TDM.TypeDetail) 
					FROM ItemTypeDetailMaster SL1
						LEFT JOIN TypeDetailMaster TDM ON TDM.Id=SL1.F_TypeDetailMaster
					WHERE SL1.F_ItemMaster = IM.Id AND ISNULL(SL1.IsSelected,0)=1
					ORDER BY TDM.Id
					FOR XML PATH('')), 1, 2, '')
		FROM ItemMaster IM
		LEFT JOIN ItemGroupMaster IGM ON IGM.Id = IM.F_ItemGroupMaster
		LEFT JOIN ItemSalesTaxMaster ISTM ON ISTM.F_ItemMaster = IM.Id
		WHERE (CHARINDEX(@SearchName ,IM.Name)>0
				OR CHARINDEX(@SearchName ,IGM.Name)>0
				OR @SearchName='')
			AND 1=(CASE WHEN(@MRP=0) THEN 1 ELSE (CASE WHEN(ISTM.MRP=@MRP) THEN 1 ELSE 0 END) END)
			AND IM.F_CompanyMaster = @F_CompanyMaster
			AND 1=(CASE WHEN(@F_ItemBrandMaster =0) THEN 1 ELSE (CASE WHEN(IM.F_ItemBrandMaster = @F_ItemBrandMaster) THEN 1 ELSE 0 END) END)
			AND 1=(CASE WHEN(@IsActive =-1) THEN 1 ELSE (CASE WHEN(IM.IsActive = @IsActive) THEN 1 ELSE 0 END) END)
			AND ISTM.SDate>=20170701
		GROUP BY IGM.Name,IM.Name,ISTM.SDate,ISTM.MRP,IM.Id
		ORDER BY IGM.Name,IM.Name,ISTM.SDate DESC,ISTM.MRP DESC,IM.Id
	END
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_LedgerMaster]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











CREATE PROCEDURE [dbo].[Select_LedgerMaster]
	@Id numeric(18,0),
	@F_CompanyMaster numeric(18,0),
	@F_CompanyYears numeric(18,0)
AS
BEGIN
	DECLARE @Id_LedgerMasterL numeric(18,0)
	DECLARE @F_UserMaster numeric(18,0)
	DECLARE @F_LedgerGroupMaster numeric(18,0)
	
	SET @F_UserMaster = ISNULL((SELECT F_UserMaster FROM LedgerMaster WHERE Id=@Id),0)
	SET @F_LedgerGroupMaster = ISNULL((SELECT F_LedgerGroupMaster FROM LedgerMaster WHERE Id=@Id),0)
	SET @Id_LedgerMasterL = ISNULL((SELECT Id FROM LedgerMasterL WHERE F_LedgerMaster = @Id AND F_CompanyYears = @F_CompanyYears),0)
	
	IF(@Id_LedgerMasterL =0)
		BEGIN
			INSERT INTO [dbo].[LedgerMasterL]([F_LedgerMaster],[DepRate],[F_CompanyYears],[F_UserMaster],[DateOfCreation],[F_LedgerGroupMaster])
				VALUES (@Id,0,@F_CompanyYears,@F_UserMaster,GETDATE(),@F_LedgerGroupMaster)
		END

	SELECT LM.*,LML.F_LedgerGroupMaster AS F_LedgerGroupMaster_Curr
	FROM LedgerMaster LM
	LEFT JOIN LedgerMasterL LML ON LML.F_LedgerMaster = LM.Id
	WHERE LML.F_CompanyYears = @F_CompanyYears AND LM.Id = @Id AND LM.F_CompanyMaster = @F_CompanyMaster
END




































































GO
```

# [dbo].[Select_LedgerMaster_Spl]

```sql
/****** Object:  StoredProcedure [dbo].[Select_LedgerMaster_Spl]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO









CREATE PROCEDURE [dbo].[Select_LedgerMaster_Spl] 
	@F_ItemBrandMaster numeric(18,0)=0,
	@F_AreaMaster numeric(18,0)=0,
	@F_LedgerMaster numeric(18,0)=0,
	@F_LedgerGroupMaster numeric(18,0)=0,
	@F_CityMaster numeric(18,0)=0,
	@AreaIds nvarchar(max)='',
	@Select nvarchar(50)='',
	@F_CompanyMaster numeric(18,0)=1,
	@WithOne SMALLINT=1
AS
BEGIN
	IF(@Select='')
		BEGIN
			IF(@WithOne=1)
				BEGIN
					SELECT '[None]' AS Name,0 AS Id
					UNION ALL
					SELECT LM.Name,LM.Id
					FROM LedgerMaster LM
					WHERE 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE CASE WHEN(F_LedgerGroupMaster = @F_LedgerGroupMaster) THEN 1 ELSE 0 END END)
					AND 1=(CASE WHEN(@F_CityMaster=0) THEN 1 ELSE CASE WHEN(F_CityMaster = @F_CityMaster) THEN 1 ELSE 0 END END)
					ORDER BY Name,Id
				END
			ELSE
				BEGIN
					SELECT LM.Name,LM.Id
					FROM LedgerMaster LM
					WHERE 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE CASE WHEN(F_LedgerGroupMaster = @F_LedgerGroupMaster) THEN 1 ELSE 0 END END)
					AND 1=(CASE WHEN(@F_CityMaster=0) THEN 1 ELSE CASE WHEN(F_CityMaster = @F_CityMaster) THEN 1 ELSE 0 END END)
					ORDER BY Name,Id
				END
		END
	ELSE IF(@Select='Brand')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[ItemBrandCompanyMaster] IBM
			LEFT JOIN LedgerMaster LM ON LM.Id=IBM.F_LedgerMaster
			WHERE 1=
				(CASE WHEN(@F_ItemBrandMaster=-1) THEN 1 ELSE 
					CASE WHEN(F_ItemBrandMaster = @F_ItemBrandMaster) THEN 1 ELSE 0 END
				END)	
			ORDER BY Name,Id
		END
	ELSE IF(@Select='Other')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT 
			(LM.Name
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')
			)		
			,LM.Id
			FROM [dbo].[OtherLedgerMaster] LM
				LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=LM.F_CityMaster
				LEFT JOIN [dbo].[StateMaster] SM ON SM.Id=CM.F_StateMaster	
			WHERE F_LedgerMaster = @F_LedgerMaster
			ORDER BY Name,Id
		END
	ELSE IF(@Select='Purchase' OR @Select='TAXABLE EXPENSE INVOICE (INWARD)')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (6,31,40,10)
			ORDER BY Name,Id
		END
	ELSE IF(@Select ='DEBIT NOTE')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			--WHERE LGM.GroupType IN (6,31,32)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='PurchaseAc')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (25)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='Sales' OR @Select ='CREDIT NOTE' OR @Select ='SERVICE TAX INVOICE')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (1,6,31,32,40,10)
				AND 1 = (CASE WHEN(@F_AreaMaster=0) THEN 1 ELSE
							(CASE WHEN(@F_AreaMaster = LM.F_AreaMaster) THEN 1 ELSE 0 END)
						END)
				AND 1=(CASE WHEN (@AreaIds ='') THEN 1 ELSE (CASE WHEN (LM.F_AreaMaster  IN(SELECT * FROM Split(@AreaIds,',')) ) THEN 1 ELSE 0 END) END)
				AND 1=(CASE WHEN(@F_LedgerGroupMaster=0) THEN 1 ELSE CASE WHEN(F_LedgerGroupMaster = @F_LedgerGroupMaster) THEN 1 ELSE 0 END END)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='Sales-Debitor')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (32,40)
				AND 1 = (CASE WHEN(@F_AreaMaster=0) THEN 1 ELSE
							(CASE WHEN(@F_AreaMaster = LM.F_AreaMaster) THEN 1 ELSE 0 END)
						END)
				AND 1=(CASE WHEN (@AreaIds ='') THEN 1 ELSE (CASE WHEN (LM.F_AreaMaster  IN(SELECT * FROM Split(@AreaIds,',')) ) THEN 1 ELSE 0 END) END)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='Sales_WO')
		BEGIN
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (6,32,40)
				AND 1 = (CASE WHEN(@F_AreaMaster=0) THEN 1 ELSE
							(CASE WHEN(@F_AreaMaster = F_AreaMaster) THEN 1 ELSE 0 END)
						END)
				AND 1=(CASE WHEN (@AreaIds ='') THEN 1 ELSE (CASE WHEN (F_AreaMaster  IN(SELECT * FROM Split(@AreaIds,',')) ) THEN 1 ELSE 0 END) END)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='SalesAc')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE 1=CASE WHEN(@F_LedgerGroupMaster=0) THEN (CASE WHEN(LGM.GroupType IN (28)) THEN 1 ELSE 0 END)
				ELSE
					(CASE WHEN(@F_LedgerGroupMaster=LGM.GroupType) THEN 1 ELSE 0 END)
			END
			ORDER BY Name,Id
		END
	ELSE IF(@Select='SalesPurchaseAc')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (25,28)
			ORDER BY Name,Id
		END
	ELSE IF(@Select ='RCM')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (12) AND LM.TaxType=11
			ORDER BY Name,Id
		END
	ELSE IF(@Select ='RCM_Tax')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (12)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='Tax')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (10,12,13,14,16,17,18,19,35,36)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='RECEIPT' OR @Select='PAYMENT')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (6,1,2,3,29)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='RECEIPT-AGAINST' OR @Select='PAYMENT-AGAINST')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE 1=
				(CASE WHEN(@F_LedgerGroupMaster=0) THEN
					(CASE WHEN (LGM.GroupType NOT IN (6,1,3)) THEN 1 ELSE 0 END)
				ELSE
					(CASE WHEN(LM.F_LedgerGroupMaster = @F_LedgerGroupMaster) THEN 1 ELSE 0 END)
				END)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='JOURNAL')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id		
			ORDER BY Name,Id
		END
	ELSE IF(@Select='JOURNAL-AGAINST')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LM.Id NOT IN (@F_LedgerMaster)	
			ORDER BY Name,Id
		END
	ELSE IF(@Select='CONTRA-CASH')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (6)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='CONTRA-BANK')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (1,2,3,29)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='PURCHASE VOUCHER' OR @Select='SALES VOUCHER')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (6,31,32,40)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='VOUCHER TAX')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM
				LEFT JOIN LedgerGroupMaster LGM ON LGM.Id=LM.F_LedgerGroupMaster
			WHERE LGM.GroupType IN (7,10,12,13,14,15,16,17,18,19,25,28,35,36)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='Expenses')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM 
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (10,13,14,16,17,18,19,35,36)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='SGST')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LM.TaxType IN (6) AND LGM.GroupType IN (12)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='CGST')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LM.TaxType IN (7) AND LGM.GroupType IN (12)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='IGST')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LM.TaxType IN (8) AND LGM.GroupType IN (12)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='UGST')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LM.TaxType IN (9) AND LGM.GroupType IN (12)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='TDS')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LM.TaxType IN (4) AND LGM.GroupType IN (12)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='RCM')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LM.TaxType IN (11) AND LGM.GroupType IN (12)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='VAT')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LM.TaxType IN (0) AND LGM.GroupType IN (12)
			ORDER BY Name,Id
		END
	ELSE IF(@Select='PARTIES')
		BEGIN
			SELECT '[None]' AS Name,0 AS Id
			UNION ALL
			SELECT LM.Name,LM.Id
			FROM [dbo].[LedgerMaster] LM
			LEFT JOIN LedgerGroupMaster LGM ON LM.F_LedgerGroupMaster=LGM.Id
			WHERE LGM.GroupType IN (31,32,40)
			ORDER BY Name,Id
		END
END
GO
```

# [dbo].[Select_LoadingSheet]

```sql
/****** Object:  StoredProcedure [dbo].[Select_LoadingSheet]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO


CREATE PROCEDURE [dbo].[Select_LoadingSheet] 	
	@InvoiceIds nvarchar(max),
	@TableNameH nvarchar(50)='SaleInvoiceH',
	@TableNameL nvarchar(50)='SaleInvoiceL',
	@Status smallint=-1
AS
BEGIN
--
	DECLARE @ESTIMATE_WHERE NVARCHAR(MAX)
	IF(@TableNameH='EstimateH')
		BEGIN
			IF(@Status=-1)
				BEGIN
					SET @ESTIMATE_WHERE=''
				END
			ELSE IF(@Status=0) --NOT CONVERTED
				BEGIN
					--SET @ESTIMATE_WHERE=' AND GIL.ID NOT IN(SELECT ISNULL(F_EstimateL,0) FROM SaleInvoiceL)'
					SET @ESTIMATE_WHERE=' AND GIH.ID NOT IN(SELECT ISNULL(F_EstimateH,0) FROM SaleInvoiceH)'
				END
			ELSE IF(@Status=1) --CONVERTED
				BEGIN
					--SET @ESTIMATE_WHERE=' AND GIL.ID IN(SELECT ISNULL(F_EstimateL,0) FROM SaleInvoiceL)'
					SET @ESTIMATE_WHERE=' AND GIH.ID IN(SELECT ISNULL(F_EstimateH,0) FROM SaleInvoiceH)'
				END
		END
	EXEC('
	SELECT TAB1.Item,TAB1.MRP,SUM(TAB1.Qty) AS Qty,SUM(TAB1.Free) AS Free,SUM(TAB1.Qty + TAB1.Free) AS TotalQty
		,(SUM(TAB1.TaxAmt) + SUM(TAB1.FinalAmt)) AS NetAmt
		,Invoices = STUFF(			
				(SELECT '', '' + GIH1.VoucherNo
					FROM ' + @TableNameH + ' GIH1
					WHERE 1= 
						CASE WHEN (''' + @InvoiceIds + '''='''') THEN 1 ELSE 
						(CASE WHEN (CAST(GIH1.Id AS NVARCHAR) IN(SELECT * FROM Split(''' + @InvoiceIds + ''','',''))) THEN 1 ELSE 0 END) END
					ORDER BY CAST(VoucherNo AS NUMERIC)
				FOR XML PATH('''')), 1, 2, '''')
		,TypeDetails = STUFF(			
			(SELECT   '', '' + (TDM.TypeDetail + ''-'' + RTRIM(LTRIM(STR(SUM(SL1.Qty))))) 
			FROM ' + @TableNameL + '_TypeDetailMaster SL1
				LEFT JOIN TypeDetailMaster TDM ON TDM.Id=SL1.F_TypeDetailMaster
			WHERE 1= 
					CASE WHEN (''' + @InvoiceIds + '''='''') THEN 1 ELSE 
					(CASE WHEN (CAST(SL1.F_' + @TableNameH + ' AS NVARCHAR) IN(SELECT * FROM Split(''' + @InvoiceIds + ''','',''))) THEN 1 ELSE 0 END) END					 
				AND SL1.Qty>0
				AND SL1.F_ItemMaster=TAB1.F_ItemMaster
				AND SL1.TmpTypeId IN (SELECT TmpTypeId FROM ' + @TableNameL + ' SL2 WHERE SL2.F_' + @TableNameH + ' IN(SELECT * FROM Split(''' + @InvoiceIds + ''','','')) AND SL2.MRP=TAB1.MRP)
				
			GROUP BY TDM.TypeDetail,TDM.Id
			ORDER BY TDM.Id 
			FOR XML PATH('''')), 1, 2, '''')
	FROM
	(
		SELECT IM.Name AS Item,GIL.MRP,SUM(GIL.MinQty) AS Qty,0 AS Free,GIL.F_ItemMaster,GIL.TmpTypeId,SUM(GIL.TaxAmt) AS TaxAmt,SUM(GIL.TaxableAmt) AS FinalAmt
		FROM [dbo].[' + @TableNameL + '] GIL
			LEFT JOIN ' + @TableNameH + ' GIH ON GIH.Id=GIL.F_' + @TableNameH + '		
			LEFT JOIN ItemMaster IM ON IM.Id = GIL.F_ItemMaster
		WHERE 1= CASE WHEN (''' + @InvoiceIds + '''='''') THEN 1 ELSE (CASE WHEN (CAST(GIH.Id AS NVARCHAR) IN(SELECT * FROM Split(''' + @InvoiceIds + ''','',''))) THEN 1 ELSE 0 END) END
			 ' + @ESTIMATE_WHERE + '
		GROUP BY IM.Name,GIL.MRP,GIL.F_ItemMaster,GIL.TmpTypeId
		
		UNION
		
		SELECT IMF.Name AS Item,GIL.MRP
			,0 AS Qty,SUM(GIL.MinQtyF) AS Free,GIL.F_ItemMasterF AS F_ItemMaster,GIL.TmpTypeId,SUM(GIL.TaxAmt) AS TaxAmt,SUM(GIL.TaxableAmt) AS FinalAmt
		FROM [dbo].[' + @TableNameL + '] GIL
			LEFT JOIN ' + @TableNameH + ' GIH ON GIH.Id=GIL.F_' + @TableNameH + '		
			LEFT JOIN ItemMaster IMF ON IMF.Id = GIL.F_ItemMasterF
		WHERE 1= CASE WHEN (''' + @InvoiceIds + '''='''') THEN 1 ELSE (CASE WHEN (CAST(GIH.Id AS NVARCHAR) IN(SELECT * FROM Split(''' + @InvoiceIds + ''','',''))) THEN 1 ELSE 0 END) END
			AND GIL.F_ItemMasterF>0 ' + @ESTIMATE_WHERE + '
		GROUP BY IMF.Name,GIL.MRP,GIL.F_ItemMasterF,GIL.TmpTypeId
	) AS TAB1
		GROUP BY TAB1.Item,TAB1.MRP,TAB1.F_ItemMaster
	ORDER BY TAB1.Item,TAB1.MRP DESC')
END







































GO
```

# [dbo].[Select_MoneyReceiptBankL]

```sql
/****** Object:  StoredProcedure [dbo].[Select_MoneyReceiptBankL]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



























CREATE PROCEDURE [dbo].[Select_MoneyReceiptBankL] 
	@F_MoneyReceiptBankH numeric(18,0)=0,
	@OrderBy nvarchar(50)='GIL.Id',
	@F_SaleInvoiceH numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@F_LedgerMaster NUMERIC(18,0)=0,
	@ChqCancel smallint=0,
	@IsDebitNote smallint=0
AS
BEGIN
	DECLARE @SQL nvarchar(max)
	DECLARE @WHERE NVARCHAR(MAX)
	IF(@F_SaleInvoiceH>0)
		BEGIN
			IF(@IsDebitNote=1)
				SET @WHERE = ' WHERE GIL.F_SaleInvoiceH = ' + CAST(@F_SaleInvoiceH AS NVARCHAR) + ' AND ISNULL(GIL.IsDebitNote,0)=' + CAST(@IsDebitNote AS NVARCHAR)
			ELSE
				SET @WHERE = ' WHERE GIL.F_SaleInvoiceH = ' + CAST(@F_SaleInvoiceH AS NVARCHAR) + ' AND ISNULL(GIH.IsChequeCancel,0)=' + CAST(@ChqCancel AS NVARCHAR)
		END
	ELSE IF(@F_MoneyReceiptBankH>0)
		BEGIN
			IF(@ChqCancel>0)
				SET @WHERE = ' WHERE GIL.F_MoneyReceiptBankH = ' + CAST(@F_MoneyReceiptBankH AS NVARCHAR) + ' AND ISNULL(GIH.IsChequeCancel,0)=' + CAST(@ChqCancel AS NVARCHAR)
			ELSE
				SET @WHERE = ' WHERE GIL.F_MoneyReceiptBankH = ' + CAST(@F_MoneyReceiptBankH AS NVARCHAR)
		END
	ELSE IF(@F_LedgerMaster>0)
		BEGIN
			SET @WHERE = ' WHERE SH.VoucherDate < ' + @FromDate + ' AND SH.F_LedgerMaster = ' + CAST(@F_LedgerMaster AS NVARCHAR) + ' AND ISNULL(GIH.IsChequeCancel,0)=' + CAST(@ChqCancel AS NVARCHAR)
		END

	SET @SQL = 'SELECT 
					(CASE WHEN(ISNULL(GIL.IsDebitNote,0)=0) THEN
						((CASE WHEN(ISNULL(SH.VoucherNoPrefix,'''')<>'''') THEN SH.VoucherNoPrefix + '' '' ELSE '''' END)+SH.VoucherNo)
					ELSE
						((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'''')<>'''') THEN VH.VoucherNoPrefix + '' '' ELSE '''' END)+VH.VoucherNo)
					END) AS VoucherNo
					,
					(CASE WHEN(ISNULL(GIL.IsDebitNote,0)=0) THEN
						(SUBSTRING(SH.VoucherDate,7,2)+''/''+SUBSTRING(SH.VoucherDate,5,2)+''/''+SUBSTRING(SH.VoucherDate,1,4))
					ELSE
						(SUBSTRING(VH.VoucherDate,7,2)+''/''+SUBSTRING(VH.VoucherDate,5,2)+''/''+SUBSTRING(VH.VoucherDate,1,4))
					END) AS VoucherDate
					,GIL.Amount
					,GIL.F_SaleInvoiceH
					,GIL.Id,GIL.F_MoneyReceiptBankH,GIH.VoucherNo AS ReceiptNo,ISNULL(GIL.IsDebitNote,0) AS IsDebitNote
				FROM [dbo].[MoneyReceiptBankL] GIL
					LEFT JOIN MoneyReceiptBankH GIH ON GIH.Id=GIL.F_MoneyReceiptBankH
					LEFT JOIN SaleInvoiceH SH ON SH.Id=GIL.F_SaleInvoiceH
					LEFT JOIN VoucherH VH ON VH.Id = GIL.F_SaleInvoiceH AND ISNULL(GIL.IsDebitNote,0) =1
				' + @WHERE + 
				' ORDER BY ' + @OrderBy
	EXEC (@SQL)
END























GO
```

# [dbo].[Select_MoneyReceiptCashL]

```sql
/****** Object:  StoredProcedure [dbo].[Select_MoneyReceiptCashL]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO



























CREATE PROCEDURE [dbo].[Select_MoneyReceiptCashL] 
	@F_MoneyReceiptCashH numeric(18,0)=0,
	@OrderBy nvarchar(50)='GIL.Id',
	@F_SaleInvoiceH numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@F_LedgerMaster NUMERIC(18,0)=0,
	@IsDebitNote smallint=0
AS
BEGIN
	DECLARE @SQL nvarchar(max)
	DECLARE @WHERE NVARCHAR(MAX)
	IF(@F_SaleInvoiceH>0)
		BEGIN
			IF(@IsDebitNote=1)
				SET @WHERE = ' WHERE GIL.F_SaleInvoiceH = ' + CAST(@F_SaleInvoiceH AS NVARCHAR) + ' AND ISNULL(GIL.IsDebitNote,0)=' + CAST(@IsDebitNote AS NVARCHAR)
			ELSE
				SET @WHERE = ' WHERE GIL.F_SaleInvoiceH = ' + CAST(@F_SaleInvoiceH AS NVARCHAR)
		END
	ELSE IF(@F_MoneyReceiptCashH>0)
		BEGIN
			SET @WHERE = ' WHERE GIL.F_MoneyReceiptCashH = ' + CAST(@F_MoneyReceiptCashH AS NVARCHAR)
		END
	ELSE IF(@F_LedgerMaster>0)
		BEGIN
			SET @WHERE = ' WHERE SH.VoucherDate < ' + @FromDate + ' AND SH.F_LedgerMaster = ' + CAST(@F_LedgerMaster AS NVARCHAR)
		END

	SET @SQL = 'SELECT 
					CASE WHEN(ISNULL(GIL.IsDebitNote,0)=0) THEN
						((CASE WHEN(ISNULL(SH.VoucherNoPrefix,'''')<>'''') THEN SH.VoucherNoPrefix + '' '' ELSE '''' END)+SH.VoucherNo)
					ELSE
						((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'''')<>'''') THEN VH.VoucherNoPrefix + '' '' ELSE '''' END)+VH.VoucherNo)
					END
					AS VoucherNo
					,
					CASE WHEN(ISNULL(GIL.IsDebitNote,0)=0) THEN
						(SUBSTRING(SH.VoucherDate,7,2)+''/''+SUBSTRING(SH.VoucherDate,5,2)+''/''+SUBSTRING(SH.VoucherDate,1,4))
					ELSE
						(SUBSTRING(VH.VoucherDate,7,2)+''/''+SUBSTRING(VH.VoucherDate,5,2)+''/''+SUBSTRING(VH.VoucherDate,1,4))
					END
					AS VoucherDate
					,GIL.Amount
					,GIL.F_SaleInvoiceH
					,GIL.Id,GIL.F_MoneyReceiptCashH,ISNULL(GIL.IsDebitNote,0) AS IsDebitNote
				FROM MoneyReceiptCashL GIL
					LEFT JOIN MoneyReceiptCashH GIH ON GIH.Id=GIL.F_MoneyReceiptCashH
					LEFT JOIN SaleInvoiceH SH ON SH.Id=GIL.F_SaleInvoiceH
					LEFT JOIN VoucherH VH ON VH.Id = GIL.F_SaleInvoiceH
				' + @WHERE + 
				' ORDER BY ' + @OrderBy
	EXEC (@SQL)
END
























GO
```

# [dbo].[Select_PaidAmountBy_MoneyReceipt]

```sql
/****** Object:  StoredProcedure [dbo].[Select_PaidAmountBy_MoneyReceipt]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO













CREATE PROCEDURE [dbo].[Select_PaidAmountBy_MoneyReceipt] 
	@TableName nvarchar(50)='MoneyReceiptCashH',
	@F_SaleInvoiceH numeric(18,0)
AS
BEGIN
	SELECT (
			ISNULL(
				(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptCashL MRL LEFT JOIN MoneyReceiptCashH MRH ON MRH.Id=MRL.F_MoneyReceiptCashH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)
			,0)
			+
			ISNULL(
				(SELECT SUM(MRL.Amount)
							FROM MoneyReceiptBankL MRL LEFT JOIN MoneyReceiptBankH MRH ON MRH.Id=MRL.F_MoneyReceiptBankH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0 AND ISNULL(MRH.IsChequeCancel,0)=0)
			,0)
			+
			ISNULL(
				(SELECT SUM(VH1.TotalNetAmt)
							FROM SaleInvoiceH_CreditNote SIC LEFT JOIN VoucherH VH1 ON VH1.Id=SIC.F_VoucherH
							WHERE SIC.F_SaleInvoiceH = VH.Id AND ISNULL(VH1.Cancelled,0)=0)
			,0)
			+
			ISNULL(
				(SELECT SUM(MRL.Amount)
							FROM SaleReturnL_InvoiceDetails MRL LEFT JOIN SaleReturnH MRH ON MRH.Id=MRL.F_SaleReturnH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)
			,0)
			+
			ISNULL(
				(SELECT SUM(MRL.Amount)
							FROM VoucherL_BillDetails MRL LEFT JOIN VoucherH VH1 ON VH1.Id=MRL.F_VoucherH
							WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(VH1.Cancelled,0)=0)
			,0)
			) AS Paid
		,VH.VoucherDate,VH.TotalNetAmt
		,(VH.TotalNetAmt-(
			ISNULL(
				(SELECT SUM(MRL.Amount)
						FROM MoneyReceiptCashL MRL LEFT JOIN MoneyReceiptCashH MRH ON MRH.Id=MRL.F_MoneyReceiptCashH
						WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)
			,0)
			+
			ISNULL(
				(SELECT SUM(MRL.Amount)
						FROM MoneyReceiptBankL MRL LEFT JOIN MoneyReceiptBankH MRH ON MRH.Id=MRL.F_MoneyReceiptBankH
						WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0 AND ISNULL(MRH.IsChequeCancel,0)=0)
			,0)
			+
			ISNULL(
				(SELECT SUM(VH1.TotalNetAmt)
						FROM SaleInvoiceH_CreditNote SIC LEFT JOIN VoucherH VH1 ON VH1.Id=SIC.F_VoucherH
						WHERE SIC.F_SaleInvoiceH = VH.Id AND ISNULL(VH1.Cancelled,0)=0)
			,0)
			+
			ISNULL(
				(SELECT SUM(MRL.Amount)
						FROM SaleReturnL_InvoiceDetails MRL LEFT JOIN SaleReturnH MRH ON MRH.Id=MRL.F_SaleReturnH
						WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(MRH.Cancelled,0)=0)
			,0)
			+
			ISNULL(
				(SELECT SUM(MRL.Amount)
						FROM VoucherL_BillDetails MRL LEFT JOIN VoucherH VH1 ON VH1.Id=MRL.F_VoucherH
						WHERE MRL.F_SaleInvoiceH = VH.Id AND ISNULL(VH1.Cancelled,0)=0)
			,0)
			)) AS DueAmt,
			(
				LM.Name
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')
			) AS Ledger
	FROM SaleInvoiceH VH
	LEFT JOIN LedgerMaster LM ON LM.Id = VH.F_LedgerMaster
	LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=LM.F_CityMaster
	LEFT JOIN StateMaster SM ON SM.Id=CM.F_StateMaster
	WHERE VH.Id =@F_SaleInvoiceH
END























GO
```

# [dbo].[Select_PendingEstimates]

```sql
/****** Object:  StoredProcedure [dbo].[Select_PendingEstimates]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE PROCEDURE [dbo].[Select_PendingEstimates] 	
	@FromDate NVARCHAR(50),
	@ToDate NVARCHAR(50)
AS
BEGIN	
	SELECT EH.Id,
		EH.VoucherNo,
		(SUBSTRING(EH.VoucherDate,7,2)+'/'+SUBSTRING(EH.VoucherDate,5,2)+'/'+SUBSTRING(EH.VoucherDate,1,4)) AS VoucherDate,
		((CASE WHEN(LM.Name<>'') THEN LM.Name ELSE EH.LedgerName END)
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')
				+ ISNULL((CASE WHEN(ISNULL(OLM.Name,'') <> '') THEN ' (' 
					+ CAST(OLM.Name AS NVARCHAR) 
						+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address1 AS NVARCHAR) END),'')
						+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address2 AS NVARCHAR) END),'')
						+ ISNULL((CASE WHEN(ISNULL(OCM.Name,'') <> '') THEN ', ' + OCM.Name END),'')
					+ ')' END),'')
				) AS Ledger,
		(CASE WHEN (EH.IsVATIncluded=0) THEN '' ELSE 'Yes' END) AS VATIncluded,
		EH.TotalNetAmt,
		LM.Id AS LedgerId
	FROM EstimateH EH
	LEFT JOIN LedgerMaster LM ON LM.Id=EH.F_LedgerMaster
	LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=LM.F_CityMaster
	LEFT JOIN StateMaster SM ON SM.Id=CM.F_StateMaster
	
	LEFT JOIN OtherLedgerMaster OLM ON OLM.Id = EH.F_OtherLedgerMaster
	LEFT JOIN CityMaster OCM ON OCM.Id=OLM.F_CityMaster
	WHERE EH.VoucherDate >= @FromDate AND EH.VoucherDate <= @ToDate
	ORDER BY CAST(EH.VoucherNo AS NUMERIC)
END



















































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_Purchase_ImportH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO










CREATE PROCEDURE [dbo].[Select_Purchase_ImportH]
	@InvoiceNo nvarchar(50)='',
	@InvoiceDate nvarchar(50)='',
	@IsIncRate SMALLINT=0,
	@LotNo numeric(18,2)=0
AS
BEGIN
	DECLARE @SQL NVARCHAR(MAX)
	DECLARE @FIELDS NVARCHAR(MAX)
	DECLARE @FIELDS2 NVARCHAR(MAX)
	DECLARE @WHERE NVARCHAR(MAX)
	DECLARE @LEFT_JOIN NVARCHAR(MAX)
	DECLARE @ORDER_BY NVARCHAR(MAX)
	DECLARE @GROUP_BY NVARCHAR(MAX)
	IF(@LotNo>0)
		BEGIN
			SET @LEFT_JOIN='
				OUTER APPLY(SELECT SUM(GIL.Qty_CS) AS LotNo
					FROM Transport_ImportH GIL
					WHERE GIL.InvoiceNo = VL.InvoiceNo
					GROUP BY InvoiceNo,InvoiceDate
				) AS LotDetails
				OUTER APPLY(SELECT SUM(GIL.Qty) AS Rec_Qty
					FROM TransportStockInL GIL
					WHERE GIL.PurVoucherNo = VL.InvoiceNo AND GIL.F_ItemMaster = IM.Id
				) AS RecDetails'
			SET @FIELDS=',MAX(LotDetails.LotNo) AS LotNo'
			SET @FIELDS2=',(SUM(VL.[Qty_CS])-MAX(ISNULL(RecDetails.Rec_Qty,0))) AS BalQty,'''' AS Rec_Qty'
			SET @WHERE='WHERE LotDetails.LotNo=' + CAST(@LotNo AS NVARCHAR)
			IF(@InvoiceNo<>'')
				BEGIN
					SET @WHERE = @WHERE + ' AND VL.InvoiceNo=''' + @InvoiceNo + ''''
				END
			IF(@InvoiceDate<>'')
				BEGIN
					SET @WHERE = @WHERE + ' AND VL.InvoiceDate=''' + @InvoiceDate + ''''
				END
			SET @GROUP_BY=' GROUP BY VL.InvoiceNo,VL.InvoiceDate,IM.Id,IM.Name,TIMM.MRP'
			SET @ORDER_BY=' ORDER BY IM.Name,TIMM.MRP'
		END
	ELSE
		BEGIN
			SET @LEFT_JOIN=''
			SET @FIELDS=''
			SET @FIELDS2=''
			SET @WHERE='WHERE VL.InvoiceNo=''' + @InvoiceNo + ''''
			IF(@InvoiceDate<>'')
				BEGIN
					SET @WHERE = @WHERE + ' AND VL.InvoiceDate=''' + @InvoiceDate + ''''
				END
			SET @GROUP_BY=' GROUP BY VL.InvoiceNo,VL.InvoiceDate,IM.Id,IM.Name,TIMM.MRP,(CASE WHEN(VL.TaxableAmt=0) THEN VL.TaxableAmt ELSE NULL END)
			,LastRate.Rate,LastRate.IncRate,VL.DiscRate'
			SET @ORDER_BY=' ORDER BY IM.Name,TIMM.MRP,CASE WHEN(VL.TaxableAmt=0) THEN VL.TaxableAmt ELSE NULL END'
		END
	SET @SQL='
	SELECT VL.[InvoiceNo],
		(SUBSTRING(VL.InvoiceDate,7,2)+''/''+SUBSTRING(VL.InvoiceDate,5,2)+''/''+SUBSTRING(VL.InvoiceDate,1,4)) AS InvoiceDate
		,''Saleable'' AS Saleable
		,IM.Name AS Map_Item,TIMM.MRP
		,(CASE WHEN(' + CAST(@IsIncRate AS NVARCHAR) + '=0) THEN MAX(LastRate.Rate) ELSE MAX(LastRate.IncRate) END) AS Rate
		' + @FIELDS + '
		,SUM(VL.[Qty_CS]) AS Qty_CS' + @FIELDS2 + ',SUM(VL.[Qty_Pcs]) AS Qty_Pcs
		,MAX(VL.DiscRate) AS DiscRate,SUM(VL.DiscAmt) AS DiscAmt
		,SUM(VL.TaxableAmt) AS TaxableAmt
		,SUM(VL.NetAmt) AS NetAmt,SUM(VL.TotalTax) AS TotalTax,MAX(VL.CGSTRate) AS CGSTRate,SUM(VL.CGSTAmt) AS CGSTAmt,MAX(VL.SGSTRate) AS SGSTRate,SUM(VL.SGSTAmt) AS SGSTAmt
		,IM.Id AS F_ItemMaster
	FROM [dbo].[Transport_ImportH] VL
		LEFT JOIN TransportItemMapMaster TIMM ON TIMM.TransportItemId = VL.Item_Id
		LEFT JOIN ItemMaster IM ON IM.Id = TIMM.F_ItemMaster
		OUTER APPLY(SELECT TOP 1 GIL.Rate,GIL.IncRate,GIL.ProductDiscRate,GIL.IsManualProDis
			FROM [dbo].[PurchaseInvoiceL] GIL LEFT JOIN PurchaseInvoiceH GIH ON GIH.Id=GIL.F_PurchaseInvoiceH
			WHERE GIL.F_ItemMaster = IM.Id AND GIL.MRP=TIMM.MRP AND GIH.F_LedgerMaster=912
			ORDER BY GIH.VoucherDate DESC
		) AS LastRate' + @LEFT_JOIN + '
	' + @WHERE + @GROUP_BY + @ORDER_BY
	EXEC(@SQL)
END



GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_PurchaseBillNo_DateWise]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO























































CREATE PROCEDURE [dbo].[Select_PurchaseBillNo_DateWise] 
	@F_CompanyMaster numeric(18,0),
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Id numeric(18,0)=0
AS
BEGIN
	SELECT Name,Id,VoucherNo1 FROM (
		SELECT 0 AS RowNumber,'' AS Name,0 AS Id,' [None]' AS VoucherNo1
		
		UNION ALL
		
		SELECT ROW_NUMBER() OVER (order by CAST(VH.VoucherNo AS NUMERIC) ASC) as RowNumber,
		(VH.PurVoucherNo + ' ' + (SUBSTRING(VH.PurVoucherDate,7,2)+'/'+SUBSTRING(VH.PurVoucherDate,5,2)+'/'+SUBSTRING(VH.PurVoucherDate,1,4)) + ' ' + LM.Name) AS Name
		,VH.Id,VH.PurVoucherNo AS VoucherNo1
			FROM PurchaseInvoiceH VH
				LEFT JOIN LedgerMaster LM ON VH.F_LedgerMaster=LM.Id
			WHERE VH.F_CompanyMaster =@F_CompanyMaster
				AND 1=
					CASE WHEN(@FromDate='') THEN 1 ELSE
						(CASE WHEN(VH.PurVoucherDate >= @FromDate AND VH.PurVoucherDate <=@ToDate) THEN 1 ELSE 0 END)
					END		
		) AS Tab1
	ORDER BY RowNumber,Name,Id
END





















































GO
```

# [dbo].[Select_PurchaseL]

```sql
/****** Object:  StoredProcedure [dbo].[Select_PurchaseL]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
































CREATE PROCEDURE [dbo].[Select_PurchaseL] 	
	@F_ItemMaster numeric(18,0),	
	@FromDate nvarchar(50),	
	@ToDate nvarchar(50),
	@MRP numeric(18,2),
	@Auto smallint=1,
	@ClStockDate nvarchar(50)=''
AS
BEGIN
	IF(@Auto=1)
		BEGIN
			SELECT IM.LandingCost, GIL.Rate,GIL.MinQty,GIH.PurVoucherNo AS VoucherNo,GIH.PurVoucherDate AS VoucherDate
			FROM PurchaseInvoiceL GIL
				LEFT JOIN PurchaseInvoiceH GIH ON GIH.Id=GIL.F_PurchaseInvoiceH
				LEFT JOIN ItemMaster IM ON IM.Id=GIL.F_ItemMaster
			WHERE GIH.VoucherDate>=@FromDate AND GIH.VoucherDate<=@ToDate AND GIL.F_ItemMaster = @F_ItemMaster
				AND GIL.MRP=@MRP
			ORDER BY GIH.VoucherDate
		END
	ELSE IF(@Auto=2)
		BEGIN
			SELECT IM.LandingCost, GIL.Rate,GIL.MinQty,GIH.PurVoucherNo AS VoucherNo,GIH.PurVoucherDate AS VoucherDate,LM.Name AS Ledger,LM.TINNo
			FROM PurchaseInvoiceL GIL
				LEFT JOIN PurchaseInvoiceH GIH ON GIH.Id=GIL.F_PurchaseInvoiceH
				LEFT JOIN ItemMaster IM ON IM.Id=GIL.F_ItemMaster
				LEFT JOIN LedgerMaster LM ON LM.Id = GIH.F_LedgerMaster
			WHERE GIL.F_ItemMaster = @F_ItemMaster AND GIL.MRP=@MRP AND GIH.VoucherDate<=@ClStockDate
			ORDER BY GIH.VoucherDate DESC
		END
	ELSE
		BEGIN
			SELECT TOP 1 GIL.LandingCost,GIL.LandingCost AS Rate,GIL.Qty AS MinQty,CAST(ROUND(GIL.LandingCost*GIL.Qty,2) AS NUMERIC(18,2)) AS TaxableAmt
			FROM StockInL GIL
				LEFT JOIN StockInH GIH ON GIH.Id=GIL.F_StockInH
				LEFT JOIN ItemMaster IM ON IM.Id=GIL.F_ItemMaster
			WHERE GIH.VoucherDate>=@FromDate AND GIH.VoucherDate<=@ToDate AND GIL.F_ItemMaster = @F_ItemMaster AND GIL.MRP=@MRP
			ORDER BY GIH.VoucherDate DESC
		END
	--
END


































GO
```

# [dbo].[Select_RetailerFootPathL]

```sql
/****** Object:  StoredProcedure [dbo].[Select_RetailerFootPathL]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
























































CREATE PROCEDURE [dbo].[Select_RetailerFootPathL] 
	@F_RetailerFootPathH numeric(18,0)
AS
BEGIN
	SELECT 
		(CASE WHEN(GIL.F_LedgerMaster>0) THEN (IM.Name	
					+ ISNULL((CASE WHEN(ISNULL(CAST(IM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(IM.Address1 AS NVARCHAR) END),'')
					+ ISNULL((CASE WHEN(ISNULL(CAST(IM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(IM.Address2 AS NVARCHAR) END),'')
					+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),''))					
		
			ELSE (OIM.Name 
					+ ISNULL((CASE WHEN(ISNULL(CAST(OIM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OIM.Address1 AS NVARCHAR) END),'')
					+ ISNULL((CASE WHEN(ISNULL(CAST(OIM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OIM.Address2 AS NVARCHAR) END),'')
					+ ISNULL((CASE WHEN(ISNULL(OCM.Name,'') <> '') THEN ', ' + OCM.Name END),''))
		END)
		 AS Ledger
		,GIL.F_LedgerMaster,GIL.F_OtherLedgerMaster,
		GIL.Remark,
		--(CASE WHEN(GIL.F_LedgerMaster=0) THEN OIM.Remark ELSE IM.Remark END) AS 
		GIL.Brands, GIL.BrandIds,
		(CASE WHEN(GIL.F_LedgerMaster>0) THEN '' ELSE 'Other' END) AS Other,GIH.AreaGroupName
	FROM [dbo].[RetailerFootPathL] GIL
		LEFT JOIN RetailerFootPathH GIH ON GIH.Id=GIL.F_RetailerFootPathH
		
		LEFT JOIN LedgerMaster IM ON IM.Id = GIL.F_LedgerMaster
		LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=IM.F_CityMaster
		
		LEFT JOIN OtherLedgerMaster OIM ON OIM.Id = GIL.F_OtherLedgerMaster		
		LEFT JOIN [dbo].[CityMaster] OCM ON OCM.Id=OIM.F_CityMaster
		
	WHERE GIL.F_RetailerFootPathH = @F_RetailerFootPathH
	ORDER BY GIL.Id
END




















































GO
```

# [dbo].[Select_rptInvoice]

```sql
/****** Object:  StoredProcedure [dbo].[Select_rptInvoice]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO























CREATE PROCEDURE [dbo].[Select_rptInvoice] 
	@Id numeric(18,0),
	@TableNameH nvarchar(50),--SaleInvoiceH
	@TableNameL nvarchar(50) --SaleInvoiceL
AS
BEGIN
	DECLARE @TINNo NVARCHAR(MAX)
	SET @TINNo = '(CASE WHEN(ISNULL(LM.TINNo3,'''') <>'''') THEN (CASE WHEN(SH.VoucherDate>=ISNULL(LM.TINNoFrom3,'''')) THEN ISNULL(LM.TINNo3,'''') ELSE
												(CASE WHEN(SH.VoucherDate>=ISNULL(LM.TINNoFrom2,'''')) THEN ISNULL(LM.TINNo2,'''') ELSE ISNULL(LM.TINNo,'''') END) END)
					ELSE (CASE WHEN(ISNULL(LM.TINNo2,'''') <>'''') THEN (CASE WHEN(SH.VoucherDate>=ISNULL(LM.TINNoFrom2,'''')) THEN ISNULL(LM.TINNo2,'''') ELSE ISNULL(LM.TINNo,'''') END)
										ELSE ISNULL(LM.TINNo,'''') END)
									 END) AS TINNo,'
	

	EXEC('SELECT 
		ROW_NUMBER() OVER (ORDER BY CAST(SL.Id AS NUMERIC) DESC) AS SNo,
		((CASE WHEN(ISNULL(SH.VoucherNoPrefix,'''')<>'''') THEN SH.VoucherNoPrefix + '' '' ELSE '''' END)+ SH.VoucherNo) AS VoucherNo,
		(SUBSTRING(SH.VoucherDate,7,2)+''/''+SUBSTRING(SH.VoucherDate,5,2)+''/''+SUBSTRING(SH.VoucherDate,1,4)) AS VoucherDate,
		(CASE WHEN(ISNULL(OLM.Name,'''')='''') THEN LM.Name
			ELSE OLM.Name END)
		 AS Ledger,
		(CASE WHEN(ISNULL(OLM.Name,'''')='''') THEN LM.Address1
			ELSE OLM.Address1 END)
		AS Address1,
		' + @TINNo + '
		CM.Name AS City,
		IM.Name AS Item,SL.Remark,
		TypeDetails = STUFF(			
			(SELECT   '', '' + (TDM.TypeDetail + ''-'' + RTRIM(LTRIM(STR(SL1.Qty)))) 
			FROM ' + @TableNameL + '_TypeDetailMaster SL1
				LEFT JOIN TypeDetailMaster TDM ON TDM.Id=SL1.F_TypeDetailMaster
			WHERE SL1.F_' + @TableNameH + ' = SH.Id AND SL1.Qty>0 AND SL1.F_ItemMaster=SL.F_ItemMaster
				AND SL1.TmpTypeId=SL.TmpTypeId
			ORDER BY TDM.Id 
			FOR XML PATH('''')), 1, 2, ''''),
		SL.TaxRate,SL.TaxRate2,SL.TaxRate3,
		SL.MRP,
		SL.Qty,
		UMQ.Name AS UnitQ,
		SL.Free,IMF.Name AS ItemF,
		(CASE WHEN (SL.Free>0) THEN substring(UMF.Name,1,3) ELSE '''' END)
		 AS UnitF,
		SL.ProductDiscAmt,
		SL.Rate,
		SL.QtyxRate,
		SL.Amount,
		SH.TotalProDisAmt,
		SH.TotalDisplayAmt,
		SH.TotalCashDisAmt,
		SH.TotalReplacementAmt,
		SH.TotalGrossAmt,
		SH.TotalTaxationAmt,
		SH.TotalLessSpecialAmt,
		SH.TotalNetAmt,SH.TotalQtyxRate,
		Taxs = STUFF(
			(SELECT   CHAR(10) + (LM.Name) 
			FROM ' + @TableNameL + 'Tax SLTax
				LEFT JOIN LedgerMaster LM ON LM.Id=SLTax.F_LedgerMaster
			WHERE SLTax.F_' + @TableNameH + ' = SH.Id
			ORDER BY SLTax.Id
			FOR XML PATH('''')), 1, 1, ''''),
		AddAmts = STUFF(
			(SELECT   CHAR(10) + CAST(SLTax.AddAmt AS NVARCHAR) 
			FROM ' + @TableNameL + 'Tax SLTax
				LEFT JOIN LedgerMaster LM ON LM.Id=SLTax.F_LedgerMaster
			WHERE SLTax.F_' + @TableNameH + ' = SH.Id
			ORDER BY SLTax.Id
			FOR XML PATH('''')), 1, 1, ''''),
		TaxableAmts = STUFF(
			(SELECT   CHAR(10) + CAST(SLTax.TaxableAmt AS NVARCHAR)
			FROM ' + @TableNameL + 'Tax SLTax
				LEFT JOIN LedgerMaster LM ON LM.Id=SLTax.F_LedgerMaster
			WHERE SLTax.F_' + @TableNameH + ' = SH.Id
			ORDER BY SLTax.Id
			FOR XML PATH('''')), 1, 1, '''')
	FROM ' + @TableNameH + ' SH
	LEFT JOIN ' + @TableNameL + ' SL ON SH.Id=SL.F_' + @TableNameH + '
	LEFT JOIN LedgerMaster LM ON LM.Id=SH.F_LedgerMaster
	LEFT JOIN OtherLedgerMaster OLM ON OLM.Id=SH.F_OtherLedgerMaster
	LEFT JOIN CityMaster CM ON CM.Id=LM.F_CityMaster
	LEFT JOIN ItemMaster IM ON IM.Id=SL.F_ItemMaster
	LEFT JOIN ItemMaster IMF ON IMF.Id = SL.F_ItemMasterF
	LEFT JOIN UnitMaster UMQ ON UMQ.Id=SL.F_UnitMasterQ
	LEFT JOIN UnitMaster UMF ON UMF.Id=SL.F_UnitMasterF
	WHERE SH.Id=' + @Id + '
	ORDER BY SH.VoucherDate,SH.VoucherNo,SL.Id')
END




















































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_rptInvoice_GST]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO









 

CREATE PROCEDURE [dbo].[Select_rptInvoice_GST] 
	@Id numeric(18,0),
	@TableNameH nvarchar(50),--SaleInvoiceH
	@TableNameL nvarchar(50) --SaleInvoiceL
AS
BEGIN
	DECLARE @TINNo NVARCHAR(MAX)
	SET @TINNo = '(CASE WHEN(ISNULL(LM.TINNo3,'''') <>'''') THEN (CASE WHEN(SH.VoucherDate>=ISNULL(LM.TINNoFrom3,'''')) THEN ISNULL(LM.TINNo3,'''') ELSE
												(CASE WHEN(SH.VoucherDate>=ISNULL(LM.TINNoFrom2,'''')) THEN ISNULL(LM.TINNo2,'''') ELSE ISNULL(LM.TINNo,'''') END) END)
					ELSE (CASE WHEN(ISNULL(LM.TINNo2,'''') <>'''') THEN (CASE WHEN(SH.VoucherDate>=ISNULL(LM.TINNoFrom2,'''')) THEN ISNULL(LM.TINNo2,'''') ELSE ISNULL(LM.TINNo,'''') END)
										ELSE ISNULL(LM.TINNo,'''') END)
									 END) AS TINNo,'
	
	DECLARE @GSTNo NVARCHAR(MAX)
	SET @GSTNo = '(CASE WHEN(SH.VoucherDate>=LM.GSTRegDate AND (ISNULL(LM.IsStopGST,0)=0 OR (ISNULL(LM.IsStopGST,0)=1 AND SH.VoucherDate<=LM.GSTStopDate))) THEN LM.GSTNo ELSE '''' END) AS TINNo,'

	EXEC('SELECT 
		ROW_NUMBER() OVER (ORDER BY CAST(SL.Id AS NUMERIC) DESC) AS SNo,
		((CASE WHEN(ISNULL(SH.VoucherNoPrefix,'''')<>'''') THEN SH.VoucherNoPrefix + '' '' ELSE '''' END)+ SH.VoucherNo) AS VoucherNo,
		(SUBSTRING(SH.VoucherDate,7,2)+''/''+SUBSTRING(SH.VoucherDate,5,2)+''/''+SUBSTRING(SH.VoucherDate,1,4)) AS VoucherDate,
		(CASE WHEN(ISNULL(OLM.Name,'''')='''') THEN LM.Name
			ELSE OLM.Name END)
		 AS Ledger,
		(CASE WHEN(ISNULL(OLM.Name,'''')='''') THEN LM.Address1
			ELSE OLM.Address1 END)
		AS Address1,
		' + @GSTNo + '
		CM.Name AS City,
		(IM.Name + ''('' + ISNULL(HSN.Name,'''') + '')'') AS Item,
		(IM.Name) AS Item1,
		SL.Remark,
		TypeDetails = STUFF(
			(SELECT   '', '' + (TDM.TypeDetail + ''-'' + RTRIM(LTRIM(STR(SL1.Qty)))) 
			FROM ' + @TableNameL + '_TypeDetailMaster SL1
				LEFT JOIN TypeDetailMaster TDM ON TDM.Id=SL1.F_TypeDetailMaster
			WHERE SL1.F_' + @TableNameH + ' = SH.Id AND SL1.Qty>0 AND SL1.F_ItemMaster=SL.F_ItemMaster
				AND SL1.TmpTypeId=SL.TmpTypeId
			ORDER BY TDM.Id 
			FOR XML PATH('''')), 1, 2, ''''),
		SL.TaxRate,SL.TaxRate2,SL.TaxRate3,
		SL.MRP,
		SL.Qty,
		UMQ.Name AS UnitQ,
		SL.Free,
		(IMF.Name + ''('' + ISNULL(HSNF.Name,'''') + '')'') AS ItemF,
		(CASE WHEN (SL.Free>0) THEN substring(UMF.Name,1,3) ELSE '''' END)
		 AS UnitF,
		SL.ProductDiscAmt,
		SL.CashDiscAmt,
		SL.Rate,
		SL.QtyxRate,
		SL.Amount,
		SL.FinalAmt,
		SH.TotalProDisAmt,
		SH.TotalDisplayAmt,
		SH.TotalCashDisAmt,
		SH.TotalReplacementAmt,
		SH.TotalGrossAmt,
		SH.TotalTaxationAmt,
		SH.TotalLessSpecialAmt,
		SH.TotalNetAmt,SH.TotalQtyxRate,
		Taxs = STUFF(
			(SELECT   CHAR(10) + (LM.Name) 
			FROM ' + @TableNameL + 'Tax SLTax
				LEFT JOIN LedgerMaster LM ON LM.Id=SLTax.F_LedgerMaster
			WHERE SLTax.F_' + @TableNameH + ' = SH.Id
			ORDER BY SLTax.Id
			FOR XML PATH('''')), 1, 1, ''''),
		AddAmts = STUFF(
			(SELECT   CHAR(10) + CAST(SLTax.AddAmt AS NVARCHAR) 
			FROM ' + @TableNameL + 'Tax SLTax
				LEFT JOIN LedgerMaster LM ON LM.Id=SLTax.F_LedgerMaster
			WHERE SLTax.F_' + @TableNameH + ' = SH.Id
			ORDER BY SLTax.Id
			FOR XML PATH('''')), 1, 1, ''''),
		TaxableAmts = STUFF(
			(SELECT   CHAR(10) + CAST(SLTax.TaxableAmt AS NVARCHAR)
			FROM ' + @TableNameL + 'Tax SLTax
				LEFT JOIN LedgerMaster LM ON LM.Id=SLTax.F_LedgerMaster
			WHERE SLTax.F_' + @TableNameH + ' = SH.Id
			ORDER BY SLTax.Id
			FOR XML PATH('''')), 1, 1, ''''),
		ISNULL(HSN.Name,'''') AS HSNCode,
		SH.TotalAdditionalTaxAmt,
		(SELECT TOP 1 LM1.Name
			FROM ' + @TableNameH + '_AdditionalTaxation LSTM
			LEFT JOIN LedgerMaster LM1 ON LSTM.F_LedgerMaster_Tax = LM1.Id
			WHERE LSTM.F_' + @TableNameH + '=' + @Id + ') AS LedgerAdditionalTax,
		EAH.IRN,EAH.SignedQRCode,EAH.DocNo,EAH.DocDate AS DocDate8
	FROM ' + @TableNameH + ' SH
	LEFT JOIN ' + @TableNameL + ' SL ON SH.Id=SL.F_' + @TableNameH + '
	LEFT JOIN LedgerMaster LM ON LM.Id=SH.F_LedgerMaster
	LEFT JOIN OtherLedgerMaster OLM ON OLM.Id=SH.F_OtherLedgerMaster
	LEFT JOIN CityMaster CM ON CM.Id=LM.F_CityMaster
	LEFT JOIN ItemMaster IM ON IM.Id=SL.F_ItemMaster
	LEFT JOIN ItemMaster IMF ON IMF.Id = SL.F_ItemMasterF
	LEFT JOIN UnitMaster UMQ ON UMQ.Id=SL.F_UnitMasterQ
	LEFT JOIN UnitMaster UMF ON UMF.Id=SL.F_UnitMasterF
	LEFT JOIN HSNCodeMaster HSN ON HSN.Id = IM.F_HSNCodeMaster
	LEFT JOIN HSNCodeMaster HSNF ON HSNF.Id = IMF.F_HSNCodeMaster

	LEFT JOIN EInvoiceAknowledgementH EAH ON EAH.DocNo = SH.VoucherNo AND EAH.DocDate = SH.VoucherDate
	WHERE SH.Id=' + @Id + '
	ORDER BY SH.VoucherDate,SH.VoucherNo,SL.Id')
END

GO
```

# [dbo].[Select_rptTransportStockIn]

```sql
/****** Object:  StoredProcedure [dbo].[Select_rptTransportStockIn]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO










CREATE PROCEDURE [dbo].[Select_rptTransportStockIn] 
	@Id nvarchar(50)
AS
BEGIN
	SELECT DSH.Id,DSH.VoucherNo,
		(SUBSTRING(DSH.VoucherDate,7,2)+'/'+SUBSTRING(DSH.VoucherDate,5,2)+'/'+SUBSTRING(DSH.VoucherDate,1,4)) AS VoucherDate,
		TM.Name AS Transport,DSH.DriverName,DSH.TotalQty,
		(CASE WHEN (ISNULL(DSH.Cancelled,0)=1) THEN 'Cancelled' ELSE '' END) AS Cancelled,
		IM.Name AS Item,DSL.Remark,UMQ.Name AS UnitQ
		,DSL.Qty,DSL.LotSize,DSL.PurVoucherNo
		,(SUBSTRING(DSL.PurVoucherDate,7,2)+'/'+SUBSTRING(DSL.PurVoucherDate,5,2)+'/'+SUBSTRING(DSL.PurVoucherDate,1,4)) AS PurVoucherDate
		,GM.Name AS Godown
		,DSL.F_ItemMaster,UMQ.Id AS F_UnitMaster,DSL.F_GodownMaster,DSL.Id AS LineId
	FROM TransportStockInH DSH
		LEFT JOIN TransportStockInL DSL ON DSH.Id=DSL.F_TransportStockInH
		LEFT JOIN TransportMaster TM ON TM.Id = DSH.F_TransportMaster
		LEFT JOIN ItemMaster IM ON IM.Id = DSL.F_ItemMaster
		LEFT JOIN UnitMaster UMQ ON UMQ.Id = DSL.F_UnitMaster
		LEFT JOIN GodownMaster GM ON GM.Id = DSL.F_GodownMaster
	WHERE 
		1=(CASE WHEN (@Id='') THEN 1 ELSE (CASE WHEN (DSH.Id IN (SELECT * FROM Split(@Id,','))) THEN 1 ELSE 0 END) END)
	GROUP BY DSH.Id,DSH.VoucherNo,DSH.VoucherDate,TM.Name,DSH.DriverName,DSH.TotalQty,DSH.Cancelled,IM.Name,DSL.Remark,UMQ.Name
		,DSL.Qty,DSL.LotSize,DSL.PurVoucherNo,DSL.PurVoucherDate,GM.Name,DSL.F_ItemMaster,UMQ.Id,DSL.F_GodownMaster,DSL.Id
	ORDER BY DSH.VoucherDate,CAST(DSH.VoucherNo AS NUMERIC),DSH.Id,TM.Name
END




















































GO
```

# [dbo].[Select_SaleBillDetails]

```sql
/****** Object:  StoredProcedure [dbo].[Select_SaleBillDetails]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






































CREATE PROCEDURE [dbo].[Select_SaleBillDetails] 
	@Id numeric(18,0)
AS
BEGIN
	SELECT (SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) AS VoucherDate
		,(LM.Name
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address1 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CAST(LM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(LM.Address2 AS NVARCHAR) END),'')
				+ ISNULL((CASE WHEN(ISNULL(CM.Name,'') <> '') THEN ', ' + CM.Name END),'')
				+ ISNULL((CASE WHEN(ISNULL(OLM.Name,'') <> '') THEN ' ('
					+ CAST(OLM.Name AS NVARCHAR) 
						+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address1 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address1 AS NVARCHAR) END),'')
						+ ISNULL((CASE WHEN(ISNULL(CAST(OLM.Address2 AS NVARCHAR),'') <> '') THEN ', ' + CAST(OLM.Address2 AS NVARCHAR) END),'')
						+ ISNULL((CASE WHEN(ISNULL(OCM.Name,'') <> '') THEN ', ' + OCM.Name END),'')
					+ ')' END),'')
				) AS Ledger
		,VH.TotalNetAmt,LM.Id AS LedgerId,VH.F_OtherLedgerMaster		
	FROM SaleInvoiceH VH
	LEFT JOIN LedgerMaster LM ON LM.Id=VH.F_LedgerMaster
	LEFT JOIN [dbo].[CityMaster] CM ON CM.Id=LM.F_CityMaster
	LEFT JOIN StateMaster SM ON SM.Id=CM.F_StateMaster
	
	LEFT JOIN OtherLedgerMaster OLM ON OLM.Id = VH.F_OtherLedgerMaster
	LEFT JOIN CityMaster OCM ON OCM.Id=OLM.F_CityMaster
	WHERE VH.Id=@Id
END




















































GO
```

# [dbo].[Select_SaleBillNo_DateWise]

```sql
/****** Object:  StoredProcedure [dbo].[Select_SaleBillNo_DateWise]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











CREATE PROCEDURE [dbo].[Select_SaleBillNo_DateWise] 
	@F_CompanyMaster numeric(18,0),
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@Id numeric(18,0)=0,
	@IsEstimate smallint=0
AS
BEGIN
	IF(@IsEstimate=0)
		BEGIN
			SELECT Name,Id,VoucherNo1 FROM (
				SELECT 0 AS RowNumber,'' AS Name,0 AS Id,' [None]' AS VoucherNo1
				
				UNION ALL
				
				SELECT ROW_NUMBER() OVER (order by CAST(VH.VoucherNo AS NUMERIC) ASC) as RowNumber,
				(
					(CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+	
				VH.VoucherNo + '    ' + (SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4))) AS Name
				,VH.Id,
				((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+ VH.VoucherNo) AS VoucherNo1
					FROM SaleInvoiceH VH
						LEFT JOIN LedgerMaster LM ON VH.F_LedgerMaster=LM.Id
					WHERE VH.F_CompanyMaster =@F_CompanyMaster
						AND 1=
							CASE WHEN(@FromDate='') THEN 1 ELSE
								(CASE WHEN(VH.VoucherDate >= @FromDate AND VH.VoucherDate <=@ToDate) THEN 1 ELSE 0 END)
							END		
				) AS Tab1
			ORDER BY RowNumber,Name,Id
		END
	ELSE
		BEGIN
			SELECT Name,Id,VoucherNo1 FROM (
				SELECT 0 AS RowNumber,'' AS Name,0 AS Id,' [None]' AS VoucherNo1
				
				UNION ALL
				
				SELECT ROW_NUMBER() OVER (order by CAST(VH.VoucherNo AS NUMERIC) ASC) as RowNumber,
				(
					(CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+	
				VH.VoucherNo + '    ' + (SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4))) AS Name
				,VH.Id,
				((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+ VH.VoucherNo) AS VoucherNo1
					FROM EstimateH VH
						LEFT JOIN LedgerMaster LM ON VH.F_LedgerMaster=LM.Id
					WHERE VH.F_CompanyMaster =@F_CompanyMaster
						AND 1=
							CASE WHEN(@FromDate='') THEN 1 ELSE
								(CASE WHEN(VH.VoucherDate >= @FromDate AND VH.VoucherDate <=@ToDate) THEN 1 ELSE 0 END)
							END		
				) AS Tab1
			ORDER BY RowNumber,Name,Id		
		END
END




























GO
```

# [dbo].[Select_SaleBillNo_YearWise]

```sql
/****** Object:  StoredProcedure [dbo].[Select_SaleBillNo_YearWise]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






































CREATE PROCEDURE [dbo].[Select_SaleBillNo_YearWise] 
	@F_CompanyMaster numeric(18,0),	
	@Id numeric(18,0)=0,
	@F_CompanyYears numeric(18,0)
AS
BEGIN
	SELECT Name,Id,VoucherNo1 FROM (
		SELECT 0 AS RowNumber,'' AS Name,0 AS Id,' [None]' AS VoucherNo1
		
		UNION ALL
		
		SELECT ROW_NUMBER() OVER (order by CAST(VH.VoucherNo AS NUMERIC) ASC) as RowNumber,
		(
			(CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+	
		VH.VoucherNo + ' ' + (SUBSTRING(VH.VoucherDate,7,2)+'/'+SUBSTRING(VH.VoucherDate,5,2)+'/'+SUBSTRING(VH.VoucherDate,1,4)) + ' ' + LM.Name) AS Name
		,VH.Id,
		((CASE WHEN(ISNULL(VH.VoucherNoPrefix,'')<>'') THEN VH.VoucherNoPrefix + ' ' ELSE '' END)+ VH.VoucherNo) AS VoucherNo1
			FROM SaleInvoiceH VH
				LEFT JOIN LedgerMaster LM ON VH.F_LedgerMaster=LM.Id
				LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=VH.F_CompanyMaster AND CY.Id=@F_CompanyYears
			WHERE VH.F_CompanyMaster =@F_CompanyMaster				
				AND VH.VoucherDate>=CY.FinYearFrom AND VH.VoucherDate<=CY.FinYearTo
		) AS Tab1
	ORDER BY RowNumber,Name,Id
END




















































GO
```

# [dbo].[Select_SaleInvoice_ImportH]

```sql
/****** Object:  StoredProcedure [dbo].[Select_SaleInvoice_ImportH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE PROCEDURE [dbo].[Select_SaleInvoice_ImportH]
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@OrderBy nvarchar(max)='',
	@InvoiceNo nvarchar(50)='',
	@TOP1 NVARCHAR(50)='',
	@VoucherDate_DESC NVARCHAR(50)='',
	@IsSummary SMALLINT=0,
	@FilterBy SMALLINT=0
AS
BEGIN
	DECLARE @WHERE AS NVARCHAR(MAX)

	IF(@FromDate='')
		SET @WHERE = ''
	ELSE
		SET @WHERE = 'WHERE (VL.SaleDate >= ' + @FromDate + ' AND VL.SaleDate<= ' + @ToDate + ')'

	IF(@InvoiceNo<>'')
		BEGIN
			IF(@WHERE <>'')
				SET @WHERE = @WHERE + ' AND '
			ELSE
				SET @WHERE = @WHERE + ' WHERE '
			SET @WHERE = @WHERE + ' VL.BillNo=''' + @InvoiceNo + ''''
		END

-- Converted / Non-converted
	IF(@WHERE <>'')
		SET @WHERE = @WHERE + ' AND '
	ELSE
		SET @WHERE = @WHERE + ' WHERE '
	IF(@FilterBy<>0)
		SET @WHERE = @WHERE + ' VL.BillNo IN ( SELECT BizomInvoiceNo FROM EstimateH WHERE ISNULL(BizomInvoiceNo,'''') <> '''' )'
	ELSE
		SET @WHERE = @WHERE + ' VL.BillNo NOT IN ( SELECT BizomInvoiceNo FROM EstimateH WHERE ISNULL(BizomInvoiceNo,'''') <> '''' )'
--
	IF(@TOP1<>'')
		BEGIN
			SELECT TOP 1 (SUBSTRING(VL.SaleDate,7,2)+'/'+SUBSTRING(VL.SaleDate,5,2)+'/'+SUBSTRING(VL.SaleDate,1,4)) AS InvoiceDate,
				VL.SaleDate AS InvoiceDate1
			FROM SaleInvoice_ImportH VL
			ORDER BY VL.SaleDate DESC
		END
	ELSE IF(@IsSummary=0)
		BEGIN
			EXEC('
				SELECT ' + @TOP1 + ' VL.[Id],
					VL.BuyerName,VL.BuyerCode,VL.BuyerAddress,
					VL.BillNo,
					(SUBSTRING(VL.SaleDate,7,2)+''/''+SUBSTRING(VL.SaleDate,5,2)+''/''+SUBSTRING(VL.SaleDate,1,4)) AS InvoiceDate,
					VL.SkuCode,VL.SkuName,VL.MRP, 
					IM.Name + (CASE WHEN(ISNULL(TDM.TypeDetail,'''')<>'''') THEN '' ('' + ISNULL(TDM.TypeDetail,'''') + '')'' ELSE '''' END) AS Map_Item,
					ISTM.MRP AS Map_MRP,
					VL.[Qty_Pcs],VL.Rate,VL.[F_CompanyYears],
					VL.Scheme_Value,VL.QPS_Scheme_Value,VL.Cash_Discount,
					VL.Taxable_Amount,VL.Total_Tax,VL.Total_Amount,IM.Id AS F_ItemMaster,VL.[BillNo] AS BillNo1,VL.SaleDate AS InvoiceDate1,ISTM.F_ItemTypeMaster,ISTM.F_TypeDetailMaster
				FROM [dbo].[SaleInvoice_ImportH] VL
					LEFT JOIN SaleInvoiceItemMapMaster ISTM ON ISTM.SaleInvoiceItemId = VL.SkuCode
					LEFT JOIN ItemMaster IM ON IM.Id = ISTM.F_ItemMaster
					LEFT JOIN TypeDetailMaster TDM ON TDM.Id = ISTM.F_TypeDetailMaster					
				' + @WHERE + '
				ORDER BY ' + @OrderBy + ' VL.SaleDate ' + @VoucherDate_DESC + '
					,CASE WHEN IsNumeric(VL.BillNo) = 1 THEN Right(Replicate(''0'',21) + VL.BillNo, 20)
						  WHEN IsNumeric(VL.BillNo) = 0 THEN Left(VL.BillNo + Replicate('''',21), 20)
						 ELSE VL.BillNo
					 END,VL.[Id]')
		END
	ELSE
		BEGIN
			EXEC('
				SELECT MAX(VL.[Id]) AS Id,VL.[InvoiceNo],
					(SUBSTRING(VL.SaleDate,7,2)+''/''+SUBSTRING(VL.SaleDate,5,2)+''/''+SUBSTRING(VL.SaleDate,1,4)) AS SaleDate,
					SUM(VL.[Qty_CS]) AS Qty_CS,SUM(VL.[Qty_Pcs]) AS Qty_Pcs,
					SUM(VL.TaxableAmt) AS TaxableAmt,SUM(VL.NetAmt) AS NetAmt,SUM(VL.TotalTax) AS TotalTax,
					SUM(VL.CGSTAmt) AS CGSTAmt,SUM(VL.SGSTAmt) AS SGSTAmt,SUM(VL.DiscAmt) AS DiscAmt
				FROM [dbo].[Transport_ImportH] VL
				' + @WHERE + '
				GROUP BY VL.InvoiceNo,VL.SaleDate
				ORDER BY VL.SaleDate
					,CASE WHEN IsNumeric([InvoiceNo]) = 1 THEN Right(Replicate(''0'',21) + [InvoiceNo], 20)
						  WHEN IsNumeric([InvoiceNo]) = 0 THEN Left([InvoiceNo] + Replicate('''',21), 20)
						 ELSE [InvoiceNo]
					 END')
		END
END




GO
```

# [dbo].[Select_Scheme]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Scheme]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





















































CREATE PROCEDURE [dbo].[Select_Scheme] 
	@VoucherDate nvarchar(50),
	@F_ItemMaster numeric(18,0),
	@Qty numeric(18,0),
	@Amount numeric(18,0)
AS
BEGIN
	SELECT SL.*,UM.Name AS UnitF,IM.Name AS ItemF
	FROM SchemeH SH
		LEFT JOIN SchemeL SL ON SL.F_SchemeH=SH.Id
		LEFT JOIN UnitMaster UM ON UM.Id=SL.F_UnitMasterF
		LEFT JOIN ItemMaster IM ON IM.Id=SL.F_ItemMasterF
	WHERE (@VoucherDate >= SL.FromDate AND @VoucherDate <=SL.ToDate)
		AND F_ItemMaster = @F_ItemMaster
		AND
		(
			(
				MinQty>0 AND
				@Qty>=MinQty AND 1= 
					(CASE WHEN(SL.MaxQty>0) THEN
						(CASE WHEN(@Qty<=SL.MaxQty) THEN 1 ELSE 0 END)
					ELSE 1
					END)
			)
			OR
			(
				MinAmt>0 AND
				@Amount>=MinAmt AND 1= 
					(CASE WHEN(SL.MaxAmt>0) THEN
						(CASE WHEN(@Amount<=SL.MaxAmt) THEN 1 ELSE 0 END)
					ELSE 1
					END)
			)
			OR
			(
			  ((MinQty>0 AND MaxQty>0 AND (MinQty=MaxQty)) OR
				(MinAmt>0 AND MaxAmt>0 AND (MinAmt=MaxAmt))
			   ) AND
			  (@Qty>=MinQty OR @Amount>=MinAmt)
			)
		)
		AND ISNULL(SH.Cancelled,0)=0
	ORDER BY Free DESC
END

















































GO
```

# [dbo].[Select_Scheme_QPS]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Scheme_QPS]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO






















































CREATE PROCEDURE [dbo].[Select_Scheme_QPS] 
	@VoucherDate nvarchar(50),
	@ItemIds nvarchar(50),
	@Qty numeric(18,0),
	@Amount numeric(18,0)
AS
BEGIN
	SELECT SL.*,UM.Name AS UnitF,IM.Name AS ItemF
	FROM SchemeH SH
		LEFT JOIN SchemeL SL ON SL.F_SchemeH=SH.Id
		LEFT JOIN UnitMaster UM ON UM.Id=SL.F_UnitMasterF
		LEFT JOIN ItemMaster IM ON UM.Id=SL.F_ItemMasterF
		LEFT JOIN SchemeBatchMaster SBM ON SBM.Id=SL.F_SchemeBatchMaster
	WHERE (@VoucherDate >= SL.FromDate AND @VoucherDate <=SL.ToDate)
		AND SH.Id 
			IN (SELECT * FROM Split((STUFF(			
			(SELECT DISTINCT ', ' + STR(SBL.F_SchemeH)
				FROM SchemeL_BatchItems SBL
				WHERE SBL.F_SchemeH = SL.F_SchemeH
					AND SBL.F_ItemMaster IN (SELECT * FROM Split(@ItemIds,','))
			FOR XML PATH('')), 1, 2, '')),','))
			
		AND
		(
			(
				MinQty>0 AND
				@Qty>=MinQty AND 1= 
					(CASE WHEN(SL.MaxQty>0) THEN
						(CASE WHEN(@Qty<=SL.MaxQty) THEN 1 ELSE 0 END)
					ELSE 1
					END)
			)
			OR
			(
				MinAmt>0 AND
				@Amount>=MinAmt AND 1= 
					(CASE WHEN(SL.MaxAmt>0) THEN
						(CASE WHEN(@Amount<=SL.MaxAmt) THEN 1 ELSE 0 END)
					ELSE 1
					END)
			)
			--OR
			--(
			--  ((Free>0 AND FreeOnQty>0) OR (Free>0 AND FreeOnAmt>0)) AND
			--  (@Qty>=FreeOnQty OR @Amount>=FreeOnAmt)
			--)
		)
	ORDER BY FreeOnQty DESC,FreeOnAmt DESC
END


















































GO
```

# [dbo].[Select_SchemeDisplay]

```sql
/****** Object:  StoredProcedure [dbo].[Select_SchemeDisplay]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





















































CREATE PROCEDURE [dbo].[Select_SchemeDisplay] 
	@VoucherDate nvarchar(50),
	@F_ItemMaster numeric(18,0)	
AS
BEGIN
	SELECT SBM.Name AS Batch
		--,'' AS Items
		,Items = STUFF(			
			(SELECT DISTINCT ', ' + IMT.Name
			FROM SchemeL_BatchItems ST
				LEFT JOIN ItemMaster IMT ON IMT.Id=ST.F_ItemMaster
			WHERE ST.F_SchemeH = GIL.F_SchemeH AND ST.F_SchemeBatchMaster=GIL.F_SchemeBatchMaster
			FOR XML PATH('')), 1, 2, '')
		,IM.Name AS Item
		,(SUBSTRING(GIL.FromDate,7,2)+'/'+SUBSTRING(GIL.FromDate,5,2)+'/'+SUBSTRING(GIL.FromDate,1,4)) AS FromDate
		,(SUBSTRING(GIL.ToDate,7,2)+'/'+SUBSTRING(GIL.ToDate,5,2)+'/'+SUBSTRING(GIL.ToDate,1,4)) AS ToDate
		,GIL.MinAmt,GIL.MaxAmt,GIL.MinQty,GIL.MaxQty,GIL.DiscRate,GIL.DiscAmt
		,GIL.FreeOnAmt,GIL.FreeOnQty,GIL.Free,UMF.Name AS UnitF,IMF.Name AS ItemF
		,GIL.F_SchemeBatchMaster,GIL.F_ItemMaster,GIL.F_UnitMasterF,GIL.F_ItemMasterF
	FROM SchemeH SH
		LEFT JOIN SchemeL GIL ON GIL.F_SchemeH=SH.Id
		LEFT JOIN SchemeBatchMaster SBM ON SBM.Id=GIL.F_SchemeBatchMaster
		LEFT JOIN ItemMaster IM ON IM.Id = GIL.F_ItemMaster
		LEFT JOIN UnitMaster UMF ON UMF.Id = GIL.F_UnitMasterF
		LEFT JOIN ItemMaster IMF ON IMF.Id = GIL.F_ItemMasterF
	WHERE (@VoucherDate >= GIL.FromDate AND @VoucherDate <=GIL.ToDate)
		AND F_ItemMaster = @F_ItemMaster
		AND ISNULL(SH.Cancelled,0)=0
	ORDER BY FreeOnQty DESC,FreeOnAmt DESC
END

















































GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Select_StockInL]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





















CREATE PROCEDURE [dbo].[Select_StockInL] 
	@F_StockInH numeric(18,0)=-1,
	@F_ItemMaster numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@MRP numeric(18,2)=0
AS
BEGIN
	DECLARE @FromDateLast nvarchar(50)
	SET @FromDateLast = SUBSTRING(@FromDate,1,4)+ '0331'
	
	SELECT IM.Name AS Item,UMQ.Name AS UnitQ
		,GIL.Qty,GIL.MRP,GIL.Margin,GIL.SellingRate,GIL.DMargin,GIL.Cost,LM.Name AS Tax,GIL.BeforeVATRate
		,GIL.SchRate,GIL.SchAmt,GIL.LandingCost
		,GIL.F_ItemMaster,UMQ.Id AS F_UnitMaster,GIL.F_LedgerMaster_Tax
		,GIL.IsManualScheme,GIL.IsManualDMargin,GIL.IsManualMargin
	FROM [dbo].[StockInL] GIL
		LEFT JOIN StockInH GIH ON GIH.Id=GIL.F_StockInH
		LEFT JOIN ItemMaster IM ON IM.Id = GIL.F_ItemMaster
		LEFT JOIN UnitMaster UMQ ON UMQ.Id = GIL.F_UnitMaster
		LEFT JOIN LedgerMaster LM ON LM.Id = GIL.F_LedgerMaster_Tax
	WHERE
		1=(CASE WHEN(@F_StockInH=-1) THEN 1 ELSE
			(CASE WHEN(GIL.F_StockInH = @F_StockInH) THEN 1 ELSE 0 END)
		   END)
		AND 1=(CASE WHEN(@F_ItemMaster=0) THEN 1 ELSE
				(CASE WHEN(@F_ItemMaster=GIL.F_ItemMaster) THEN 1 ELSE 0 END)
			   END)
		AND 1=(CASE WHEN(@FromDate='') THEN 1 ELSE
				(CASE WHEN(GIH.VoucherDate = @FromDateLast) THEN 1 ELSE 0 END)
			   END)
		AND 1=(CASE WHEN(@MRP=0) THEN 1 ELSE
				(CASE WHEN(GIL.MRP=@MRP) THEN 1 ELSE 0 END)
			   END)
	ORDER BY GIL.Id
END



















































GO
```

# [dbo].[Select_StockInL2]

```sql
/****** Object:  StoredProcedure [dbo].[Select_StockInL2]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO























CREATE PROCEDURE [dbo].[Select_StockInL2] 
	@F_StockInH numeric(18,0)=-1,
	@F_ItemMaster numeric(18,0)=0,
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@MRP numeric(18,2)=0
AS
BEGIN
	DECLARE @FromDateLast nvarchar(50)
	SET @FromDateLast = SUBSTRING(@FromDate,1,4)+ '0331'
	
	SELECT IM.Name AS Item
		,HSN.Name AS HSN
		,UMQ.Name AS UnitQ
		,GIL.Qty,GIL.MRP,GIL.Margin,GIL.SellingRate,GIL.DMargin,GIL.Cost
		,ISTM_Tax.TaxName
		,GIL.BeforeVATRate
		,GIL.SchRate,GIL.SchAmt
		,GIL.LandingCost,CAST(ROUND(GIL.LandingCost*GIL.Qty,2) AS NUMERIC(18,2)) AS LandingAmount
		,GIL.F_ItemMaster,UMQ.Id AS F_UnitMaster,GIL.F_LedgerMaster_Tax
		,GIL.IsManualScheme,GIL.IsManualDMargin,GIL.IsManualMargin
		,'' AS PurVoucherNo,'' AS PurVoucherDate,'' AS PurQty,'' AS DiffQty,'' AS Ledger,'' AS TINNo
	FROM [dbo].[StockInL] GIL
		LEFT JOIN StockInH GIH ON GIH.Id=GIL.F_StockInH
		LEFT JOIN ItemMaster IM ON IM.Id = GIL.F_ItemMaster
		LEFT JOIN HSNCodeMaster HSN ON HSN.Id = IM.F_HSNCodeMaster
		LEFT JOIN UnitMaster UMQ ON UMQ.Id = GIL.F_UnitMaster
		OUTER APPLY (SELECT TOP 1 LM.Name AS TaxName,LM.Id AS TaxId
				FROM ItemSalesTaxMaster ISTM LEFT JOIN LedgerMaster LM ON ISTM.F_LedgerMaster_Tax = LM.Id
				WHERE ISTM.SDate<=20170630 AND ISTM.F_ItemMaster = IM.Id
				ORDER BY ISTM.SDate DESC,LM.Name,LM.Id
		) ISTM_Tax
	WHERE
		1=(CASE WHEN(@F_StockInH=-1) THEN 1 ELSE
			(CASE WHEN(GIL.F_StockInH = @F_StockInH) THEN 1 ELSE 0 END)
		   END)
		AND 1=(CASE WHEN(@F_ItemMaster=0) THEN 1 ELSE
				(CASE WHEN(@F_ItemMaster=GIL.F_ItemMaster) THEN 1 ELSE 0 END)
			   END)
		AND 1=(CASE WHEN(@FromDate='') THEN 1 ELSE
				(CASE WHEN(GIH.VoucherDate = @FromDateLast) THEN 1 ELSE 0 END)
			   END)
		AND 1=(CASE WHEN(@MRP=0) THEN 1 ELSE
				(CASE WHEN(GIL.MRP=@MRP) THEN 1 ELSE 0 END)
			   END)
	ORDER BY GIL.Id
END





















































GO
```

# [dbo].[Select_Summary_DueDelivery_DayWise]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Summary_DueDelivery_DayWise]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
























CREATE PROCEDURE [dbo].[Select_Summary_DueDelivery_DayWise]
	@F_CompanyMaster numeric(18,0),
	@F_CompanyYears numeric(18,0),
	@CurrentDate nvarchar(50)	
AS
BEGIN
	SELECT TAB2.[MonthName],TAB2.Total_InProcess,TAB2.Total_Pending
	FROM (
		SELECT 1 AS ROWNUMBER, '0 - 7 Days' AS [MonthName],
				ISNULL(SUM(TAB1.InProcessId),0) AS Total_InProcess,ISNULL(SUM(TAB1.PendingId),0) AS Total_Pending
		FROM (

			--IN PROCESSS
			SELECT GIH.VoucherDate,GIH.VoucherNo,GIH.TotalNetAmt,1 AS InProcessId, 0 AS PendingId
			FROM SaleInvoiceH GIH
				LEFT JOIN DeliveryProcessL DPL ON DPL.F_SaleInvoiceH = GIH.Id
				LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DPL.F_DeliveryStatusMaster
				LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
			WHERE ISNULL(DPL.Id,0)<>0 AND (DPL.F_DeliveryStatusMaster = 0 )
				AND ISNULL(GIH.Cancelled,0)=0 AND GIH.F_CompanyMaster = @F_CompanyMaster
				AND GIH.VoucherDate>=CY.FinYearFrom AND GIH.VoucherDate<=CY.FinYearTo
			UNION ALL
			--PENDING
			SELECT GIH.VoucherDate,GIH.VoucherNo,GIH.TotalNetAmt,0 AS InProcessId, 1 AS PendingId
			FROM SaleInvoiceH GIH
				LEFT JOIN DeliveryProcessL DPL ON DPL.F_SaleInvoiceH = GIH.Id
				LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DPL.F_DeliveryStatusMaster
				LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
			WHERE ISNULL(DPL.Id,0)=0
				AND ISNULL(GIH.Cancelled,0)=0 AND GIH.F_CompanyMaster = @F_CompanyMaster
				AND GIH.VoucherDate>=CY.FinYearFrom AND GIH.VoucherDate<=CY.FinYearTo
		) AS TAB1	
		WHERE 1=CASE WHEN(7 >= DATEDIFF(day,CONVERT(datetime,TAB1.VoucherDate),CONVERT(datetime,@CurrentDate))) THEN 1 ELSE 0 END			
		
		UNION ALL
		
		SELECT 2 AS ROWNUMBER, '8 - 15 Days' AS [MonthName],
				ISNULL(SUM(TAB1.InProcessId),0) AS Total_InProcess,ISNULL(SUM(TAB1.PendingId),0) AS Total_Pending
		FROM (

			--IN PROCESSS
			SELECT GIH.VoucherDate,GIH.VoucherNo,GIH.TotalNetAmt,1 AS InProcessId, 0 AS PendingId
			FROM SaleInvoiceH GIH
				LEFT JOIN DeliveryProcessL DPL ON DPL.F_SaleInvoiceH = GIH.Id
				LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DPL.F_DeliveryStatusMaster
				LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
			WHERE ISNULL(DPL.Id,0)<>0 AND (DPL.F_DeliveryStatusMaster = 0 )
				AND ISNULL(GIH.Cancelled,0)=0 AND GIH.F_CompanyMaster = @F_CompanyMaster
				AND GIH.VoucherDate>=CY.FinYearFrom AND GIH.VoucherDate<=CY.FinYearTo
			UNION ALL
			--PENDING
			SELECT GIH.VoucherDate,GIH.VoucherNo,GIH.TotalNetAmt,0 AS InProcessId, 1 AS PendingId
			FROM SaleInvoiceH GIH
				LEFT JOIN DeliveryProcessL DPL ON DPL.F_SaleInvoiceH = GIH.Id
				LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DPL.F_DeliveryStatusMaster
				LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
			WHERE ISNULL(DPL.Id,0)=0
				AND ISNULL(GIH.Cancelled,0)=0 AND GIH.F_CompanyMaster = @F_CompanyMaster
				AND GIH.VoucherDate>=CY.FinYearFrom AND GIH.VoucherDate<=CY.FinYearTo

		) AS TAB1	
		WHERE 1=CASE WHEN(
					15 >= DATEDIFF(day,CONVERT(datetime,TAB1.VoucherDate),CONVERT(datetime,@CurrentDate))
					AND
					8 <= DATEDIFF(day,CONVERT(datetime,TAB1.VoucherDate),CONVERT(datetime,@CurrentDate))
				) THEN 1 ELSE 0 END			
		
		UNION ALL
		
		SELECT 3 AS ROWNUMBER, 'Above 15 Days' AS [MonthName],
				ISNULL(SUM(TAB1.InProcessId),0) AS Total_InProcess,ISNULL(SUM(TAB1.PendingId),0) AS Total_Pending
		FROM (

			--IN PROCESSS
			SELECT GIH.VoucherDate,GIH.VoucherNo,GIH.TotalNetAmt,1 AS InProcessId, 0 AS PendingId
			FROM SaleInvoiceH GIH
				LEFT JOIN DeliveryProcessL DPL ON DPL.F_SaleInvoiceH = GIH.Id
				LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DPL.F_DeliveryStatusMaster
				LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
			WHERE ISNULL(DPL.Id,0)<>0 AND (DPL.F_DeliveryStatusMaster = 0 )
				AND ISNULL(GIH.Cancelled,0)=0 AND GIH.F_CompanyMaster = @F_CompanyMaster
				AND GIH.VoucherDate>=CY.FinYearFrom AND GIH.VoucherDate<=CY.FinYearTo
			UNION ALL
			--PENDING
			SELECT GIH.VoucherDate,GIH.VoucherNo,GIH.TotalNetAmt,0 AS InProcessId, 1 AS PendingId
			FROM SaleInvoiceH GIH
				LEFT JOIN DeliveryProcessL DPL ON DPL.F_SaleInvoiceH = GIH.Id
				LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DPL.F_DeliveryStatusMaster
				LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
			WHERE ISNULL(DPL.Id,0)=0
				AND ISNULL(GIH.Cancelled,0)=0 AND GIH.F_CompanyMaster = @F_CompanyMaster
				AND GIH.VoucherDate>=CY.FinYearFrom AND GIH.VoucherDate<=CY.FinYearTo
		) AS TAB1	
		WHERE 1=CASE WHEN(15 < DATEDIFF(day,CONVERT(datetime,TAB1.VoucherDate),CONVERT(datetime,@CurrentDate))) THEN 1 ELSE 0 END				
	) AS TAB2
	WHERE (TAB2.Total_InProcess>0 OR TAB2.Total_Pending>0)
	ORDER BY TAB2.ROWNUMBER
END




































































GO
```

# [dbo].[Select_Summary_DueDelivery_MonthWise]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Summary_DueDelivery_MonthWise]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO























CREATE PROCEDURE [dbo].[Select_Summary_DueDelivery_MonthWise]
	@F_CompanyMaster numeric(18,0),
	@F_CompanyYears numeric(18,0)
AS
BEGIN
	SELECT (MAX(SUBSTRING(TAB1.VoucherDate,1,4)) + ' - ' +  SUBSTRING(DateName(mm,DATEADD(mm,CAST(MAX(SUBSTRING(TAB1.VoucherDate,5,2)) AS NUMERIC)-1,0)),0,4)) AS [MonthName],
			SUM(TAB1.InProcessId) AS Total_InProcess,SUM(TAB1.PendingId) AS Total_Pending,
			SUBSTRING(TAB1.VoucherDate,1,6) AS SYear_Month
	FROM (

		--IN PROCESSS
		SELECT GIH.VoucherDate,GIH.VoucherNo,GIH.TotalNetAmt,1 AS InProcessId, 0 AS PendingId
		FROM SaleInvoiceH GIH
			LEFT JOIN DeliveryProcessL DPL ON DPL.F_SaleInvoiceH = GIH.Id
			LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DPL.F_DeliveryStatusMaster
			LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
		WHERE ISNULL(DPL.Id,0)<>0 AND (DPL.F_DeliveryStatusMaster = 0 )
				AND ISNULL(GIH.Cancelled,0)=0 AND GIH.F_CompanyMaster = @F_CompanyMaster
				AND GIH.VoucherDate>=CY.FinYearFrom AND GIH.VoucherDate<=CY.FinYearTo
		UNION ALL
		--PENDING
		SELECT GIH.VoucherDate,GIH.VoucherNo,GIH.TotalNetAmt,0 AS InProcessId, 1 AS PendingId
		FROM SaleInvoiceH GIH
			LEFT JOIN DeliveryProcessL DPL ON DPL.F_SaleInvoiceH = GIH.Id
			LEFT JOIN DeliveryStatusMaster DSM ON DSM.Id=DPL.F_DeliveryStatusMaster
			LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
		WHERE ISNULL(DPL.Id,0)=0
				AND ISNULL(GIH.Cancelled,0)=0 AND GIH.F_CompanyMaster = @F_CompanyMaster
				AND GIH.VoucherDate>=CY.FinYearFrom AND GIH.VoucherDate<=CY.FinYearTo
	) AS TAB1	
	GROUP BY SUBSTRING(TAB1.VoucherDate,1,6)
	ORDER BY SUBSTRING(TAB1.VoucherDate,1,6)
END




































































GO
```

# [dbo].[Select_Summary_DueEstimates_MonthWise]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Summary_DueEstimates_MonthWise]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO




CREATE PROCEDURE [dbo].[Select_Summary_DueEstimates_MonthWise]
--declare
	@F_CompanyMaster numeric(18,0)=1,
	@F_CompanyYears numeric(18,0)=10
AS
BEGIN
	;WITH Sales AS (
	  SELECT SH.F_EstimateH,
			 COUNT(*) AS SaleCount,
			 SUM(SH.TotalNetAmt) AS SaleTotalNetAmt
	  FROM dbo.SaleInvoiceH SH
	  GROUP BY SH.F_EstimateH
	),
	Estimates AS (
	  SELECT
		ym = CONVERT(char(6), GIH.VoucherDate, 112),
		COUNT(*) AS EstimateCount,
		SUM(GIH.TotalNetAmt) AS TotalNetAmt
	  FROM dbo.EstimateH GIH
	  JOIN dbo.LedgerMaster LM ON LM.Id = GIH.F_LedgerMaster
	  JOIN dbo.LedgerGroupMaster LGM ON LGM.Id = LM.F_LedgerGroupMaster AND LGM.GroupType <> 6
	  JOIN dbo.CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
	  WHERE ISNULL(GIH.Cancelled,0)=0
		AND GIH.F_CompanyMaster = @F_CompanyMaster
		AND GIH.VoucherDate >= CY.FinYearFrom
		AND GIH.VoucherDate <= CY.FinYearTo
	  GROUP BY CONVERT(char(6), GIH.VoucherDate, 112)
	)
	SELECT
	  [MonthName] = LEFT(S.ym,4) + ' - ' + LEFT(DATENAME(mm, DATEFROMPARTS(LEFT(S.ym,4), RIGHT(S.ym,2), 1)),3),
	  (E.EstimateCount - ISNULL(S.SaleCount,0)) AS TotalEstimates,
	  (E.TotalNetAmt   - ISNULL(S.SaleTotalNetAmt,0)) AS TotalNetAmt,
	  SYear_Month = S.ym
	FROM Estimates E
	LEFT JOIN (
	   SELECT CONVERT(char(6), GIH.VoucherDate,112) AS ym,
			  SUM(Sales.SaleCount) AS SaleCount,
			  SUM(Sales.SaleTotalNetAmt) AS SaleTotalNetAmt
	   FROM dbo.EstimateH GIH
	   JOIN Sales ON Sales.F_EstimateH = GIH.Id
	   GROUP BY CONVERT(char(6), GIH.VoucherDate,112)
	) S
	  ON S.ym = E.ym
	WHERE (E.EstimateCount - ISNULL(S.SaleCount,0)) > 0
	ORDER BY S.ym;
END

GO
```

# [dbo].[Select_Summary_DueInvoices_MonthWise]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Summary_DueInvoices_MonthWise]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
























CREATE PROCEDURE [dbo].[Select_Summary_DueInvoices_MonthWise]
	@F_CompanyMaster numeric(18,0),
	@F_CompanyYears numeric(18,0)
AS
BEGIN
	--DECLARE @DefaultExpenditurePostingLedger NUMERIC(18,0)
	--SET @DefaultExpenditurePostingLedger = (SELECT DefaultExpenditurePostingLedger FROM Options)

	SELECT (MAX(SUBSTRING(TAB1.VoucherDate,1,4)) + ' - ' +  SUBSTRING(DateName(mm,DATEADD(mm,CAST(MAX(SUBSTRING(TAB1.VoucherDate,5,2)) AS NUMERIC)-1,0)),0,4)) as [MonthName],
			COUNT(TAB1.VoucherNo) AS TotalInvoices
			,SUM((TAB1.TotalNetAmt-(ISNULL(TAB1.MReceiptCashAmount,0)+ISNULL(TAB1.MReceiptBankAmount,0)+ISNULL(TAB1.CreditNoteAmount,0)+ISNULL(TAB1.SaleReturnAmount,0)))) AS TotalNetAmt
			,SUBSTRING(TAB1.VoucherDate,1,6) AS SYear_Month
	FROM (
			SELECT GIH.VoucherDate,GIH.VoucherNo,GIH.TotalNetAmt
				,(SELECT SUM(MRL.Amount)
						FROM MoneyReceiptCashL MRL LEFT JOIN MoneyReceiptCashH MRH ON MRH.Id=MRL.F_MoneyReceiptCashH
						WHERE MRL.F_SaleInvoiceH = GIH.Id AND ISNULL(MRH.Cancelled,0)=0) AS MReceiptCashAmount
				,(SELECT SUM(MRL.Amount)
						FROM MoneyReceiptBankL MRL LEFT JOIN MoneyReceiptBankH MRH ON MRH.Id=MRL.F_MoneyReceiptBankH
						WHERE MRL.F_SaleInvoiceH = GIH.Id AND ISNULL(MRH.Cancelled,0)=0 AND ISNULL(MRH.IsChequeCancel,0)=0) AS MReceiptBankAmount
				,(SELECT SUM(VH.TotalNetAmt)
						FROM SaleInvoiceH_CreditNote SIC LEFT JOIN VoucherH VH ON VH.Id=SIC.F_VoucherH
						WHERE SIC.F_SaleInvoiceH = GIH.Id AND ISNULL(GIH.Cancelled,0)=0 AND ISNULL(VH.Cancelled,0)=0) AS CreditNoteAmount
				,(SELECT SUM(MRL.Amount)
						FROM SaleReturnL_InvoiceDetails MRL LEFT JOIN MoneyReceiptBankH MRH ON MRH.Id=MRL.F_SaleReturnH
						WHERE MRL.F_SaleInvoiceH = GIH.Id AND ISNULL(MRH.Cancelled,0)=0) AS SaleReturnAmount
			FROM SaleInvoiceH GIH
				LEFT JOIN LedgerMaster LM ON LM.Id = GIH.F_LedgerMaster
				LEFT JOIN LedgerGroupMaster LGM ON LGM.Id = LM.F_LedgerGroupMaster
				LEFT JOIN CompanyYears CY ON CY.F_CompanyMaster=GIH.F_CompanyMaster AND CY.Id=@F_CompanyYears
			WHERE ISNULL(GIH.Cancelled,0)=0 AND GIH.F_CompanyMaster = @F_CompanyMaster
				AND GIH.VoucherDate>=CY.FinYearFrom AND GIH.VoucherDate<=CY.FinYearTo
				AND LGM.GroupType NOT IN(6)
	) AS TAB1
	WHERE ((TAB1.TotalNetAmt-(ISNULL(TAB1.MReceiptCashAmount,0)+ISNULL(TAB1.MReceiptBankAmount,0)+ISNULL(TAB1.CreditNoteAmount,0)+ISNULL(TAB1.SaleReturnAmount,0))))>0
	GROUP BY SUBSTRING(TAB1.VoucherDate,1,6)
	ORDER BY SUBSTRING(TAB1.VoucherDate,1,6)
END


















GO
```

# [dbo].[Select_tmp_GSTR2A]

```sql
/****** Object:  StoredProcedure [dbo].[Select_tmp_GSTR2A]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











CREATE PROCEDURE [dbo].[Select_tmp_GSTR2A]
	@GSTNo nvarchar(50),
	@Status SMALLINT=0,
	@F_CompanyYears numeric(18,0),
	@FromDate nvarchar(50),
	@ToDate nvarchar(50)
AS
BEGIN
	
	DECLARE @WHERE_Status NVARCHAR(MAX)
	IF(@Status=0)				--All
		SET @WHERE_Status=' '
	ELSE IF(@Status=1)			--Matched
		SET @WHERE_Status=' AND ISNULL(VL.Status,0)=1'
	ELSE IF(@Status=2)			--Not Exist
		SET @WHERE_Status=' AND ISNULL(VL.Status,0)=2'
	ELSE IF(@Status=3)			--Not Matched
		SET @WHERE_Status=' AND ISNULL(VL.Status,0)=0'

	EXEC('
	SELECT VL.[Id],VL.[Supplier],VL.[GSTN],VL.[InvoiceNo],
		(SUBSTRING(VL.[InvoiceDate],1,2) + ''-'' + LEFT(DATENAME(MONTH, (SUBSTRING(VL.[InvoiceDate],7,4)+''/''+SUBSTRING(VL.[InvoiceDate],4,2)+''/''+SUBSTRING(VL.[InvoiceDate],1,2))),3) + ''-'' + SUBSTRING(VL.[InvoiceDate],9,2)) AS [InvoiceDate],
		VL.[TaxableValue],
		VL.[IGST],VL.[CGST],VL.[SGST],VL.[InvoiceValue],G_S.Name AS [Status],G_S.Name AS [PrevStatus],VL.[InvoiceDate] AS InvoiceDate2
	FROM [dbo].[tmp_GSTR2A] VL
		LEFT JOIN GSTR2A_StatusMaster G_S ON G_S.Id=ISNULL(VL.Status,0)
	WHERE 1=(CASE WHEN(''' + @GSTNo + '''='''') THEN 1 ELSE (CASE WHEN([GSTN]=''' + @GSTNo + ''') THEN 1 ELSE 0 END) END)
		AND (VL.InvoiceDate8>=' + @FromDate + ' AND VL.InvoiceDate8<=' + @ToDate +')
		AND VL.F_CompanyYears = '+@F_CompanyYears
		+ @WHERE_Status + '
	ORDER BY [GSTN],SUBSTRING([InvoiceDate],7,4)+SUBSTRING([InvoiceDate],4,2)+SUBSTRING([InvoiceDate],1,2)
		,CASE WHEN IsNumeric([InvoiceNo]) = 1 then Right(Replicate(''0'',21) + [InvoiceNo], 20)
			  WHEN IsNumeric([InvoiceNo]) = 0 then Left([InvoiceNo] + Replicate('''',21), 20)
			 ELSE [InvoiceNo]
		 END')
END











GO
```

# [dbo].[Select_Transport_ImportH]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Transport_ImportH]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











CREATE PROCEDURE [dbo].[Select_Transport_ImportH]
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@OrderBy nvarchar(max)='',
	@InvoiceNo nvarchar(50)='',
	@TOP1 NVARCHAR(50)='',
	@VoucherDate_DESC NVARCHAR(50)='',
	@IsSummary SMALLINT=0
AS
BEGIN
	DECLARE @WHERE AS NVARCHAR(MAX)

	IF(@FromDate='')
		SET @WHERE = ''
	ELSE
		SET @WHERE = 'WHERE (VL.InvoiceDate >= ' + @FromDate + ' AND VL.InvoiceDate<= ' + @ToDate + ')'

	IF(@InvoiceNo<>'')
		BEGIN
			IF(@WHERE <>'')
				SET @WHERE = @WHERE + ' AND '
			ELSE
				SET @WHERE = @WHERE + ' WHERE '
			SET @WHERE = @WHERE + ' VL.InvoiceNo=''' + @InvoiceNo + ''''
		END
	IF(@IsSummary=0)
		BEGIN
			EXEC('
				SELECT ' + @TOP1+ ' VL.[Id],VL.[InvoiceNo],
					(SUBSTRING(VL.InvoiceDate,7,2)+''/''+SUBSTRING(VL.InvoiceDate,5,2)+''/''+SUBSTRING(VL.InvoiceDate,1,4)) AS InvoiceDate,
					VL.[Item_Id],VL.[Item_Details],
					IM.Name AS Map_Item,ISTM.MRP,
					VL.[Qty_CS],VL.[Qty_Pcs],VL.[F_CompanyYears],
					VL.TaxableAmt,VL.NetAmt,VL.TotalTax,VL.CGSTRate,VL.CGSTAmt,VL.SGSTRate,VL.SGSTAmt
					,IM.Id AS F_ItemMaster,VL.[InvoiceNo] AS InvoiceNo1,VL.InvoiceDate AS InvoiceDate1
					,VL.DiscRate,VL.DiscAmt
				FROM [dbo].[Transport_ImportH] VL
					LEFT JOIN TransportItemMapMaster ISTM ON ISTM.TransportItemId = VL.Item_Id
					LEFT JOIN ItemMaster IM ON IM.Id = ISTM.F_ItemMaster
				' + @WHERE + '
				ORDER BY ' + @OrderBy + ' VL.InvoiceDate ' + @VoucherDate_DESC + '
					,CASE WHEN IsNumeric([InvoiceNo]) = 1 THEN Right(Replicate(''0'',21) + [InvoiceNo], 20)
						  WHEN IsNumeric([InvoiceNo]) = 0 THEN Left([InvoiceNo] + Replicate('''',21), 20)
						 ELSE [InvoiceNo]
					 END')
		END
	ELSE
		BEGIN
			EXEC('
				SELECT MAX(VL.[Id]) AS Id,VL.[InvoiceNo],
					(SUBSTRING(VL.InvoiceDate,7,2)+''/''+SUBSTRING(VL.InvoiceDate,5,2)+''/''+SUBSTRING(VL.InvoiceDate,1,4)) AS InvoiceDate,
					SUM(VL.[Qty_CS]) AS Qty_CS,SUM(VL.[Qty_Pcs]) AS Qty_Pcs,
					SUM(VL.TaxableAmt) AS TaxableAmt,SUM(VL.NetAmt) AS NetAmt,SUM(VL.TotalTax) AS TotalTax,
					SUM(VL.CGSTAmt) AS CGSTAmt,SUM(VL.SGSTAmt) AS SGSTAmt,SUM(VL.DiscAmt) AS DiscAmt
				FROM [dbo].[Transport_ImportH] VL
				' + @WHERE + '
				GROUP BY VL.InvoiceNo,VL.InvoiceDate
				ORDER BY VL.InvoiceDate
					,CASE WHEN IsNumeric([InvoiceNo]) = 1 THEN Right(Replicate(''0'',21) + [InvoiceNo], 20)
						  WHEN IsNumeric([InvoiceNo]) = 0 THEN Left([InvoiceNo] + Replicate('''',21), 20)
						 ELSE [InvoiceNo]
					 END')
		END
END




GO
```

# [dbo].[Select_Transport_PurVoucherNo_List]

```sql
/****** Object:  StoredProcedure [dbo].[Select_Transport_PurVoucherNo_List]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











CREATE PROCEDURE [dbo].[Select_Transport_PurVoucherNo_List]
	@LotSize numeric(18,2)
AS
BEGIN
	SELECT TAB1.InvoiceNo,TAB1.InvoiceDate,TAB1.Name
	FROM(
		SELECT InvoiceNo,InvoiceDate,SUM(Qty_CS) AS LotSize,
		(InvoiceNo + ' (' + (SUBSTRING(InvoiceDate,7,2)+'/'+SUBSTRING(InvoiceDate,5,2)+'/'+SUBSTRING(InvoiceDate,1,4)) + ')' ) AS Name
		FROM [Transport_ImportH]
		GROUP BY InvoiceNo,InvoiceDate
	) AS TAB1
	WHERE TAB1.LotSize=@LotSize
	ORDER BY CASE WHEN IsNumeric(TAB1.[InvoiceNo]) = 1 THEN Right(Replicate('0',21) + TAB1.[InvoiceNo], 20)
			  WHEN IsNumeric(TAB1.[InvoiceNo]) = 0 THEN Left(TAB1.[InvoiceNo] + Replicate('',21), 20)
				ELSE TAB1.[InvoiceNo]
			 END
END




GO
```

# [dbo].[Select_VoucherL_InvoiceRollover]

```sql
/****** Object:  StoredProcedure [dbo].[Select_VoucherL_InvoiceRollover]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
























CREATE PROCEDURE [dbo].[Select_VoucherL_InvoiceRollover] 
	@F_VoucherTypeMaster numeric(18,0)=0,
	@Id numeric(18,0)=0,
	@F_LedgerMaster numeric(18,0)=0,
	@Return smallint=0,
	@MatchField nvarchar(50)='F_MoneyReceiptBankH'
AS
BEGIN
	IF(@Return=0)
		BEGIN
			IF(@MatchField='F_MoneyReceiptBankH')
				BEGIN
					SELECT (CASE WHEN(GIH.NetAmount=0) THEN GIH.ChequeCancelCharges ELSE
							(CASE WHEN(GIH.F_VoucherH = GIH.F_VoucherH_Journal) THEN GIH.ChequeCancelCharges ELSE GIH.NetAmount END)
						END) AS Amount, GIH.Id AS F_MoneyReceiptBankH,VL.VoucherNo,VL.F_LedgerMaster,0 AS F_SaleReturnH
					FROM MoneyReceiptBankH GIH
					LEFT JOIN VoucherL VL ON GIH.F_VoucherH=VL.F_VoucherH 
					WHERE VL.F_VoucherH=@Id AND VL.F_VoucherTypeMaster=@F_VoucherTypeMaster AND VL.F_LedgerMaster =@F_LedgerMaster
				END
			ELSE
				BEGIN
					SELECT (CASE WHEN(GIH.NetAmount=0) THEN GIH.ChequeCancelCharges ELSE
							(CASE WHEN(ISNULL(GBH.F_VoucherH,0) = GIH.F_VoucherH_Journal) THEN GIH.ChequeCancelCharges ELSE GIH.NetAmount END)
						END) AS Amount, GIH.Id AS F_MoneyReceiptCashH,VL.VoucherNo,VL.F_LedgerMaster,0 AS F_SaleReturnH
					FROM MoneyReceiptCashH GIH
					LEFT JOIN VoucherL VL ON GIH.F_VoucherH_Journal=VL.F_VoucherH
					OUTER APPLY (SELECT GBH.F_VoucherH FROM MoneyReceiptBankH GBH WHERE GBH.F_VoucherH = GIH.F_VoucherH_Journal) AS GBH
					WHERE VL.F_VoucherH=@Id AND VL.F_VoucherTypeMaster=@F_VoucherTypeMaster AND VL.F_LedgerMaster =@F_LedgerMaster
				END
		END
	ELSE
		BEGIN
			SELECT VH.VoucherNo,VH.Id
			FROM VoucherH VH
			LEFT JOIN MoneyReceiptBankH MRB ON MRB.F_VoucherH=VH.Id
			WHERE VH.F_LedgerMaster =@F_LedgerMaster AND MRB.IsChequeCancel=1
			AND VH.Id NOT IN
			(
				SELECT F_VoucherH_Journal FROM MoneyReceiptCashH WHERE ISNULL(F_VoucherH_Journal,0)>0
				UNION ALL
				SELECT F_VoucherH_Journal FROM MoneyReceiptBankH WHERE ISNULL(F_VoucherH_Journal,0)>0
			)
			OR VH.Id=@Id
		END
END



















GO
```

# [dbo].[SelectWithNone_Master]

```sql
/****** Object:  StoredProcedure [dbo].[SelectWithNone_Master]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO















CREATE PROCEDURE [dbo].[SelectWithNone_Master]
	@TableName nvarchar(50),
	@FieldName nvarchar(50) ='Name',
	@MatchFieldName1 nvarchar(50)='',
	@MatchFieldValue1 nvarchar(50)='',
	@MatchFieldName2 nvarchar(50)='',
	@MatchFieldValue2 nvarchar(50)='',
	@SelectionName nvarchar(50)=' [None]',
	@OrderBy nvarchar(50)='Name,Id',
	@F_CompanyMaster numeric(18,0)=1,
	@Ids nvarchar(max)='',
	@IdsField nvarchar(50)='Id',
	@IdsField_Display nvarchar(max)='Id',
	@FromDate nvarchar(50)='',
	@ToDate nvarchar(50)='',
	@VoucherDateField nvarchar(50)='VoucherDate',
	@GroupBy nvarchar(max)=',Id'
AS
BEGIN
	DECLARE @SQL nvarchar(max)

	DECLARE @WHERE NVARCHAR(MAX)
	SET @WHERE=''

	IF EXISTS(SELECT * FROM SYS.COLUMNS
				WHERE Name = N'F_CompanyMaster' AND Object_ID = Object_ID(@TableName))    
	BEGIN
		-- Column Exists
		IF(@MatchFieldName1<>'')
			SET @WHERE = ' AND F_CompanyMaster = ' + CAST(@F_CompanyMaster AS NVARCHAR)
		ELSE
			SET @WHERE = ' WHERE F_CompanyMaster = ' + CAST(@F_CompanyMaster AS NVARCHAR)
	END

	IF(@Ids<>'')
		BEGIN
			IF(@MatchFieldName1<>'')
				SET @WHERE = @WHERE + ' AND 1=(CASE WHEN (''' + @Ids + '''='''') THEN 1 ELSE (CASE WHEN (' + @IdsField + ' IN (SELECT * FROM Split(''' + @Ids + ''','',''))) THEN 1 ELSE 0 END) END)'
			ELSE				
				SET @WHERE = @WHERE + ' WHERE 1=(CASE WHEN (''' + @Ids + '''='''') THEN 1 ELSE (CASE WHEN (' + @IdsField + ' IN (SELECT * FROM Split(''' + @Ids + ''','',''))) THEN 1 ELSE 0 END) END)'
		END
	
	IF(@FromDate<>'')
		BEGIN
			IF(@MatchFieldName1<>'')
				SET @WHERE= @WHERE + ' AND (' + @VoucherDateField + '>=' + @FromDate + ' AND ' + @VoucherDateField + '<=' + @ToDate + ')'
			ELSE
				SET @WHERE= @WHERE + ' WHERE (' + @VoucherDateField + '>=' + @FromDate + ' AND ' + @VoucherDateField + '<=' + @ToDate + ')'
		END

	SET @SQL = 'SELECT ''0'' AS Id,''' + @SelectionName + ''' AS Name UNION ALL
			SELECT ' + @IdsField_Display + ',' + @FieldName + ' AS Name FROM ' + + @TableName +
			CASE WHEN(@MatchFieldName1='') THEN '' ELSE
			' WHERE ' + @MatchFieldName1 + ' = ' + @MatchFieldValue1
			END +
			CASE WHEN(@MatchFieldName2='') THEN '' ELSE
			' AND ' + @MatchFieldName2 + ' = ' + @MatchFieldValue2
			END
			+ @WHERE +
			+ ' GROUP BY ' + @FieldName + @GroupBy
			+ ' ORDER BY ' + @OrderBy

	EXEC (@SQL)
END













GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Update_GSTR2A_Status]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











CREATE PROCEDURE [dbo].[Update_GSTR2A_Status]
	@F_VoucherH numeric(18,0),
	@F_VoucherTypeMaster numeric(18,0)=0,
	@Status nvarchar(50),
	@VoucherNo nvarchar(50),
	@VoucherDate nvarchar(50),
	@IsUpdateVoucherL smallint=0,
	@GSTN NVARCHAR(50)=''
AS
BEGIN
	DECLARE @StatusValue SMALLINT
	IF(@Status='Matched')
		SET @StatusValue=1
	ELSE IF(@Status='Not Exist')
		SET @StatusValue=2
	ELSE IF(@Status='Not Claim')
		SET @StatusValue=4
	ELSE
		SET @StatusValue=0
--Updateing VoucherL
	IF(@IsUpdateVoucherL=0)
		BEGIN
			UPDATE VoucherL
			SET [Status]=@StatusValue
			WHERE F_VoucherH = @F_VoucherH AND F_VoucherTypeMaster = @F_VoucherTypeMaster AND VoucherNo=@VoucherNo --AND VoucherDate = @VoucherDate
		END
	ELSE
		BEGIN
			UPDATE tmp_GSTR2A
			SET [Status]=@StatusValue
			WHERE GSTN=@GSTN AND InvoiceNo=@VoucherNo AND InvoiceDate=@VoucherDate AND Id=@F_VoucherH
		END
END











GO
```

# [dbo].[Update_LedgerMaster]

```sql
/****** Object:  StoredProcedure [dbo].[Update_LedgerMaster]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO











CREATE PROCEDURE [dbo].[Update_LedgerMaster]
	@Id numeric(18, 0),
	@F_UserMaster numeric(18, 0),
	@OpBal numeric(18,2)=0,
	@DrCr smallint=0,
	@CreditLimit_NoOfDays numeric(18,0)=0,
	@CreditLimit_NoOfBills numeric(18,0)=0,
	@UpdateType smallint=0,
	@F_CompanyYears numeric(18,0)=0
AS
BEGIN
	DECLARE @Id_LedgerMasterL numeric(18,0)
	DECLARE @F_LedgerGroupMaster numeric(18,0)
	IF(@UpdateType=0)
		BEGIN
			--UPDATE LedgerMaster
			--SET OpeningBalance=@OpBal,DrCr=@DrCr
			--WHERE Id=@Id
			SET @Id_LedgerMasterL = ISNULL((SELECT Id FROM LedgerMasterL WHERE F_LedgerMaster = @Id AND F_CompanyYears = @F_CompanyYears),0)
			SET @F_LedgerGroupMaster = ISNULL((SELECT F_LedgerGroupMaster FROM LedgerMaster WHERE Id=@Id),0)
			IF(@Id_LedgerMasterL =0)
				BEGIN
					INSERT INTO [dbo].[LedgerMasterL]([F_LedgerMaster],[DepRate],[F_CompanyYears],[F_UserMaster],[DateOfCreation],[F_LedgerGroupMaster],[OpeningBalance],[DrCr])
						VALUES (@Id,0,@F_CompanyYears,@F_UserMaster,GETDATE(),@F_LedgerGroupMaster,@OpBal,@DrCr)
				END
			ELSE
				BEGIN
					UPDATE LedgerMasterL
					SET OpeningBalance=@OpBal,DrCr=@DrCr,LastOpeningBalance=@OpBal,LastDrCr=@DrCr
					WHERE F_LedgerMaster=@Id AND F_CompanyYears=@F_CompanyYears
				END
		END
	ELSE IF(@UpdateType=1)
		BEGIN
			UPDATE LedgerMaster
			SET [CreditLimit_NoOfDays] = @CreditLimit_NoOfDays,
				[CreditLimit_NoOfBills] = @CreditLimit_NoOfBills
			WHERE Id=@Id
		END
	ELSE IF(@UpdateType=2)	--Split
		BEGIN
			DECLARE @FinYearFrom NVARCHAR(50)
			DECLARE @CompanyYearId NUMERIC(18,0)
			SET @FinYearFrom = (SELECT FinYearFrom FROM CompanyYears WHERE Id=@F_CompanyYears)
			SET @CompanyYearId = (SELECT TOP 1 Id FROM CompanyYears WHERE FinYearFrom>@FinYearFrom)

			SET @Id_LedgerMasterL = ISNULL((SELECT Id FROM LedgerMasterL WHERE F_LedgerMaster = @Id AND F_CompanyYears = @CompanyYearId),0)
			SET @F_LedgerGroupMaster = ISNULL((SELECT F_LedgerGroupMaster FROM LedgerMaster WHERE Id=@Id),0)
			IF(@Id_LedgerMasterL =0)
				BEGIN
					INSERT INTO [dbo].[LedgerMasterL]([F_LedgerMaster],[DepRate],[F_CompanyYears],[F_UserMaster],[DateOfCreation],[F_LedgerGroupMaster],[OpeningBalance],[DrCr])
						VALUES (@Id,0,@CompanyYearId,@F_UserMaster,GETDATE(),@F_LedgerGroupMaster,@OpBal,@DrCr)
				END
			ELSE
				BEGIN
					UPDATE LedgerMasterL
					SET OpeningBalance=@OpBal,DrCr=@DrCr,LastOpeningBalance=@OpBal,LastDrCr=@DrCr
					WHERE F_LedgerMaster=@Id AND F_CompanyYears=@CompanyYearId
				END
		END
END





GO
```

```sql
/****** Object:  StoredProcedure [dbo].[Update_Master]    Script Date: 20/09/2025 13:30:46 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO





















































CREATE PROCEDURE [dbo].[Update_Master]
	@TableName nvarchar(50),
	@Id numeric(18,0),
	@F_UserMaster numeric(18,0),
	@F_CompanyMaster numeric(18,0),
	@Name nvarchar(150),
	@FieldName nvarchar(150)='Name'
AS
BEGIN
	DECLARE @SQL nvarchar(max)
	SET @SQL = 'UPDATE ' + + @TableName + ' SET ' +
		'[LastUpdateOn] = ' + '''' +  convert(nvarchar(50),GETDATE()) + '''' + 
		', [F_UserMaster] = ' + '''' + convert(nvarchar,@F_UserMaster) + '''' + 
		', [F_CompanyMaster] = ' + '''' + convert(nvarchar,@F_CompanyMaster) + '''' + 
		', ' + @FieldName + ' = ' + '''' + convert(nvarchar,@Name) + '''' + 
		' WHERE [Id] = ' + '''' + convert(nvarchar,@Id) + '''' 
	EXEC (@SQL)	

END



















































GO
```
